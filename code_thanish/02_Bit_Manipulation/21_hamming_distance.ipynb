To solve the problem of calculating the **Hamming distance** between two integers `x` and `y`, we need to determine how many bits are different between the two numbers.

### Problem Analysis:

- **Hamming distance** is defined as the number of bit positions at which two numbers differ.
- The most efficient way to find the number of differing bits between two integers is by using the **XOR** operation (`^`).
  - XOR between two bits produces `1` if the bits are different and `0` if they are the same.
  - Thus, XORing the two integers will highlight the differing bits.

### Steps:

1. Perform a bitwise XOR between `x` and `y`. This will result in a number where each bit is `1` if the corresponding bits of `x` and `y` differ, and `0` otherwise.
2. Count the number of `1`s in the result of the XOR operation. This count represents the Hamming distance.

### Approach:

1. **XOR** the two numbers: `diff = x ^ y`.
2. Count the number of `1` bits in the binary representation of `diff`. This is equivalent to counting the differing bit positions between `x` and `y`.
3. We can use a built-in method to count the number of `1` bits, such as `bit_count()` in Rust (which has an O(1) time complexity due to the fixed bit-width of integers).

### Time Complexity:

- **O(1)**: The time complexity is constant since the number of bits in the integers is fixed at 32 bits (for a 32-bit integer).

### Space Complexity:

- **O(1)**: We are using a constant amount of space regardless of the input size.

### Rust Code Implementation:

```rust
impl Solution {
    pub fn hamming_distance(x: i32, y: i32) -> i32 {
        // Perform XOR operation between x and y
        let diff = x ^ y;

        // Count the number of 1 bits in the result
        diff.count_ones() as i32
    }
}

fn main() {
    // Test Case 1
    let x1 = 1;
    let y1 = 4;
    println!("Hamming distance between {} and {}: {}", x1, y1, Solution::hamming_distance(x1, y1));
    // Output: 2

    // Test Case 2

    let y2 = 1;
    println!("Hamming distance between {} and {}: {}", x2, y2, Solution::hamming_distance(x2, y2));
    // Output: 1
}
```

### Explanation of the Code:

1. **XOR Operation**:
   - `let diff = x ^ y;`: This line performs the XOR operation between `x` and `y`. For each bit position, the result is `1` if the bits are different, and `0` if they are the same.
2. **Counting the `1`s**:

   - `diff.count_ones()`: This is a built-in Rust function that counts the number of `1` bits in the binary representation of the number. It's highly optimized and operates in constant time since we're dealing with a 32-bit integer.
   - We cast the result to `i32` to match the return type of the function.

3. **Main function**:
   - We test the function with two sets of inputs:
     - In the first test case, `x = 1` and `y = 4`. The XOR result will be `5` (`0101`), which has two `1` bits, so the output is `2`.
     - In the second test case, `x = 3` and `y = 1`. The XOR result will be `2` (`0010`), which has one `1` bit, so the output is `1`.

````rust
pub struct Solution;

impl Solution {
    pub fn hamming_distance(x: i32, y: i32) -> i32 {
        // Perform XOR operation between x and y
        let diff = x ^ y;

        // Count the number of 1 bits in the result
        diff.count_ones() as i32
    }
}

fn main() {
    // Test Case 1
    let x1 = 1;
    let y1 = 4;
    println!("Hamming distance between {} and {}: {}", x1, y1, Solution::hamming_distance(x1, y1)); 
    // Output: 2

    // Test Case 2
    let x2 = 3;
    let y2 = 1;
    println!("Hamming distance between {} and {}: {}", x2, y2, Solution::hamming_distance(x2, y2)); 
    // Output: 1
}


main()
````
