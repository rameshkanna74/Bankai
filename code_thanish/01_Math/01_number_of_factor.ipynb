/// Returns the total number of factors of a given integer `n`.
///
/// # Arguments
/// * `n` - An integer in the range [1, 100000].
///
/// # Returns
/// * The count of positive divisors of `n`.
pub fn count_factors(n: u32) -> u32 {
    let mut count = 0;
    let sqrt_n = (n as f64).sqrt() as u32;

    for i in 1..=sqrt_n {
        if n % i == 0 {
            count += if i == n / i { 1 } else { 2 };
        }
    }

    count
}

````rust
fn count_factors(n: i32) -> i32 {
    // Take the absolute value of n to ensure we are counting divisors of positive numbers
    let n = n.abs();
    let mut divisor_count = 0;

    println!("Finding divisors of: {}", n);

    // Loop through all integers from 1 to sqrt(N)
    for i in 1..=((n as f64).sqrt() as i32) {
        if n % i == 0 {
            println!("Divisor found: {}", i);
            // i is a divisor
            divisor_count += 1;

            // Check if N / i is a different divisor
            if i != n / i {
                println!("Pairing divisor found: {}", n / i);
                divisor_count += 1;
            }
        }
    }

    println!("Total divisors: {}", divisor_count);
    divisor_count
}

fn main() {
    // Example 1
    let n1 = -255;
    println!("\nExample 1: n = {}", n1);
    println!("Number of divisors of {}: {}", n1, count_factors(n1)); // Expected output: 8

    // Example 2
    let n2 = 96;
    println!("\nExample 2: n = {}", n2);
    println!("Number of divisors of {}: {}", n2, count_factors(n2)); // Expected output: 12
}
````
