To solve the problem of finding the number of **unique binary search trees** (BSTs) that can be made with `n` unique nodes, we can utilize **Dynamic Programming (DP)**. This problem is a classic example that uses the **Catalan number** sequence, which gives the number of unique BSTs that can be made from `n` nodes.

### Key Insights:

- **Binary Search Tree Properties**: For any given node, if it is chosen as the root, all nodes less than it will be in the left subtree, and all nodes greater than it will be in the right subtree. This leads to a recursive structure.
- The number of unique BSTs for `n` nodes can be calculated as the sum of products of the number of BSTs formed by splitting the nodes at each possible root. For any root `i` (where `1 ≤ i ≤ n`), the left subtree will have `i - 1` nodes, and the right subtree will have `n - i` nodes. Hence, the number of unique BSTs with `i` as the root is the product of the number of BSTs of the left and right subtrees.

The recursive formula for the number of unique BSTs with `n` nodes is:
\[
dp[n] = \sum\_{i=1}^{n} dp[i-1] \times dp[n-i]
\]
Where `dp[i]` represents the number of unique BSTs that can be formed with `i` nodes.

### Base Case:

- `dp[0] = 1`: There is 1 BST that can be formed with 0 nodes (the empty tree).
- `dp[1] = 1`: There is 1 BST that can be formed with 1 node (just the node itself).

### Approach:

1. We will maintain a DP array `dp` where `dp[i]` will store the number of unique BSTs that can be formed with `i` nodes.
2. For each number of nodes `i`, we will compute the number of unique BSTs by considering each number from `1` to `i` as the root and using the recursive formula.

### Rust Code Implementation:

```rust
impl Solution {
    // Function to calculate the number of unique BSTs for n nodes
    pub fn num_trees(n: i32) -> i32 {
        // Create a dp array to store the number of unique BSTs for each number of nodes
        let mut dp = vec![0; (n + 1) as usize];

        // Base case: 1 way to make a BST with 0 nodes or 1 node
        dp[0] = 1; // 1 BST with 0 nodes (empty tree)
        dp[1] = 1; // 1 BST with 1 node

        // Fill the dp array for all cases from 2 to n
        for nodes in 2..=n {
            // For each number of nodes, calculate the number of unique BSTs
            for root in 1..=nodes {
                // dp[root - 1] is the number of BSTs in the left subtree
                // dp[nodes - root] is the number of BSTs in the right subtree
                dp[nodes as usize] += dp[(root - 1) as usize] * dp[(nodes - root) as usize];
            }
        }

        // The result will be stored in dp[n]
        dp[n as usize]
    }
}

fn main() {
    // Example 1: n = 3
    let n = 3;
    let result = Solution::num_trees(n);
    println!("The number of unique BSTs with {} nodes is: {}", n, result);
    // Expected output: 5

    // Example 2: n = 1
    let n = 1;
    let result = Solution::num_trees(n);
    println!("The number of unique BSTs with {} nodes is: {}", n, result);
    // Expected output: 1
}
```

### Explanation of the Code:

1. **Dynamic Programming Array (`dp`)**:
   - The array `dp` stores the number of unique BSTs that can be formed with a specific number of nodes. For example, `dp[i]` represents the number of unique BSTs that can be formed with `i` nodes.
2. **Base Cases**:

   - `dp[0] = 1`: There is exactly 1 BST that can be formed with 0 nodes, which is the empty tree.
   - `dp[1] = 1`: There is exactly 1 BST that can be formed with 1 node, which is the single node tree.

3. **Filling the DP Array**:

   - We loop through all possible values of `nodes` (from 2 to `n`), and for each `nodes`, we check every possible root node (from 1 to `nodes`).
   - For each root `i`, the number of unique BSTs formed is the product of:
     - The number of BSTs that can be formed by the left subtree (`dp[i - 1]`).
     - The number of BSTs that can be formed by the right subtree (`dp[n - i]`).
   - We accumulate these products into `dp[n]`.

4. **Returning the Result**:
   - After filling the DP table, `dp[n]` will contain the number of unique BSTs that can be formed with `n` nodes.

### Time Complexity:

- **Time Complexity**: \( O(n^2) \), where `n` is the input number. This is because for each `nodes` from 2 to `n`, we iterate over all `root` values (which leads to an inner loop running `n` times for each `nodes`).
- **Space Complexity**: \( O(n) \), where `n` is the input number. We use a `dp` array of size `n+1` to store the number of unique BSTs for each number of nodes.
