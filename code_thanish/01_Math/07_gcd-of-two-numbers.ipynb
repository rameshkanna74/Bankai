To solve the problem of finding the **Greatest Common Divisor (GCD)** of two numbers efficiently, we can use the **Euclidean algorithm**, which has a time complexity of \( O(\log(\min(a, b))) \). This is the most optimal approach for this problem.

### Key Idea:
The Euclidean algorithm for GCD works on the principle that:

\[
\text{gcd}(a, b) = \text{gcd}(b, a \% b)
\]

This continues until \( b = 0 \), at which point \( a \) will be the GCD of the original two numbers.

### Algorithm:
1. If \( b = 0 \), the GCD is \( a \).
2. Otherwise, compute \( a \% b \) and call the function recursively with \( b \) and \( a \% b \).

This process ensures that at each step, we reduce the size of the numbers and quickly reach the GCD.

### Rust Code Implementation:

```rust
impl Solution {
    // Function to find the GCD of two numbers using the Euclidean algorithm
    pub fn gcd(a: i32, b: i32) -> i32 {
        let mut x = a;
        let mut y = b;
        
        // Apply the Euclidean algorithm until one number becomes zero
        while y != 0 {
            let temp = y;
            y = x % y; // Update y to be the remainder of x divided by y
            x = temp;  // Update x to be the old y
        }
        
        x // When y becomes 0, x is the GCD
    }
}

fn main() {
    // Test the GCD function with example inputs
    
    let a = 3;
    let b = 6;
    let result = Solution::gcd(a, b);
    println!("GCD of {} and {} is: {}", a, b, result); // Expected output: 3
    
    let a = 1;
    let b = 1;
    let result = Solution::gcd(a, b);
    println!("GCD of {} and {} is: {}", a, b, result); // Expected output: 1
}
```
