# ✅ Final Interview-Ready Solution

```rust
impl Solution {
    /// Returns the count of prime numbers less than `n`.
    /// Time Complexity: O(n log log n)
    /// Space Complexity: O(n)
    pub fn count_primes(n: i32) -> i32 {
        if n <= 2 {
            return 0;
        }
        
        let n = n as usize;
        let mut is_prime = vec![true; n];
        is_prime[0] = false;
        is_prime[1] = false;

        let limit = (n as f64).sqrt() as usize + 1;

        for i in 2..limit {
            if is_prime[i] {
                for multiple in (i * i..n).step_by(i) {
                    is_prime[multiple] = false;
                }
            }
        }

        is_prime.iter().filter(|&&prime| prime).count() as i32
    }
}
```

# ✅ Algorithm Explanation

### 🔍 Core DSA Concepts:
- **Sieve of Eratosthenes** (Prime number generation).
- **Mathematical optimization** (eliminate redundant checks).
- **Arrays and iteration** (standard collections and loop techniques).

---

### 🔍 Algorithm Design:
1. **Edge Case Handling**: For `n <= 2`, return `0` immediately.
2. **Sieve Initialization**: Use a boolean vector `is_prime[n]`, where `true` denotes a prime candidate.
3. **Sieve Process**:
   - Iterate from `2` to `sqrt(n)` and eliminate multiples of each prime.
   - Start marking non-primes from `i * i` to avoid redundant work.
4. **Count Primes**: Count all entries in `is_prime` that remain `true`.

---

### 🔍 Time Complexity:
- **\( O(n \log \log n) \)**: This is the optimal time for generating all primes less than `n`.
  - Every prime `p` removes about \( n/p \) numbers.
  - The harmonic series of primes sums to \( \log \log n \).

---

### 🔍 Space Complexity:
- **\( O(n) \)**: We maintain a boolean array of length `n`.

---

### 🔍 Trade-offs:
| Approach              | Time Complexity | Space Complexity | Reason for rejection |
|-----------------------|-----------------|------------------|----------------------|
| Trial Division        | \( O(n\sqrt{n}) \) | \( O(1) \)        | Too slow for large `n` |
| Segmented Sieve       | \( O(n \log \log n) \) | Lower \( O(\sqrt{n}) \)  | Overkill for single query |
| Bit-packed Sieve      | \( O(n \log \log n) \) | \( O(n/8) \)      | Slightly complex, but a future optimization |
