To solve the problem of checking whether an array is sorted in non-decreasing order, we can iterate through the array once and compare each element with the next one. If at any point an element is greater than the next, the array is not sorted in non-decreasing order, and we return `false`. If we go through the entire array without finding such a pair, we return `true`.

### Time and Space Complexity:
- **Time Complexity**: The time complexity is **O(n)**, where `n` is the size of the array, since we make a single pass through the array.
- **Space Complexity**: The space complexity is **O(1)**, because we only need a few variables to track the comparison, and we don't require extra space proportional to the size of the array.

### Rust Code:

```rust
fn is_sorted(arr: &[i32]) -> bool {
    // Loop through the array from the first element to the second-last element
    for i in 0..arr.len() - 1 {
        // Compare the current element with the next one
        if arr[i] > arr[i + 1] {
            // If any element is greater than the next, return false
            return false;
        }
    }
    // If no such pair was found, return true
    true
}

fn main() {
    // Example 1
    let arr1 = vec![10, 20, 30, 40, 50];
    println!("Is arr1 sorted? {}", is_sorted(&arr1));  // Expected: true
    
    // Example 2
    let arr2 = vec![90, 80, 100, 70, 40, 30];
    println!("Is arr2 sorted? {}", is_sorted(&arr2));  // Expected: false
}
```
