## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn move_zeroes(nums: &mut Vec<i32>) {
        let mut last_non_zero = 0; // Pointer to track the position of non-zero elements

        // Move non-zero elements to the front
        for i in 0..nums.len() {
            if nums[i] != 0 {
                nums.swap(i, last_non_zero);
                last_non_zero += 1;
            }
        }
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `nums`.
  - We iterate through the array once (`O(N)`) and perform at most `O(N)` swaps.
  - **Overall Complexity**: **O(N)** (linear time).

### **Space Complexity:**

- **O(1) extra space**.
  - We modify the array **in-place**.
  - Only a few integer variables (`last_non_zero`) are used.
  - **No additional data structures** are used.

---

## **Key Optimizations**

1. **Two-Pointer Approach**

   - `last_non_zero` tracks the position where the next non-zero element should be placed.
   - Swapping ensures all **non-zero elements are shifted left**, and zeros remain at the end.

2. **Minimizing Swap Operations**

   - This approach performs at most **N swaps**.
   - **Avoids unnecessary writes**, reducing memory write overhead.

3. **Handles Edge Cases Efficiently**
   - Single-element arrays
   - Already sorted input
   - All zeros or no zeros cases
