## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn longest_ones(nums: Vec<i32>, k: i32) -> i32 {
        let mut left = 0;
        let mut zero_count = 0;
        let mut max_len = 0;

        for right in 0..nums.len() {
            if nums[right] == 0 {
                zero_count += 1;
            }

            while zero_count > k {
                if nums[left] == 0 {
                    zero_count -= 1;
                }
                left += 1;
            }

            max_len = max_len.max(right - left + 1);
        }

        max_len as i32
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `nums`.
- Each element is processed **at most twice**: once by `right` expanding and once by `left` shrinking.

### **Space Complexity:**

- **O(1)** since only a few integer variables (`left`, `zero_count`, `max_len`) are used.

---

## **Key Optimizations**

1. **Sliding Window for O(N) Complexity**

   - Expand the `right` pointer to increase the window.
   - Adjust the `left` pointer when more than `k` zeroes exist.

2. **No Extra Data Structures**

   - Operates in **constant space** without auxiliary arrays or hash maps.

3. **Early Exit for Edge Cases**
   - If `k == nums.len()`, return `nums.len()` immediately (all can be flipped).

---

## **Edge Cases Considered**

1. **Minimum Input (`nums.len() = 1`)**

   - Example: `nums = [0], k = 1`
   - ‚úÖ Returns `1`.

2. **All Ones (No Need to Flip)**

   - Example: `nums = [1,1,1,1,1], k = 2`

3. **All Zeroes (Flip All)**
