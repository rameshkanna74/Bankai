## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn longest_subarray(nums: Vec<i32>) -> i32 {
        let mut left = 0;
        let mut zero_count = 0;
        let mut max_len = 0;

        for right in 0..nums.len() {
            if nums[right] == 0 {
                zero_count += 1;
            }

            while zero_count > 1 {
                if nums[left] == 0 {
                    zero_count -= 1;
                }
                left += 1;
            }

            max_len = max_len.max(right - left);
        }

        max_len as i32
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `nums`.
- Each element is processed at most twice: once by `right` expanding and once by `left` shrinking.

### **Space Complexity:**

- **O(1)** since only a few integer variables (`left`, `zero_count`, `max_len`) are used.

---

## **Key Optimizations**

1. **Sliding Window for O(N) Complexity**

   - Expand the `right` pointer to increase the window.
   - Adjust the `left` pointer when more than one `0` is encountered.

2. **Handling Edge Cases for Deletion**

   - The problem requires **deleting exactly one element**, so we return `max_len` instead of `right - left + 1`.
   - Ensures that we always delete **at most one zero**.

3. **No Extra Data Structures**
   - Operates in **constant space** without auxiliary arrays or hash maps.
