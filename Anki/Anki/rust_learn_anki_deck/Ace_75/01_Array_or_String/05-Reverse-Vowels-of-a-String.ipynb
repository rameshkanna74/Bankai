## **Optimized Rust Solution**
```rust
impl Solution {
    pub fn reverse_vowels(s: String) -> String {
        let mut chars: Vec<char> = s.chars().collect();
        let vowels: std::collections::HashSet<char> = 
            ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'].iter().cloned().collect();

        let (mut left, mut right) = (0, chars.len() - 1);

        while left < right {
            while left < right && !vowels.contains(&chars[left]) {
                left += 1;
            }
            while left < right && !vowels.contains(&chars[right]) {
                right -= 1;
            }
            if left < right {
                chars.swap(left, right);
                left += 1;
                right -= 1;
            }
        }
        
        chars.into_iter().collect()
    }
}
```

## **Algorithm Analysis**
### **Time Complexity:**
- **O(N)**, where `N` is the length of the string.
  - The **two-pointer technique** ensures each character is checked at most **once**.
  - Swapping vowels is **O(1)**.
  - `HashSet::contains()` is **O(1)** on average.

### **Space Complexity:**
- **O(N)** (due to the mutable character array `Vec<char>`).
  - This is required since Rust's `String` is immutable.
  - However, no **extra auxiliary space** is used, only input storage.

---

## **Key Optimizations**
1. **Two-Pointer Approach:**
   - The **left** pointer moves forward until it finds a vowel.
   - The **right** pointer moves backward until it finds a vowel.
   - If both are vowels, **swap** and continue.
   - This ensures **O(N) time complexity**.

2. **Fast Vowel Lookup with `HashSet`:**
   - Using `HashSet` allows **O(1) average-time lookups** for vowel checking.
   - This is **faster** than using `.contains()` on an array.

3. **In-Place Swapping:**
   - Instead of **storing** vowels separately and reversing, we **swap in-place**.
   - Reduces extra **memory usage**.
