## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn compress(chars: &mut Vec<char>) -> i32 {
        let mut write = 0; // Position to write the compressed characters
        let mut read = 0;  // Position to read characters

        while read < chars.len() {
            let current_char = chars[read];
            let mut count = 0;

            // Count occurrences of the current character
            while read < chars.len() && chars[read] == current_char {
                read += 1;
                count += 1;
            }

            // Write the character
            chars[write] = current_char;
            write += 1;

            // Write the count if greater than 1
            if count > 1 {
                for digit in count.to_string().chars() {
                    chars[write] = digit;
                    write += 1;
                }
            }
        }

        write as i32 // Return the new length of the compressed array
    }
}
```

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `chars`.
  - We traverse the array **once** with a `read` pointer.
  - Each character is written at most **once** with a `write` pointer.
  - **Total Complexity**: **O(N)** (linear time).

### **Space Complexity:**

- **O(1) extra space**.
  - We modify the input array **in-place**.
  - We only use a few **integer variables** (`write`, `read`, `count`).
  - No additional data structures are used.

---

## **Key Optimizations**

1. **Two-Pointer Approach**

   - **`read` pointer** scans the input array.
   - **`write` pointer** modifies the array in place.
   - Ensures an **in-place O(1) space solution**.

2. **Efficient Character Encoding**

   - Converts **group counts (e.g., `12`) to characters (`'1'`, `'2'`)**.
   - Uses `count.to_string().chars()` to extract digits efficiently.

3. **Early Exit for Efficiency**
   - The loop **stops early** if `read >= chars.len()`, reducing unnecessary iterations.
