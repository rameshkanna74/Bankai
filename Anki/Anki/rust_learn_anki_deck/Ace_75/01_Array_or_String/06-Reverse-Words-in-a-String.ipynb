Given an input string s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.

Return a string of the words in reverse order concatenated by a single space.

Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should only have a single space separating the words. Do not include any extra spaces.

Example 1:

Input: s = "the sky is blue"
Output: "blue is sky the"

Example 2:

Input: s = " hello world "
Output: "world hello"
Explanation: Your reversed string should not contain leading or trailing spaces.

Example 3:

Input: s = "a good example"
Output: "example good a"
Explanation: You need to reduce multiple spaces between two words to a single space in the reversed string.

Constraints:

    1 <= s.length <= 104
    s contains English letters (upper-case and lower-case), digits, and spaces ' '.
    There is at least one word in s.

## **Optimized Rust Solution**
```rust
impl Solution {
    pub fn reverse_words(s: String) -> String {
        s.split_whitespace()
            .rev()
            .collect::<Vec<&str>>()
            .join(" ")
    }
}
```

---

## **Algorithm Analysis**
### **Time Complexity:**
- **O(N)**, where `N` is the length of the string.
  - **Splitting (`split_whitespace`)**: `O(N)`, as Rust scans the string once.
  - **Reversing (`rev()`)**: `O(W)`, where `W` is the number of words (`W ≤ N`).
  - **Joining (`join(" ")`)**: `O(N)`, as it constructs a new string.
  - **Total Complexity**: `O(N)`, which is optimal.

### **Space Complexity:**
- **O(N) (extra space for `Vec<&str>` and output string)**
  - The function does not modify `s` in-place due to Rust's immutability.
  - Requires a temporary `Vec<&str>` to store words, which is `O(W)`.
  - The final result is stored in a new `String`, which is `O(N)`.
  - **Trade-off**: Uses `Vec<&str>`, but avoids unnecessary string operations.

---

## **Key Optimizations**
1. **Efficient Splitting using `split_whitespace()`**
   - Automatically **removes extra spaces** (leading, trailing, multiple spaces).
   - **More efficient** than manual space trimming.

2. **Reversing in O(W) time**
   - Uses `.rev()` which **swaps elements in-place**.
   - Avoids unnecessary looping or index manipulations.

3. **Minimal Memory Overhead**
   - Uses `Vec<&str>` instead of allocating new `String` for each word.
   - `join(" ")` is optimized for **string concatenation**.

---

## **Edge Cases Considered**
1. **Leading and Trailing Spaces:**  
   - Example: `"  hello world  "` → `"world hello"`
   - `split_whitespace()` **removes** these automatically.

2. **Multiple Spaces Between Words:**  
   - Example: `"a  good   example"` → `"example good a"`
   - Ensures **single spaces** in the output.

3. **Single Word Input:**  
   - Example: `"hello"` → `"hello"`  
   - Ensures **no extra spaces added**.

4. **Only Spaces (No Words):**  
   - Example: `"    "` → `""`  
   - Returns an **empty string**.

5. **Mix of Uppercase, Lowercase, Digits:**  
   - Example: `"Hello123 World456"` → `"World456 Hello123"`
   - Preserves **original word order**.
