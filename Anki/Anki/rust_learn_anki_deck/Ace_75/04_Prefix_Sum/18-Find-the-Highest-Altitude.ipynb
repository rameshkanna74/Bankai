## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn largest_altitude(gain: Vec<i32>) -> i32 {
        let mut max_altitude = 0;
        let mut current_altitude = 0;

        for g in gain {
            current_altitude += g;
            max_altitude = max_altitude.max(current_altitude);
        }

        max_altitude
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `gain`.
- We iterate through `gain` once, performing O(1) operations per element.

### **Space Complexity:**

- **O(1)** since we use only two integer variables (`max_altitude` and `current_altitude`).
- No extra data structures are used.

---

## **Key Optimizations**

1. **Single Pass Computation (O(N))**

   - We **accumulate altitude changes** in a **single loop**, keeping track of the maximum.
   - Avoids unnecessary array allocations or prefix sum storage.

2. **No Extra Data Structures**
   - Directly computes the **highest altitude** in **O(1) space**.
