Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

Input: n = 2
Output: [0,1,1]

Input: n = 5

3 --> 11
4 --> 100
5 --> 101

Constraints:

    0 <= n <= 105

### Key Observations:

1. **Binary Representation**: The binary representation of numbers exhibits a regular pattern. If we know the number of set bits for a number `i`, we can compute it for `i + 1` using the previous result.

2. **Optimization Insight**: Instead of counting set bits for each number individually (which would be inefficient), we can use the relationship:
   - `countBits(i) = countBits(i >> 1) + (i & 1)`
   - This works because `i >> 1` essentially removes the least significant bit (LSB), and `(i & 1)` checks whether the LSB is 1. Thus, we can compute the number of set bits of `i` based on the result for `i >> 1`.

### Approach:

- We start with an array `ans` of size `n + 1` initialized with zeros.
- For each number `i` from `1` to `n`, we compute `ans[i]` using the relation `ans[i] = ans[i >> 1] + (i & 1)`.
- This results in a **linear time complexity** solution: \(O(n)\), since we're iterating through the numbers once and using bitwise operations, which are constant time.

### Time and Space Complexity:

- **Time Complexity**: \(O(n)\), since we iterate through the numbers from `0` to `n` and each operation takes constant time.
- **Space Complexity**: \(O(n)\), for the result array that stores the number of set bits for each number from `0` to `n`.

impl Solution {
    pub fn count_bits(n: i32) -> Vec<i32> {
        // Create a vector to store the number of set bits for each number from 0 to n
        let mut ans = vec![0; (n + 1) as usize];

        // Iterate over each number from 1 to n
        for i in 1..=n {
            // The number of set bits in i is the same as the number of set bits in i >> 1,
            // plus the current least significant bit (i & 1) if it's 1
            ans[i as usize] = ans[(i >> 1) as usize] + (i & 1);
        }

        ans
    }
}
