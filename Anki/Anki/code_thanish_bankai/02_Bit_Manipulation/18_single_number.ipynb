To solve the problem of finding the single number in an array where every other number appears twice, we can use a very efficient method using the **XOR (exclusive or)** bitwise operator. This solution is optimal in both **time complexity** and **space complexity**, as it meets the constraints of \(O(n)\) time and \(O(1)\) space.

### Key Observations:

1. **XOR properties**:
   - \( a \oplus a = 0 \) (XOR of a number with itself results in 0)
   - \( a \oplus 0 = a \) (XOR of a number with 0 results in the number itself)
   - XOR is both **commutative** and **associative**, meaning the order of operations doesn’t matter.
2. **How this helps**:
   - If every number appears twice except for one, the numbers that appear twice will cancel each other out (because \(x \oplus x = 0\)).
   - The only number that won’t cancel out is the single number, because XORing with 0 leaves it unchanged.

### Plan:

1. Iterate through the array and XOR all the numbers.
2. Since the XOR of two identical numbers results in 0, all pairs will cancel out, and we’ll be left with the number that doesn’t repeat.

### Time Complexity:

- **O(n)**, where `n` is the number of elements in the array. We only need one pass through the array.

### Space Complexity:

- **O(1)** because we are using only a constant amount of extra space (just one integer to store the result).

### Rust Code Implementation:

```rust
impl Solution {
    pub fn single_number(nums: Vec<i32>) -> i32 {
        // Initialize the result variable to 0
        let mut result = 0;

        // Iterate through each number in the array and apply XOR
        for num in nums {
            result ^= num; // XOR the current number with the result
        }

        // After iterating through all numbers, result holds the single number
        result
    }
}

fn main() {
    // Test cases
    let nums1 = vec![2, 2, 1];
    let nums2 = vec![4, 1, 2, 1, 2];
    let nums3 = vec![1];

    // Call the function and print results
    println!("Single number in nums1: {}", Solution::single_number(nums1)); // Output: 1
    println!("Single number in nums2: {}", Solution::single_number(nums2)); // Output: 4
    println!("Single number in nums3: {}", Solution::single_number(nums3)); // Output: 1
}
```

### Explanation:

1. **`let mut result = 0;`**:

   - We initialize a variable `result` to 0. This will be used to store the cumulative XOR result.

2. **`for num in nums { result ^= num; }`**:

   - We iterate through the array `nums`. For each element `num`, we perform an XOR operation between `result` and `num`.
   - Initially, `result` is 0. After the first XOR with a number, it will be that number, then the next XOR with a repeated number will cancel out, and so on.

3. **Final result**:
   - Once all numbers have been XORed, `result` will contain the single number that does not have a pair.

### Example Walkthrough:

#### Example 1:

Input: `nums = [2, 2, 1]`

1. **Initial result = 0**
2. **First iteration**: `result = 0 ^ 2 = 2`
3. **Second iteration**: `result = 2 ^ 2 = 0` (cancel out)
4. **Third iteration**: `result = 0 ^ 1 = 1`

Final result: `1` (the single number).

#### Example 2:

Input: `nums = [4, 1, 2, 1, 2]`

1. **Initial result = 0**
2. **First iteration**: `result = 0 ^ 4 = 4`
3. **Second iteration**: `result = 4 ^ 1 = 5`
4. **Third iteration**: `result = 5 ^ 2 = 7`
5. **Fourth iteration**: `result = 7 ^ 1 = 6`
6. **Fifth iteration**: `result = 6 ^ 2 = 4`

Final result: `4` (the single number).

#### Example 3:

Input: `nums = [1]`

1. **Initial result = 0**
2. **First iteration**: `result = 0 ^ 1 = 1`

Final result: `1` (the single number).

### Conclusion:

This approach is highly efficient, and it takes constant space and linear time to find the single number in the array. The use of XOR ensures that the solution is optimal in both time and space.


````rust
pub struct Solution;

impl Solution {
    pub fn single_number(nums: Vec<i32>) -> i32 {
        // Initialize the result variable to 0
        let mut result = 0;

        // Iterate through each number in the array and apply XOR
        for num in nums {
            result ^= num; // XOR the current number with the result
        }

        // After iterating through all numbers, result holds the single number
        result
    }
}

fn main() {
    // Test cases
    let nums1 = vec![2, 2, 1];
    let nums2 = vec![4, 1, 2, 1, 2];
    let nums3 = vec![1];

    // Call the function and print results
    println!("Single number in nums1: {}", Solution::single_number(nums1)); // Output: 1
    println!("Single number in nums2: {}", Solution::single_number(nums2)); // Output: 4
    println!("Single number in nums3: {}", Solution::single_number(nums3)); // Output: 1
}

main()
````
