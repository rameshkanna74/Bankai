````python
import pandas as pd


def markdown_to_dataframe(markdown_text):
    # Split the markdown text by line breaks
    lines = markdown_text.strip().split("\n")

    # Extract the header and separator lines
    header_line = lines[0]
    separator_line = lines[1]

    # Extract the rest of the data rows
    data_lines = lines[2:]

    # Process the header line (remove leading/trailing spaces)
    headers = [header.strip() for header in header_line.split("|") if header.strip()]

    # Process the data rows
    data = []
    for line in data_lines:
        if not line.strip():
            continue  # Skip empty lines
        # Split and clean the data row (remove leading/trailing spaces)
        row = [col.strip() for col in line.split("|") if col.strip()]
        if len(row) == len(headers):  # Only add rows that match header length
            data.append(row)

    # Create a DataFrame from the parsed headers and data
    df = pd.DataFrame(data, columns=headers)

    return df
````
