```python
import pandas as pd
from io import StringIO

def markdown_to_dataframe(markdown_str):
    # Replace the markdown table delimiter with a pandas-compatible delimiter
    df = pd.read_csv(StringIO(markdown_str), sep="|", engine="python")

    # Drop any empty columns and reset the index
    df = df.dropna(axis=1, how="all").reset_index(drop=True)

    # Clean up the column names
    df.columns = [col.strip() for col in df.columns]

    return df
```
