\[
nCr = \frac{n!}{r!(n-r)!}

impl Solution {
    pub fn ncr(n: u32, r: u32) -> u32 {
        if r > n {
            return 0;
        }
        if r == 0 || r == n {
            return 1;
        }

        let r = r.min(n - r); // Optimize by reducing computations
        let mut num = 1;
        let mut den = 1;

        for i in 0..r {
            num *= n - i;
            den *= i + 1;
        }

        num / den
    }
}
