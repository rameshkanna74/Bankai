## **Optimal Approach**

The problem requires checking whether \( n \) is prime in **\( O(\sqrt{n}) \) time complexity**. The optimal approach leverages the following optimizations:

1. **Eliminate trivial cases**:

   - Numbers \( n \leq 1 \) are not prime.
   - \( n = 2 \) and \( n = 3 \) are prime (smallest prime numbers).
   - Even numbers (\( n \% 2 == 0 \)) greater than 2 are **not prime**.
   - Numbers divisible by 3 (\( n \% 3 == 0 \)) greater than 3 are **not prime**.

2. **Iterate up to \( \sqrt{n} \) efficiently**:
   - Instead of checking all numbers, we check divisibility by numbers of the form **\( 6k \pm 1 \)** up to \( \sqrt{n} \).
   - This works because all primes \( > 3 \) are of the form \( 6k \pm 1 \).
   - Reduces the number of iterations significantly.

fn is_prime(n: u32) -> u32 {
    if n <= 1 {
        return 0;
    }
    if n == 2 || n == 3 {
        return 1;
    }
    if n % 2 == 0 || n % 3 == 0 {
        return 0;
    }

    let mut i = 5;
    while i * i <= n {
        if n % i == 0 || n % (i + 2) == 0 {
            return 0;
        }
        i += 6;
    }

    1
}
