## **Optimal Approach**

The key observation is that factors of \( N \) always appear in pairs:  
For example, for \( N = 36 \), the factors are:
\[
1 \times 36, 2 \times 18, 3 \times 12, 4 \times 9, 6 \times 6
\]
Thus, iterating up to \( \sqrt{N} \) is sufficient to find all factors.

- If \( i \) divides \( N \) evenly, both \( i \) and \( N/i \) are factors.
- Special case: If \( i \times i = N \), count \( i \) only once to avoid double-counting.
- This results in an **O(âˆšN) time complexity**, which is optimal for factorization.

## **Rust Code (Interview-Ready)**

```rust
fn count_factors(n: u32) -> u32 {
    let mut count = 0;
    let sqrt_n = (n as f64).sqrt() as u32;

    for i in 1..=sqrt_n {
        if n % i == 0 {
            count += if i == n / i { 1 } else { 2 };
        }
    }

    count
}
```

## **Algorithm Explanation**

1. **Iterate up to \( \sqrt{N} \)**:

   - If \( i \) is a factor of \( N \), then \( N/i \) is also a factor.
   - This prevents iterating up to \( N \), reducing complexity to \( O(\sqrt{N}) \).

2. **Avoid double-counting**:
   - If \( i \times i = N \) (i.e., perfect square), count it only once.
