# **Optimized Rust Solution for Unique Binary Search Trees (Catalan Number)**

## **Approach**

The number of unique Binary Search Trees (BSTs) with \( n \) nodes follows the **Catalan Number** formula:

\[
C(n) = \sum\_{i=1}^{n} C(i-1) \times C(n-i)
\]

Where:

- Each \( i \) represents the **root node**.
- The left subtree contains **\( i-1 \) nodes**.
- The right subtree contains **\( n-i \) nodes**.

This recurrence can be solved using **Dynamic Programming (DP)** or using the **direct Catalan number formula**.

---

## **Optimized Rust Code (Production-Quality & Interview-Ready)**

### **Approach 1: Dynamic Programming (DP) - \( O(n^2) \)**

```rust
impl Solution {
    pub fn num_trees(n: i32) -> i32 {
        let n = n as usize;
        let mut dp = vec![0; n + 1];
        dp[0] = 1;
        dp[1] = 1;

        for nodes in 2..=n {
            for root in 1..=nodes {
                dp[nodes] += dp[root - 1] * dp[nodes - root];
            }
        }

        dp[n] as i32
    }
}
```

## **Alternative Approach: Using Direct Catalan Formula**

Instead of DP, we can compute **\( C(n) \) using a direct mathematical formula**:

\[
C(n) = \frac{(2n)!}{(n+1)! n!}
\]

Using integer division, the iterative form is:

\[
C(n) = \frac{C(n-1) \times 2(2n - 1)}{n + 1}
\]

### **Optimized Rust Code - \( O(n) \)**

```rust
impl Solution {
    pub fn num_trees(n: i32) -> i32 {
        let mut c = 1;
        for i in 0..n {
            c = c * 2 * (2 * i + 1) / (i + 2);
        }
        c
    }
}
```

## **Algorithm Explanation**

### **Approach 1: Dynamic Programming (DP)**

1. **Base Cases**:

   - \( dp[0] = 1 \) (An empty tree has one structure).
   - \( dp[1] = 1 \) (A single-node tree has one structure).

2. **DP Recurrence Relation**:
   - Iterate from \( 2 \) to \( n \).
   - For each **root node** at \( i \), compute:
     \[
     dp[nodes] += dp[left] \times dp[right]
     \]
   - Store results to avoid recomputation.

#### **Time Complexity: \( O(n^2) \)**

- Two nested loops:
  - Outer loop: \( O(n) \)
  - Inner loop: \( O(n) \)
- Total: \( O(n^2) \).

#### **Space Complexity: \( O(n) \)**

- Uses an array of size \( n+1 \).

---

### **Approach 2: Direct Catalan Number Calculation**

1. Uses:
   \[
   C(n) = \frac{C(n-1) \times 2(2n - 1)}{n + 1}
   \]
2. Computes in **\( O(n) \) time** using integer division.

#### **Time Complexity: \( O(n) \)**

- Uses only a single loop.

#### **Space Complexity: \( O(1) \)**

- Uses only a few integer variables.
