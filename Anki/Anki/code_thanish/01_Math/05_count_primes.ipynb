# **Optimized Rust Solution for Counting Primes (Leetcode 204)**

## **Approach: Optimized Sieve of Eratosthenes**

We use the **Sieve of Eratosthenes** algorithm, which efficiently finds all primes **strictly less than \( n \)** in **\( O(n \log \log n) \)** time.

### **Why Sieve of Eratosthenes?**

- **Naive approach** (checking each numberâ€™s primality up to \( \sqrt{n} \)) would take **\( O(n \sqrt{n}) \)**, which is too slow for large \( n \) (\( 5 \times 10^6 \)).
- **Sieve of Eratosthenes** runs in **\( O(n \log \log n) \)**, making it the best known method for this problem.

```rust
impl Solution {
    pub fn count_primes(n: i32) -> i32 {
        if n < 2 {
            return 0;
        }

        let n = n as usize;
        let mut is_prime = vec![true; n];
        is_prime[0] = false;
        is_prime[1] = false;

        let mut count = 0;

        for p in 2..n {
            if is_prime[p] {
                count += 1;
                let mut multiple = p * p;
                while multiple < n {
                    is_prime[multiple] = false;
                    multiple += p;
                }
            }
        }

        count
    }
}
```

## **Algorithm Explanation**

### **1. Initialize a Boolean Array**

- We use a boolean array `is_prime` where `is_prime[i] = true` means \( i \) is assumed to be prime.
- Mark **0 and 1 as non-prime** (`false`).

### **2. Iterate Over Numbers from 2 to \( n-1 \)**

- If `is_prime[p]` is `true`, \( p \) is **a prime number**.
- **Increase the count** of primes.
- **Mark all multiples of \( p \) starting from \( p^2 \) as non-prime** to optimize performance.

---

## **Time & Space Complexity Analysis**

### **Time Complexity: \( O(n \log \log n) \)**

- Each number is marked as **non-prime at most once per prime factor**.
- The number of prime factors up to \( N \) grows at a rate of \( \frac{N}{\log N} \), leading to **\( O(n \log \log n) \)** complexity.

### **Space Complexity: \( O(n) \)**

- The boolean array `is_prime` takes **\( O(n) \)** space.
- **No additional allocations**, making this solution optimal.
