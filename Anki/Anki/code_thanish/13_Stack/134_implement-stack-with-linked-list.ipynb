# **Optimized Rust Solution for Stack Implementation Using Linked List**

## **Approach**

We implement a **stack using a singly linked list**, ensuring **O(1) time complexity** for all operations:

1. **push(value):** Inserts an element at the top of the stack.
2. **pop():** Removes the top element if the stack is not empty.
3. **get_top():** Returns the top element; returns `-1` if empty.
4. **is_empty():** Returns `true` if the stack is empty.
5. **get_size():** Returns the size of the stack.

---

## **Time & Space Complexity**

âœ… **Time Complexity:**

- **Each operation (Push, Pop, Get Top, Is Empty, Get Size) runs in `O(1)`.**
- This is achieved using a **linked list** with a head pointer.

âœ… **Space Complexity:**

- **O(N) space for storing elements**, where `N` is the number of elements in the stack.
- **O(1) extra space** since we only store the `size` and `head` pointer.

---

## **Rust Implementation**

```rust
/// Represents a node in the stack (singly linked list).
struct Node {
    value: i32,
    next: Option<Box<Node>>,
}

/// Stack implemented using a singly linked list.
pub struct Stack {
    head: Option<Box<Node>>, // Points to the top element
    size: usize,             // Stores the current stack size
}

impl Stack {
    /// Creates a new empty stack.
    pub fn new() -> Self {
        Self { head: None, size: 0 }
    }

    /// Returns the current size of the stack.
    pub fn get_size(&self) -> usize {
        self.size
    }

    /// Returns true if the stack is empty, false otherwise.
    pub fn is_empty(&self) -> bool {
        self.head.is_none()
    }

    /// Pushes an element onto the stack.
    pub fn push(&mut self, value: i32) {
        let new_node = Box::new(Node {
            value,
            next: self.head.take(),
        });
        self.head = Some(new_node);
        self.size += 1;
    }

    /// Removes the top element of the stack.
    pub fn pop(&mut self) {
        if self.head.is_some() {
            self.head = self.head.take().unwrap().next;
            self.size -= 1;
        }
    }

    /// Returns the top element of the stack or `-1` if empty.
    pub fn get_top(&self) -> i32 {
        match &self.head {
            Some(node) => node.value,
            None => -1,
        }
    }
}
```

    fn test_stack_operations() {
        let mut stack = Stack::new();

        // Initially, the stack is empty
        assert_eq!(stack.is_empty(), true);
        assert_eq!(stack.get_size(), 0);
        assert_eq!(stack.get_top(), -1);

        // Push elements
        stack.push(5);
        stack.push(4);

        // Check size and top element
        assert_eq!(stack.get_size(), 2);
        assert_eq!(stack.is_empty(), false);
        assert_eq!(stack.get_top(), 4);

        // Pop an element
        stack.pop();
        assert_eq!(stack.get_size(), 1);
        assert_eq!(stack.get_top(), 5);

        // Pop last element
        stack.pop();
        assert_eq!(stack.get_size(), 0);
        assert_eq!(stack.is_empty(), true);
        assert_eq!(stack.get_top(), -1);

        // Popping from an empty stack should do nothing
        stack.pop();
        assert_eq!(stack.get_size(), 0);
    }
}
