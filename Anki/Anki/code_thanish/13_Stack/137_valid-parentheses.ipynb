# **Optimized Rust Solution for Valid Parentheses**

## **Approach**

We need to validate if a given string of parentheses is **balanced**:

- Every **open bracket** must be **closed** by the same type of bracket.
- Brackets must be **closed in the correct order**.

### **Efficient Solution (Using a Stack)**

- We use a **stack** to keep track of open brackets.
- When we encounter a **closing bracket**, we check if it matches the **most recent open bracket** (top of the stack).
- If not, the string is **invalid**.
- At the end, if the stack is **empty**, the string is **valid**.

### **Example Walkthrough**

| Step | Character | Stack         | Valid?            |
| ---- | --------- | ------------- | ----------------- |
| 1    | `(`       | `[`(`)`       | âœ…                |
| 2    | `{`       | `[`(`, `{``]` | âœ…                |
| 3    | `}`       | `[`(``]`      | âœ… (matched `{}`) |
| 4    | `)`       | `[]`          | âœ… (matched `()`) |
| 5    | `]`       | `empty`       | âœ… (matched `[]`) |

## **Rust Implementation**

```rust
use std::collections::HashMap;

impl Solution {
    pub fn is_valid(s: String) -> bool {
        let mut stack = Vec::new();
        let matching_pairs: HashMap<char, char> = [(')', '('), (']', '['), ('}', '{')].iter().cloned().collect();

        for c in s.chars() {
            if matching_pairs.contains_key(&c) {
                // If stack is empty or top element doesn't match, it's invalid
                if stack.pop() != Some(*matching_pairs.get(&c).unwrap()) {
                    return false;
                }
            } else {
                stack.push(c);
            }
        }
        stack.is_empty()
    }
}
```
