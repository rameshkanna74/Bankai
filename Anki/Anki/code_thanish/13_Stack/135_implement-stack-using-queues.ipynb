# **Optimized Rust Solution for Stack Using Queues**

## **Approach**

We implement a **stack (LIFO) using two queues (FIFO)**. There are two main approaches:

1. **Push-efficient approach (O(1) push, O(N) pop & top)**

   - Insert elements into **queue1** directly.
   - To pop/top, move all elements except the last to **queue2**, swap queues, and return the last element.

2. **Pop-efficient approach (O(N) push, O(1) pop & top) âœ… (Preferred for Interview)**
   - Always keep the topmost element at the front of **queue1** by reversing the order at every push.
   - This ensures O(1) pop and top operations.

We'll implement the **pop-efficient** approach since **interviewers often prefer O(1) pop and top operations for efficiency**.

---

## **Time & Space Complexity**

âœ… **Time Complexity:**

- `push(x) â†’ O(N)`, because we reverse elements when pushing.
- `pop() â†’ O(1)`, since we remove the front element.
- `top() â†’ O(1)`, since we access the front element.
- `empty() â†’ O(1)`, since we check if `queue1` is empty.

âœ… **Space Complexity:**

- **O(N)** for storing elements in `queue1`.

## **Rust Implementation**

```rust
use std::collections::VecDeque;

/// Stack implemented using a single queue
pub struct MyStack {
    queue1: VecDeque<i32>, // Primary queue maintaining stack order
}

impl MyStack {
    /// Creates a new empty stack.
    pub fn new() -> Self {
        Self {
            queue1: VecDeque::new(),
        }
    }

    /// Pushes an element onto the stack.
    /// This operation ensures the last pushed element is always at the front.
    pub fn push(&mut self, x: i32) {
        let size = self.queue1.len();
        self.queue1.push_back(x); // Insert new element

        // Rotate elements to maintain stack order
        for _ in 0..size {
            if let Some(val) = self.queue1.pop_front() {
                self.queue1.push_back(val);
            }
        }
    }

    /// Removes and returns the top element.
    /// Returns -1 if stack is empty.
    pub fn pop(&mut self) -> i32 {
        self.queue1.pop_front().unwrap_or(-1)
    }

    /// Returns the top element without removing it.
    /// Returns -1 if stack is empty.
    pub fn top(&self) -> i32 {
        *self.queue1.front().unwrap_or(&-1)
    }

    /// Returns true if the stack is empty.
    pub fn empty(&self) -> bool {
        self.queue1.is_empty()
    }
}
```
