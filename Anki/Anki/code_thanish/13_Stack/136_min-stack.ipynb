# **Optimized Rust Solution for Min Stack**

## **Approach**

We need to implement a stack that supports:

1. **push(x)** â†’ Inserts `x`. âœ… O(1)
2. **pop()** â†’ Removes the top element. âœ… O(1)
3. **top()** â†’ Returns the top element. âœ… O(1)
4. **getMin()** â†’ Returns the **minimum** element in O(1). âœ… O(1)

### **Efficient Solution (Using an Auxiliary Stack)**

- We maintain **two stacks**:
  1. **stack** â†’ Stores all elements.
  2. **min_stack** â†’ Stores the minimum value at that point.

### **How it works?**

| Operation | stack        | min_stack |
| --------- | ------------ | --------- |
| push(3)   | [3]          | [3]       |
| push(5)   | [3, 5]       | [3]       |
| push(2)   | [3, 5, 2]    | [3, 2]    |
| push(1)   | [3, 5, 2, 1] | [3, 2, 1] |
| getMin()  | -            | 1         |
| pop()     | [3, 5, 2]    | [3, 2]    |
| getMin()  | -            | 2         |

- **Push:** If `x` is the smallest seen so far, push it onto `min_stack`.
- **Pop:** If `x` was the smallest (top of `min_stack`), remove it.
- **getMin():** Simply return `min_stack.top()`.

## **Rust Implementation**

```rust
use std::vec::Vec;

/// Stack that supports retrieving the minimum element in O(1) time.
pub struct MinStack {
    stack: Vec<i32>,
    min_stack: Vec<i32>,
}

impl MinStack {
    /// Initializes an empty MinStack.
    pub fn new() -> Self {
        Self {
            stack: Vec::new(),
            min_stack: Vec::new(),
        }
    }

    /// Pushes a new element onto the stack.
    /// If it is the new minimum, it is also pushed to `min_stack`.
    pub fn push(&mut self, val: i32) {
        self.stack.push(val);
        if self.min_stack.is_empty() || val <= *self.min_stack.last().unwrap() {
            self.min_stack.push(val);
        }
    }

    /// Removes the top element from the stack.
    /// If the popped value is the current minimum, it is also removed from `min_stack`.
    pub fn pop(&mut self) {
        if let Some(top) = self.stack.pop() {
            if top == *self.min_stack.last().unwrap() {
                self.min_stack.pop();
            }
        }
    }

    /// Returns the top element of the stack.
    /// Returns -1 if stack is empty.
    pub fn top(&self) -> i32 {
        *self.stack.last().unwrap_or(&-1)
    }

    /// Retrieves the minimum element from the stack.
    /// Returns -1 if stack is empty.
    pub fn get_min(&self) -> i32 {
        *self.min_stack.last().unwrap_or(&-1)
    }
}
```
