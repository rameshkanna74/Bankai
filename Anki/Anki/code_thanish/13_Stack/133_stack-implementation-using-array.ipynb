# **Optimized Rust Solution for Stack Implementation Using Array**

## **Approach**

We implement a stack **using a fixed-size array (Vec in Rust)** with the following operations:

1. **Push(num):** Adds an element to the stack if space is available.
2. **Pop():** Removes and prints the top element; prints `-1` if empty.
3. **Top():** Prints the top element; prints `-1` if empty.
4. **isEmpty():** Prints `1` if the stack is empty, otherwise `0`.
5. **isFull():** Prints `1` if the stack is full, otherwise `0`.

---

## **Time & Space Complexity**

âœ… **Time Complexity:**

- **Each operation (Push, Pop, Top, isEmpty, isFull) runs in `O(1)` constant time.**
- Using an array-based approach ensures **fast indexing and no additional memory overhead** (unlike linked lists).

âœ… **Space Complexity:**

- **O(N) space for storing elements**, where `N` is the stackâ€™s capacity.
- **O(1) extra space** since we use a fixed-sized array without additional structures.

## **Rust Implementation**

```rust
pub struct Stack {
    data: Vec<i32>,  // Internal storage
    capacity: usize,  // Fixed capacity
}

impl Stack {
    /// Creates a new stack with a fixed capacity.
    pub fn new(capacity: usize) -> Self {
        Self {
            data: Vec::with_capacity(capacity),
            capacity,
        }
    }

    /// Pushes an element onto the stack if it's not full.
    pub fn push(&mut self, num: i32) {
        if self.data.len() < self.capacity {
            self.data.push(num);
        }
    }

    /// Pops and prints the top element. Prints `-1` if empty.
    pub fn pop(&mut self) {
        if let Some(val) = self.data.pop() {
            println!("{}", val);
        } else {
            println!("-1");
        }
    }

    /// Prints the top element of the stack. Prints `-1` if empty.
    pub fn top(&self) {
        if let Some(&val) = self.data.last() {
            println!("{}", val);
        } else {
            println!("-1");
        }
    }

    /// Prints `1` if the stack is empty, otherwise prints `0`.
    pub fn is_empty(&self) {
        println!("{}", if self.data.is_empty() { 1 } else { 0 });
    }

    /// Prints `1` if the stack is full, otherwise prints `0`.
    pub fn is_full(&self) {
        println!("{}", if self.data.len() == self.capacity { 1 } else { 0 });
    }
}
```
