# **Optimized Rust Solution for Reversing a Stack Without Extra Space**

## **Approach**

To **reverse a stack without extra space**, we must use **recursion**.  
Since function call stack is used for recursion, it acts as an **implicit stack**, thus avoiding an explicit auxiliary data structure.

### **Steps to Reverse a Stack**

1. **Recursively pop all elements** from the stack until it becomes empty.
2. **Insert each popped element at the bottom** of the stack while returning from the recursion.

### **Key Recursive Functions**

1. **reverse_stack(stack: &mut Vec<i32>)**
   - Base Case: If stack is empty, return.
   - Recursively remove the top element.
   - Call itself to reverse the remaining stack.
   - Insert the removed element at the bottom using `insert_at_bottom()`.
2. **insert_at_bottom(stack: &mut Vec<i32>, element: i32)**
   - Base Case: If stack is empty, push `element` at the bottom.
   - Otherwise, remove the top element, insert `element` recursively, then push the removed element back.

### **Example Walkthrough**

#### **Initial Stack**

```
[1, 2, 3, 4]
```

#### **Recursive Steps**

1. `4` is removed â†’ Call `reverse_stack([1, 2, 3])`
2. `3` is removed â†’ Call `reverse_stack([1, 2])`
3. `2` is removed â†’ Call `reverse_stack([1])`
4. `1` is removed â†’ Stack is now **empty**
5. `1` is inserted at the bottom
6. `2` is inserted at the bottom â†’ `[2, 1]`
7. `3` is inserted at the bottom â†’ `[3, 2, 1]`
8. `4` is inserted at the bottom â†’ `[4, 3, 2, 1]`

### **Final Reversed Stack**

```
[4, 3, 2, 1]
```

## **Rust Implementation**

```rust
struct Stack {
    stack: Vec<i32>,
}

impl Stack {
    fn new() -> Self {
        Stack { stack: Vec::new() }
    }

    fn push(&mut self, val: i32) {
        self.stack.push(val);
    }

    fn pop(&mut self) -> Option<i32> {
        self.stack.pop()
    }

    fn is_empty(&self) -> bool {
        self.stack.is_empty()
    }

    fn reverse(&mut self) {
        if !self.is_empty() {
            let top = self.pop().unwrap();
            self.reverse();
            self.insert_at_bottom(top);
        }
    }

    fn insert_at_bottom(&mut self, val: i32) {
        if self.is_empty() {
            self.push(val);
        } else {
            let top = self.pop().unwrap();
            self.insert_at_bottom(val);
            self.push(top);
        }
    }

    fn get_stack(&self) -> Vec<i32> {
        self.stack.clone()
    }
}
```
