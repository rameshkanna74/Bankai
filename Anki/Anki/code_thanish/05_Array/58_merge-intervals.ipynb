```rust
impl Solution {
    pub fn merge(mut intervals: Vec<Vec<i32>>) -> Vec<Vec<i32>> {
        if intervals.is_empty() {
            return vec![];
        }

        // Step 1: Sort intervals by start time
        intervals.sort_by(|a, b| a[0].cmp(&b[0]));

        // Step 2: Merge intervals
        let mut merged = Vec::new();
        merged.push(intervals[0].clone());

        for interval in intervals.into_iter().skip(1) {
            let last = merged.last_mut().unwrap(); // Safe because merged is never empty

            if interval[0] <= last[1] {
                // Merge overlapping intervals
                last[1] = last[1].max(interval[1]);
            } else {
                // No overlap, push new interval
                merged.push(interval);
            }
        }

        merged
    }
}
```

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Sorting + Greedy Merge (`O(N log N)`)**

1. **Sort intervals** by `start` time.

   - Ensures we process intervals **in order**.
   - Sorting takes **`O(N log N)`** time.

2. **Iterate through sorted intervals**, merging overlapping ones.
   - If **overlap**, merge: `last[1] = max(last[1], current[1])`.
   - If **no overlap**, add the new interval to the result.
