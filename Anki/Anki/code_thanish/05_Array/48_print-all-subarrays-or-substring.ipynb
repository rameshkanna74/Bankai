impl Solution {
    pub fn generate_subarrays(arr: Vec<i32>) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let n = arr.len();

        for start in 0..n {
            for end in start..n {
                result.push(arr[start..=end].to_vec());
            }
        }

        result
    }
}
