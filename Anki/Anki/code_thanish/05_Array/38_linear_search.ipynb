````python
# ğŸš€ **Optimized, Production-Quality Rust Solution for Linear Search**  

---

## **ğŸ“Œ Problem Breakdown**
### **Task**
- Find the **first occurrence** of an integer `X` in an array `arr`.
- Return **index of first occurrence** if found, else return `-1`.
- **Use Linear Search (Sequential Search).**

### **Constraints**
1. **Optimal Time Complexity** â†’ `O(N)` (best we can do for unsorted arrays).
2. **Optimal Space Complexity** â†’ `O(1)`, modifying the array in-place.
3. **Handling Edge Cases**:
   - `X` is present in the **first element** (`[5, 2, 3], X=5` â†’ `0`).
   - `X` is present in the **last element** (`[1, 2, 3], X=3` â†’ `2`).
   - `X` is **not present** (`[1, 2, 3], X=9` â†’ `-1`).
   - **Empty array** (`[], X=5` â†’ `-1`).
   - **Single-element array** (`[5], X=5` â†’ `0` or `[5], X=9` â†’ `-1`).
   - **Multiple occurrences** (`[1, 3, 3, 7], X=3` â†’ `1`).

---

## **ğŸ“Š Time & Space Complexity Analysis**
| **Metric**       | **Value** |
|-----------------|-----------|
| **Time Complexity**  | `O(N)`, since each element is checked at most once. |
| **Space Complexity** | `O(1)`, as no extra space is used. |

âœ… **This is the most efficient approach in terms of both time and space.**

---

## **ğŸ“Œ Optimized Rust Code**
```rust
impl Solution {
    pub fn linear_search(arr: &[i32], x: i32) -> i32 {
        for (i, &num) in arr.iter().enumerate() {
            if num == x {
                return i as i32;
            }
        }
        -1
    }
}

#[cfg(test)]
mod tests {
    use super::Solution;

    #[test]
    fn test_linear_search() {
        assert_eq!(Solution::linear_search(&[2, 13, 4, 1, 3, 6, 28], 3), 4);
        assert_eq!(Solution::linear_search(&[2, 13, 4, 1, 3, 6, 28], 9), -1);
        assert_eq!(Solution::linear_search(&[7, 8, 5, 9, 5], 5), 2);
        assert_eq!(Solution::linear_search(&[], 5), -1); // Empty array
        assert_eq!(Solution::linear_search(&[5], 5), 0); // Single-element found
        assert_eq!(Solution::linear_search(&[5], 9), -1); // Single-element not found
        assert_eq!(Solution::linear_search(&[1, 2, 3, 4, 5], 5), 4); // Last element match
        assert_eq!(Solution::linear_search(&[5, 2, 3, 5, 6], 5), 0); // First occurrence match
    }
}
```

---

## **ğŸ“Œ Algorithm Explanation**
### **ğŸ”¹ Brute Force Approach (`O(N)`)**
- **Key Idea**: Iterate over the array **sequentially**.
- **How it works**:
  1. Use `.iter().enumerate()` to iterate with both **index** and **value**.
  2. If `num == x`, return **index** immediately.
  3. If the loop completes, return `-1`.

### **Step-by-Step Example**
#### **Input:**  
`arr = [2, 13, 4, 1, 3, 6, 28], x = 3`

#### **Execution Trace:**
| **Iteration** | **Index (i)** | **Value (num)** | **Action** |
|--------------|--------------|----------------|------------|
| 1           | 0            | `2`            | Continue   |
| 2           | 1            | `13`           | Continue   |
| 3           | 2            | `4`            | Continue   |
| 4           | 3            | `1`            | Continue   |
| 5           | 4            | `3`            | **Return 4** âœ… |

#### **Final Output:**  
`4` âœ…

---

## **ğŸ“Œ Edge Cases Considered**
âœ… **Empty array** â†’ Return `-1`.  
âœ… **Single-element match** â†’ Return `0`.  
âœ… **Single-element non-match** â†’ Return `-1`.  
âœ… **Last element match** â†’ Return last index.  
âœ… **First element match** â†’ Return `0`.  
âœ… **Multiple occurrences** â†’ Return index of first occurrence.  

---

## **ğŸ“Œ Alternative Approaches**
### **1ï¸âƒ£ Iterator Approach (`O(N) Time, O(1) Space`)**
Using `.position()` for a **more idiomatic Rust solution**:
```rust
pub fn linear_search(arr: &[i32], x: i32) -> i32 {
    arr.iter().position(|&num| num == x).map(|idx| idx as i32).unwrap_or(-1)
}
```
âœ… **Pros:** More idiomatic and concise.  
âŒ **Cons:** Slightly less explicit than the loop version.  

---

### **2ï¸âƒ£ Parallel Linear Search for Large Arrays**
If the array size is very large (`N > 10â¶`), **multi-threading** can help process large chunks in parallel:
```rust
use rayon::prelude::*;

pub fn parallel_linear_search(arr: &[i32], x: i32) -> i32 {
    arr.par_iter()
        .enumerate()
        .find_any(|&(_, &num)| num == x)
        .map(|(i, _)| i as i32)
        .unwrap_or(-1)
}
```
âœ… **Pros:** Faster for large datasets using multi-threading.  
âŒ **Cons:** Overhead for smaller arrays.  

---

## **ğŸ“Œ Follow-Up Enhancements**
### **1ï¸âƒ£ Streaming Input Adaptation**
- If the input **arrives as a stream**, we can process it **lazily**:
```rust
pub fn stream_search<I>(iter: I, x: i32) -> i32
where
    I: IntoIterator<Item = i32>,
{
    iter.into_iter()
        .enumerate()
        .find(|&(_, num)| num == x)
        .map(|(i, _)| i as i32)
        .unwrap_or(-1)
}
```
âœ… **Handles data dynamically from streams!**  

---

### **2ï¸âƒ£ Real-World Use Cases**
- **Database Indexing:** Searching for records in unsorted datasets.
- **Log Analysis:** Finding specific error codes in system logs.
- **Packet Filtering:** Searching for specific headers in network packets.

---

## **âœ… Final Verdict**
| **Criteria** | **Linear Search (`O(N)`)** |
|-------------|--------------------------|
| **Time Complexity** | âœ… `O(N)` |
| **Space Complexity** | âœ… `O(1)` |
| **In-Place Modification** | âœ… Yes |
| **Handles Large Inputs** | âœ… Yes (`10âµ` elements) |
| **Scalability & Parallelization** | âœ… Rayon-supported |

âœ… **Best for:** **Interview-ready, optimal, first-occurrence linear search.** ğŸš€
````
