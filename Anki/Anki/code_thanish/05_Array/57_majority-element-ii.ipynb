```rust
impl Solution {
    pub fn majority_element(nums: Vec<i32>) -> Vec<i32> {
        let (mut candidate1, mut candidate2) = (None, None);
        let (mut count1, mut count2) = (0, 0);

        // Phase 1: Find potential candidates
        for &num in &nums {
            if Some(num) == candidate1 {
                count1 += 1;
            } else if Some(num) == candidate2 {
                count2 += 1;
            } else if count1 == 0 {
                candidate1 = Some(num);
                count1 = 1;
            } else if count2 == 0 {
                candidate2 = Some(num);
                count2 = 1;
            } else {
                count1 -= 1;
                count2 -= 1;
            }
        }

        // Phase 2: Verify candidates
        let (mut count1, mut count2) = (0, 0);
        for &num in &nums {
            if Some(num) == candidate1 {
                count1 += 1;
            } else if Some(num) == candidate2 {
                count2 += 1;
            }
        }

        let mut result = Vec::new();
        let threshold = nums.len() / 3;

        if count1 > threshold {
            result.push(candidate1.unwrap());
        }
        if count2 > threshold {
            result.push(candidate2.unwrap());
        }

        result
    }
}
```

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Boyer-Moore Voting Algorithm (`O(N)`)**

1. **Find up to 2 potential majority candidates** (since at most 2 numbers can appear `> n/3` times).

   - Use two variables (`candidate1`, `candidate2`) and counters (`count1`, `count2`).
   - If a number matches a candidate, increment its count.
   - If a counter is `0`, replace the candidate.
   - Otherwise, decrement both counters.

2. **Verify the candidates** in a second pass to check if they appear more than `n/3` times.

#### **Example Walkthrough**

üìå **Input:** `nums = [3,2,3]`  
üìå **Threshold:** `n / 3 = 3 / 3 = 1`

**Phase 1: Find Candidates**
| Num | `candidate1` | `count1` | `candidate2` | `count2` |
|-----|-------------|---------|-------------|---------|
| 3 | 3 | 1 | None | 0 |
| 2 | 3 | 1 | 2 | 1 |
| 3 | 3 | 2 | 2 | 1 |

**Phase 2: Count Occurrences**
| Candidate | Count |
|-----------|-------|
| 3 | 2 |
| 2 | 1 |

‚úÖ `3` appears `> 1` times ‚Üí **Final Answer: `[3]`**

## **üìå Edge Cases Considered**

‚úÖ **Single Element (`[1]`)** ‚Üí Always returns `[1]`.  
‚úÖ **All Unique (`[1,2,3,4,5]`)** ‚Üí No element appears `> n/3`, returns `[]`.  
‚úÖ **All Duplicates (`[1,1,1,1]`)** ‚Üí Returns `[1]`.  
‚úÖ **Two Majority Elements (`[1,1,2,2,3,3,1,2]`)** ‚Üí Returns `[1, 2]`.  
‚úÖ **No Majority (`[1,2,3,4,5,6]`)** ‚Üí Returns `[]`.
