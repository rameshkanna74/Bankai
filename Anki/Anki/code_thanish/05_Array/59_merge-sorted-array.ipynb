## **üìä Time & Space Complexity Analysis**

| **Approach**                                       | **Time Complexity** | **Space Complexity** | **Why?**                                            |
| -------------------------------------------------- | ------------------- | -------------------- | --------------------------------------------------- |
| **Sort after merging (`O((m+n) log(m+n))`)** ‚ùå    | `O((m+n) log(m+n))` | `O(1)`               | Sorting the entire array is **suboptimal**.         |
| **Two-pointer merge from the front (`O(m+n)`)** ‚ùå | `O(m+n)`            | `O(m+n)`             | Requires extra space for merging.                   |

- **Merging from the back avoids shifting elements** ‚Üí **Efficient in-place merge**.
- **Time Complexity:** `O(m + n)`

impl Solution {
    pub fn merge(nums1: &mut Vec<i32>, mut m: i32, nums2: &mut Vec<i32>, mut n: i32) {
        let mut i = (m + n - 1) as usize; // Pointer for the last position in nums1
        let mut p1 = (m - 1) as isize;    // Pointer for last element in nums1
        let mut p2 = (n - 1) as isize;    // Pointer for last element in nums2

        while p2 >= 0 {
            if p1 >= 0 && nums1[p1 as usize] > nums2[p2 as usize] {
                nums1[i] = nums1[p1 as usize];
                p1 -= 1;
            } else {
                nums1[i] = nums2[p2 as usize];
                p2 -= 1;
            }
            i -= 1;
        }
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Two-Pointer Merge from the Back (`O(m + n)`)**

1. **Start from the end of `nums1`** (at `m + n - 1`) since it has enough space.
2. **Use two pointers (`p1`, `p2`)** to track the last elements of `nums1` and `nums2`.
3. **Compare `nums1[p1]` and `nums2[p2]`**:
   - If `nums1[p1] > nums2[p2]`, move `nums1[p1]` to the end.
   - Else, move `nums2[p2]` to the end.
4. **Repeat until `nums2` is fully merged**.

### **Why Merge from the Back?**

- **Avoids unnecessary element shifts.**
- **Works in-place (`O(1)` extra space).**

## **üìå Edge Cases Considered**

‚úÖ **`nums2` is empty (`n = 0`)** ‚Üí No changes to `nums1`.  
‚úÖ **`nums1` is empty (`m = 0`)** ‚Üí Copy `nums2` into `nums1`.  
‚úÖ **All elements of `nums1` are greater than `nums2`** ‚Üí Append `nums2` at the start.  
‚úÖ **All elements of `nums2` are greater than `nums1`** ‚Üí `nums1` remains unchanged.  
‚úÖ **Interleaved elements (`[1,2,4,5,6]` + `[3]`)** ‚Üí Proper merging.  
‚úÖ **`nums1` already contains `nums2` elements (`[1,2,3,0,0,0]` + `[2,5,6]`)** ‚Üí Correct placement.
