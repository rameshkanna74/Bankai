Note that you must do this in-place without making a copy of the array.

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

Example 2:

Input: nums = [0]
Output: [0]

Constraints:

    1 <= nums.length <= 104
    -231 <= nums[i] <= 231 - 1

Follow up: Could you minimize the total number of operations done?

```rust
impl Solution {
    pub fn move_zeroes(nums: &mut Vec<i32>) {
        let mut last_non_zero = 0;

        // Move non-zero elements to the front
        for i in 0..nums.len() {
            if nums[i] != 0 {
                nums.swap(i, last_non_zero);
                last_non_zero += 1;
            }
        }
    }
}
```

## **üìå Algorithm Explanation**

### **üîπ Two-Pointer Approach (Efficient `O(N)`)**

- **Key Idea**: Maintain a **non-zero position pointer** (`last_non_zero`).
- **How it works**:
  1. Iterate over the array.
  2. Whenever a **non-zero element** is found, **swap it** with `last_non_zero` position.
  3. Increment `last_non_zero`.

### **Step-by-Step Example**

#### **Input:**

`[0, 1, 0, 3, 12]`

#### **Execution Trace:**

| **Iteration** | **Index (i)** | **Value (nums[i])** | **last_non_zero** | **Action**  | **Array State**    |
| ------------- | ------------- | ------------------- | ----------------- | ----------- | ------------------ |
| 1             | 0             | `0`                 | `0`               | No action   | `[0, 1, 0, 3, 12]` |
| 2             | 1             | `1`                 | `0`               | Swap (0,1)  | `[1, 0, 0, 3, 12]` |
| 3             | 2             | `0`                 | `1`               | No action   | `[1, 0, 0, 3, 12]` |
| 4             | 3             | `3`                 | `1`               | Swap (0,3)  | `[1, 3, 0, 0, 12]` |
| 5             | 4             | `12`                | `2`               | Swap (0,12) | `[1, 3, 12, 0, 0]` |

#### **Final Output:**

## **üìå Alternative Approaches**

### **1Ô∏è‚É£ Extra Array Approach (`O(N) Time, O(N) Space`)**

- Create a new array, copy non-zero elements first, then append zeros.

```rust
fn move_zeroes(nums: &mut Vec<i32>) {
    let mut temp: Vec<i32> = nums.iter().filter(|&&x| x != 0).copied().collect();
    temp.resize(nums.len(), 0);
    *nums = temp;
}
```

‚úÖ **Pros:** Simpler logic  
‚ùå **Cons:** Uses `O(N)` extra space

---

### **2Ô∏è‚É£ Two-Pass Approach (`O(N) Time, O(1) Space`)**

- First pass: Shift non-zero elements forward.
- Second pass: Fill remaining elements with zeros.

```rust
fn move_zeroes(nums: &mut Vec<i32>) {
    let mut index = 0;

    for &num in nums.iter() {
        if num != 0 {
            nums[index] = num;
            index += 1;
        }
    }

    for i in index..nums.len() {
        nums[i] = 0;
    }
}
```

‚úÖ **Pros:** No unnecessary swaps  
‚ùå **Cons:** Slightly less optimal due to two-pass approach
