# 🚀 **Optimized, Production-Quality Rust Solution for "Max Consecutive Ones" (Leetcode 485)**

## **📌 Problem Breakdown**

### **Task**

- Given a **binary array** `nums` (only containing `0`s and `1`s), **find the longest consecutive sequence of `1`s**.

### **Constraints**

- `1 ≤ nums.length ≤ 10⁵`
- `nums[i]` is either `0` or `1`.

### **Key Observations**

- We need **O(N) time complexity** (single pass) and **O(1) space complexity**.
- **Sorting or extra data structures are unnecessary**.

---

## **📊 Time & Space Complexity Analysis**

| **Approach**                      | **Time Complexity** | **Space Complexity** | **Reasoning**                 |
| --------------------------------- | ------------------- | -------------------- | ----------------------------- |
| **Sorting (`O(N log N)`)**        | ❌ `O(N log N)`     | ✅ `O(1)`            | Sorting is unnecessary        |

| **Greedy / Single Pass (`O(N)`)** | ✅ `O(N)`           | ✅ `O(1)`            | **Best solution**             |

✅ **Best Choice**:

- **Greedy / Single Pass Approach** (`O(N)` time, `O(1)` space).

## **📌 Optimized Rust Code**

### **1️⃣ Greedy Approach**

```rust
impl Solution {
    pub fn find_max_consecutive_ones(nums: Vec<i32>) -> i32 {
        let mut max_count = 0;
        let mut current_count = 0;

        for num in nums {
            if num == 1 {
                current_count += 1;
                max_count = max_count.max(current_count);
            } else {
                current_count = 0;
            }
        }

        max_count
    }
}
```

## **📌 Algorithm Explanation**

### **Greedy Single Pass (`O(N)`)**

- **Use a single loop** to traverse `nums[]`.
- Track:
  - `current_count`: Consecutive `1`s so far.
  - `max_count`: Maximum consecutive `1`s seen.
- If `num == 1`, **increment `current_count`**.
- If `num == 0`, **reset `current_count` to `0`**.
- **Update `max_count` whenever `current_count` increases**.

#### **Example**

📌 **Input:** `nums = [1,1,0,1,1,1]`

| **Index** | **num** | **current_count** | **max_count** |
| --------- | ------- | ----------------- | ------------- |
| `0`       | `1`     | `1`               | `1`           |
| `1`       | `1`     | `2`               | `2`           |
| `2`       | `0`     | `0` (reset)       | `2`           |
| `3`       | `1`     | `1`               | `2`           |
| `4`       | `1`     | `2`               | `2`           |
| `5`       | `1`     | `3`               | `3` ✅        |

✅ **Final Answer:** `3`.
