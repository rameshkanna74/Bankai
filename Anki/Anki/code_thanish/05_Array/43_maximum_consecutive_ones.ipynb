impl Solution {
    pub fn find_max_consecutive_ones(nums: Vec<i32>) -> i32 {
        let mut max_count = 0;
        let mut current_count = 0;

        for num in nums {
            if num == 1 {
                current_count += 1;
                max_count = max_count.max(current_count);
            } else {
                current_count = 0;
            }
        }

        max_count
    }
}

## **ğŸ“Œ Algorithm Explanation**

### **Greedy Single Pass (`O(N)`)**

- **Use a single loop** to traverse `nums[]`.
- Track:
  - `current_count`: Consecutive `1`s so far.
  - `max_count`: Maximum consecutive `1`s seen.
- If `num == 1`, **increment `current_count`**.
- If `num == 0`, **reset `current_count` to `0`**.
- **Update `max_count` whenever `current_count` increases**.
