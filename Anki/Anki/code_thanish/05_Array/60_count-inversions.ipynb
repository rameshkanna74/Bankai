## **üìå Problem Breakdown**

### **Task**

Given an array `arr` of size `n`, count the **number of inversions**, where a pair `(i, j)` is an **inversion** if:

- `arr[i] > arr[j]`
- `i < j`

### **Examples**

| **Input**         | **Output** | **Explanation**                                   |
| ----------------- | ---------- | ------------------------------------------------- |
| `[4, 3, 2, 1]`    | `6`        | Pairs: `(4,3), (4,2), (4,1), (3,2), (3,1), (2,1)` |
| `[1, 2, 3, 4, 5]` | `0`        | Already sorted, no inversions.                    |
| `[10, 10, 10]`    | `0`        | All elements are equal, no inversions.            |
| `[5, 4, 3, 2, 1]` | `10`       | Maximum inversions (reverse sorted).              |

- **Time Complexity:** `O(n log n)` (efficient divide-and-conquer).

- **Avoids brute-force nested loops** ‚Üí **Fast for large `n`**.

impl Solution {
    pub fn inversion_count(arr: Vec<i32>) -> i64 {
        let mut arr = arr.clone(); // Clone to avoid modifying input
        let mut temp = vec![0; arr.len()]; // Temporary array for merging
        Self::merge_sort_and_count(&mut arr, &mut temp, 0, arr.len() - 1)
    }

    fn merge_sort_and_count(arr: &mut Vec<i32>, temp: &mut Vec<i32>, left: usize, right: usize) -> i64 {
        if left >= right {
            return 0;
        }
        let mid = left + (right - left) / 2;
        let mut count = 0;

        count += Self::merge_sort_and_count(arr, temp, left, mid);
        count += Self::merge_sort_and_count(arr, temp, mid + 1, right);
        count += Self::merge_and_count(arr, temp, left, mid, right);

        count
    }

    fn merge_and_count(arr: &mut Vec<i32>, temp: &mut Vec<i32>, left: usize, mid: usize, right: usize) -> i64 {
        let mut i = left; // Left subarray pointer
        let mut j = mid + 1; // Right subarray pointer
        let mut k = left; // Sorted merge index
        let mut count = 0;

        while i <= mid && j <= right {
            if arr[i] <= arr[j] {
                temp[k] = arr[i];
                i += 1;
            } else {
                temp[k] = arr[j];
                count += (mid - i + 1) as i64; // Count inversions
                j += 1;
            }
            k += 1;
        }

        while i <= mid {
            temp[k] = arr[i];
            i += 1;
            k += 1;
        }

        while j <= right {
            temp[k] = arr[j];
            j += 1;
            k += 1;
        }

        arr[left..=right].copy_from_slice(&temp[left..=right]); // Copy sorted elements back
        count
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Modified Merge Sort (`O(n log n)`)**

1. **Divide the array** recursively into two halves until single elements remain.
2. **Merge subarrays** while counting inversions.
3. **Count inversions efficiently** during merging:
   - If `arr[i] <= arr[j]`, no inversion.
   - If `arr[i] > arr[j]`, all elements from `arr[i]` onward are greater ‚Üí **Inversions = (mid - i + 1)**.

### **2Ô∏è‚É£ Why Use Merge Sort?**

- **Merging preserves order efficiently**.
- **Avoids brute-force pair comparisons (`O(n¬≤)`)**.
- **Balanced recursion keeps it `O(n log n)`**.

### **3Ô∏è‚É£ Why `i64` for the Inversion Count?**

- The inversion count **can be large**, especially for **large `n`**.
- Using `i64` prevents **integer overflow**.

‚úÖ **Single element (`[1]`)** ‚Üí No inversions (`0`).  
‚úÖ **Already sorted (`[1,2,3,4,5]`)** ‚Üí No inversions (`0`).
