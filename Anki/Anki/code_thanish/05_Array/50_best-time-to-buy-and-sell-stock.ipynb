## **üìå Problem Breakdown**

### **Task**

- Given an array `prices`, where `prices[i]` is the price of a stock on day `i`:
  - **Buy on one day** and **sell on another later day**.
  - **Maximize profit** (`profit = sell_price - buy_price`).
  - **If no profit can be made, return `0`**.

### **Examples**

| **Input**       | **Output** | **Explanation**                   |
| --------------- | ---------- | --------------------------------- |
| `[7,1,5,3,6,4]` | `5`        | Buy at `1`, sell at `6` (`6-1=5`) |
| `[7,6,4,3,1]`   | `0`        | Prices only decrease; no profit.  |

## **üìå Optimized Rust Code: One-Pass (`O(N)`)**

```rust
impl Solution {
    pub fn max_profit(prices: Vec<i32>) -> i32 {
        let mut min_price = i32::MAX; // Store lowest price seen so far
        let mut max_profit = 0; // Track maximum profit

        for price in prices {
            min_price = min_price.min(price); // Update minimum price
            max_profit = max_profit.max(price - min_price); // Update max profit
        }

        max_profit
    }
}
```

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ One-Pass Greedy Approach (`O(N)`)**

- **Tracks the minimum price seen so far (`min_price`)**.
- **Calculates the potential profit (`price - min_price`)** at each step.
- **Updates `max_profit` whenever a higher profit is found**.

#### **Example Walkthrough**

üìå **Input:** `prices = [7,1,5,3,6,4]`

| Day | Price | `min_price` (Lowest so far) | Profit (`price - min_price`) | `max_profit` |
| --- | ----- | --------------------------- | ---------------------------- | ------------ |
| 1   | `7`   | `7`                         | `7-7=0`                      | `0`          |
| 2   | `1`   | `1`                         | `1-1=0`                      | `0`          |
| 3   | `5`   | `1`                         | `5-1=4`                      | `4`          |
| 4   | `3`   | `1`                         | `3-1=2`                      | `4`          |
| 5   | `6`   | `1`                         | `6-1=5`                      | `5`          |
| 6   | `4`   | `1`                         | `4-1=3`                      | `5`          |

‚úÖ **Final Answer:** `5` (Buy at `1`, sell at `6`).
