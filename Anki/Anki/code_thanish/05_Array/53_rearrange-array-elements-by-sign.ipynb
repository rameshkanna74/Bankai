You are given a 0-indexed integer array nums of even length consisting of an equal number of positive and negative integers.

You should return the array of nums such that the the array follows the given conditions:

    Every consecutive pair of integers have opposite signs.
    For all integers with the same sign, the order in which they were present in nums is preserved.
    The rearranged array begins with a positive integer.

Return the modified array after rearranging the elements to satisfy the aforementioned conditions.

Example 1:

Input: nums = [3,1,-2,-5,2,-4]
Output: [3,-2,1,-5,2,-4]
Explanation:
The positive integers in nums are [3,1,2]. The negative integers are [-2,-5,-4].
The only possible way to rearrange them such that they satisfy all conditions is [3,-2,1,-5,2,-4].
Other ways such as [1,-2,2,-5,3,-4], [3,1,2,-2,-5,-4], [-2,3,-5,1,-4,2] are incorrect because they do not satisfy one or more conditions.

Example 2:

Input: nums = [-1,1]
Output: [1,-1]
Explanation:
1 is the only positive integer and -1 the only negative integer in nums.
So nums is rearranged to [1,-1].

Constraints:

    2 <= nums.length <= 2 * 105
    nums.length is even
    1 <= |nums[i]| <= 105
    nums consists of equal number of positive and negative integers.

It is not required to do the modifications in-place.

# 🚀 **Optimized, Production-Quality Rust Solution for Rearranging Array by Sign**

---

## **📌 Problem Breakdown**

### **Task**

- Given an **even-length** array of integers `nums`, where there are **equal numbers of positive and negative numbers**, **rearrange the array** such that:
  1. **Every consecutive pair** of numbers has opposite signs.
  2. **Order is preserved** for both positive and negative numbers.
  3. **First element is positive**.

### **Examples**

| **Input**          | **Output**         | **Explanation**                                                                                   |
| ------------------ | ------------------ | ------------------------------------------------------------------------------------------------- |
| `[3,1,-2,-5,2,-4]` | `[3,-2,1,-5,2,-4]` | Positive numbers `[3,1,2]`, negative numbers `[-2,-5,-4]` are interleaved while preserving order. |
| `[-1,1]`           | `[1,-1]`           | `1` is positive, `-1` is negative, so `[1,-1]` is the only valid output.                          |

### **Constraints**

- `2 ≤ nums.len() ≤ 2 × 10⁵`
- `nums.len()` is **even**.
- `1 ≤ |nums[i]| ≤ 10⁵`
- `nums` contains **equal positive and negative numbers**.

---

## **📊 Time & Space Complexity Analysis**

| **Approach**                 | **Time Complexity** | **Space Complexity** | **Why?**                                              |
| ---------------------------- | ------------------- | -------------------- | ----------------------------------------------------- |
| **Brute Force (`O(N²)`)** ❌ | `O(N²)`             | `O(N)`               | Slow due to shifting elements during rearrangement.   |
| **Two-Pointer (`O(N)`)** ✅  | ✅ `O(N)`           | ✅ `O(N)`            | One pass collects positives and negatives separately. |

✅ **Best Choice: Two-Pointer (`O(N)`)**

- **Single pass (`O(N)`) to collect positive & negative numbers.**
- **Second pass (`O(N)`) to merge alternately.**
- **Uses `O(N)` extra space for output but avoids inefficient shifting.**

## **📌 Optimized Rust Code: Two-Pointer (`O(N)`)**

```rust
impl Solution {
    pub fn rearrange_array(nums: Vec<i32>) -> Vec<i32> {
        let mut pos = Vec::with_capacity(nums.len() / 2);
        let mut neg = Vec::with_capacity(nums.len() / 2);

        // Collect positive and negative numbers separately
        for &num in &nums {
            if num > 0 {
                pos.push(num);
            } else {
                neg.push(num);
            }
        }

        // Interleave positive and negative numbers
        let mut result = Vec::with_capacity(nums.len());
        for i in 0..pos.len() {
            result.push(pos[i]);
            result.push(neg[i]);
        }

        result
    }
}
```

## **📌 Algorithm Explanation**

### **1️⃣ Two-Pointer Collection (`O(N)`)**

- **Pass 1:** Store positives in `pos` and negatives in `neg` (`O(N)`).
- **Pass 2:** Interleave from both lists into `result` (`O(N)`).

#### **Example Walkthrough**

📌 **Input:** `[3,1,-2,-5,2,-4]`

**Step 1: Split into positives and negatives**

```plaintext
Positives: [3, 1, 2]
Negatives: [-2, -5, -4]
```

**Step 2: Interleave alternately**

```plaintext
Result: [3, -2, 1, -5, 2, -4]
```

✅ **Final Answer:** `[3, -2, 1, -5, 2, -4]`.

## **📌 Follow-Up Enhancements**

### **1️⃣ Handling Streaming Input**

- Instead of collecting separately, **use two indices** (`pos_idx` and `neg_idx`) to process the array **in-place**.

### **2️⃣ In-Place Modification (`O(1)` Space)**

- **Use a two-pointer approach** with swapping instead of auxiliary arrays.

### **3️⃣ Parallelizing the Algorithm**

- **Split `N` elements across multiple threads**.
- **Each thread processes** half of `nums` using separate buffers.
- Merge results **in `O(P log P)` time**.

### **4️⃣ Distributed Systems Approach**

- Works well with **MapReduce**:
  - **Mapper:** Splits numbers into positives & negatives.
  - **Reducer:** Merges them into pairs.
