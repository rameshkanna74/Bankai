impl Solution {
    /// Removes duplicates from a sorted array in-place.
    /// Returns the count of unique elements.
    pub fn remove_duplicates(nums: &mut Vec<i32>) -> i32 {
        if nums.is_empty() {
            return 0;
        }

        let mut k = 1; // First unique element is always at index 0

        for i in 1..nums.len() {
            if nums[i] != nums[k - 1] {
                nums[k] = nums[i]; // Move unique element forward
                k += 1;
            }
        }

        k as i32
    }
}

## **üìå Algorithm Explanation**

### **Concept Used: Two-Pointer Technique**

- **Pointer `k` (slow pointer):** Tracks the next position for a **unique** element.
- **Pointer `i` (fast pointer):** Scans the array to find unique elements.

### **Step-by-Step Approach**

1. **Edge Case Handling:** If `nums` is empty, return `0`.
2. **Initialize `k = 1`** (the first element is always unique).
3. **Iterate `i` from `1` to `nums.len()`**:
   - If `nums[i]` **is different** from `nums[k-1]`, it means we found a new unique element.
   - Place `nums[i]` at index `k` and **increment `k`**.
4. **Return `k`** (number of unique elements).
