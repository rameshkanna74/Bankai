# 🚀 **Optimized, Production-Quality Rust Solution for Grouping Anagrams**

---

## **📌 Problem Breakdown**

### **Task**

- Given an array of strings `strs`, **group** all **anagrams** together.
- An **anagram** consists of words that contain the **same characters** with **different orders**.

### **Examples**

| **Input**                               | **Output**                                    | **Explanation**                                                |
| --------------------------------------- | --------------------------------------------- | -------------------------------------------------------------- |
| `["eat","tea","tan","ate","nat","bat"]` | `[["bat"],["nat","tan"],["ate","eat","tea"]]` | `"eat", "tea", "ate"` are anagrams, and so are `"tan", "nat"`. |
| `[""]`                                  | `[[""]]`                                      | Empty string is a valid anagram of itself.                     |
| `["a"]`                                 | `[["a"]]`                                     | Single-character words are anagrams of themselves.             |

### **Constraints**

- `1 ≤ strs.len() ≤ 10⁴`
- `0 ≤ strs[i].len() ≤ 100`
- `strs[i]` consists of lowercase English letters.

---

## **📊 Time & Space Complexity Analysis**

| **Approach**                                 | **Time Complexity** | **Space Complexity** | **Why?**                                                                    |
| -------------------------------------------- | ------------------- | -------------------- | --------------------------------------------------------------------------- |
| **Brute Force (`O(N²M)`)** ❌                | `O(N²M)`            | `O(NM)`              | Compares each string to all others, inefficient.                            |
| **Sorting-Based Hashing (`O(NM log M)`)** ✅ | ✅ `O(NM log M)`    | ✅ `O(NM)`           | Sorting each word (`O(M log M)`) and using a hashmap for grouping (`O(N)`). |
| **Frequency Count (`O(NM)`)** 🔥✅           | 🔥✅ `O(NM)`        | ✅ `O(NM)`           | Uses character frequency as a key, avoids sorting. **Fastest approach.**    |

✅ **Best Choice: Frequency Count (`O(NM)`)**

- **Avoids sorting overhead (`O(M log M)`).**
- **Ideal for large constraints (`N = 10⁴, M = 100`).**
- **Uses `HashMap` for `O(1)` lookups.**

---

## **📌 Optimized Rust Code: Frequency Count (`O(NM)`)**

```rust
use std::collections::HashMap;

impl Solution {
    pub fn group_anagrams(strs: Vec<String>) -> Vec<Vec<String>> {
        let mut anagram_map: HashMap<[u8; 26], Vec<String>> = HashMap::new();

        for s in strs {
            let mut char_count = [0u8; 26]; // Frequency count for 26 lowercase letters
            for c in s.bytes() {
                char_count[(c - b'a') as usize] += 1;
            }
            anagram_map.entry(char_count).or_default().push(s);
        }

        anagram_map.into_values().collect()
    }
}
```

## **📌 Algorithm Explanation**

### **1️⃣ Frequency Count (`O(NM)`)**

- Each string is **converted into a frequency array** (`O(M)`).
- Uses a **fixed-size array `[u8; 26]`** for efficiency (instead of sorting).
- **Groups words in a HashMap** where:
  - **Key:** Character frequency signature.
  - **Value:** List of words sharing that signature.
- Finally, the grouped words are **collected into a result vector**.

#### **Example Walkthrough**

📌 **Input:** `["eat", "tea", "tan", "ate", "nat", "bat"]`

| Word    | Frequency Signature       | Mapped Group            |
| ------- | ------------------------- | ----------------------- |
| `"eat"` | `[1, 0, 0, ..., 1, 1, 0]` | `["eat"]`               |
| `"tea"` | `[1, 0, 0, ..., 1, 1, 0]` | `["eat", "tea"]`        |
| `"tan"` | `[1, 1, 0, ..., 1, 0, 0]` | `["tan"]`               |
| `"ate"` | `[1, 0, 0, ..., 1, 1, 0]` | `["eat", "tea", "ate"]` |
| `"nat"` | `[1, 1, 0, ..., 1, 0, 0]` | `["tan", "nat"]`        |
| `"bat"` | `[1, 0, 0, ..., 0, 1, 0]` | `["bat"]`               |

✅ **Final Answer:** `[["bat"], ["nat", "tan"], ["ate", "eat", "tea"]]`.
