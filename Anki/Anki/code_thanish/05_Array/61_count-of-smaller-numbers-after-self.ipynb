impl Solution {
    pub fn count_smaller(nums: Vec<i32>) -> Vec<i32> {
        let n = nums.len();
        let mut counts = vec![0; n]; // Stores result
        let mut indices: Vec<usize> = (0..n).collect(); // Track original indices

        Self::merge_sort_and_count(&nums, &mut indices, &mut counts, 0, n - 1);
        counts
    }

    fn merge_sort_and_count(
        nums: &Vec<i32>, indices: &mut Vec<usize>, counts: &mut Vec<i32>, left: usize, right: usize
    ) {
        if left >= right {
            return;
        }

        let mid = left + (right - left) / 2;
        Self::merge_sort_and_count(nums, indices, counts, left, mid);
        Self::merge_sort_and_count(nums, indices, counts, mid + 1, right);
        Self::merge_and_count(nums, indices, counts, left, mid, right);
    }

    fn merge_and_count(
        nums: &Vec<i32>, indices: &mut Vec<usize>, counts: &mut Vec<i32>, left: usize, mid: usize, right: usize
    ) {
        let mut temp = Vec::with_capacity(right - left + 1);
        let mut i = left; // Left partition
        let mut j = mid + 1; // Right partition
        let mut right_count = 0; // Count of smaller numbers seen in right partition

        while i <= mid && j <= right {
            if nums[indices[i]] <= nums[indices[j]] {
                counts[indices[i]] += right_count;
                temp.push(indices[i]);
                i += 1;
            } else {
                right_count += 1;
                temp.push(indices[j]);
                j += 1;
            }
        }

        while i <= mid {
            counts[indices[i]] += right_count;
            temp.push(indices[i]);
            i += 1;
        }

        while j <= right {
            temp.push(indices[j]);
            j += 1;
        }

        indices.splice(left..=right, temp);
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Merge Sort + Indexed Counting (`O(n log n)`)**

1. **Sort while maintaining original indices**.
2. **Count how many smaller elements appear in the right partition**:
   - If `nums[i] > nums[j]` (right), increment count.
   - Otherwise, merge normally.
3. **Efficiently counts smaller elements without nested loops**.

### **2Ô∏è‚É£ Why Use Merge Sort?**

- **Efficient merging** ‚Üí avoids `O(n¬≤)` brute-force.
- **Sorted merging keeps counts accurate**.
