
# 🚀 **Optimized, Production-Quality Rust Solution for Merging Two Sorted Arrays**

---

## **📌 Problem Breakdown**

### **Task**

- Given **two sorted arrays** `a` and `b`, return their **union** (merged sorted array with unique elements).
- The output **must remain sorted** and **contain no duplicates**.
- **Time Complexity:** `O(N + M)` (since both arrays are already sorted).
- **Space Complexity:** `O(N + M)` (to store the result).

### **Constraints**

1. **Optimal Time Complexity** → `O(N + M)` using a **two-pointer approach**.
2. **Optimal Space Complexity** → `O(N + M)`, as output storage is necessary.
3. **Handling Edge Cases**:
   - **Disjoint Arrays** (`a = [1, 2, 3]`, `b = [4, 5, 6]` → `[1, 2, 3, 4, 5, 6]`).
   - **Identical Arrays** (`a = [1, 2, 3]`, `b = [1, 2, 3]` → `[1, 2, 3]`).
   - **All elements duplicated** (`a = [1, 1, 1]`, `b = [1, 1]` → `[1]`).
   - **One array empty** (`a = []`, `b = [1, 2, 3]` → `[1, 2, 3]`).
   - **Large arrays (10⁵ elements each)**.

## **📌 Optimized Rust Code**

```rust
impl Solution {
    pub fn merge_sorted_arrays(a: &[i32], b: &[i32]) -> Vec<i32> {
        let mut result = Vec::with_capacity(a.len() + b.len()); // Preallocate space
        let (mut i, mut j) = (0, 0);

        while i < a.len() && j < b.len() {
            match a[i].cmp(&b[j]) {
                std::cmp::Ordering::Less => {
                    if result.last() != Some(&a[i]) {
                        result.push(a[i]);
                    }
                    i += 1;
                }
                std::cmp::Ordering::Greater => {
                    if result.last() != Some(&b[j]) {
                        result.push(b[j]);
                    }
                    j += 1;
                }
                std::cmp::Ordering::Equal => {
                    if result.last() != Some(&a[i]) {
                        result.push(a[i]);
                    }
                    i += 1;
                    j += 1;
                }
            }
        }

        // Append remaining elements from a
        while i < a.len() {
            if result.last() != Some(&a[i]) {
                result.push(a[i]);
            }
            i += 1;
        }

        // Append remaining elements from b
        while j < b.len() {
            if result.last() != Some(&b[j]) {
                result.push(b[j]);
            }
            j += 1;
        }

        result
    }
}

#[cfg(test)]
mod tests {
    use super::Solution;

    #[test]
    fn test_merge_sorted_arrays() {
        assert_eq!(Solution::merge_sorted_arrays(&[1, 2, 3, 4, 6], &[2, 3, 5]), [1, 2, 3, 4, 5, 6]);
        assert_eq!(Solution::merge_sorted_arrays(&[1, 2, 3, 3], &[2, 2, 4]), [1, 2, 3, 4]);
        assert_eq!(Solution::merge_sorted_arrays(&[1, 2, 3], &[4, 5, 6]), [1, 2, 3, 4, 5, 6]);
        assert_eq!(Solution::merge_sorted_arrays(&[], &[1, 2, 3]), [1, 2, 3]); // One empty array
        assert_eq!(Solution::merge_sorted_arrays(&[1, 2, 3], &[]), [1, 2, 3]); // One empty array
        assert_eq!(Solution::merge_sorted_arrays(&[], &[]), []); // Both empty arrays
        assert_eq!(Solution::merge_sorted_arrays(&[1, 1, 1], &[1, 1]), [1]); // All elements duplicated
        assert_eq!(Solution::merge_sorted_arrays(&[-5, -2, 0, 3], &[-3, 0, 2, 4]), [-5, -3, -2, 0, 2, 3, 4]); // Negative numbers
    }
}
```

## **📌 Algorithm Explanation**

### **🔹 Two-Pointer Approach (`O(N + M)`)**

- Since **both arrays are sorted**, we use two pointers (`i` for `a`, `j` for `b`).
- **Iterate through both arrays** while comparing elements:
  1. If `a[i] < b[j]`, add `a[i]` and move `i++`.
  2. If `a[i] > b[j]`, add `b[j]` and move `j++`.
  3. If `a[i] == b[j]`, add only **one** occurrence and move both pointers.
- **Finally, append any remaining elements** from either array.

### **Step-by-Step Example**

#### **Input:**

`a = [1, 2, 3, 4, 6]`, `b = [2, 3, 5]`

#### **Execution Trace:**

| **Step** | **Pointer (i, j)** | **Comparison** | **Action** | **Result**           |
| -------- | ------------------ | -------------- | ---------- | -------------------- |
| 1        | `(0, 0)`           | `1 < 2`        | Add `1`    | `[1]`                |
| 2        | `(1, 0)`           | `2 == 2`       | Add `2`    | `[1, 2]`             |
| 3        | `(2, 1)`           | `3 == 3`       | Add `3`    | `[1, 2, 3]`          |
| 4        | `(3, 2)`           | `4 < 5`        | Add `4`    | `[1, 2, 3, 4]`       |
| 5        | `(4, 2)`           | `6 > 5`        | Add `5`    | `[1, 2, 3, 4, 5]`    |
| 6        | `(4, 3)`           | `6 remains`    | Add `6`    | `[1, 2, 3, 4, 5, 6]` |

#### **Final Output:**

`[1, 2, 3, 4, 5, 6]` ✅
