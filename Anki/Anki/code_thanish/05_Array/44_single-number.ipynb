    pub fn single_number(nums: Vec<i32>) -> i32 {
        nums.iter().fold(0, |acc, &num| acc ^ num)  // XOR all elements
    }
}

## **üìå Algorithm Explanation**

### **XOR Approach (`O(N)`)**

- **XOR Properties**:
  - `a ^ a = 0` (a number XORed with itself cancels out).
  - `0 ^ b = b` (XOR with `0` keeps the number).
  - XOR is **commutative and associative**, meaning order doesn't matter.

#### **Example**

üìå **Input:** `nums = [4, 1, 2, 1, 2]`

| **Step** | \*\*XOR Accumulator (`acc`) | Current `num` | XOR Result |
| -------- | --------------------------- | ------------- | ---------- |
| `0`      | `0`                         | `4`           | `4`        |
| `1`      | `4`                         | `1`           | `5`        |
| `2`      | `5`                         | `2`           | `7`        |
| `3`      | `7`                         | `1`           | `6`        |
| `4`      | `6`                         | `2`           | `4`        |

‚úÖ **Final Answer:** `4` (since all duplicates cancel out).
