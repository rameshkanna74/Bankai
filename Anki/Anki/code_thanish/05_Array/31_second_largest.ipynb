/// Finds the second largest distinct element in the given array.
/// Returns `Some(value)` if a second largest exists, otherwise returns `None`.
pub fn find_second_largest(arr: &[u32]) -> Option<u32> {
    if arr.len() < 2 {
        return None;
    }

    let mut largest = u32::MIN;
    let mut second_largest = u32::MIN;

    for &num in arr {
        if num > largest {
            second_largest = largest;
            largest = num;
        } else if num > second_largest && num < largest {
            second_largest = num;
        }
    }

    if second_largest == u32::MIN {
        None
    } else {
        Some(second_largest)
    }
}
