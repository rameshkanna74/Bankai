use std::collections::HashSet;

impl Solution {
    pub fn min_abs_difference(nums: Vec<i32>, goal: i32) -> i32 {
        let n = nums.len();
        let (left_part, right_part) = nums.split_at(n / 2);

        // Generate all subset sums for both halves
        let left_sums = Self::generate_subset_sums(left_part);
        let mut right_sums = Self::generate_subset_sums(right_part);

        // Sort right sums for binary search
        right_sums.sort_unstable();

        let mut min_diff = i32::MAX;

        // Find the closest sum
        for &sum in &left_sums {
            let remaining = goal - sum;
            if let Ok(pos) = right_sums.binary_search(&remaining) {
                return 0; // Exact match found
            } else {
                let insert_pos = right_sums.partition_point(|&x| x < remaining);

                if insert_pos < right_sums.len() {
                    min_diff = min_diff.min((remaining - right_sums[insert_pos]).abs());
                }
                if insert_pos > 0 {
                    min_diff = min_diff.min((remaining - right_sums[insert_pos - 1]).abs());
                }
            }
        }

        min_diff
    }

    fn generate_subset_sums(nums: &[i32]) -> Vec<i32> {
        let mut sums = HashSet::new();
        sums.insert(0);

        for &num in nums {
            let mut new_sums = sums.clone();
            for &sum in &sums {
                new_sums.insert(sum + num);
            }
            sums = new_sums;
        }

        sums.into_iter().collect()
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Meet-in-the-Middle (`O(2^(n/2) log(2^(n/2)))`)**

- **Step 1:** Divide `nums` into **two halves** (`left_part` & `right_part`).
- **Step 2:** Compute **all possible subset sums** for each half.
- **Step 3:** Sort the right-half sums (`O(2^(n/2) log(2^(n/2)))`).
- **Step 4:** For each sum in the left-half:
  - **Binary search** for closest sum in the right-half.
  - **Update min absolute difference**.

### **2Ô∏è‚É£ Why Meet-in-the-Middle?**

- **Exponential `O(2^n)` is infeasible** for `n = 40`.
- **Reducing `n` to `n/2` makes it feasible.**
- **Binary search speeds up finding closest sum**.
