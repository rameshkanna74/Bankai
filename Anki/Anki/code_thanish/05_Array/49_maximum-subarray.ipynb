# 🚀 **Optimized, Production-Quality Rust Solution for Maximum Subarray (Kadane's Algorithm)**

---

## **📌 Problem Breakdown**

### **Task**

- Given an integer array `nums`, find the **contiguous subarray** with the largest sum.
- **Return** the sum of that subarray.

### **Examples**

| **Input**                 | **Output** | **Explanation**             |
| ------------------------- | ---------- | --------------------------- |
| `[-2,1,-3,4,-1,2,1,-5,4]` | `6`        | `[4,-1,2,1]` has sum `6`    |
| `[1]`                     | `1`        | `[1]` has sum `1`           |
| `[5,4,-1,7,8]`            | `23`       | `[5,4,-1,7,8]` has sum `23` |

### **Constraints**

- `1 ≤ nums.length ≤ 10⁵`
- `-10⁴ ≤ nums[i] ≤ 10⁴`

---

## **📊 Time & Space Complexity Analysis**

| **Approach**                           | **Time Complexity** | **Space Complexity** | **Why?**                                 |
| -------------------------------------- | ------------------- | -------------------- | ---------------------------------------- |
| **Brute Force (`O(N²)`)** ❌           | `O(N²)`             | `O(1)`               | Checks all subarrays. Too slow.          |
| **Kadane’s Algorithm (`O(N)`)** ✅     | ✅ `O(N)`           | ✅ `O(1)`            | Linear time, best choice for interviews. |
| **Divide & Conquer (`O(N log N)`)** ✅ | `O(N log N)`        | `O(log N)`           | Suboptimal but useful follow-up.         |

✅ **Best Choice: Kadane’s Algorithm (`O(N)`)**

- **Linear time complexity** – ideal for `N = 10⁵`.
- **Constant space (`O(1)`)** – only a few variables.
- **Greedy & DP hybrid** – optimal for contiguous subarrays.

## **📌 Optimized Rust Code: Kadane’s Algorithm (`O(N)`)**

```rust
impl Solution {
    pub fn max_sub_array(nums: Vec<i32>) -> i32 {
        let mut max_sum = nums[0]; // Store max sum found so far
        let mut current_sum = nums[0]; // Track current subarray sum

        for &num in nums.iter().skip(1) {
            current_sum = current_sum.max(0) + num; // Extend or start fresh
            max_sum = max_sum.max(current_sum); // Update global max
        }

        max_sum
    }
}
```

## **📌 Algorithm Explanation**

### **1️⃣ Kadane’s Algorithm (`O(N)`)**

- **Tracks the "best subarray so far"** while iterating through `nums`.
- **Uses a Greedy + DP approach:**
  - `current_sum = max(0, current_sum) + num`
  - `max_sum = max(max_sum, current_sum)`

#### **Example Walkthrough**

📌 **Input:** `nums = [-2,1,-3,4,-1,2,1,-5,4]`

| `num` | `current_sum` | `max_sum` |
| ----- | ------------- | --------- |
| `-2`  | `-2`          | `-2`      |
| `1`   | `1`           | `1`       |
| `-3`  | `-2`          | `1`       |
| `4`   | `4`           | `4`       |
| `-1`  | `3`           | `4`       |
| `2`   | `5`           | `5`       |
| `1`   | `6`           | `6`       |
| `-5`  | `1`           | `6`       |
| `4`   | `5`           | `6`       |

✅ **Output:** `6`

✅ **All Negative Numbers (`[-1, -2, -3] → -1`)**  
✅ **Single Element (`[3] → 3`)**  
✅ **All Positive Numbers (`[1, 2, 3] → 6`)**  
✅ **Large Input (`N = 10⁵`)** – Efficient `O(N)`.

## **📌 Follow-Up: Divide & Conquer (`O(N log N)`)**

**Idea**:

- Split array into two halves recursively.
- **Three possible max subarrays**:
  - Left half (`max_sub_array(left)`).
  - Right half (`max_sub_array(right)`).
  - **Crossing the middle** (must include elements from both halves).

```rust
impl Solution {
    pub fn max_sub_array(nums: Vec<i32>) -> i32 {
        fn divide_and_conquer(nums: &[i32]) -> i32 {
            let n = nums.len();
            if n == 1 { return nums[0]; }

            let mid = n / 2;
            let left_max = divide_and_conquer(&nums[..mid]);
            let right_max = divide_and_conquer(&nums[mid..]);

            // Find max crossing sum
            let mut left_sum = i32::MIN;
            let mut sum = 0;
            for &num in nums[..mid].iter().rev() {
                sum += num;
                left_sum = left_sum.max(sum);
            }

            let mut right_sum = i32::MIN;
            sum = 0;
            for &num in &nums[mid..] {
                sum += num;
                right_sum = right_sum.max(sum);
            }

            left_max.max(right_max).max(left_sum + right_sum)
        }

        divide_and_conquer(&nums)
    }
}
```

✅ **Time Complexity:** `O(N log N)`  
✅ **Space Complexity:** `O(log N)` (recursive calls)  
✅ **Useful for interviews, but `O(N)` is better.**

## **📌 Follow-Up Enhancements**

### **1️⃣ Handling Streaming Input**

- Keep track of `current_sum` and `max_sum` using a **rolling window**.
- **Example:** If data is arriving continuously, we process chunks **on the fly**.

### **2️⃣ Parallelizing Kadane’s Algorithm**

- Divide `nums` into `P` chunks and **apply Kadane’s** to each chunk.
- Merge results **in `O(P)` time**.

### **3️⃣ Distributed Systems Approach**

- Kadane’s **fits MapReduce**:
  - **Mapper:** Finds local `max_subarray` for each partition.
  - **Reducer:** Merges results across partitions.
