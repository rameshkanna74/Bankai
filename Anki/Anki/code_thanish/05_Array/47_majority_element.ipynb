impl Solution {
    pub fn majority_element(nums: Vec<i32>) -> i32 {
        let (mut candidate, mut count) = (0, 0);

        for &num in &nums {
            if count == 0 {
                candidate = num;
            }
            count += if num == candidate { 1 } else { -1 };
        }

        candidate // Guaranteed to be correct (per problem statement).
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Boyer-Moore Voting Algorithm (`O(N)`)**

- Maintains a **candidate** and a **count**.
- **Iterates through the array once** (`O(N)`) and:
  - If `count == 0`, pick a new **candidate**.
  - If the current number matches the **candidate**, **increment count**.
  - Otherwise, **decrement count**.
- **Why does this work?**
  - If a majority element exists, it **must** be the last surviving candidate.

#### **Example Walkthrough**

üìå **Input:** `nums = [2,2,1,1,1,2,2]`

| Step | Num | Candidate | Count | Action                     |
| ---- | --- | --------- | ----- | -------------------------- |
| 1    | 2   | 2         | 1     | Init candidate             |
| 2    | 2   | 2         | 2     | Increment                  |
| 3    | 1   | 2         | 1     | Decrement                  |
| 4    | 1   | 2         | 0     | Decrement, Reset candidate |
| 5    | 1   | 1         | 1     | New candidate              |
| 6    | 2   | 1         | 0     | Decrement                  |
| 7    | 2   | 2         | 1     | New candidate              |

‚úÖ **Output:** `2`

### **1Ô∏è‚É£ Handling Streaming Input**

- If input is coming in chunks, use a **sliding window** or maintain a **running candidate**.

### **2Ô∏è‚É£ Parallelizing the Majority Count**

- The Boyer-Moore **is inherently sequential**, but we can:
  - **Split the array** into chunks.
  - Compute local majority **in parallel**.
  - Combine results in a **reduction step**.

```rust
use rayon::prelude::*;

fn parallel_majority_element(nums: &[i32]) -> i32 {
    let local_counts = nums.par_chunks(nums.len() / 4)
        .map(|chunk| Solution::majority_element(chunk.to_vec()))
        .collect::<Vec<_>>();

    Solution::majority_element(local_counts)
}
```
