# 🚀 **Optimized, Production-Quality Rust Solution for 4Sum (`O(N³)`)**

---

## **📌 Problem Breakdown**

### **Task**

- Given an integer array `nums` and an integer `target`, return **all unique quadruplets** `[nums[a], nums[b], nums[c], nums[d]]` such that:
  - `a, b, c, d` are distinct indices.
  - `nums[a] + nums[b] + nums[c] + nums[d] == target`.
- **The output must contain only unique quadruplets**.

### **Examples**

| **Input**                       | **Output**                            | **Explanation**                    |
| ------------------------------- | ------------------------------------- | ---------------------------------- |
| `[1,0,-1,0,-2,2]`, `target = 0` | `[[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]` | Valid quadruplets that sum to `0`. |
| `[2,2,2,2,2]`, `target = 8`     | `[[2,2,2,2]]`                         | Only one unique quadruplet.        |

### **Constraints**

- `1 ≤ nums.length ≤ 200`
- `-10⁹ ≤ nums[i] ≤ 10⁹`
- `-10⁹ ≤ target ≤ 10⁹`

## **📊 Time & Space Complexity Analysis**

| **Approach**                            | **Time Complexity** | **Space Complexity**   | **Why?**                                              |
| --------------------------------------- | ------------------- | ---------------------- | ----------------------------------------------------- |
| **Brute Force (`O(N⁴)`)** ❌            | `O(N⁴)`             | `O(1)`                 | Too slow for `N = 200`, needs four nested loops.      |
| **Sorting + Two Pointers (`O(N³)`)** ✅ | ✅ `O(N³)`          | ✅ `O(N)` (for output) | Best for this problem, avoids duplicates efficiently. |

✅ **Best Choice: Sorting + Two Pointers (`O(N³)`)**

- **Sort the array** to efficiently remove duplicates.
- **Use a two-pointer approach** inside a nested loop.
- **Avoid unnecessary iterations** to optimize performance.

## **📌 Optimized Rust Code: Sorting + Two Pointers (`O(N³)`)**

```rust
impl Solution {
    pub fn four_sum(mut nums: Vec<i32>, target: i32) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let n = nums.len();

        if n < 4 {
            return result;
        }

        // Step 1: Sort the array to easily avoid duplicates
        nums.sort();

        // Step 2: Iterate through the array with two fixed indices
        for i in 0..n - 3 {
            // Skip duplicate elements
            if i > 0 && nums[i] == nums[i - 1] {
                continue;
            }

            for j in i + 1..n - 2 {
                // Skip duplicate elements
                if j > i + 1 && nums[j] == nums[j - 1] {
                    continue;
                }

                let mut left = j + 1;
                let mut right = n - 1;

                while left < right {
                    let sum = nums[i] as i64 + nums[j] as i64 + nums[left] as i64 + nums[right] as i64;

                    if sum == target as i64 {
                        result.push(vec![nums[i], nums[j], nums[left], nums[right]]);

                        // Move left and right pointers while avoiding duplicates
                        while left < right && nums[left] == nums[left + 1] {
                            left += 1;
                        }
                        while left < right && nums[right] == nums[right - 1] {
                            right -= 1;
                        }

                        left += 1;
                        right -= 1;
                    } else if sum < target as i64 {
                        left += 1; // Increase sum
                    } else {
                        right -= 1; // Decrease sum
                    }
                }
            }
        }

        result
    }
}
```

## **📌 Algorithm Explanation**

### **1️⃣ Sorting + Two-Pointer Technique (`O(N³)`)**

1. **Sort the array** (`O(N log N)`) to enable efficient duplicate removal.
2. **Fix two indices (`i` and `j`)** (`O(N²)` loop).
3. **Use two-pointer search (`left`, `right`)** to find pairs that sum to `target - (nums[i] + nums[j])`.
4. **Skip duplicate values** efficiently to ensure unique quadruplets.

#### **Example Walkthrough**

📌 **Input:** `nums = [1,0,-1,0,-2,2]`, `target = 0`  
📌 **Sorted:** `[-2, -1, 0, 0, 1, 2]`

| i   | j   | Left | Right | Sum                      | Result             |
| --- | --- | ---- | ----- | ------------------------ | ------------------ |
| 0   | 1   | `0`  | `2`   | `-2 + (-1) + 0 + 2 = -1` | Move `left`        |
| 0   | 1   | `0`  | `1`   | `-2 + (-1) + 0 + 1 = -2` | Move `left`        |
| 0   | 2   | `0`  | `2`   | `-2 + 0 + 0 + 2 = 0`     | ✅ `[-2, 0, 0, 2]` |
| 1   | 2   | `0`  | `1`   | `-1 + 0 + 0 + 1 = 0`     | ✅ `[-1, 0, 0, 1]` |

✅ **Final Answer:** `[[-2, -1, 1, 2], [-2, 0, 0, 2], [-1, 0, 0, 1]]`
