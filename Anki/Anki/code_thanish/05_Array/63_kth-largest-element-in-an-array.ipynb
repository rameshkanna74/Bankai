### **QuickSelect Approach (`O(n)`)**

```rust
impl Solution {
    pub fn find_kth_largest(mut nums: Vec<i32>, k: i32) -> i32 {
        let k = nums.len() - k as usize; // Convert k-th largest to index
        Self::quickselect(&mut nums, 0, nums.len() - 1, k)
    }

    fn quickselect(nums: &mut [i32], left: usize, right: usize, k: usize) -> i32 {
        if left == right { return nums[left]; }

        let pivot_index = Self::partition(nums, left, right);

        match pivot_index.cmp(&k) {
            std::cmp::Ordering::Equal => nums[k],
            std::cmp::Ordering::Less => Self::quickselect(nums, pivot_index + 1, right, k),
            std::cmp::Ordering::Greater => Self::quickselect(nums, left, pivot_index - 1, k),
        }
    }

    fn partition(nums: &mut [i32], left: usize, right: usize) -> usize {
        let pivot = nums[right];
        let mut store_index = left;

        for i in left..right {
            if nums[i] < pivot {
                nums.swap(store_index, i);
                store_index += 1;
            }
        }

        nums.swap(store_index, right);
        store_index
    }
}
```

## **üìå Alternative Approach: Min-Heap (`O(n log k)`)**

```rust
use std::collections::BinaryHeap;
use std::cmp::Reverse;

impl Solution {
    pub fn find_kth_largest(nums: Vec<i32>, k: i32) -> i32 {
        let k = k as usize;
        let mut min_heap = BinaryHeap::with_capacity(k);

        for &num in &nums {
            if min_heap.len() < k {
                min_heap.push(Reverse(num));
            } else if min_heap.peek().unwrap().0 < num {
                min_heap.pop();
                min_heap.push(Reverse(num));
            }
        }

        min_heap.peek().unwrap().0
    }
}
```

‚úÖ **Time Complexity:** `O(n log k)`  
‚úÖ **Space Complexity:** `O(k)`

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ QuickSelect (`O(n)`)**

- **Step 1:** Convert `k-th` largest to an **index**: `nums.len() - k`.
- **Step 2:** **Partition using a pivot**:
  - Elements **smaller** than pivot go left.
  - Elements **greater** than pivot go right.
- **Step 3:** **Recursively QuickSelect** the correct partition.

### **2Ô∏è‚É£ Min-Heap (`O(n log k)`)**

- Maintain a **min-heap of size `k`**.
- **Only store the top `k` elements**.
- **Pop the smallest** when a larger element arrives.
