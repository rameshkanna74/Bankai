## ‚úÖ Problem Statement Recap

You are given the `root` of a binary tree. Return the **preorder traversal** (Root ‚Üí Left ‚Üí Right) of its nodes' values.

---

## üîç Algorithm Strategy

We provide the **iterative approach** here (since recursive is trivial) using a **stack**, which simulates the call stack of recursion.

### üß† DSA Concepts Used

* **Binary Tree Traversal**
* **Depth-First Search (DFS)**
* **Stack (LIFO)** for traversal state

---

## üßÆ Time & Space Complexity

### ‚è± Time Complexity: `O(n)`

* We visit each node **exactly once**.

### üß† Space Complexity: `O(h)` worst-case

* Where `h` is the height of the tree (log n for balanced trees, `n` in the worst case if skewed).
* We maintain an explicit `Vec` as a **stack**.

## ‚úÖ Rust Code (Production-Quality, Interview-Ready)

```rust
use std::cell::RefCell;
use std::rc::Rc;

/// Definition for a binary tree node.
#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode {
            val,
            left: None,
            right: None,
        }
    }
}

pub struct Solution;

impl Solution {
    /// Preorder traversal (iterative DFS): Root -> Left -> Right
    pub fn preorder_traversal(root: Option<Rc<RefCell<TreeNode>>>) -> Vec<i32> {
        let mut result = Vec::new();
        let mut stack = Vec::new();

        if let Some(node) = root {
            stack.push(node);
        }

        while let Some(node_rc) = stack.pop() {
            let node = node_rc.borrow();
            result.push(node.val);

            // Push right first so left is processed first
            if let Some(right) = node.right.clone() {
                stack.push(right);
            }
            if let Some(left) = node.left.clone() {
                stack.push(left);
            }
        }

        result
    }
}
```

mod tests {
    use super::*;

    fn build_tree_from_vec(values: &[Option<i32>]) -> Option<Rc<RefCell<TreeNode>>> {
        use std::collections::VecDeque;

        if values.is_empty() || values[0].is_none() {
            return None;
        }

        let root = Rc::new(RefCell::new(TreeNode::new(values[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));

        let mut index = 1;

        while index < values.len() {
            if let Some(current) = queue.pop_front() {
                if let Some(Some(val)) = values.get(index) {
                    let left = Rc::new(RefCell::new(TreeNode::new(*val)));
                    current.borrow_mut().left = Some(Rc::clone(&left));
                    queue.push_back(left);
                }
                index += 1;

                if let Some(Some(val)) = values.get(index) {
                    let right = Rc::new(RefCell::new(TreeNode::new(*val)));
                    current.borrow_mut().right = Some(Rc::clone(&right));
                    queue.push_back(right);
                }
                index += 1;
            }
        }

        Some(root)
    }
