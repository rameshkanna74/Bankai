## ‚úÖ Problem Summary

* **Goal:** Return the **preorder traversal** (Root ‚Üí Left ‚Üí Right) of a binary tree using **iterative** method.
* **Input:** Binary tree represented by `Option<Rc<RefCell<TreeNode>>>`
* **Output:** `Vec<i32>` with node values in preorder

---

## üß† Core Algorithm (Iterative Preorder DFS using Stack)

### Strategy:

Use a **stack** to simulate the recursion stack in preorder DFS traversal.

### Steps:

1. Initialize an empty stack.
2. Push the root node if it exists.
3. While the stack is not empty:

   * Pop a node, process it (`val ‚Üí result`).
   * Push right child (if any).
   * Push left child (if any).

### Why push right before left?

Because **stack is LIFO**, so we want left to be processed before right.

---

## ‚è± Time & Space Complexity

* **Time Complexity:** `O(n)` ‚Äî each node is visited once.
* **Space Complexity:** `O(h)` ‚Äî at most `h` nodes on the stack, where `h` is the height of the tree (`log n` for balanced, `n` for skewed).

## üß¨ Clean & Idiomatic Rust Solution

```rust
use std::cell::RefCell;
use std::rc::Rc;

/// Definition for a binary tree node.
#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode {
            val,
            left: None,
            right: None,
        }
    }
}

pub struct Solution;

impl Solution {
    /// Iterative Preorder Traversal: Root -> Left -> Right
    pub fn preorder_traversal(root: Option<Rc<RefCell<TreeNode>>>) -> Vec<i32> {
        let mut result = Vec::new();
        let mut stack = Vec::new();

        if let Some(node) = root {
            stack.push(node);
        }

        while let Some(curr) = stack.pop() {
            let node = curr.borrow();
            result.push(node.val);

            // Push right first so that left is processed first
            if let Some(ref right) = node.right {
                stack.push(Rc::clone(right));
            }
            if let Some(ref left) = node.left {
                stack.push(Rc::clone(left));
            }
        }

        result
    }
}
```

````rust
use std::cell::RefCell;
use std::rc::Rc;

/// Definition for a binary tree node.
#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode {
            val,
            left: None,
            right: None,
        }
    }
}

pub struct Solution;

impl Solution {
    /// Iterative Preorder Traversal: Root -> Left -> Right
    pub fn preorder_traversal(root: Option<Rc<RefCell<TreeNode>>>) -> Vec<i32> {
        let mut result = Vec::new();
        let mut stack = Vec::new();

        if let Some(node) = root {
            stack.push(node);
        }

        while let Some(curr) = stack.pop() {
            let node = curr.borrow();
            result.push(node.val);

            // Push right first so that left is processed first
            if let Some(ref right) = node.right {
                stack.push(Rc::clone(right));
            }
            if let Some(ref left) = node.left {
                stack.push(Rc::clone(left));
            }
        }

        result
    }
}
````
