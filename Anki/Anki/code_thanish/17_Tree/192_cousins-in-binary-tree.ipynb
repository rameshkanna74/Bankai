## Problem Recap

Given a binary tree with unique node values, and two values `x` and `y`, determine if the nodes corresponding to `x` and `y` are cousins.

**Cousins** means:

* Same depth
* Different parents

## Rust Code

```rust
use std::rc::Rc;
use std::cell::RefCell;

#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode { val, left: None, right: None }
    }
}

pub struct Solution;

impl Solution {
    /// Returns true if nodes with values x and y are cousins in the tree.
    ///
    /// Cousins = same depth but different parents.
    pub fn is_cousins(root: Option<Rc<RefCell<TreeNode>>>, x: i32, y: i32) -> bool {
        // Helper function to find depth and parent of a given target node
        fn find_depth_and_parent(
            node: &Option<Rc<RefCell<TreeNode>>>, 
            target: i32, 
            depth: i32, 
            parent: Option<i32>
        ) -> Option<(i32, Option<i32>)> {
            if let Some(rc_node) = node {
                let node_ref = rc_node.borrow();
                if node_ref.val == target {
                    return Some((depth, parent));
                }
                // Search left subtree
                if let Some(res) = find_depth_and_parent(&node_ref.left, target, depth + 1, Some(node_ref.val)) {
                    return Some(res);
                }
                // Search right subtree
                if let Some(res) = find_depth_and_parent(&node_ref.right, target, depth + 1, Some(node_ref.val)) {
                    return Some(res);
                }
            }
            None
        }

        if x == y {
            return false; // distinct nodes required
        }

        let x_info = find_depth_and_parent(&root, x, 0, None);
        let y_info = find_depth_and_parent(&root, y, 0, None);

        match (x_info, y_info) {
            (Some((dx, px)), Some((dy, py))) => dx == dy && px != py,
            _ => false,
        }
    }
}
```

## Explanation

### Core DSA Concepts

* **Tree traversal (DFS)**
* **Binary Tree properties**

### Algorithm Design

* We want to find the **depth** and **parent** of both nodes `x` and `y`.
* Implement a DFS helper that returns `(depth, parent)` for a target node.
* Once both `(depth_x, parent_x)` and `(depth_y, parent_y)` are known, check:

  * `depth_x == depth_y` (same level)
  * `parent_x != parent_y` (different parents)
* If both hold, return `true`; else `false`.

This is a straightforward, single-pass DFS approach.

            return None;

        let root = Rc::new(RefCell::new(TreeNode::new(nodes[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));
        let mut i = 1;

        while i < nodes.len() {
            if let Some(curr) = queue.pop_front() {

                    curr.borrow_mut().left = Some(Rc::clone(&left));
                    queue.push_back(left);
