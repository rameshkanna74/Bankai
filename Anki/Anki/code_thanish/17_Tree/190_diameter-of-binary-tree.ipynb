# Rust Code

```rust
use std::rc::Rc;
use std::cell::RefCell;

#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode { val, left: None, right: None }
    }
}

pub struct Solution;

impl Solution {
    /// Returns the diameter of the binary tree.
    ///
    /// Diameter is the length (number of edges) of the longest path between any two nodes.
    pub fn diameter_of_binary_tree(root: Option<Rc<RefCell<TreeNode>>>) -> i32 {
        /// Helper function to compute depth and update diameter.
        /// Returns the height of the current node.
        fn depth(node: &Option<Rc<RefCell<TreeNode>>>, diameter: &mut i32) -> i32 {
            match node {
                Some(rc_node) => {
                    let node_ref = rc_node.borrow();
                    let left_depth = depth(&node_ref.left, diameter);
                    let right_depth = depth(&node_ref.right, diameter);

                    // Diameter at this node is sum of left + right depths
                    *diameter = (*diameter).max(left_depth + right_depth);

                    // Return max depth + 1 for current node
                    1 + left_depth.max(right_depth)
                }
                None => 0,
            }
        }

        let mut diameter = 0;
        depth(&root, &mut diameter);
        diameter
    }
}
```

# Explanation

### Core Concepts

* **Depth-first search (DFS)**
* **Post-order traversal**
* **Tree diameter logic**: The diameter at a node = longest path through that node = `left subtree height + right subtree height`.
* Track and update the maximum diameter globally.

### Algorithm Design

* Define a helper recursive function `depth`:

  * Returns the height of a subtree rooted at the current node.
  * Updates a mutable reference `diameter` with the maximum diameter found so far.
* Recursively compute depths for left and right subtrees.
* For each node, potential diameter = left\_depth + right\_depth (edges count).
* Keep global max diameter updated.
* Return the max diameter at the end.

### Why is this optimal?

* We visit each node exactly once â†’ **O(n) time**.
* No repeated computations â†’ no extra overhead.
* Uses recursion stack proportional to tree height â†’ **O(h) space**.

# Edge Cases

* Single node tree â†’ diameter = 0.
* Skewed tree â†’ diameter = number of nodes - 1.
* Balanced tree â†’ diameter = longest path crossing root or any node.
* Empty tree â†’ (not in constraints, but logically diameter = 0).

#[cfg(test)]
mod tests {
    use super::*;
    use std::collections::VecDeque;

    /// Builds tree from level order vector representation.
    fn build_tree(nodes: &[Option<i32>]) -> Option<Rc<RefCell<TreeNode>>> {
        if nodes.is_empty() || nodes[0].is_none() {
            return None;
        }

        let root = Rc::new(RefCell::new(TreeNode::new(nodes[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));
        let mut i = 1;

        while i < nodes.len() {
            if let Some(curr) = queue.pop_front() {
                if let Some(Some(val)) = nodes.get(i).cloned() {
                    let left = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().left = Some(Rc::clone(&left));
                    queue.push_back(left);
                }
                i += 1;

                if let Some(Some(val)) = nodes.get(i).cloned() {
                    let right = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().right = Some(Rc::clone(&right));
                    queue.push_back(right);
                }
                i += 1;
            }
        }
        Some(root)
    }

    #[test]
    fn test_example_1() {
        let root = build_tree(&[Some(1), Some(2), Some(3), Some(4), Some(5)]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 3);
    }

    #[test]
    fn test_example_2() {
        let root = build_tree(&[Some(1), Some(2)]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 1);
    }

    #[test]
    fn test_single_node() {
        let root = build_tree(&[Some(42)]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 0);
    }

    #[test]
    fn test_skewed_left() {
        let root = build_tree(&[Some(1), Some(2), None, Some(3), None, Some(4), None]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 3);
    }

    #[test]
    fn test_skewed_right() {
        let root = build_tree(&[Some(1), None, Some(2), None, Some(3), None, Some(4)]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 3);
    }

    #[test]
    fn test_empty_tree() {
        let root = build_tree(&[]);
        assert_eq!(Solution::diameter_of_binary_tree(root), 0);
    }
