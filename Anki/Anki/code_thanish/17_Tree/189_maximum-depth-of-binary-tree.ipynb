# ✅ 104. Maximum Depth of Binary Tree

## ✅ Final Rust Code (Recursive DFS)

```rust
use std::cell::RefCell;
use std::rc::Rc;

#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode { val, left: None, right: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn max_depth(root: Option<Rc<RefCell<TreeNode>>>) -> i32 {
        fn dfs(node: &Option<Rc<RefCell<TreeNode>>>) -> i32 {
            match node {
                Some(rc_node) => {
                    let node_ref = rc_node.borrow();
                    let left_depth = dfs(&node_ref.left);
                    let right_depth = dfs(&node_ref.right);
                    1 + left_depth.max(right_depth)
                }
                None => 0,
            }
        }

        dfs(&root)
    }
}
```

mod tests {

    fn build_tree(nodes: &[Option<i32>]) -> Option<Rc<RefCell<TreeNode>>> {
        if nodes.is_empty() || nodes[0].is_none() {
            return None;
        }

        let root = Rc::new(RefCell::new(TreeNode::new(nodes[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));
        let mut i = 1;

        while i < nodes.len() {
            if let Some(curr) = queue.pop_front() {
                if let Some(Some(val)) = nodes.get(i).map(|x| x.clone()) {
                    let left = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().left = Some(Rc::clone(&left));
                    queue.push_back(left);
                }
                i += 1;

                if let Some(Some(val)) = nodes.get(i).map(|x| x.clone()) {
                    let right = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().right = Some(Rc::clone(&right));
                    queue.push_back(right);
                }
                i += 1;
            }
        }

        Some(root)
    }

## 🔍 Algorithm Explanation

### 🔹 Core Concepts

* **Depth-First Search (DFS)**
* **Recursion**
* Tree traversal

### 🔹 Strategy

We recursively:

* Visit each node
* Compute the depth of left and right subtrees
* Take the maximum of both and add `1` for the current node

### 🔹 Edge Cases Handled

* Empty tree → returns 0
* Skewed tree → depth = number of nodes
* Balanced tree → accurate depth

| Complexity Type      | Value | Justification                                             |

| ⏱️ Time              | O(n)  | Each node visited once                                    |
| 🧠 Space (recursion) | O(h)  | h = height of the tree (log n for balanced, n for skewed) |

We avoid extra space; only recursion stack is used.

## 🚀 Follow-up Enhancements

### ✅ Iterative (Stack-based DFS)

If recursion depth is a concern (e.g., embedded systems), use stack-based DFS:

```rust
pub fn max_depth(mut root: Option<Rc<RefCell<TreeNode>>>) -> i32 {
    use std::collections::VecDeque;
    if root.is_none() { return 0; }

    let mut stack = vec![(root, 1)];
    let mut max_depth = 0;

    while let Some((Some(node), depth)) = stack.pop() {
        let node_ref = node.borrow();
        max_depth = max_depth.max(depth);
        if node_ref.left.is_some() {
            stack.push((node_ref.left.clone(), depth + 1));
        }
        if node_ref.right.is_some() {
            stack.push((node_ref.right.clone(), depth + 1));
        }
    }

    max_depth
}
```

### 🔹 Parallel Execution

* Each subtree depth can be calculated independently
* Ideal for **parallel map-reduce**
