## Problem Summary

* Find the maximum sum of any path in the binary tree.
* A path is any sequence of nodes connected by edges, not necessarily passing through root.
* Nodes can appear only once in the path.
* Node values can be negative.

## Rust Code

```rust
use std::rc::Rc;
use std::cell::RefCell;

#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode { val, left: None, right: None }
    }
}

pub struct Solution;

impl Solution {
    /// Returns the maximum path sum in the binary tree.
    ///
    /// The path can be any sequence of connected nodes and doesn't need to pass through the root.
    pub fn max_path_sum(root: Option<Rc<RefCell<TreeNode>>>) -> i32 {
        /// Recursive helper returns max gain from this node to leaf path (single branch)
        /// Also updates global max path sum (can include left + node + right).
        fn max_gain(node: &Option<Rc<RefCell<TreeNode>>>, max_sum: &mut i32) -> i32 {
            if let Some(rc_node) = node {
                let node_ref = rc_node.borrow();

                // Recursively get max gain from left and right subtrees, ignore negatives by max with 0
                let left_gain = max_gain(&node_ref.left, max_sum).max(0);
                let right_gain = max_gain(&node_ref.right, max_sum).max(0);

                // Price of the new path using this node as the root of that path
                let price_newpath = node_ref.val + left_gain + right_gain;

                // Update global max sum if this path is better
                *max_sum = (*max_sum).max(price_newpath);

                // Return max gain if continuing the same path upward (only one side)
                node_ref.val + left_gain.max(right_gain)
            } else {
                0
            }
        }

        let mut max_sum = i32::MIN;
        max_gain(&root, &mut max_sum);
        max_sum
    }
}
```

## Explanation

### Core DSA Concepts

* **Depth-First Search (DFS) / Postorder traversal**
* **Divide and Conquer**
* **Greedy / Dynamic Programming (top-down with memoization implicitly through recursion)**

### Algorithm Design

* Use a helper function `max_gain` that computes the maximum gain from the current node down to any leaf through one child only.
* The function returns the max gain of continuing the path upward to its parent.
* It also calculates the maximum path sum that **includes** the current node and both children (forming a "path root" at current node).
* Update a global max path sum when the current node is root of a potential max path.
* Ignore negative gains by taking `max(0, gain)` to avoid hurting the path sum.
* Finally, return the global max sum.

### Why is this Optimal?

* Single DFS traversal â†’ **O(n) time**.
* No duplicated computations â†’ linear recursion.
* Space proportional to recursion stack â†’ **O(h)**, where h is tree height.

#[cfg(test)]
mod tests {
    use super::*;
    use std::collections::VecDeque;

    /// Builds tree from level order vector representation.
    fn build_tree(nodes: &[Option<i32>]) -> Option<Rc<RefCell<TreeNode>>> {
        if nodes.is_empty() || nodes[0].is_none() {
            return None;
        }

        let root = Rc::new(RefCell::new(TreeNode::new(nodes[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));
        let mut i = 1;

        while i < nodes.len() {
            if let Some(curr) = queue.pop_front() {
                if let Some(Some(val)) = nodes.get(i).cloned() {
                    let left = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().left = Some(Rc::clone(&left));
                    queue.push_back(left);
                }
                i += 1;

                if let Some(Some(val)) = nodes.get(i).cloned() {
                    let right = Rc::new(RefCell::new(TreeNode::new(val)));
                    curr.borrow_mut().right = Some(Rc::clone(&right));
                    queue.push_back(right);
                }
                i += 1;
            }
        }
        Some(root)
    }

    #[test]
    fn test_example_1() {
        let root = build_tree(&[Some(1), Some(2), Some(3)]);

        let root = build_tree(&[Some(-10), Some(9), Some(20), None, None, Some(15), Some(7)]);
        assert_eq!(Solution::max_path_sum(root), 42);
    }

    #[test]
    fn test_single_node() {
        let root = build_tree(&[Some(5)]);

    fn test_mixed_values() {
        let root = build_tree(&[Some(2), Some(-1)]);
        assert_eq!(Solution::max_path_sum(root), 2);
    }

    #[test]
    fn test_large_balanced_tree() {
        // Test with a balanced tree of depth 4 or more
