## âœ… Final Rust Code

```rust
use std::rc::Rc;
use std::cell::RefCell;

#[derive(Debug, PartialEq, Eq)]
pub struct TreeNode {
    pub val: i32,
    pub left: Option<Rc<RefCell<TreeNode>>>,
    pub right: Option<Rc<RefCell<TreeNode>>>,
}

impl TreeNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        TreeNode { val, left: None, right: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn good_nodes(root: Option<Rc<RefCell<TreeNode>>>) -> i32 {
        fn dfs(node: Option<Rc<RefCell<TreeNode>>>, max_so_far: i32) -> i32 {
            if let Some(rc_node) = node {
                let node_ref = rc_node.borrow();
                let curr_val = node_ref.val;
                let new_max = max_so_far.max(curr_val);
                let is_good = (curr_val >= max_so_far) as i32;
                is_good
                    + dfs(node_ref.left.clone(), new_max)
                    + dfs(node_ref.right.clone(), new_max)
            } else {
                0
            }
        }

        if let Some(rc_root) = root {
            dfs(Some(rc_root), i32::MIN)
        } else {
            0
        }
    }
}
```

```rust
        let root = Rc::new(RefCell::new(TreeNode::new(nodes[0].unwrap())));
        let mut queue = VecDeque::new();
        queue.push_back(Rc::clone(&root));

        let mut i = 1;
        while i < nodes.len() {
            if let Some(curr) = queue.pop_front() {
                if i < nodes.len() {
                    if let Some(val) = nodes[i] {
                        let left = Rc::new(RefCell::new(TreeNode::new(val)));
                        curr.borrow_mut().left = Some(Rc::clone(&left));
                        queue.push_back(left);
                    }
                    i += 1;
                }

                if i < nodes.len() {

                        let right = Rc::new(RefCell::new(TreeNode::new(val)));
                        curr.borrow_mut().right = Some(Rc::clone(&right));

                }

        }

        Some(root)
```

## ðŸ§  Algorithm Explanation

### ðŸ”¹ **Core Concepts**

* **DFS (Depth-First Search)** traversal
* **Tracking path maximum**
* **Recursive tree traversal with state propagation**

### ðŸ”¹ **Approach**

* A node is *"good"* if its value is **â‰¥ maximum value seen from root to that node**.
* We carry the `max_so_far` during DFS.
* If the current nodeâ€™s value â‰¥ `max_so_far`, it's good.
* Recurse into left and right subtrees with updated `max_so_far`.

### ðŸ”¹ **Edge Cases**

* Tree with:

  * All values equal â†’ All nodes are good
  * Decreasing values â†’ Only root is good
  * Increasing path â†’ All nodes are good
