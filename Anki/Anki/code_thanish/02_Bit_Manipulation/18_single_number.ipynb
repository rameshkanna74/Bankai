# **Optimized Rust Solution for "Single Number" (Leetcode 136) üöÄ**

## **Approach**

The problem requires us to find the unique number in an array where every other number appears **exactly twice**. We must ensure:  
‚úÖ **O(n) time complexity** (Linear)  
‚úÖ **O(1) space complexity** (Constant extra space)

### **Optimal Approach: XOR (Bitwise Operator)**

- The **XOR operation (‚äï)** is the most efficient way to solve this problem.
- Properties of XOR:
  1. `a ‚äï a = 0` (Any number XORed with itself is 0)
  2. `a ‚äï 0 = a` (Any number XORed with 0 remains unchanged)
  3. XOR is **commutative and associative**: `a ‚äï b ‚äï a = (a ‚äï a) ‚äï b = 0 ‚äï b = b`

Using this property, **all duplicate numbers cancel out, leaving only the single number**.

## **Optimized Rust Code (Bitwise XOR)**

```rust
impl Solution {
    pub fn single_number(nums: Vec<i32>) -> i32 {
        nums.into_iter().fold(0, |acc, num| acc ^ num)
    }
}
```

## **Algorithm Explanation**

1. **Initialize accumulator (`acc = 0`)**.
2. **Iterate through `nums`** using `fold()`:
   - Apply `acc ^= num`, effectively XORing all elements.
   - Since **all duplicates cancel out**, only the single element remains.
3. **Return the final `acc` value**, which is the unique number.

‚úÖ **Output:** `4` (as expected)
