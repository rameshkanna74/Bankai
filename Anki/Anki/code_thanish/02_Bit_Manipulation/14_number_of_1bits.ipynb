# **Optimized Rust Solution for Counting Set Bits (Hamming Weight) 🚀**

## **Approach**

To count the number of **set bits (1s) in the binary representation of an integer `n`**, we use **bitwise operations**.

### **Optimal Method: Brian Kernighan’s Algorithm**

Instead of checking every bit (`O(log n)` with naive shifting), we use:

```rust
n = n & (n - 1);
```

This **removes the rightmost set bit** in `O(1)` time.

### **Why is it Efficient?**

- Each operation **eliminates one `1` bit**, so it runs in **O(k)**, where `k` is the number of set bits.
- Worst case: **All bits are `1`** → `O(log n)`, which is optimal.

---

## **Optimized Rust Code**

```rust
pub fn count_set_bits(mut n: u32) -> u32 {
    let mut count = 0;
    while n > 0 {
        n &= n - 1; // Drops the lowest set bit
        count += 1;
    }
    count
}
```

## **Algorithm Explanation**

### **Brian Kernighan’s Algorithm**

1. **Key Operation**:

   ```rust
   n &= (n - 1);
   ```

   - Removes **rightmost 1-bit** efficiently.

2. **How It Works (Example `n = 6` → `110` in binary)**
   ```
   n = 110
   Step 1: 110 & 101 = 100 (count = 1)
   Step 2: 100 & 011 = 000 (count = 2)
   ```

### **Time Complexity Analysis**

✅ **Best Case**: \( O(1) \) (if `n = 0`)  
✅ **Worst Case**: \( O(\log n) \) (when all bits are `1`)  
✅ **Expected Case**: \( O(k) \), where `k` is the number of set bits

### **Space Complexity Analysis**

✅ **\( O(1) \)** (constant extra space)
