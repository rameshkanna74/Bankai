## **Optimized Rust Solution for Sorting a Linked List of 0s, 1s, and 2s**

### **Approach**

We can solve this problem optimally using **Dutch National Flag Algorithm**, which involves **three pointers (zero, one, and two buckets)** to sort the list in a single pass.  
The goal is to rearrange the list so that **all 0s come first, followed by 1s, and then 2s** without changing the node values.

### **Optimal Strategy**

- **Count-Based Sorting (O(N), O(1))**

  - Traverse the list once to count occurrences of `0`, `1`, and `2`.
  - Modify the existing linked list nodes in-place based on the counts.

- **Three-Pointer Approach (O(N), O(1))**
  - Maintain three dummy nodes for `0s`, `1s`, and `2s` and merge them in the end.
  - This method avoids modifying node values (follow-up requirement).

### **Rust Implementation**

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    /// Sorts the linked list consisting of 0s, 1s, and 2s in O(N) time and O(1) space.
    pub fn sort_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut zero_head = ListNode::new(0); // Dummy node for 0s
        let mut one_head = ListNode::new(0); // Dummy node for 1s
        let mut two_head = ListNode::new(0); // Dummy node for 2s

        let (mut zero_tail, mut one_tail, mut two_tail) = (&mut zero_head, &mut one_head, &mut two_head);

        // Partition the list into 0s, 1s, and 2s
        while let Some(mut node) = head {
            head = node.next.take(); // Move the head forward
            match node.val {
                0 => { zero_tail.next = Some(node); zero_tail = zero_tail.next.as_mut().unwrap(); }
                1 => { one_tail.next = Some(node); one_tail = one_tail.next.as_mut().unwrap(); }
                _ => { two_tail.next = Some(node); two_tail = two_tail.next.as_mut().unwrap(); }
            }
        }

        // Merge the three partitions
        zero_tail.next = one_head.next.or(two_head.next); // Link 0s â†’ 1s â†’ 2s
        one_tail.next = two_head.next;

        zero_head.next
    }
}
```

## **Algorithm Explanation**

### **Core Concepts**

1. **Three-Pointer Approach (Dutch National Flag Algorithm)**
   - Separate the linked list into three sublists:
     - **Zero List** (stores all `0s`)
     - **One List** (stores all `1s`)
     - **Two List** (stores all `2s`)
   - Maintain **three tail pointers** to link the sublists correctly.
   - Merge them to get the sorted linked list.

### **Time & Space Complexity**

âœ… **Time Complexity:**

- **O(N)**, since we traverse the list **once** to partition it and **once** to merge.

âœ… **Space Complexity:**

- **O(1) extra space**, since we modify the list **in-place** without using extra memory.
