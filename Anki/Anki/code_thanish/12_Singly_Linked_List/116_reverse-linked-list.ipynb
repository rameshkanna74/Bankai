# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

## 🧠 **Problem Breakdown**

We need to **reverse** a **singly linked list** and return the new head.

### **Key Observations**

1. **Iterative Approach (`O(n)`)**

   - Use three pointers: `prev`, `current`, and `next`.
   - Traverse the list while **reversing the links** at each step.

2. **Recursive Approach (`O(n)`)**
   - Recursively reach the end of the list.
   - As recursion unwinds, **reverse the links**.

Both approaches are **optimal**, and we will implement both.

---

## ⏳ **Time and Space Complexity**

✅ **Time Complexity:** `O(n)`, since we traverse each node once.  
✅ **Space Complexity:**

- **Iterative:** `O(1)` (constant extra space).
- **Recursive:** `O(n)` (stack space due to recursion).

---

## 🚀 **Optimized Rust Implementation**

### **🟢 Iterative Approach (Best for Production)**

```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }

    pub fn reverse_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev = None;
        while let Some(mut curr) = head {
            head = curr.next.take();
            curr.next = prev;
            prev = Some(curr);
        }
        prev
    }
}
```

### **🟢 Recursive Approach (Alternative)**

```rust
impl ListNode {
    pub fn reverse_list_recursive(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        fn helper(mut head: Option<Box<ListNode>>, prev: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
            match head {
                Some(mut node) => {
                    let next = node.next.take();
                    node.next = prev;
                    helper(next, Some(node))
                }
                None => prev,
            }
        }
        helper(head, None)
    }
}
```

---

## 🔥 **Why is this the Best Solution?**

✅ **Iterative (`O(n)`) is best for production** (constant space).  
✅ **Recursive (`O(n)`) is elegant but uses extra stack space**.  
✅ **Uses `Option<Box<T>>` to maintain Rust safety**.  
✅ **Avoids unnecessary `unwrap()` or `expect()`**.  
✅ **Handles edge cases (`None`, single node, large lists) correctly**.
