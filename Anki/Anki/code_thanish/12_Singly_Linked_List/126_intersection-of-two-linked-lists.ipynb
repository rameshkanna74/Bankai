## **Optimized Rust Solution for Finding Intersection of Two Linked Lists**

### **Approach**
We can solve this problem efficiently in **O(m + n) time complexity** and **O(1) space complexity** using **two-pointer traversal**.

---

### **Optimal Strategy**
**Two-Pointer Approach (O(m + n) time, O(1) space)**
1. Initialize two pointers, `p1` and `p2`, at `headA` and `headB` respectively.
2. Traverse the lists. When a pointer reaches the end, redirect it to the head of the other list.
3. If the lists intersect, the pointers will eventually meet at the intersection node.
4. If they do not intersect, both pointers will become `None` at the same time, ensuring termination.

#### **Why This Works?**
- By switching heads, both pointers **travel equal total distances** (`m + n`).
- If thereâ€™s an intersection, they will meet at the common node.
- If not, both will simultaneously reach `None`.

## **Rust Implementation**
```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    /// Finds the intersection node of two linked lists using O(m + n) time and O(1) space.
    pub fn get_intersection_node(
        head_a: Option<Box<ListNode>>,
        head_b: Option<Box<ListNode>>,
    ) -> Option<Box<ListNode>> {
        let (mut p1, mut p2) = (head_a.as_ref(), head_b.as_ref());

        while p1 != p2 {
            p1 = if p1.is_some() { p1.unwrap().next.as_ref() } else { head_b.as_ref() };
            p2 = if p2.is_some() { p2.unwrap().next.as_ref() } else { head_a.as_ref() };
        }
        p1.cloned()
    }
}
```
