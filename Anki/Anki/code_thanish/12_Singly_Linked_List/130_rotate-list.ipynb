## **Optimized Rust Solution for Rotating a Linked List**

### **Approach**

The problem requires us to **rotate a linked list to the right by `k` places**. Instead of performing `k` rotations one by one, which would be inefficient, we use an **optimal two-pass approach**:

1. **Find the Length of the List (`n`)**: Traverse the list once to determine the number of nodes.
2. **Optimize `k`**: Since rotating a list by `n` positions results in the same list (`k %= n`), we reduce redundant rotations.
3. **Identify the New Tail**: The new tail will be at position `(n - k - 1)`, and the new head will be `(n - k)`.
4. **Perform the Rotation**:
   - Connect the current tail to the head (forming a circular linked list).
   - Break the circular link at the new tail to get the rotated list.

This approach ensures an **O(n) time complexity** with **O(1) extra space**.

## **Optimized Rust Solution**

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    /// Rotates a linked list to the right by k positions.
    pub fn rotate_right(
        mut head: Option<Box<ListNode>>,
        k: i32,
    ) -> Option<Box<ListNode>> {
        if head.is_none() || k == 0 {
            return head;
        }

        // Step 1: Find length of the linked list
        let mut length = 1;
        let mut tail = &mut head;
        while let Some(node) = tail.as_mut().unwrap().next.as_mut() {
            length += 1;
            tail = &mut node.next;
        }

        // Step 2: Optimize k (avoid unnecessary full rotations)
        let k = k as usize % length;
        if k == 0 {
            return head; // No rotation needed
        }

        // Step 3: Identify the new tail (at position n-k-1) and new head (n-k)
        let split_pos = length - k;
        let mut new_tail = &mut head;
        for _ in 1..split_pos {
            new_tail = &mut new_tail.as_mut().unwrap().next;
        }

        let mut new_head = new_tail.as_mut().unwrap().next.take();
        tail.as_mut().unwrap().next = head;
        head = new_head;

        head
    }
}
```

## **Algorithm Explanation**

### **Core Concepts Used**

1. **Linked List Traversal**: We traverse the list to **count nodes**.
2. **Modular Arithmetic (`k %= n`)**: Optimizes redundant rotations.
3. **Two-Pointer Technique**: Finds the **new tail** efficiently.
4. **Circular Linked List Trick**: Temporarily connects the **tail to the head**, then breaks the link.
