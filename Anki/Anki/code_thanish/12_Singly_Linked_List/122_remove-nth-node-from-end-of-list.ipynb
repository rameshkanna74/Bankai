### **Optimized Rust Solution for "Remove Nth Node From End of List"**

This problem can be efficiently solved using the **two-pointer technique**, which allows us to find and remove the nth node from the end in a **single pass** (`O(sz)`). Below is the optimized, idiomatic Rust implementation:

---

### **Rust Implementation**

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    pub fn remove_nth_from_end(head: Option<Box<ListNode>>, n: i32) -> Option<Box<ListNode>> {
        let mut dummy = Some(Box::new(ListNode { val: 0, next: head }));
        let mut first = dummy.as_ref();
        let mut second = dummy.as_mut();

        // Move first pointer n+1 steps ahead
        for _ in 0..=n {
            if let Some(node) = first {
                first = node.next.as_ref();
            }
        }

        // Move both first and second pointers
        while let Some(node) = first {
            first = node.next.as_ref();
            second = second.unwrap().next.as_mut();
        }

        // Remove the nth node from the end
        if let Some(node) = second {
            node.next = node.next.take().and_then(|next_node| next_node.next);
        }

        dummy.unwrap().next
    }
}
```

---

## **Algorithm Explanation**

### **Core Concepts:**

- **Two-pointer technique:** One pointer (`first`) moves `n+1` steps ahead, ensuring that when it reaches the end, the other pointer (`second`) is exactly at the node **before** the node to be deleted.
- **Dummy node usage:** This simplifies edge cases, such as removing the first node.
- **Safe `Option<Box<ListNode>>` manipulation:** Uses `.take()` and `.and_then()` to avoid unnecessary clones and `unwrap()`.

### **Time Complexity:**

- **O(sz)**: We traverse the linked list at most **once**.

### **Space Complexity:**

- **O(1)**: Only a few extra pointers are used.
