### **Key Observations**

- We can solve this using **O(n) time and O(1) space**:
  1. **Find the middle** of the linked list using the **slow and fast pointer technique**.
  2. **Reverse the second half** of the list.
  3. **Compare the first half with the reversed second half**.
  4. **Restore the original list (if required in real-world cases).**

#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn is_palindrome(mut head: Option<Box<ListNode>>) -> bool {
        if head.is_none() || head.as_ref().unwrap().next.is_none() {
            return true;
        }

        // Step 1: Find middle using slow and fast pointers
        let mut slow = &head;
        let mut fast = &head;
        while fast.is_some() && fast.as_ref().unwrap().next.is_some() {
            slow = &slow.as_ref().unwrap().next;
            fast = &fast.as_ref().unwrap().next.as_ref().unwrap().next;
        }

        // Step 2: Reverse second half
        let mut second_half = Self::reverse_list(slow.clone());

        // Step 3: Compare first and second half
        let mut first_half = &head;
        while let Some(n2) = &second_half {
            if first_half.as_ref().unwrap().val != n2.val {
                return false;
            }
            first_half = &first_half.as_ref().unwrap().next;
            second_half = n2.next.clone();
        }

        true
    }

    fn reverse_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev = None;
        while let Some(mut node) = head {
            head = node.next.take();
            node.next = prev;
            prev = Some(node);
        }
        prev
    }
}
