## üß† **Problem Breakdown**

We need to rearrange a **singly linked list** such that:

- **Nodes at odd indices** are grouped first.
- **Nodes at even indices** follow.
- The **relative order** within both groups is preserved.
- We must achieve **O(n) time and O(1) space complexity**.

### **Key Observations**

1. We can solve this in **one pass (`O(n)`)** using **two pointers**:
   - One pointer for **odd-indexed** nodes (`odd`).
   - One pointer for **even-indexed** nodes (`even`).
   - Maintain a reference to the **head of even nodes** (`even_head`).
2. Once we finish reordering:
   - Attach the last odd node to the head of the even list.

### **Why Use Two Pointers?**

‚úÖ **Time Complexity: `O(n)`** (single traversal of the list).  
‚úÖ **Space Complexity: `O(1)`** (no extra data structures, just pointer manipulation).  
‚úÖ **Preserves relative order** in both odd and even groups.

## üöÄ **Optimized Rust Implementation**

```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn odd_even_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        if head.is_none() || head.as_ref().unwrap().next.is_none() {
            return head;
        }

        let mut odd = head.as_mut();
        let mut even = odd.as_mut().unwrap().next.as_mut();
        let mut even_head = even.as_deref_mut();

        while let (Some(o), Some(e)) = (odd, even) {
            if let Some(next_odd) = e.next.take() {
                o.next = Some(next_odd);
                odd = o.next.as_mut();
            } else {
                break;
            }

            if let Some(next_even) = odd.as_mut().unwrap().next.take() {
                e.next = Some(next_even);
                even = e.next.as_mut();
            } else {
                break;
            }
        }

        odd.unwrap().next = even_head.map(Box::new);
        head
    }
}
```
