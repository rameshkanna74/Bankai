### **Optimized Rust Solution for "Delete the Middle Node of a Linked List"**

This problem is efficiently solved using the **fast & slow pointer technique (two-pointer method)** in a **single pass** (`O(n)`). Below is the optimal and idiomatic Rust solution:

---

### **Rust Implementation**
```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    pub fn delete_middle(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        // Edge case: If there's only one node, return None
        if head.as_ref()?.next.is_none() {
            return None;
        }

        let mut dummy = Some(Box::new(ListNode { val: 0, next: head }));
        let mut slow = dummy.as_mut();
        let mut fast = slow.as_ref()?.next.as_ref();

        // Move fast pointer twice as fast as slow
        while let Some(next_fast) = fast {
            fast = next_fast.next.as_ref().and_then(|n| n.next.as_ref());
            slow = slow.unwrap().next.as_mut();
        }

        // Remove the middle node
        if let Some(node) = slow {
            node.next = node.next.take().and_then(|next_node| next_node.next);
        }

        dummy.unwrap().next
    }
}
```

## **Algorithm Explanation**
### **Core Concepts:**
- **Two-pointer technique (fast & slow pointers)**: 
  - `fast` moves **twice** as fast as `slow`, ensuring `slow` is at the node **before** the middle node when `fast` reaches the end.
  - This allows removal of the middle node in **one pass** (`O(n)`).
- **Dummy node usage**: 
  - Handles edge cases cleanly (e.g., deleting the first node).

### **Time Complexity:**
- **O(n)**: We traverse the linked list **only once**.

### **Space Complexity:**
- **O(1)**: Only a few extra pointers (`dummy`, `slow`, `fast`) are used.
