# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

## 🧠 **Problem Breakdown**
We need to find the **middle node** of a **singly linked list**:
- If the length is **odd**, return the **exact middle**.
- If the length is **even**, return the **second middle node**.

### **Key Observations**
1. **Optimal Approach (`O(n)`)**  
   - Use the **Two Pointer (Tortoise and Hare) approach**:
     - `slow` moves **one step** at a time.
     - `fast` moves **two steps** at a time.
     - When `fast` reaches the end, `slow` is at the middle.

2. **Why is this optimal?**  
   - **Single pass (`O(n)`)**, compared to alternative approaches that may scan the list twice.
   - **Constant space (`O(1)`)**, since we only use two pointers.

---

## ⏳ **Time and Space Complexity**
✅ **Time Complexity:** `O(n)`, since we traverse the linked list once.  
✅ **Space Complexity:** `O(1)`, as we use only two pointers.

## 🚀 **Optimized Rust Implementation**
```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }

    pub fn middle_node(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut slow = &head;
        let mut fast = &head;

        while let Some(fast_node) = fast {
            if let Some(next_fast) = &fast_node.next {
                fast = &next_fast.next;
                slow = &slow.as_ref().unwrap().next;
            } else {
                break;
            }
        }
        
        slow.clone()
    }
}
```

✅ **Uses references (`&`) to avoid unnecessary memory moves.**  
✅ **Avoids `unwrap()` or `expect()`, preventing panics.**  
✅ **Handles all edge cases gracefully.**
