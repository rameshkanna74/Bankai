## üß† **Problem Breakdown**

We need to **detect a cycle** in a singly linked list and **find the starting node of the cycle**.

### **Key Observations**

- We use **Floyd‚Äôs Cycle Detection Algorithm (Tortoise and Hare)** to first detect if a cycle exists.
- If a cycle is found, we use a **mathematical property** to find the starting node of the cycle:
  - When the slow and fast pointers meet inside the loop, we place another pointer at the head and move both at the same speed. The node where they meet again is the start of the cycle.

### **Why Floyd‚Äôs Algorithm?**

‚úÖ **Time Complexity:** `O(n)`, since both pointers traverse at most `n` nodes.  
‚úÖ **Space Complexity:** `O(1)`, as it uses only two pointers.  
‚úÖ **Most optimal** for cycle detection and finding its starting node.

## üöÄ **Optimized Rust Implementation**

```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn detect_cycle(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let (mut slow, mut fast) = (&head, &head);

        // Step 1: Detect cycle using Floyd's Algorithm
        while let Some(f_node) = fast {
            if let Some(f_next) = &f_node.next {
                slow = &slow.as_ref().unwrap().next;
                fast = &f_next.next;

                if slow == fast {
                    return Self::find_cycle_start(&head, slow);
                }
            } else {
                return None;
            }
        }
        None
    }

    fn find_cycle_start(
        mut head: &Option<Box<ListNode>>,
        mut meeting_point: &Option<Box<ListNode>>,
    ) -> Option<Box<ListNode>> {
        while head != meeting_point {
            head = &head.as_ref().unwrap().next;
            meeting_point = &meeting_point.as_ref().unwrap().next;
        }
        head.clone()
    }
}
```
