## 🧠 **Problem Breakdown**

We need to **detect a cycle in a singly linked list**.

- If a cycle exists, return `true`.
- Otherwise, return `false`.

### **Key Observations**

- **Floyd’s Cycle Detection Algorithm (Tortoise and Hare)**
  - Uses **two pointers**:
    - **Slow pointer (`tortoise`)** moves **one step** at a time.
    - **Fast pointer (`hare`)** moves **two steps** at a time.
  - If there is a cycle, they **will meet** inside the cycle.
  - If there is **no cycle**, `hare` will reach the end (`None`).

### **Why Floyd’s Algorithm?**

✅ **Time Complexity:** `O(n)`, since both pointers traverse at most `n` nodes.  
✅ **Space Complexity:** `O(1)`, as it uses only two pointers.  
✅ **Most optimal** for detecting cycles in a linked list.

## 🚀 **Optimized Rust Implementation**

```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn has_cycle(mut head: Option<Box<ListNode>>) -> bool {
        let (mut slow, mut fast) = (&head, &head);

        while let Some(f_node) = fast {
            if let Some(f_next) = &f_node.next {
                slow = &slow.as_ref().unwrap().next;
                fast = &f_next.next;
                if slow == fast {
                    return true;
                }
            } else {
                return false;
            }
        }
        false
    }
}
```
