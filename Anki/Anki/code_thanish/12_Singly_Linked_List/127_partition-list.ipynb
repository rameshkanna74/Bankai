## **Optimized Rust Solution for Partitioning a Linked List**

### **Approach**

We solve this problem optimally using **two separate linked lists**:

- One list (`less_head`) contains nodes with values **less than `x`**.
- Another list (`greater_head`) contains nodes with values **greater than or equal to `x`**.
- We then **connect** the two lists while preserving their relative order.

This ensures a **stable partitioning** in **O(n) time** and **O(1) extra space** (excluding the output list).

---

## **Optimal Rust Solution**

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    /// Partitions the linked list so that all nodes less than x come before nodes greater than or equal to x.
    pub fn partition(head: Option<Box<ListNode>>, x: i32) -> Option<Box<ListNode>> {
        let (mut less_dummy, mut greater_dummy) = (ListNode::new(0), ListNode::new(0));
        let (mut less, mut greater) = (&mut less_dummy, &mut greater_dummy);
        let mut current = head;

        while let Some(mut node) = current {
            current = node.next.take(); // Move node ownership, avoid unnecessary clones

            if node.val < x {
                less.next = Some(node);
                less = less.next.as_mut().unwrap();
            } else {
                greater.next = Some(node);
                greater = greater.next.as_mut().unwrap();
            }
        }

        // Connect less list to greater list
        less.next = greater_dummy.next;
        less_dummy.next
    }
}
```

## **Algorithm Explanation**

### **Core Concepts Used**

- **Two-Pointer Technique**: Two separate pointers (`less` and `greater`) manage two partitions.
- **Dummy Head Nodes**: Simplifies merging lists and avoids extra condition checks.
- **Ownership & Borrowing**: We use `.take()` to efficiently **move ownership** and avoid unnecessary copies.
