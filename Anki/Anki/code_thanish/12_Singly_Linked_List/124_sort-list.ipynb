### **Optimized Rust Solution for "Sort List" (LeetCode 148)**  

This problem is best solved using **Merge Sort** because it guarantees an **O(n log n) time complexity** and works well with linked lists due to its **efficient merging process**.

---

### **Rust Implementation (Merge Sort)**
```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    pub fn sort_list(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        // Base case: If the list is empty or has a single node, return as it is already sorted
        if head.is_none() || head.as_ref()?.next.is_none() {
            return head;
        }

        // Step 1: Split the list into two halves
        let (mut left, mut right) = Self::split_list(head);

        // Step 2: Recursively sort both halves
        left = Self::sort_list(left);
        right = Self::sort_list(right);

        // Step 3: Merge the sorted halves
        Self::merge_sorted_lists(left, right)
    }

    /// Splits the linked list into two halves and returns (left_half, right_half)
    fn split_list(mut head: Option<Box<ListNode>>) -> (Option<Box<ListNode>>, Option<Box<ListNode>>) {
        let mut slow = &mut head;
        let mut fast = &mut head;

        while fast.is_some() && fast.as_ref()?.next.is_some() {
            fast = &mut fast.as_mut()?.next.as_mut()?.next;
            slow = &mut slow.as_mut()?.next;
        }

        // Split the list into two halves
        let right_half = slow.as_mut()?.next.take();
        (head, right_half)
    }

    /// Merges two sorted linked lists into one sorted list
    fn merge_sorted_lists(
        l1: Option<Box<ListNode>>,
        l2: Option<Box<ListNode>>,
    ) -> Option<Box<ListNode>> {
        let mut dummy = ListNode::new(0);
        let mut tail = &mut dummy;

        let (mut l1, mut l2) = (l1, l2);

        while let (Some(mut node1), Some(mut node2)) = (l1, l2) {
            if node1.val < node2.val {
                l1 = node1.next.take();
                tail.next = Some(node1);
            } else {
                l2 = node2.next.take();
                tail.next = Some(node2);
            }
            tail = tail.next.as_mut().unwrap();
        }

        // Append remaining nodes
        tail.next = l1.or(l2);
        dummy.next
    }
}
```

## **Algorithm Explanation**
### **Core Concepts**
1. **Merge Sort (O(n log n))**
   - **Divide:** Recursively split the linked list into two halves.
   - **Conquer:** Sort each half recursively.
   - **Merge:** Combine the two sorted halves efficiently.
   
2. **Fast & Slow Pointer (Splitting List)**
   - The **slow pointer** moves one step at a time.
   - The **fast pointer** moves two steps at a time.
   - When the fast pointer reaches the end, the slow pointer is at the middle.

3. **Merge Two Sorted Lists (O(n))**
   - Uses a dummy node to track the merged result.
   - Compares elements from both lists and appends the smaller one.
