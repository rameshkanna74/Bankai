## **Optimized Rust Solution for Adding Two Numbers in Linked Lists**

### **Approach**

We solve this problem optimally using **element-wise addition with carry propagation**:

- Traverse both linked lists **simultaneously**.
- At each step, **add corresponding digits + carry** and store the result in a new linked list.
- If one list is shorter, treat missing digits as **0**.
- If there is a **carry after the last digit**, create an extra node.
- The sum is stored in **reverse order**, so we process nodes from **least significant to most significant**.

This ensures an **O(n) time complexity** and **O(1) extra space** (excluding output).

## **Optimal Rust Solution**

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    /// Adds two numbers represented as reversed linked lists.
    pub fn add_two_numbers(
        l1: Option<Box<ListNode>>,
        l2: Option<Box<ListNode>>,
    ) -> Option<Box<ListNode>> {
        let mut dummy_head = ListNode::new(0);
        let mut current = &mut dummy_head;
        let (mut l1, mut l2) = (l1, l2);
        let mut carry = 0;

        while l1.is_some() || l2.is_some() || carry > 0 {
            let sum = carry
                + l1.as_ref().map_or(0, |node| node.val)
                + l2.as_ref().map_or(0, |node| node.val);
            carry = sum / 10;

            // Append new node to result
            current.next = Some(Box::new(ListNode::new(sum % 10)));
            current = current.next.as_mut().unwrap();

            // Move to next nodes
            l1 = l1.and_then(|node| node.next);
            l2 = l2.and_then(|node| node.next);
        }

        dummy_head.next
    }
}
```
