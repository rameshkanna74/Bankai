# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

---

## 🧠 **Problem Breakdown**
We are given a **singly linked list** and a **key**. The goal is to **check if the key exists in the list**.

### **Key Observations**
1. **Naïve Approach (`O(n)`)**  
   - Traverse the list and compare each node's value to the key.
   - If found, return `true`; otherwise, return `false`.
   - This is **already optimal** (`O(n)`) since we must check each node.

2. **Optimal Approach (`O(n)`)**  
   - Iterate through the linked list in a **single pass**.
   - Use **efficient memory management** (`Box<T>`, borrowing).

---

## ⏳ **Time and Space Complexity**
✅ **Time Complexity:** `O(n)`, since we traverse the linked list once.  
✅ **Space Complexity:** `O(1)`, as we use no extra memory apart from a pointer.  

---

## 🚀 **Optimized Rust Implementation**
```rust
#[derive(PartialEq, Eq, Debug)]
struct ListNode {
    val: i32,
    next: Option<Box<ListNode>>,
}

impl ListNode {
    fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }

    fn search(head: Option<&Box<ListNode>>, key: i32) -> bool {
        let mut current = head;
        while let Some(node) = current {
            if node.val == key {
                return true;
            }
            current = node.next.as_ref();
        }
        false
    }
}
```

---

## 🔥 **Why is this the Best Solution?**
✅ **Runs in `O(n)`, the best possible for linked list traversal.**  
✅ **Uses `Option<Box<T>>` for safe, memory-managed linked list handling.**  
✅ **Avoids `unwrap()` or `expect()`, preventing panics.**  
✅ **Handles edge cases gracefully.**
