## üß† **Problem Breakdown**

We need to **detect a cycle (loop) in a singly linked list** and determine its **length**.

- If a loop is present, return the **number of nodes in the loop**.
- Otherwise, return `0`.

### **Key Observations**

- We use **Floyd‚Äôs Cycle Detection Algorithm (Tortoise and Hare)** to detect the loop.
- If a loop exists, the slow and fast pointers will **meet inside the loop**.
- To find the **length of the loop**, we keep moving one pointer inside the cycle until it returns to the meeting point.

### **Why Floyd‚Äôs Algorithm?**

‚úÖ **Time Complexity:** `O(n)`, since both pointers traverse at most `n` nodes.  
‚úÖ **Space Complexity:** `O(1)`, as it uses only two pointers.  
‚úÖ **Most optimal** for detecting cycles and measuring their lengths.

## üöÄ **Optimized Rust Implementation**

```rust
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    pub fn count_nodes_in_loop(mut head: Option<Box<ListNode>>) -> usize {
        let (mut slow, mut fast) = (&head, &head);

        // Detect cycle using Floyd's Algorithm
        while let Some(f_node) = fast {
            if let Some(f_next) = &f_node.next {
                slow = &slow.as_ref().unwrap().next;
                fast = &f_next.next;

                if slow == fast {
                    return Self::count_loop_length(slow);
                }
            } else {
                return 0;
            }
        }
        0
    }

    fn count_loop_length(mut node: &Option<Box<ListNode>>) -> usize {
        let mut count = 1;
        let start = node.clone();

        while let Some(n) = node {
            node = &n.next;
            count += 1;
            if node == &start {
                break;
            }
        }
        count
    }
}
```
