## ðŸ§  **Algorithm Explanation**

### **ðŸ”¹ Observations**

1. **Brute Force Approach (`O(m^(1/n))`)**:

   - Check all numbers `x` from `1` to `m`, verifying if `x^n == m`.
   - **Downside**: Inefficient for large `m`.

2. **Binary Search Approach (`O(log m)`)**:
   - The **nth root of `m` lies between `1` and `m`**.
   - **Steps**:
     1. Initialize `low = 1`, `high = m`.
     2. Compute `mid = (low + high) / 2`.
     3. Compute `mid^n` (handling overflow using `i64`).
     4. If `mid^n == m`, return `mid`.
     5. If `mid^n < m`, move **low** to `mid + 1`.
     6. If `mid^n > m`, move **high** to `mid - 1`.
     7. If no integer root is found, return `-1`.

- **Time Complexity**: **O(log m)** (Binary Search halves the range in each step).
- **Space Complexity**: **O(1)** (Only integer variables are used).

impl Solution {
    pub fn nth_root(n: i32, m: i32) -> i32 {
        let (mut low, mut high) = (1, m);

        while low <= high {
            let mid = low + (high - low) / 2;
            let power = Self::power(mid, n, m); // Compute mid^n safely

            if power == m as i64 {
                return mid;
            } else if power < m as i64 {
                low = mid + 1;
            } else {
                high = mid - 1;
            }
        }

        -1
    }

    fn power(base: i32, exp: i32, limit: i32) -> i64 {
        let mut result: i64 = 1;
        let base = base as i64;
        let limit = limit as i64;

        for _ in 0..exp {
            result *= base;
            if result > limit {
                return limit + 1; // Prevent overflow
            }
        }

        result
    }
}
