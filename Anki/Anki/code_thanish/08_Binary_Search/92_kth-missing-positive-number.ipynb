# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

---

## 🧠 **Problem Breakdown**

We need to find the **kth missing positive integer** from a sorted array `arr`.  
A **missing positive integer** is a number that is **not present in `arr` but should be in a strictly increasing sequence starting from 1**.

- **Example Walkthrough**:  
  Given `arr = [2,3,4,7,11]`, the missing numbers are `[1,5,6,8,9,10,12,13,...]`.  
  If `k = 5`, the answer is **9**.

- **Key Observations**:
  - If `arr = [1,2,3,4]`, no numbers are missing until `5`.
  - The number of missing integers **before `arr[i]`** is `arr[i] - (i + 1)`.
  - We need the first **k** missing numbers.

💡 **Binary search is ideal because `arr[i] - (i + 1)` is monotonically increasing.**

---

## 🚀 **Algorithm Explanation**

### **🔹 Approach 1: Linear Scan (O(n))**

1. Iterate through `arr` while tracking missing numbers.
2. If a number is missing, decrement `k`.
3. Stop when `k == 0`.

🔴 **Downside:** `O(n)` is acceptable for `n ≤ 1000`, but **binary search is faster**.

---

### **🔹 Approach 2: Binary Search (O(log n))**

1. **Binary search for the first index `mid` where**  
   `arr[mid] - (mid + 1) >= k`.

   - This means the `k`th missing number occurs **before `arr[mid]`**.
   - Otherwise, continue searching right.

2. **Compute the missing number** as:
   \[
   \text{kth missing} = k + \text{index found}
   \]

3. **Return the result**.

---

## ⏳ **Time and Space Complexity**

✅ **Binary Search:** `O(log n)`  
✅ **Space Complexity:** `O(1)` (no extra storage)

---

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_kth_positive(arr: Vec<i32>, k: i32) -> i32 {
        let (mut left, mut right) = (0, arr.len());

        while left < right {
            let mid = left + (right - left) / 2;
            if arr[mid] - (mid as i32 + 1) >= k {
                right = mid;
            } else {
                left = mid + 1;
            }
        }

        k + left as i32
    }
}
```
