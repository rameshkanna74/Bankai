   - Select `k2 = k - k1` elements from `arr2`.

## ðŸš€ **Algorithm Explanation**

### **ðŸ”¹ Approach: Binary Search on the Smaller Array**

1. **Ensure `arr1` is the smaller array (`n â‰¤ m`)** for optimization.
2. **Use Binary Search** on `arr1` to determine how many elements `k1` should come from `arr1`.
3. **Partitioning Logic:**
   - `k1 + k2 = k`, where `k2 = k - k1`
   - **Boundary conditions:**
     - If `k1 = 0` â†’ Take all `k` elements from `arr2`.
     - If `k1 = n` â†’ Take all `k` elements from `arr1`.
4. **Median Calculation:**
   - If `arr1[k1 - 1] â‰¤ arr2[k2]` and `arr2[k2 - 1] â‰¤ arr1[k1]`, return `max(arr1[k1 - 1], arr2[k2 - 1])`.
   - Adjust `left` or `right` pointers accordingly.

impl Solution {
    pub fn kth_element(arr1: &[i32], arr2: &[i32], k: usize) -> i32 {
        let (mut a, mut b) = if arr1.len() <= arr2.len() { (arr1, arr2) } else { (arr2, arr1) };
        let (n, m) = (a.len(), b.len());

        let (mut left, mut right) = (0, n);
        while left <= right {
            let mid_a = (left + right) / 2;
            let mid_b = k as isize - mid_a as isize; // mid_b = k - mid_a

            if mid_b < 0 {
                right = mid_a - 1;
                continue;
            }
            if mid_b as usize > m {
                left = mid_a + 1;
                continue;
            }

            let l1 = if mid_a > 0 { a[mid_a - 1] } else { i32::MIN };
            let r1 = if mid_a < n { a[mid_a] } else { i32::MAX };
            let l2 = if mid_b > 0 { b[mid_b as usize - 1] } else { i32::MIN };
            let r2 = if mid_b as usize < m { b[mid_b as usize] } else { i32::MAX };

            if l1 <= r2 && l2 <= r1 {
                return l1.max(l2);
            } else if l1 > r2 {
                right = mid_a - 1;
            } else {
                left = mid_a + 1;
            }
        }
        unreachable!()
    }
}
