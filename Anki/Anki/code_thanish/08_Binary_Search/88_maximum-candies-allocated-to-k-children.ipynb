## üß† **Algorithm Explanation**

### **üîπ Observations**

1. **Brute Force Approach (`O(n * max(candies))`)**:

   - Try every possible candy count from `1` to `max(candies)`.
   - Check if it's possible to allocate `k` children each at least `x` candies.
   - **Downside**: `max(candies)` can be `10‚Å∑`, leading to **inefficient** performance.

2. **Binary Search Approach (`O(n log max(candies))`)**:
   - **Lower Bound (`low = 1`)**: The smallest allocation.
   - **Upper Bound (`high = max(candies)`)**: The largest possible allocation.
   - **Binary Search on `x` (max candies per child)**:
     1. Compute `mid = (low + high) / 2` as a candidate number of candies per child.
     2. Check if we can allocate `k` children at least `mid` candies each.
     3. If possible, **increase `x`** (`low = mid + 1`).
     4. Otherwise, **reduce `x`** (`high = mid - 1`).
     5. Continue until `low` converges to the maximum valid `x`.

  - **Total worst-case operations** ‚âà `O(10‚Åµ log 10‚Å∑) ‚âà O(10‚Å∂)`, which is efficient.

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn maximum_candies(candies: Vec<i32>, k: i64) -> i32 {
        if candies.iter().map(|&c| c as i64).sum::<i64>() < k {
            return 0; // Not enough candies to allocate at least 1 per child
        }

        let (mut low, mut high) = (1, *candies.iter().max().unwrap());
        while low < high {
            let mid = (low + high + 1) / 2; // Ceiling division to prevent infinite loop
            if Self::can_allocate(&candies, k, mid) {
                low = mid; // Try larger allocations
            } else {
                high = mid - 1; // Reduce allocation
            }
        }
        low
    }

    fn can_allocate(candies: &[i32], k: i64, x: i32) -> bool {
        candies.iter().map(|&c| (c / x) as i64).sum::<i64>() >= k
    }
}
```
