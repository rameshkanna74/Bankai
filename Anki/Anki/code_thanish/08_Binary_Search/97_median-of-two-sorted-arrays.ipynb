## 🧠 **Problem Breakdown**

We need to find the **median** of two sorted arrays `nums1` and `nums2` in **O(log (m + n))** time complexity.

### **Key Observations**

1. **Median Definition:**

   - If the combined length is **odd**, the median is the middle element.
   - If the combined length is **even**, the median is the average of the two middle elements.

2. **Optimal Approach → Binary Search on the Shorter Array:**
   - Instead of merging both arrays (`O(m + n)`) → **Use Binary Search (`O(log min(m, n))`)**.
   - Partition one array (`nums1`), find its **complementary partition** in `nums2`.
   - Ensure left partition elements ≤ right partition elements.

---

## 🚀 **Algorithm Explanation**

### **🔹 Approach: Binary Search on the Smaller Array**

1. **Ensure `nums1` is the smaller array (`m ≤ n`)** to optimize search.
2. **Use Binary Search** on `nums1` to determine its partition index.
3. **Partitioning Logic:**
   - **Left Partition (`L1`, `L2`)**: Largest element in left part of each array.
   - **Right Partition (`R1`, `R2`)**: Smallest element in right part of each array.
   - Maintain **`L1 ≤ R2` and `L2 ≤ R1`** to ensure valid partition.
4. **Median Calculation:**
   - **Odd Length (`(m + n) % 2 == 1`)** → Return `min(R1, R2)`.
   - **Even Length (`(m + n) % 2 == 0`)** → Return `(max(L1, L2) + min(R1, R2)) / 2`.

✅ **Binary Search Complexity:** `O(log min(m, n))` (Optimal)  
✅ **Partition Check Complexity:** `O(1)`  
✅ **Overall Complexity:** `O(log(min(m, n)))`  
✅ **Space Complexity:** `O(1)` (No extra storage used)

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_median_sorted_arrays(nums1: Vec<i32>, nums2: Vec<i32>) -> f64 {
        let (mut a, mut b) = if nums1.len() <= nums2.len() { (nums1, nums2) } else { (nums2, nums1) };
        let (m, n) = (a.len(), b.len());
        let total = m + n;
        let half = (total + 1) / 2;

        let (mut left, mut right) = (0, m);
        while left <= right {
            let mid_a = (left + right) / 2;
            let mid_b = half - mid_a;

            let l1 = if mid_a > 0 { a[mid_a - 1] } else { i32::MIN };
            let r1 = if mid_a < m { a[mid_a] } else { i32::MAX };
            let l2 = if mid_b > 0 { b[mid_b - 1] } else { i32::MIN };
            let r2 = if mid_b < n { b[mid_b] } else { i32::MAX };

            if l1 <= r2 && l2 <= r1 {
                return if total % 2 == 0 {
                    (f64::from(l1.max(l2)) + f64::from(r1.min(r2))) / 2.0
                } else {
                    f64::from(l1.max(l2))
                };
            } else if l1 > r2 {
                right = mid_a - 1;
            } else {
                left = mid_a + 1;
            }
        }
        unreachable!()
    }
}
```
