## ðŸ§  **Problem Breakdown**

We have an array `arr` where each element represents the number of pages in a book.  
We must distribute `n` books among `m` students while ensuring:

1. Each student gets **at least one book**.
2. Books are assigned **contiguously**.
3. The **maximum number of pages assigned to any student is minimized**.

### **Key Observations**

- **Binary Search on the answer** (`max_pages`):
  - Minimum possible `max_pages = max(arr)`, since at least one student gets the largest book.
  - Maximum possible `max_pages = sum(arr)`, if one student gets all books.
- **Checking Feasibility with a Greedy approach**:
  - Given a `max_pages` threshold, check if all books can be allocated to `m` students while respecting this limit.

---

## ðŸš€ **Algorithm Explanation**

### **ðŸ”¹ Approach: Binary Search + Greedy Check**

1. **Binary Search on `max_pages`** from `max(arr)` to `sum(arr)`.
2. **Check feasibility (`can_allocate_books`)**:
   - Assign books to students one by one.
   - If adding a book exceeds `max_pages`, allocate a new student.
   - If students exceed `m`, return `false`.
3. **Binary search for the smallest feasible `max_pages`**.

use std::io::{self, BufRead};

fn can_allocate_books(arr: &[i32], m: usize, max_pages: i32) -> bool {
    let mut students = 1;
    let mut current_pages = 0;

    for &pages in arr {
        if current_pages + pages > max_pages {
            students += 1;
            current_pages = pages;
            if students > m {
                return false;
            }
        } else {
            current_pages += pages;
        }
    }
    true
}

fn min_pages_to_allocate(arr: Vec<i32>, m: usize) -> i32 {
    if arr.len() < m {
        return -1; // Not enough books to allocate
    }

    let (mut low, mut high) = (*arr.iter().max().unwrap(), arr.iter().sum());
    let mut best = high;

    while low <= high {
        let mid = low + (high - low) / 2;
        if can_allocate_books(&arr, m, mid) {
            best = mid;
            high = mid - 1;
        } else {
            low = mid + 1;
        }
    }

    best
}

fn main() {
    let stdin = io::stdin();

    let arr: Vec<i32> = lines.next().unwrap().unwrap()
        .split_whitespace()
        .map(|s| s.parse().unwrap())
        .collect();

    println!("{}", min_pages_to_allocate(arr, m));
}
