## 🧠 **Problem Breakdown**

We need to **split** an array `nums` into `k` **contiguous subarrays**, such that the **largest sum of any subarray is minimized**.

### **Key Observations**

- **Binary Search on the answer** (`max_sum`):
  - Minimum possible `max_sum = max(nums)`, as at least one subarray must include the largest element.
  - Maximum possible `max_sum = sum(nums)`, if one subarray contains all elements.
- **Checking Feasibility with a Greedy approach**:
  - Given a `max_sum` threshold, check if `nums` can be split into **at most `k` subarrays** while ensuring **no subarray sum exceeds `max_sum`**.

---

## 🚀 **Algorithm Explanation**

### **🔹 Approach: Binary Search + Greedy Check**

1. **Binary Search on `max_sum`** from `max(nums)` to `sum(nums)`.
2. **Check feasibility (`can_split`)**:
   - Assign elements to subarrays while keeping **running sum ≤ `max_sum`**.
   - If adding an element exceeds `max_sum`, start a new subarray.
   - If more than `k` subarrays are needed, return `false`.
3. **Binary search for the smallest feasible `max_sum`**.

---

## ⏳ **Time and Space Complexity**

✅ **Binary Search:** `O(log(sum(nums)))`  
✅ **Greedy check per iteration:** `O(n)`  
✅ **Overall Complexity:** `O(n log(sum(nums)))` (Optimal for `n ≤ 10^4`)

✅ **Space Complexity:** `O(1)` (No extra space, only a few variables)

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn split_array(nums: Vec<i32>, k: i32) -> i32 {
        fn can_split(nums: &[i32], k: usize, max_sum: i32) -> bool {
            let mut subarrays = 1;
            let mut current_sum = 0;

            for &num in nums {
                if current_sum + num > max_sum {
                    subarrays += 1;
                    current_sum = num;
                    if subarrays > k {
                        return false;
                    }
                } else {
                    current_sum += num;
                }
            }
            true
        }

        let (mut low, mut high) = (*nums.iter().max().unwrap(), nums.iter().sum());
        let mut best = high;

        while low <= high {
            let mid = low + (high - low) / 2;
            if can_split(&nums, k as usize, mid) {
                best = mid;
                high = mid - 1;
            } else {
                low = mid + 1;
            }
        }

        best
    }
}
```
