## üß† **Problem Breakdown**

We need to determine the **minimum ship capacity** required to ship all packages within `days` days while ensuring that:

- **Packages are loaded in order** (no reordering).
- **Each day, the total weight does not exceed the ship's capacity**.
- **All packages must be shipped in `days` days**.

## üöÄ **Algorithm Explanation**

### **üîπ Key Idea: Binary Search on Capacity**

We perform **Binary Search** on the ship's capacity to find the **minimum capacity that allows shipping all packages within `days` days**.

1. **Lower Bound (`low`)**: The **heaviest package** (since the ship must be able to carry at least the heaviest package).
2. **Upper Bound (`high`)**: The **sum of all packages** (if we used one day to ship all packages).
3. **Binary Search**:
   - **Pick `mid = (low + high) / 2`** as a candidate ship capacity.
   - **Check if it's possible to ship within `days` days** using this capacity:
     - Simulate loading the ship while keeping track of **current weight and days used**.
     - If more than `days` are required, **increase capacity (`low = mid + 1`)**.
     - Otherwise, **try a smaller capacity (`high = mid`)**.
4. **Stopping Condition**: `low == high`, which gives the **minimum required capacity**.

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn ship_within_days(weights: Vec<i32>, days: i32) -> i32 {
        let (mut low, mut high) = (
            *weights.iter().max().unwrap(), // Minimum possible capacity
            weights.iter().sum(),           // Maximum possible capacity
        );

        while low < high {
            let mid = (low + high) / 2;
            if Self::can_ship(&weights, days, mid) {
                high = mid;
            } else {
                low = mid + 1;
            }
        }

        low
    }

    fn can_ship(weights: &[i32], days: i32, capacity: i32) -> bool {
        let (mut current_weight, mut required_days) = (0, 1);

        for &weight in weights {
            if current_weight + weight > capacity {
                required_days += 1;
                current_weight = 0;
            }
            current_weight += weight;

            if required_days > days {
                return false;
            }
        }

        true
    }
}
```
