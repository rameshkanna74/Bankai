# **Optimized Rust Solution for Postfix to Infix Conversion**

## **Approach**

To convert a **postfix** expression to **infix**, we utilize a **stack-based approach**:

1. **Push operands** (letters or numbers) directly onto the stack.
2. **When encountering an operator**:
   - Pop the last **two** elements from the stack.
   - Form a new expression **surrounded by parentheses** (`(operand1 operator operand2)`).
   - Push the newly formed expression back onto the stack.
3. **At the end**, the stack will contain only **one fully parenthesized infix expression**.

---

## **Algorithm**

1. **Initialize an empty stack** of strings.
2. **Iterate over the characters** in the postfix expression:
   - If it's an **operand (letter or digit)**, push it onto the stack.
   - If it's an **operator (`+ - * / ^`)**:
     - Pop the **top two elements** from the stack.
     - Form the **infix expression**: `"(operand1 operator operand2)"`
     - Push the new expression back onto the stack.
3. **The last element in the stack** is the final **fully parenthesized** infix expression.

---

## **Complexity Analysis**

- **Time Complexity:**

  - We iterate through `n` characters once ‚Üí **O(n)**
  - Each character is pushed/popped once from the stack ‚Üí **O(n)**
  - **Overall: O(n)**

- **Space Complexity:**
  - Stack stores expressions ‚Üí **O(n)**
  - Each operator adds a pair of parentheses ‚Üí **O(n)**
  - **Overall: O(n)** (necessary due to parentheses requirement)

## **Rust Implementation**

```rust
use std::collections::VecDeque;

pub struct PostfixToInfix;

impl PostfixToInfix {
    // Function to convert postfix expression to infix
    pub fn convert(postfix: &str) -> String {
        let mut stack: VecDeque<String> = VecDeque::new();

        for ch in postfix.chars() {
            if ch.is_alphanumeric() {
                // Operand: push directly to stack
                stack.push_back(ch.to_string());
            } else {
                // Operator: Pop two operands and form a new infix expression
                if let (Some(op2), Some(op1)) = (stack.pop_back(), stack.pop_back()) {
                    let expr = format!("({}{}{})", op1, ch, op2);
                    stack.push_back(expr);
                }
            }
        }

        // The final result is the only element left in the stack
        stack.pop_back().unwrap_or_default()
    }
}
```
