# **Optimized Rust Solution for Infix to Postfix Conversion**

## **Approach**

We will use the **Shunting Yard Algorithm** developed by Edsger Dijkstra to convert infix notation to postfix. This algorithm efficiently uses a **stack** to ensure correct operator precedence and associativity.

---

## **Algorithm**

1. **Initialize an empty stack** for operators and an empty string for output (postfix notation).
2. **Scan the expression left to right**:
   - If it's an **operand (digit or letter)**, append it to the output.
   - If it's an **operator (+, -, \*, /, ^)**:
     - Pop operators from the stack to the output **while they have higher or equal precedence** (except `^` which is right-associative).
     - Push the current operator onto the stack.
   - If it's a **left parenthesis (`(`)**, push it onto the stack.
   - If it's a **right parenthesis (`)`)**, pop from the stack until a left parenthesis is encountered.
3. **Pop remaining operators** from the stack into the output.

---

## **Operator Precedence & Associativity**

| Operator | Precedence | Associativity |
| -------- | ---------- | ------------- |
| `^`      | 3          | Right         |
| `* /`    | 2          | Left          |
| `+ -`    | 1          | Left          |

## **Complexity Analysis**

- **Time Complexity:**

  - We **iterate once** over `n` characters â†’ **O(n)**
  - Each character is **pushed and popped once from the stack** â†’ **O(n)**
  - **Overall: O(n)**

- **Space Complexity:**
  - Stack stores operators in worst case **O(n)**.
  - Output string uses **O(n)** space.
  - **Overall: O(n)** (efficient)

## **Rust Implementation**

```rust
use std::collections::VecDeque;

pub struct InfixToPostfix;

impl InfixToPostfix {
    // Function to convert infix expression to postfix
    pub fn convert(expression: &str) -> String {
        let mut output = String::new();
        let mut stack: VecDeque<char> = VecDeque::new();

        let precedence = |op: char| -> i32 {
            match op {
                '^' => 3,
                '*' | '/' => 2,
                '+' | '-' => 1,
                _ => 0,
            }
        };

        let is_left_associative = |op: char| -> bool {
            match op {
                '^' => false, // Right-associative
                _ => true, // Left-associative
            }
        };

        for ch in expression.chars() {
            if ch.is_alphanumeric() {
                // If operand, add directly to output
                output.push(ch);
            } else if ch == '(' {
                // Left parenthesis -> push to stack
                stack.push_back(ch);
            } else if ch == ')' {
                // Right parenthesis -> pop until '('
                while let Some(&top) = stack.back() {
                    if top == '(' {
                        break;
                    }
                    output.push(stack.pop_back().unwrap());
                }
                stack.pop_back(); // Remove '('
            } else {
                // Operator: Pop higher or equal precedence operators from stack
                while let Some(&top) = stack.back() {
                    if precedence(top) > precedence(ch)
                        || (precedence(top) == precedence(ch) && is_left_associative(ch))
                    {
                        output.push(stack.pop_back().unwrap());
                    } else {
                        break;
                    }
                }
                stack.push_back(ch);
            }
        }

        // Pop remaining operators
        while let Some(op) = stack.pop_back() {
            output.push(op);
        }

        output
    }
}
```
