### **Key Observations**

1. **NaÃ¯ve Approach (`O(n^2)`)**

   - Checking all substrings and counting unique characters leads to **brute force (`O(n^2)`)**.
   - Too slow for large constraints (`n â‰¤ 10^5`).

2. **Optimal Approach (`O(n)`) â€“ Sliding Window + HashMap**
   - Use a **sliding window** (two-pointer approach) with a **HashMap** to track character frequencies.
   - Expand the window until we exceed `k` unique characters.
   - Shrink from the left until we are back to `k` unique characters.
   - Keep track of the maximum valid window length.

```rust
use std::collections::HashMap;

fn longest_substring_with_k_unique(s: &str, k: usize) -> isize {
    if k == 0 { return -1; }

    let mut char_count = HashMap::new();
    let (mut left, mut max_length) = (0, -1);

    for (right, ch) in s.chars().enumerate() {
        *char_count.entry(ch).or_insert(0) += 1;

        while char_count.len() > k {
            let left_char = s.chars().nth(left).unwrap();
            if let Some(count) = char_count.get_mut(&left_char) {
                *count -= 1;
                if *count == 0 {
                    char_count.remove(&left_char);
                }
            }
            left += 1;
        }

        if char_count.len() == k {
            max_length = max_length.max((right - left + 1) as isize);
        }
    }

    max_length
}
```
