n == t.length

s and t consist of uppercase and lowercase English letters.

# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

---

## 🧠 **Problem Breakdown**
We need to **find the minimum contiguous substring in `s` that contains all characters in `t` (including duplicates).**  

### **Key Observations**
1. **Naïve Approach (`O(m * n)`)**  
   - Checking all substrings leads to **brute force (`O(m * n)`)**, which is too slow.

2. **Optimal Approach (`O(m + n)`) – Sliding Window + HashMap**  
   - Use a **sliding window** (two-pointer approach) to dynamically adjust the substring.
   - Track required character **frequencies** with a **HashMap**.
   - Expand the right pointer until all characters in `t` are present.
   - Once valid, **move the left pointer** to shrink the window while still maintaining all required characters.
   - Keep track of the **smallest valid window**.

---

## ⏳ **Time and Space Complexity**
✅ **Time Complexity:** `O(m + n)`, as each character is processed at most twice.  
✅ **Space Complexity:** `O(128) = O(1)`, since we store at most 128 ASCII characters.

## 🚀 **Optimized Rust Implementation**
```rust
use std::collections::HashMap;

impl Solution {
    pub fn min_window(s: String, t: String) -> String {
        if t.is_empty() || s.is_empty() {
            return "".to_string();
        }

        let (mut left, mut right, mut min_start, mut min_len) = (0, 0, 0, usize::MAX);
        let mut required_chars = HashMap::new();
        let mut window_chars = HashMap::new();
        let mut formed = 0;
        
        for ch in t.chars() {
            *required_chars.entry(ch).or_insert(0) += 1;
        }
        
        let required_count = required_chars.len();

        let s_chars: Vec<char> = s.chars().collect();

        while right < s_chars.len() {
            let ch = s_chars[right];
            *window_chars.entry(ch).or_insert(0) += 1;

            if required_chars.contains_key(&ch) && window_chars[&ch] == required_chars[&ch] {
                formed += 1;
            }

            while formed == required_count {
                if right - left + 1 < min_len {
                    min_len = right - left + 1;
                    min_start = left;
                }

                let left_ch = s_chars[left];
                *window_chars.entry(left_ch).or_insert(0) -= 1;

                if required_chars.contains_key(&left_ch) && window_chars[&left_ch] < required_chars[&left_ch] {
                    formed -= 1;
                }

                left += 1;
            }

            right += 1;
        }

        if min_len == usize::MAX {
            "".to_string()
        } else {
            s[min_start..min_start + min_len].to_string()
        }
    }
}
```
