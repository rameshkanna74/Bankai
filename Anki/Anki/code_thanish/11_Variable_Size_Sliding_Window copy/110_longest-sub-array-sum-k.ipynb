# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

## 🧠 **Problem Breakdown**

We need to **find the longest contiguous subarray** whose sum equals a given target `k`.

### **Key Observations**

1. **Naïve Approach (`O(n^2)`)**

   - Checking all possible subarrays leads to a brute-force approach with **`O(n^2)` complexity**.
   - Too slow for large constraints.

2. **Optimal Approach (`O(n)`) – Prefix Sum + HashMap**
   - Use a **prefix sum and a HashMap** to track seen sums.
   - If `prefix_sum - k` exists in the HashMap, it means a valid subarray of sum `k` exists.
   - The difference between indices gives the **length** of this subarray.

---

## ⏳ **Time and Space Complexity**

✅ **Time Complexity:** `O(n)`, as we traverse the array once.  
✅ **Space Complexity:** `O(n)`, as we store at most `n` prefix sums in the HashMap.

---

## 🚀 **Optimized Rust Implementation**

```rust
use std::collections::HashMap;

fn longest_subarray_with_sum_k(arr: &[i32], k: i32) -> usize {
    let mut prefix_sum = 0;
    let mut longest_length = 0;
    let mut sum_map: HashMap<i32, usize> = HashMap::new();

    for (i, &num) in arr.iter().enumerate() {
        prefix_sum += num;

        // If prefix_sum itself is k, update longest_length
        if prefix_sum == k {
            longest_length = i + 1;
        }

        // If prefix_sum - k exists, update longest_length
        if let Some(&prev_index) = sum_map.get(&(prefix_sum - k)) {
            longest_length = longest_length.max(i - prev_index);
        }

        // Store the earliest occurrence of each prefix sum
        sum_map.entry(prefix_sum).or_insert(i);
    }

    longest_length
}
```

## 🔥 **Why is this the Best Solution?**

✅ **Runs in `O(n)`, optimal for large constraints (`n = 10^6`).**  
✅ **Uses prefix sum and HashMap for efficient lookups.**  
✅ **Handles negative numbers and zero correctly.**  
✅ **Only stores one occurrence of each prefix sum (avoids overwriting early indices).**
