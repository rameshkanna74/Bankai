# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

---

## 🧠 **Problem Breakdown**

We need to **find the longest contiguous substring without repeating characters**.

### **Key Observations**

1. **Naïve Approach (`O(n^2)`)**

   - Checking all substrings and verifying uniqueness leads to **brute force (`O(n^2)`)**, which is too slow.

2. **Optimal Approach (`O(n)`) – Sliding Window + HashMap**
   - Use a **sliding window** (two-pointer approach) with a **HashMap** to track the last index of characters.
   - Expand the window while characters are unique.
   - If a character repeats, **move the left pointer** to exclude the previous occurrence.
   - Keep track of the maximum window size.

---

## ⏳ **Time and Space Complexity**

✅ **Time Complexity:** `O(n)`, as each character is processed at most twice.  
✅ **Space Complexity:** `O(128) = O(1)`, since we store at most 128 ASCII characters.

---

## 🚀 **Optimized Rust Implementation**

```rust
use std::collections::HashMap;

impl Solution {
    pub fn length_of_longest_substring(s: String) -> i32 {
        let mut char_index = HashMap::new();
        let (mut left, mut max_length) = (0, 0);

        for (right, ch) in s.chars().enumerate() {
            if let Some(&prev_index) = char_index.get(&ch) {
                left = left.max(prev_index + 1); // Move left to avoid repeating char
            }

            char_index.insert(ch, right);
            max_length = max_length.max(right - left + 1);
        }

        max_length as i32
    }
}
```
