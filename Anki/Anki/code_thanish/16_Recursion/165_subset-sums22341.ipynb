## **Optimized Rust Solution for "Subset Sums"**  

### **Algorithm Overview**  
The problem requires computing **all possible subset sums** from a given array. Since a subset can either include or exclude each element, this is a classic **Power Set / Subset Sum** problem.  

### **Optimal Approach: Backtracking (DFS)**
- **Two choices per element**: Include it or exclude it.
- **Recursively explore both cases**.
- **Time complexity** is **\(O(2^N)\)** (since each element creates two branches).
- **Sorting is unnecessary** (output can be in any order).

---

## **Time & Space Complexity**
- **Time Complexity**: **\( O(2^N) \)**  
  - Since there are `N` elements, each can either be included or not ‚Üí **\( 2^N \) subsets**.
  - `N ‚â§ 15`, so max operations ‚âà **\( 32,768 \) (manageable)**.
  
- **Space Complexity**: **\( O(2^N) \) (for storing results)**  
  - Recursion depth is **\( O(N) \)**.

This solution is **clean, efficient, and interview-ready**.

```rust
impl Solution {
    pub fn subset_sums(arr: Vec<i32>) -> Vec<i32> {
        let mut result = Vec::new();
        Self::backtrack(&arr, 0, 0, &mut result);
        result
    }

    /// Recursive backtracking function to compute all subset sums.
    fn backtrack(arr: &Vec<i32>, index: usize, current_sum: i32, result: &mut Vec<i32>) {
        if index == arr.len() {
            result.push(current_sum);
            return;
        }

        // Exclude current element
        Self::backtrack(arr, index + 1, current_sum, result);

        // Include current element
        Self::backtrack(arr, index + 1, current_sum + arr[index], result);
    }
}
```

## **Algorithm Breakdown**
### **Recursive Backtracking**
1. **Base Case**:
   - If `index == arr.len()`, push the `current_sum` to results.

2. **Recursive Case**:
   - **Exclude current element**: Don't add `arr[index]` to sum.
   - **Include current element**: Add `arr[index]` to sum.

3. **Order doesn't matter** ‚Üí No need to sort.

---

### **Edge Cases Considered**
‚úÖ **Minimal Input** (`arr = []`) ‚Üí Returns `[0]` (empty subset sum).  
‚úÖ **Single Element** (`arr = [5]`) ‚Üí Returns `[0, 5]`.  
‚úÖ **All Positive Numbers** (`arr = [1,2,3]`) ‚Üí Computes `2^3 = 8` sums.
