# **Optimized Rust Solution for "Remove Invalid Parentheses" (BFS + Pruning)**  

## **Algorithm Overview**  
Given a string `s` with **letters** and **parentheses**, we need to remove the **minimum number of invalid parentheses** to make it valid. We return all possible valid strings.

### **Optimal Approach: Breadth-First Search (BFS)**
1. **Start with the given string `s`**  
   - If it's valid, return it.
   - Otherwise, try **removing one parenthesis at a time**.

2. **Use BFS to explore all possible removals**  
   - Avoid revisiting the same state (using a **HashSet**).
   - Stop as soon as we find **valid expressions** at the current depth.

3. **Pruning to improve efficiency**
   - **Early exit:** Stop exploring deeper when valid expressions are found.
   - **Skip duplicate states:** Use **HashSet** to track visited states.

---

## **Time & Space Complexity Analysis**
### **Time Complexity: O(2^n) (Worst Case)**
- In the worst case, we may generate all `2^n` subsets of the string.
- However, **pruning and early stopping** significantly reduce unnecessary states.

### **Optimized Complexity: ~O(n * 2^d) (With BFS Pruning)**
- `d` = minimum removals required, which is much smaller than `n`.

### **Space Complexity: O(n * 2^d)**
- **O(n)** per state (string length).
- **O(2^d)** states stored in HashSet.

```rust
use std::collections::{HashSet, VecDeque};

impl Solution {
    pub fn remove_invalid_parentheses(s: String) -> Vec<String> {
        let mut result = Vec::new();
        let mut queue = VecDeque::new();
        let mut visited = HashSet::new();
        let mut found = false;

        queue.push_back(s.clone());
        visited.insert(s.clone());

        while let Some(curr) = queue.pop_front() {
            if Solution::is_valid(&curr) {
                result.push(curr.clone());
                found = true;
            }
            
            if found {
                continue; // Stop generating deeper levels once valid results are found
            }

            for i in 0..curr.len() {
                if curr.chars().nth(i).unwrap() != '(' && curr.chars().nth(i).unwrap() != ')' {
                    continue;
                }
                
                let next = format!("{}{}", &curr[..i], &curr[i + 1..]);

                if visited.insert(next.clone()) {
                    queue.push_back(next);
                }
            }
        }

        result
    }

    fn is_valid(s: &str) -> bool {
        let mut balance = 0;
        for c in s.chars() {
            match c {
                '(' => balance += 1,
                ')' => {
                    if balance == 0 {
                        return false;
                    }
                    balance -= 1;
                }
                _ => {}
            }
        }
        balance == 0
    }
}
```
