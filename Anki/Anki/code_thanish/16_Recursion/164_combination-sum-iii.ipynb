## **Optimized Rust Solution for "Combination Sum III"**

### **Algorithm Overview**

This problem requires finding **unique combinations** of `k` numbers (1-9) that sum up to `n`.

- Each number **can be used at most once**.
- The combination must be **exactly `k` elements long**.

### **Optimal Approach: Backtracking with Pruning**

- **Sort is unnecessary** (numbers are naturally ordered from 1-9).
- **Prune recursion** when:
  - The sum exceeds `n`.
  - We exceed `k` numbers.
- **Each recursive call considers the next unused number**.

---

### **Time & Space Complexity**

- **Time Complexity**: **\( O(2^9) \)** ‚Üí **(Pruned Backtracking)**

  - Without pruning, worst case is **\( O(2^9) \) (~512 recursive calls)**.
  - But pruning **drastically reduces** calls by stopping early.

- **Space Complexity**: **\( O(k) \) (Recursion Depth) + \( O(K!) \) (Result Storage)**
  - Max recursion depth is `k`.
  - At most `O(K!)` valid combinations.

impl Solution {
    pub fn combination_sum3(k: i32, n: i32) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let mut current = Vec::new();
        Self::backtrack(k, n, 1, &mut current, &mut result);
        result
    }

    /// Backtracking function to find all valid combinations.
    fn backtrack(
        k: i32,
        n: i32,
        start: i32,
        current: &mut Vec<i32>,
        result: &mut Vec<Vec<i32>>
    ) {
        if n == 0 && current.len() as i32 == k {
            result.push(current.clone());
            return;
        }

        for i in start..=9 {
            if n - i < 0 || current.len() as i32 >= k {
                break; // Prune search
            }

            current.push(i);
            Self::backtrack(k, n - i, i + 1, current, result);
            current.pop(); // Backtrack step
        }
    }
}

## **Algorithm Breakdown**

### **Backtracking with Pruning**

1. **Recursive Exploration**:
   - Start from `1`, **try adding numbers** up to `9`.
   - Keep track of the **remaining sum (`n`)** and **remaining slots (`k`)**.
2. **Base Case**:
   - If `n == 0` and `k numbers are used`, **add combination to result**.
   - If `n < 0` or `k numbers exceed`, **prune** the recursion.
3. **Pruning for Efficiency**:
   - **If `n - i < 0`, stop further recursion** (no valid sums beyond this).
   - **If `current.len() >= k`, stop adding more numbers**.
