## **Optimized Rust Solution for Merge Sort**

### **Algorithm Overview**

Merge Sort is a **Divide & Conquer** sorting algorithm that:

1. **Splits** the array into halves recursively until each sub-array contains a single element.
2. **Merges** the sorted halves back together efficiently.

**Key Properties:**

- **Stable Sort** (Maintains relative order of equal elements).
- **Guaranteed O(N log N) time complexity**, making it efficient for large inputs.
- **O(N) auxiliary space complexity** due to additional storage for merging.

---

## **Time & Space Complexity**

- **Time Complexity:** **\( O(N \log N) \)**

  - **Divide Step:** Splitting takes **\( O(\log N) \)**.
  - **Merge Step:** Merging takes **\( O(N) \)** at each level.
  - **Total:** \( O(N \log N) \), which is optimal for comparison-based sorting.

- **Space Complexity:** **\( O(N) \)**
  - Uses **extra space for merging**, making it less space-efficient than **in-place QuickSort (O(1) space)**.

impl Solution {
    pub fn merge_sort(arr: &mut Vec<i32>) {
        if arr.len() <= 1 {
            return; // Base case: already sorted
        }

        let mid = arr.len() / 2;
        let left = &mut arr[..mid].to_vec(); // Copy left half
        let right = &mut arr[mid..].to_vec(); // Copy right half

        Self::merge_sort(left);
        Self::merge_sort(right);
        Self::merge(left, right, arr);
    }

    fn merge(left: &Vec<i32>, right: &Vec<i32>, arr: &mut Vec<i32>) {
        let (mut i, mut j, mut k) = (0, 0, 0);

        // Merge sorted subarrays
        while i < left.len() && j < right.len() {
            if left[i] <= right[j] {
                arr[k] = left[i];
                i += 1;
            } else {
                arr[k] = right[j];
                j += 1;
            }
            k += 1;
        }

        // Copy any remaining elements
        while i < left.len() {
            arr[k] = left[i];
            i += 1;
            k += 1;
        }
        while j < right.len() {
            arr[k] = right[j];
            j += 1;
            k += 1;
        }
    }
}

## **Algorithm Breakdown**

### **Merge Sort Steps**

1. **Base Case:** If `arr.len() <= 1`, return (trivially sorted).
2. **Recursive Case:**
   - Split array into `left` and `right` halves.
   - Recursively sort both halves.
   - Merge them back into a sorted sequence.

### **Merge Function**

- **Two-pointer technique** efficiently merges sorted halves in **O(N)**.
- **Preserves stability** (if `left[i] == right[j]`, take from `left` first).
