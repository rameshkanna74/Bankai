### **Optimized Rust Solution for Tower of Hanoi**

#### **Algorithm Overview**

The Tower of Hanoi is a classic **recursive problem** in **divide and conquer**. Given `n` disks and three pegs (source, auxiliary, target), our goal is to move all `n` disks from the source peg to the target peg using the auxiliary peg, following these rules:

1. Only one disk can be moved at a time.
2. A disk can only be placed on an empty peg or a larger disk.
3. The minimum number of moves required is **\(2^n - 1\)**.

#### **Optimal Time & Space Complexity**

- **Time Complexity**: **O(2^n)**
  - The recurrence relation follows \(T(n) = 2T(n-1) + 1\), which solves to \(T(n) = 2^n - 1\).
  - This is the best we can do, as every move is necessary.
- **Space Complexity**: **O(n)**
  - The recursion depth reaches **\(O(n)\)** due to function call stack usage.
  - No additional data structures are used beyond function calls.

/// Solves the Tower of Hanoi problem and returns the sequence of moves.
fn tower_of_hanoi(n: u32, source: char, target: char, auxiliary: char) -> Vec<(char, char)> {
    let mut moves = Vec::new();
    solve_hanoi(n, source, target, auxiliary, &mut moves);
    moves
}

/// Recursive helper function that populates the moves vector.
fn solve_hanoi(n: u32, source: char, target: char, auxiliary: char, moves: &mut Vec<(char, char)>) {
    if n == 0 {
        return;
    }

    // Move n-1 disks from source to auxiliary, using target as auxiliary
    solve_hanoi(n - 1, source, auxiliary, target, moves);

    // Move the nth disk from source to target
    moves.push((source, target));

    // Move the n-1 disks from auxiliary to target, using source as auxiliary
    solve_hanoi(n - 1, auxiliary, target, source, moves);
}
