# **Optimized Rust Solution for Sudoku Solver (Backtracking + Bitmasking)**

## **Algorithm Overview**

The **Sudoku Solver** problem requires filling an incomplete **9x9 board** while ensuring:

- Each digit **(1-9)** appears **once per row**.
- Each digit **(1-9)** appears **once per column**.
- Each digit **(1-9)** appears **once per 3x3 subgrid**.

### **Optimized Approach**

1. **Backtracking with Constraint Propagation**

   - Try placing digits (`1-9`) in empty cells (`'.'`).
   - Use **early pruning** by skipping invalid placements.
   - If a configuration leads to a dead-end, backtrack.

2. **Bitmasking for Fast Validity Checks**
   - Maintain three **bitsets** (`u16`) for **rows, columns, and 3x3 subgrids**.
   - Instead of iterating through rows/columns, check validity in **O(1)**.
   - Reduce the number of recursive calls.

### **Time Complexity**

- **Worst-case:** **O(9‚Å∏)** (Trying all placements).
- **Optimized pruning and bitmasking** improve real-world performance significantly.
- **Average case:** **Much faster than naive approaches**.

### **Space Complexity**

- **O(1)** extra space (excluding input board).
- Uses **O(27) ‚âà O(1)** for tracking row, column, and grid constraints.

```rust
impl Solution {
    pub fn solve_sudoku(board: &mut Vec<Vec<char>>) {
        let mut rows = [0u16; 9];
        let mut cols = [0u16; 9];
        let mut grids = [0u16; 9];

        // Initialize bitmasks for pre-filled cells
        for r in 0..9 {
            for c in 0..9 {
                if let Some(d) = board[r][c].to_digit(10) {
                    let bit = 1 << (d - 1);
                    let grid_idx = (r / 3) * 3 + (c / 3);
                    rows[r] |= bit;
                    cols[c] |= bit;
                    grids[grid_idx] |= bit;
                }
            }
        }

        Self::backtrack(board, &mut rows, &mut cols, &mut grids, 0, 0);
    }

    fn backtrack(
        board: &mut Vec<Vec<char>>,
        rows: &mut [u16; 9],
        cols: &mut [u16; 9],
        grids: &mut [u16; 9],
        r: usize,
        c: usize,
    ) -> bool {
        if r == 9 {
            return true;
        }

        let (next_r, next_c) = if c == 8 { (r + 1, 0) } else { (r, c + 1) };

        if board[r][c] != '.' {
            return Self::backtrack(board, rows, cols, grids, next_r, next_c);
        }

        let grid_idx = (r / 3) * 3 + (c / 3);
        for num in 1..=9 {
            let bit = 1 << (num - 1);
            if (rows[r] & bit) == 0 && (cols[c] & bit) == 0 && (grids[grid_idx] & bit) == 0 {
                // Place number
                board[r][c] = (b'0' + num as u8) as char;
                rows[r] |= bit;
                cols[c] |= bit;
                grids[grid_idx] |= bit;

                if Self::backtrack(board, rows, cols, grids, next_r, next_c) {
                    return true;
                }

                // Backtrack
                board[r][c] = '.';
                rows[r] &= !bit;
                cols[c] &= !bit;
                grids[grid_idx] &= !bit;
            }
        }

        false
    }
}
```

## **How the Algorithm Works**

### **Bitmask Representation**

We use **bitmasks (`u16`)** for:

- `rows[i]` ‚Üí Tracks digits **in row i**.
- `cols[j]` ‚Üí Tracks digits **in column j**.
- `grids[k]` ‚Üí Tracks digits **in 3x3 grid k**.

Using **bitwise AND (`&`)**, we can check digit validity in **O(1)**.

### **Backtracking Flow**

1. **Find the next empty cell (`'.'`).**
2. **Try placing numbers (`1-9`).**
3. **Use bitmasking for O(1) validity check.**
4. **Recursively solve the board.**
5. **If a configuration fails, backtrack.**
6. **Terminate once a valid solution is found.**
