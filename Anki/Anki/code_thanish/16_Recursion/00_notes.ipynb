````python
import re

def format_title(title):
    """Convert a title to a URL-friendly format"""
    title = title.lower().strip()
    title = re.sub(r'[^a-z0-9\s-]', '', title)  # Remove special characters
    title = re.sub(r'\s+', '-', title)  # Replace spaces with hyphens
    return title

def generate_problem_dict(start_number, data):
    """Generate a dictionary mapping numbers to formatted problem names"""
    problems = data.strip().split("\n")
    output_dict = {str(start_number + i): format_title(problem) for i, problem in enumerate(problems)}
    return output_dict

# Input
start_number = 206
data = """
lowest-common-ancestor-of-a-binary-tree
maximum-width-of-binary-tree
minimum-time-to-collect-all-apples-in-a-tree
count-complete-tree-nodes
construct-binary-tree-from-preorder-and-inorder-traversal
construct-binary-tree-from-inorder-and-postorder-traversal
serialize-and-deserialize-binary-tree
flatten-binary-tree-to-linked-list
"""

# Generate Output
output = generate_problem_dict(start_number, data)
print(output)
````

````python
import os
import json


def process_ipynb(file_path):
    """Remove all cells except the first one in a Jupyter Notebook file."""
    try:
        # Check if file is empty
        if os.path.getsize(file_path) == 0:
            print(f"⚠️ Skipping empty file: {file_path}")
            return

        # Load JSON safely
        with open(file_path, "r", encoding="utf-8") as f:
            notebook = json.load(f)

        if "cells" in notebook and len(notebook["cells"]) > 1:
            notebook["cells"] = [notebook["cells"][0]]  # Keep only the first cell

            with open(file_path, "w", encoding="utf-8") as f:
                json.dump(notebook, f, indent=2, ensure_ascii=False)

            print(f"✅ Processed: {file_path}")

    except json.JSONDecodeError:
        print(f"❌ Skipping corrupt file (not valid JSON): {file_path}")
    except Exception as e:
        print(f"❌ Error processing {file_path}: {e}")


def process_folder(folder_path):
    """Recursively process all .ipynb files in a folder."""
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".ipynb"):
                process_ipynb(os.path.join(root, file))


# if __name__ == "__main__":
#     folder_name = input("Enter the folder path: ").strip()


folder_name = "/home/aurexus/Music/py_pytorch/drf/Anki/rust_learn_anki_deck"

if os.path.isdir(folder_name):
    process_folder(folder_name)
    print("\n✅ All notebooks processed successfully!")
else:
    print("❌ Error: Folder does not exist.")
````
