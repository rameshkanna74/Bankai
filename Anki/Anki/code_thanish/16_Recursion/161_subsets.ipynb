### **Optimized Rust Solution for "Subsets" (Power Set Problem)**

#### **Algorithm Overview**

The problem requires generating all subsets (the power set) of a given unique integer array `nums`.  
The best approaches are **backtracking (DFS)** or **bitwise generation**.

#### **Optimal Time & Space Complexity**

- **Time Complexity**: **\( O(2^n \cdot n) \)**

  - Each element **has two choices**: either be included or not in a subset ‚Üí \( 2^n \) total subsets.
  - Constructing each subset takes \( O(n) \) time.

- **Space Complexity**: **\( O(2^n \cdot n) \)** (output) + **\( O(n) \)** (recursion depth)
  - **Output storage** requires \( O(2^n \cdot n) \), since we store all subsets.
  - **Recursion depth is \( O(n) \)**, which is manageable given \( n \leq 10 \).

## **Optimized Rust Code**

This solution is **idiomatic, efficient, and FAANG-level interview-ready**.

```rust
impl Solution {
    pub fn subsets(nums: Vec<i32>) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let mut current = Vec::new();
        Self::backtrack(&nums, 0, &mut current, &mut result);
        result
    }

    /// Backtracking function to generate subsets.
    fn backtrack(nums: &Vec<i32>, index: usize, current: &mut Vec<i32>, result: &mut Vec<Vec<i32>>) {
        result.push(current.clone());

        for i in index..nums.len() {
            current.push(nums[i]);
            Self::backtrack(nums, i + 1, current, result);
            current.pop(); // Backtrack step
        }
    }
}
```

### **Algorithm Breakdown**

#### **Backtracking Approach**

- **Recursively build subsets** by including or excluding elements.
- **Base case**: Push `current` subset to `result` when `index == nums.len()`.
- **For loop (DFS traversal)**:
  - **Include `nums[i]`**, recurse, then backtrack (remove last element).

## **Alternative: Iterative Bitwise Approach**

Instead of recursion, we can generate subsets using **bitwise masks**.

### **Time Complexity: \( O(2^n \cdot n) \)**

- Each subset corresponds to a **binary representation** of length `n`.
- Example for `nums = [1,2,3]`:
  ```
  000 ‚Üí []
  001 ‚Üí [3]
  010 ‚Üí [2]
  011 ‚Üí [2,3]
  100 ‚Üí [1]
  101 ‚Üí [1,3]
  110 ‚Üí [1,2]
  111 ‚Üí [1,2,3]
  ```

### **Optimized Bitwise Rust Code**

```rust
impl Solution {
    pub fn subsets(nums: Vec<i32>) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let n = nums.len();
        let total = 1 << n; // 2^n subsets

        for mask in 0..total {
            let mut subset = Vec::new();
            for i in 0..n {
                if (mask & (1 << i)) != 0 {
                    subset.push(nums[i]);
                }
            }
            result.push(subset);
        }

        result
    }
}
```

### **Pros & Cons**

‚úÖ **Iterative (No recursion stack overhead).**  
‚úÖ **Faster than backtracking for large `n` due to reduced function calls.**  
‚ùå **Less intuitive than backtracking.**  
‚ùå **Bitwise operations may be less readable in interviews.**
