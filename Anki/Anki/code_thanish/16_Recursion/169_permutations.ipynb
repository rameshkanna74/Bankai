## **Optimized Rust Solution for Generating All Permutations**

### **Algorithm Overview**

The problem requires generating all possible **permutations** of a given list of **distinct integers**. Since the number of permutations for `N` elements is **N!**, we use **Backtracking + Recursion** to explore all possible arrangements efficiently.

---

## **Time & Space Complexity**

- **Time Complexity:** **O(N!)**

  - There are **N! possible permutations**.
  - Each permutation requires **O(N) operations**.
  - **Total Complexity:** **O(N √ó N!)** (but dominated by **N!**).

- **Space Complexity:** **O(N!)**
  - **O(N!) storage** for the output list.
  - **O(N) recursion depth** for function calls.

This is **optimal** for `N ‚â§ 6` (as per constraints).

---

## **Optimized Rust Code**

This solution follows **FAANG/MAANG interview standards**:

- **Uses in-place swapping (O(1) space modification) to generate permutations efficiently.**
- **Utilizes backtracking to construct valid permutations recursively.**
- **Efficiently handles all edge cases (empty list, single element, etc.).**

```rust
impl Solution {
    pub fn permute(nums: Vec<i32>) -> Vec<Vec<i32>> {
        let mut nums = nums; // Make nums mutable
        let mut result = Vec::new();
        Self::backtrack(0, &mut nums, &mut result);
        result
    }

    fn backtrack(start: usize, nums: &mut Vec<i32>, result: &mut Vec<Vec<i32>>) {
        if start == nums.len() {
            result.push(nums.clone()); // Store current permutation
            return;
        }

        for i in start..nums.len() {
            nums.swap(start, i); // Swap to choose the next element
            Self::backtrack(start + 1, nums, result);
            nums.swap(start, i); // Backtrack (undo swap)
        }
    }
}
```

---

## **How the Algorithm Works**

1. **Base Case:** If `start == nums.len()`, push the current permutation into `result`.
2. **Recursive Case:**
   - Swap each element to the current `start` position.
   - Recursively permute the remaining elements.
   - **Backtrack** (undo swap) to restore the original array order.

### **Key Features**

‚úÖ **In-place modifications ‚Üí O(1) extra space usage.**  
‚úÖ **Efficient recursion ‚Üí Generates all permutations without redundant memory copies.**  
‚úÖ **Avoids using `Vec<bool>` to track used elements (faster execution).**

### **2Ô∏è‚É£ Iterative Approach (Lexicographic Permutations)**

We can **generate permutations iteratively** using **Heap‚Äôs Algorithm** (no recursion):

```rust
impl Solution {
    pub fn permute_iterative(mut nums: Vec<i32>) -> Vec<Vec<i32>> {
        let mut result = vec![nums.clone()];
        let mut c = vec![0; nums.len()];
        let mut i = 0;

        while i < nums.len() {
            if c[i] < i {
                let swap_idx = if i % 2 == 0 { 0 } else { c[i] };
                nums.swap(swap_idx, i);
                result.push(nums.clone());
                c[i] += 1;
                i = 0;
            } else {
                c[i] = 0;
                i += 1;
            }
        }
        result
    }
}
```

‚úÖ **Avoids recursion, uses O(1) extra space!**  
‚úÖ **Lexicographic ordering of permutations.**
