## **Optimized Rust Solution for Word Search (Backtracking + Pruning)**

### **Algorithm Overview**

This problem requires checking if a given word exists in a **2D board of characters**. We must ensure:

- The word is formed by **adjacent** (horizontal/vertical) cells.
- A cell is **used only once** per word search.
- The function returns `true` if the word is found, otherwise `false`.

To solve this efficiently, we use **Backtracking with DFS (Depth-First Search)**:

1. **Start from every possible position (`board[i][j]`) that matches the first letter of `word`**.
2. **Use DFS to explore adjacent cells** while marking visited positions to avoid reuse.
3. **Backtrack when needed** (undo changes to try other paths).
4. **Use pruning (early termination) for performance optimizations**.

---

## **Time & Space Complexity**

### **Time Complexity:**

- **Worst case:** **O(m √ó n √ó 4^L)**
  - `m √ó n` ‚Üí Each cell is a starting point.
  - `4^L` ‚Üí Each letter can lead to **4 recursive calls** (up/down/left/right).
  - **L ‚â§ 15** ensures **4^15 (‚âà 1M calls)** is feasible.

### **Space Complexity:**

- **O(L)** for the recursion stack (since `L` is the word length).
- **O(1) extra space** since we **mutate the board directly** without auxiliary storage.

## **Optimized Rust Code**

This solution:

- **Uses mutable board references** to avoid unnecessary cloning.
- **Implements DFS + backtracking with pruning**.
- **Marks visited cells in-place** to avoid extra memory.

```rust
impl Solution {
    pub fn exist(mut board: Vec<Vec<char>>, word: String) -> bool {
        let rows = board.len();
        let cols = board[0].len();
        let word_chars: Vec<char> = word.chars().collect();

        // Start DFS from every cell that matches word[0]
        for i in 0..rows {
            for j in 0..cols {
                if Self::dfs(&mut board, &word_chars, 0, i, j) {
                    return true;
                }
            }
        }
        false
    }

    fn dfs(board: &mut Vec<Vec<char>>, word: &[char], index: usize, row: usize, col: usize) -> bool {
        if index == word.len() {
            return true; // Word fully matched
        }

        // Check boundaries & character match
        if row >= board.len() || col >= board[0].len() || board[row][col] != word[index] {
            return false;
        }

        // Mark as visited (temporary modification)
        let temp = board[row][col];
        board[row][col] = '#';

        // Explore all 4 directions
        let found = Self::dfs(board, word, index + 1, row.wrapping_sub(1), col)
            || Self::dfs(board, word, index + 1, row + 1, col)
            || Self::dfs(board, word, index + 1, row, col.wrapping_sub(1))
            || Self::dfs(board, word, index + 1, row, col + 1);

        // Restore original state (backtrack)
        board[row][col] = temp;

        found
    }
}
```

---

## **How the Algorithm Works**

1. **Iterate through all cells** to find potential starting points.
2. **Perform DFS recursively**, exploring **all 4 directions**.
3. **Prune unpromising paths** (if character mismatch occurs).
4. **Use backtracking** (undo cell modification after exploring a path).

```rust
use rayon::prelude::*;

impl Solution {
    pub fn exist_parallel(board: Vec<Vec<char>>, word: String) -> bool {
        let rows = board.len();
        let cols = board[0].len();
        let word_chars: Vec<char> = word.chars().collect();

        (0..rows).into_par_iter().any(|i| {
            (0..cols).into_par_iter().any(|j| {
                let mut board_clone = board.clone(); // Parallel threads need independent copies
                Self::dfs(&mut board_clone, &word_chars, 0, i, j)
            })
        })
    }
}
```

üí° **Speedup:** Distributes search across multiple CPU cores.
