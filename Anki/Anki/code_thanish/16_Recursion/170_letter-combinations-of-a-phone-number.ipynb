## **Optimized Rust Solution for Letter Combinations of a Phone Number**

### **Algorithm Overview**

This problem requires generating all possible letter combinations for a given phone number input. Since each digit maps to multiple letters (e.g., '2' → "abc"), we use **Backtracking (DFS) + Recursion** to explore all possible combinations.

---

## **Time & Space Complexity**

- **Time Complexity:** **O(3ⁿ × 4ᵐ)**

  - `n` → Number of digits mapping to **3** letters (`2, 3, 4, 5, 6, 8`).
  - `m` → Number of digits mapping to **4** letters (`7, 9`).
  - Worst-case, **4 digits → O(3⁴) = 81** possible strings.

- **Space Complexity:** **O(3ⁿ × 4ᵐ)**
  - **O(L) recursion depth**, where `L` is the number of digits.
  - **O(3ⁿ × 4ᵐ) output storage** (since we generate that many combinations).
  - **O(1) extra space** apart from output storage.

Given that `n ≤ 4`, our approach **runs efficiently within constraints**.

---

## **Optimized Rust Code**

This solution follows **FAANG/MAANG interview standards**:

- **Uses recursion + backtracking for optimal traversal.**
- **Avoids unnecessary memory copies (uses references).**
- **Handles edge cases (`""`, `"2"`, etc.).**
- **Uses precomputed mappings for O(1) lookup.**

```rust
impl Solution {
    pub fn letter_combinations(digits: String) -> Vec<String> {
        if digits.is_empty() {
            return Vec::new();
        }

        let digit_to_letters = [
            "", "", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"
        ];

        let mut result = Vec::new();
        let mut current = String::new();
        Self::backtrack(&digits, 0, &digit_to_letters, &mut current, &mut result);
        result
    }

    fn backtrack(
        digits: &str,
        index: usize,
        digit_to_letters: &[&str],
        current: &mut String,
        result: &mut Vec<String>,
    ) {
        if index == digits.len() {
            result.push(current.clone());
            return;
        }

        let digit = digits.chars().nth(index).unwrap() as usize - '0' as usize;
        for &ch in digit_to_letters[digit].chars().collect::<Vec<_>>().iter() {
            current.push(ch);
            Self::backtrack(digits, index + 1, digit_to_letters, current, result);
            current.pop(); // Backtrack
        }
    }
}
```

---

## **How the Algorithm Works**

1. **Base Case:** If `index == digits.len()`, push the built string to `result`.
2. **Recursive Case:**
   - **Extract the digit** and find corresponding letters (`digit_to_letters`).
   - **Loop through letters** and recursively explore further digits.
   - **Backtrack (pop last character)** to explore other branches.

✅ **Efficiently avoids extra memory allocation.**  
✅ **Uses in-place modification (`push`/`pop`) for performance.**  
✅ **Handles all edge cases correctly.**

### **2️⃣ Iterative Approach (BFS)**

We can **generate letter combinations iteratively** using a **BFS approach**:

```rust
impl Solution {
    pub fn letter_combinations_iterative(digits: String) -> Vec<String> {
        if digits.is_empty() {
            return Vec::new();
        }

        let digit_to_letters = [
            "", "", "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"
        ];

        let mut queue = vec![String::new()];

        for digit in digits.chars() {
            let letters = digit_to_letters[digit as usize - '0' as usize];
            let mut new_queue = Vec::new();

            for prefix in queue.iter() {
                for ch in letters.chars() {
                    new_queue.push(format!("{}{}", prefix, ch));
                }
            }

            queue = new_queue;
        }

        queue
    }
}
```

✅ **Avoids recursion, no stack overhead!**  
✅ **Breadth-First Approach → Memory efficient!**
