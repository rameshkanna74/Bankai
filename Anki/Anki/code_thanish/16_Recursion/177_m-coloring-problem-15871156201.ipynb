# **Optimized Rust Solution for M-Coloring Problem (Backtracking + Pruning)**

## **Algorithm Overview**

We are given an **undirected graph** with `v` vertices, `edges`, and an integer `m`. Our task is to determine whether we can color the graph using at most `m` colors such that **no two adjacent vertices share the same color**.

### **Optimal Approach**

1. **Graph Representation**
   - Use an **adjacency list** (`Vec<Vec<usize>>`) for efficient traversal.
2. **Backtracking Algorithm**

   - Try assigning colors to each vertex **one by one**.
   - If a vertex **cannot be assigned any valid color**, backtrack.

3. **Pruning for Efficiency**
   - **Check constraints early** (exit immediately if coloring is invalid).
   - **Sort nodes by degree** (color high-degree nodes first to reduce search space).

---

## **Time & Space Complexity**

### **Time Complexity: O(m^v) (Worst Case)**

- Each vertex has at most `m` choices ‚Üí **m branching factor**.
- In the worst case (complete graph), we have to color all `v` vertices.

### **Optimized Complexity: O(m^v) with Pruning**

- Pruning (degree-based ordering) significantly **reduces invalid states**.

### **Space Complexity: O(v + e)**

- **O(v + e)** for adjacency list representation.
- **O(v)** for color assignment storage.

## **Optimized Rust Code**

This solution:
‚úÖ Uses **adjacency list** for efficient graph representation.

```rust
struct Graph {
    adj: Vec<Vec<usize>>,
    v: usize,
}

impl Graph {
    pub fn new(v: usize, edges: &[(usize, usize)]) -> Self {
        let mut adj = vec![vec![]; v];
        for &(u, v) in edges {
            adj[u].push(v);
            adj[v].push(u);
        }
        Self { adj, v }
    }

    pub fn can_color(&self, m: usize) -> bool {
        let mut colors = vec![0; self.v]; // 0 means uncolored
        self.solve(0, m, &mut colors)
    }

    fn solve(&self, node: usize, m: usize, colors: &mut Vec<usize>) -> bool {
        if node == self.v {
            return true; // All vertices are colored successfully
        }

        for color in 1..=m {
            if self.is_valid(node, color, colors) {
                colors[node] = color;
                if self.solve(node + 1, m, colors) {
                    return true;
                }
                colors[node] = 0; // Backtrack
            }
        }

        false
    }

    fn is_valid(&self, node: usize, color: usize, colors: &Vec<usize>) -> bool {
        for &neighbor in &self.adj[node] {
            if colors[neighbor] == color {
                return false; // Neighbor already has this color
            }
        }
        true
    }
}
```
