## **Optimized Rust Solution for "Combination Sum"**

### **Algorithm Overview**

The problem requires generating all **unique** combinations of numbers from `candidates` that sum up to `target`, allowing repeated use of elements.  
The best approach is **Backtracking (DFS with pruning).**

---

### **Optimal Time & Space Complexity**

- **Time Complexity**: **\( O(2^t) \) (Exponential, but Pruned)**

  - The worst-case scenario is exploring all subsets (~\( 2^n \)), but **pruning** reduces unnecessary branches.
  - `t` is the target value, and each recursive call explores different sums.

- **Space Complexity**: **\( O(t) \) (Recursion Depth) + \( O(k) \) (Result Storage)**
  - **Recursive depth** is at most `target/min(candidates) ‚âà O(t)`.
  - **Output storage** depends on the number of valid subsets `O(k)`.

---

## **Optimized Rust Code**

This solution is **idiomatic, efficient, and FAANG-level interview-ready**.

```rust
impl Solution {
    pub fn combination_sum(candidates: Vec<i32>, target: i32) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let mut current = Vec::new();
        Self::backtrack(&candidates, target, 0, &mut current, &mut result);
        result
    }

    /// Backtracking function to find all valid combinations
    fn backtrack(
        candidates: &Vec<i32>,
        target: i32,
        start: usize,
        current: &mut Vec<i32>,
        result: &mut Vec<Vec<i32>>,
    ) {
        if target == 0 {
            result.push(current.clone());
            return;
        }

        for i in start..candidates.len() {
            if candidates[i] > target {
                continue; // Prune the search
            }
            current.push(candidates[i]);
            Self::backtrack(candidates, target - candidates[i], i, current, result);
            current.pop(); // Backtrack step
        }
    }
}
```

## **Algorithm Breakdown**

### **Backtracking Approach**

1. **Recursively explore** candidates, subtracting from `target`.
2. **Base case**:
   - If `target == 0`, a valid combination is found ‚Üí push `current` to `result`.
   - If `target < 0`, **stop (invalid path).**
3. **Pruning Optimization**:
   - **Skip candidates larger than `target`** to reduce unnecessary recursion.
   - **Use `start` index** to prevent duplicate combinations (`[2,3]` vs. `[3,2]`).

---

### **Edge Cases Considered**

‚úÖ `candidates = [2, 3, 6, 7]`, `target = 7` ‚Üí Returns `[[2,2,3],[7]]`.  
‚úÖ `candidates = [2, 3, 5]`, `target = 8` ‚Üí Returns `[[2,2,2,2],[2,3,3],[3,5]]`.  
‚úÖ `candidates = [2]`, `target = 1` ‚Üí Returns `[]` (No solution).  
‚úÖ `candidates = [7, 2, 3, 6]`, `target = 7` ‚Üí **Handles unordered input correctly**.  
‚úÖ **Scalability test** with `target = 40` runs efficiently.

## **Alternative: Dynamic Programming (Subset Sum Approach)**

### **Time Complexity: \( O(n \cdot t) \)**

- **Uses a DP table `dp[i]`** to store all combinations that sum to `i`.
- **Avoids recursion**, but **increases memory usage**.

### **Optimized DP Rust Code**

```rust
impl Solution {
    pub fn combination_sum(candidates: Vec<i32>, target: i32) -> Vec<Vec<i32>> {
        let mut dp = vec![vec![]; (target + 1) as usize];
        dp[0] = vec![vec![]];

        for &num in &candidates {
            for t in num..=target {
                let t = t as usize;
                if !dp[t - num as usize].is_empty() {
                    for subset in &dp[t - num as usize] {
                        let mut new_subset = subset.clone();
                        new_subset.push(num);
                        dp[t].push(new_subset);
                    }
                }
            }
        }

        dp[target as usize].clone()
    }
}
```

### **Pros & Cons**

‚úÖ **No recursion (iterative, bottom-up DP).**  
‚úÖ **Efficient for small target values.**  
‚ùå **Consumes extra memory `O(target * n)`.**  
‚ùå **Slower for large `target` compared to backtracking.**
