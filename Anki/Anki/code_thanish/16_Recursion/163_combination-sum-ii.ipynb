## **Optimized Rust Solution for "Combination Sum II"**

### **Algorithm Overview**

This problem is best solved using **Backtracking with Pruning**:

- Unlike **Combination Sum I**, where elements can be reused, here **each element can be used only once**.
- To **avoid duplicate combinations**, we **sort the input** and **skip duplicates** during recursion.

---

### **Optimal Time & Space Complexity**

- **Time Complexity**: **\( O(2^n) \) (Exponential, but Pruned)**

  - The worst-case scenario is exploring all subsets (~\( 2^n \)), but **sorting and duplicate skipping reduce unnecessary recursion**.
  - Sorting takes **\( O(n \log n) \)** but is negligible compared to the recursive calls.

- **Space Complexity**: **\( O(t) \) (Recursion Depth) + \( O(k) \) (Result Storage)**
  - **Recursive depth** is at most `n` (length of candidates).
  - **Output storage** depends on the number of valid subsets `O(k)`.

impl Solution {
    pub fn combination_sum2(mut candidates: Vec<i32>, target: i32) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let mut current = Vec::new();
        candidates.sort(); // Sort to handle duplicates efficiently
        Self::backtrack(&candidates, target, 0, &mut current, &mut result);
        result
    }

    /// Backtracking function to find all unique combinations
    fn backtrack(
        candidates: &Vec<i32>,
        target: i32,
        start: usize,
        current: &mut Vec<i32>,
        result: &mut Vec<Vec<i32>>,
    ) {
        if target == 0 {
            result.push(current.clone());
            return;
        }

        for i in start..candidates.len() {
            if i > start && candidates[i] == candidates[i - 1] {
                continue; // Skip duplicate elements at the same recursion level
            }

            if candidates[i] > target {
                break; // Prune the search since candidates are sorted
            }

            current.push(candidates[i]);
            Self::backtrack(candidates, target - candidates[i], i + 1, current, result);
            current.pop(); // Backtrack step
        }
    }
}

## **Algorithm Breakdown**

### **Backtracking with Pruning**

1. **Sort the input array** to make duplicate skipping easier.
2. **Recursively explore** candidates, subtracting from `target`.
3. **Base case**:
   - If `target == 0`, a valid combination is found ‚Üí push `current` to `result`.
   - If `target < 0`, **stop (invalid path).**
4. **Pruning Optimization**:
   - **If `candidates[i] > target`, break the loop** (since numbers are sorted, further numbers are too large).
   - **Skip duplicates** by checking `candidates[i] == candidates[i - 1]`.
