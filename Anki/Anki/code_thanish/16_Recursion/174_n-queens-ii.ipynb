# **Optimized Rust Solution for N-Queens II (Backtracking + Bitmasking)**

## **Algorithm Overview**

The **N-Queens II** problem requires counting the number of valid ways to place `n` queens on an `n x n` board, following these constraints:

- No two queens can be in the **same row**.
- No two queens can be in the **same column**.
- No two queens can be in the **same diagonal**.

### **Optimized Approach**

1. **Backtracking with pruning**:

   - We explore one row at a time.
   - Use **bitmasking** to track occupied **columns**, **major diagonals**, and **minor diagonals**.
   - If a position is safe, place the queen and **recursively solve the next row**.
   - If all queens are placed, increment the count.
   - Backtrack (remove the last placed queen) and continue exploring other possibilities.

2. **Bitmasking for Fast Validity Check**:
   - Instead of checking each column and diagonal manually (**O(N)** per row), we store state using **three bit vectors**:
     - `cols`: Tracks occupied **columns**.
     - `diag1`: Tracks occupied **major diagonals** (`row - col`).
     - `diag2`: Tracks occupied **minor diagonals** (`row + col`).
   - Using bitwise operations reduces the validity check to **O(1)**.

## **Time & Space Complexity**

### **Time Complexity**

- The **worst-case** backtracking explores **O(N!)** possibilities.
- Bitmasking reduces pruning checks to **O(1)** per operation.
- **Effective complexity**: **O(N!)**, but optimized compared to naive recursion.

### **Space Complexity**

- **O(N)** for `cols`, `diag1`, and `diag2`.
- No extra storage for the board.
- **Overall:** **O(N)**, making it highly space-efficient.

```rust
impl Solution {
    pub fn total_n_queens(n: i32) -> i32 {
        let n = n as usize;
        let mut count = 0;
        let mut cols = 0;
        let mut diag1 = 0;
        let mut diag2 = 0;

        Self::backtrack(n, 0, &mut count, &mut cols, &mut diag1, &mut diag2);
        count
    }

    fn backtrack(
        n: usize,
        row: usize,
        count: &mut i32,
        cols: &mut u16,
        diag1: &mut u32,
        diag2: &mut u32,
    ) {
        if row == n {
            *count += 1;
            return;
        }

        let mut available_positions = ((1 << n) - 1) & !(*cols | (*diag1 >> row) | (*diag2 >> (n - row)));

        while available_positions != 0 {
            let pos = available_positions & -available_positions; // Get the lowest set bit
            available_positions &= available_positions - 1; // Remove the bit from available positions

            let col_index = pos.trailing_zeros() as usize;

            // Place the queen
            *cols |= pos;
            *diag1 |= pos << row;
            *diag2 |= pos << (n - row);

            // Recur for the next row
            Self::backtrack(n, row + 1, count, cols, diag1, diag2);

            // Backtrack
            *cols &= !pos;
            *diag1 &= !(pos << row);
            *diag2 &= !(pos << (n - row));
        }
    }
}
```

## **How the Algorithm Works**

### **Bitwise Representation**

- **`cols` (`u16`)** â†’ Tracks occupied columns.
- **`diag1` (`u32`)** â†’ Tracks major diagonals (`row - col`).
- **`diag2` (`u32`)** â†’ Tracks minor diagonals (`row + col`).
- `available_positions = ((1 << n) - 1) & !(cols | (diag1 >> row) | (diag2 >> (n - row)))`
  - **Calculates available positions** for the current row.
  - `pos = available_positions & -available_positions` isolates the **rightmost available position**.
  - `available_positions &= available_positions - 1` removes it.
