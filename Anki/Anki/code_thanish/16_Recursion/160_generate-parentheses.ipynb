### **Optimized Rust Solution for "Generate Parentheses"**

#### **Algorithm Overview**

The problem requires generating all well-formed parentheses combinations for `n` pairs. The best approach is **backtracking (DFS with pruning)**, where we recursively build valid sequences while ensuring constraints are met.

#### **Optimal Time & Space Complexity**

- **Time Complexity**: **\( O(4^n / \sqrt{n}) \)**

  - This follows the **Catalan number** growth pattern, where the number of valid sequences is the \(n\)th Catalan number:  
    \[
    C_n = \frac{1}{n+1} \binom{2n}{n} \approx \frac{4^n}{n^{3/2}}
    \]
  - This represents the best possible complexity since every valid sequence must be considered.

- **Space Complexity**: **\( O(4^n / \sqrt{n}) \) (Output) + \( O(n) \) (Recursion Depth)**
  - The output itself takes **\( O(4^n / \sqrt{n}) \)** space.
  - The recursion stack goes **\( O(n) \)** deep, consuming extra stack space.

```rust
impl Solution {
    pub fn generate_parenthesis(n: i32) -> Vec<String> {
        let mut result = Vec::new();
        let mut current = String::with_capacity((2 * n) as usize);
        Self::backtrack(n, 0, 0, &mut current, &mut result);
        result
    }

    /// Backtracking function to generate well-formed parentheses.
    fn backtrack(n: i32, open: i32, close: i32, current: &mut String, result: &mut Vec<String>) {
        if current.len() == (2 * n) as usize {
            result.push(current.clone());
            return;
        }

        // Add '(' if we haven't used all open brackets
        if open < n {
            current.push('(');
            Self::backtrack(n, open + 1, close, current, result);
            current.pop(); // Backtrack
        }

        // Add ')' only if it wouldn't create an invalid sequence
        if close < open {
            current.push(')');
            Self::backtrack(n, open, close + 1, current, result);
            current.pop(); // Backtrack
        }
    }
}
```

### **Algorithm Breakdown**

- **Backtracking with Pruning**:
  - Start with an empty string and recursively add `(` or `)`, ensuring validity at each step.
  - **Prune invalid cases early**:
    - Only add `(` if `open < n`.
    - Only add `)` if `close < open`.
  - **Backtracking occurs by popping the last character** after recursion.
