Problem statement
You are given an array of integers. You need to sort the array in ascending order using quick sort.

Quick sort is a divide and conquer algorithm in which we choose a pivot point and partition the array into two parts i.e, left and right. The left part contains the numbers smaller than the pivot element and the right part contains the numbers larger than the pivot element. Then we recursively sort the left and right parts of the array.

## **Optimized Rust Solution for Quick Sort**

### **Algorithm Overview**

Quick Sort is a **Divide & Conquer** algorithm that:

1. **Chooses a pivot element** (typically last, first, or median).
2. **Partitions** the array such that:
   - Left subarray contains elements **less than pivot**.
   - Right subarray contains elements **greater than pivot**.
3. **Recursively sorts** both partitions.

**Key Properties:**

- **Average-case O(N log N) time complexity**, making it efficient for large datasets.
- **In-place sorting** → **O(1) extra space**, unlike Merge Sort.
- **Efficient for large inputs and commonly used in real-world applications**.

---

## **Time & Space Complexity**

- **Best & Average Case Complexity:** **\( O(N \log N) \)**

  - Partitioning at **each level** takes **O(N)**.
  - There are **O(log N) recursive calls**.
  - **Total:** \( O(N \log N) \), optimal for sorting.

- **Worst Case Complexity:** **\( O(N^2) \)**

  - Occurs if the pivot is always the **smallest or largest element**.
  - Can be **avoided using randomized pivot selection**.

- **Space Complexity:** **\( O(1) \) (in-place sorting)**
  - Uses **O(log N) recursive call stack**.

```rust
impl Solution {
    pub fn quick_sort(arr: &mut [i32]) {
        if arr.len() <= 1 {
            return; // Base case: already sorted
        }
        let len = arr.len();
        Self::quick_sort_helper(arr, 0, (len - 1) as isize);
    }

    fn quick_sort_helper(arr: &mut [i32], low: isize, high: isize) {
        if low < high {
            let pivot_index = Self::partition(arr, low, high);
            Self::quick_sort_helper(arr, low, pivot_index - 1);
            Self::quick_sort_helper(arr, pivot_index + 1, high);
        }
    }

    fn partition(arr: &mut [i32], low: isize, high: isize) -> isize {
        let pivot = arr[high as usize];
        let mut i = low - 1;

        for j in low..high {
            if arr[j as usize] <= pivot {
                i += 1;
                arr.swap(i as usize, j as usize);
            }
        }
        arr.swap((i + 1) as usize, high as usize);
        i + 1
    }
}
```

## **Algorithm Breakdown**

### **Quick Sort Steps**

1. **Base Case:** If `arr.len() <= 1`, return (already sorted).
2. **Recursive Case:**
   - Select **last element as pivot**.
   - **Partition**: Rearrange elements so that:
     - **Left** part contains elements ≤ pivot.
     - **Right** part contains elements > pivot.
   - Recursively apply Quick Sort to both halves.

### **Partition Function**

- **Uses two pointers (`i`, `j`)** to swap elements in-place.
- **O(N) partitioning** ensures no extra memory usage.
- **Preserves stability** (elements ≤ pivot go left).

---

### **Edge Cases Considered**

✅ **Empty Array** (`arr = []`) → Returns `[]`.  
✅ **Single Element** (`arr = [5]`) → Returns `[5]`.  
✅ **All Duplicates** (`arr = [2,2,2]`) → Returns `[2,2,2]`.  
✅ **Already Sorted Input** (`arr = [1,2,3]`) → Returns `[1,2,3]`.  
✅ **Reverse Sorted Input** (`arr = [5,4,3,2,1]`) → Returns `[1,2,3,4,5]`.  
✅ **Large Input (`N = 10^3`)** → Efficiently runs in `O(10^3 log 10^3)`.

### **1️⃣ Randomized Quick Sort**

- Select **pivot randomly** to **avoid worst-case `O(N²)`**.
- Improves performance for **already sorted inputs**.

```rust
use rand::Rng;

impl Solution {
    fn randomized_partition(arr: &mut [i32], low: isize, high: isize) -> isize {
        let mut rng = rand::thread_rng();
        let random_index = rng.gen_range(low..=high) as usize;
        arr.swap(random_index, high as usize);
        Self::partition(arr, low, high)
    }
}
```

```rust
use rayon::prelude::*;

impl Solution {
    pub fn parallel_quick_sort(arr: &mut [i32]) {
        if arr.len() > 1 {
            let pivot_index = Self::partition(arr, 0, (arr.len() - 1) as isize);
            let (left, right) = arr.split_at_mut(pivot_index as usize);
            rayon::join(|| Self::quick_sort(left), || Self::quick_sort(&mut right[1..]));
        }
    }
}
```
