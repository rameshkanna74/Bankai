## üß† **Problem Breakdown**

We need to find the **maximum sum of all subarrays of size `k`** where **all elements are distinct**.

### **Key Observations**

1. **Brute Force (`O(n * k)`)**:

   - Iterate over all possible subarrays of size `k`.
   - Check if all elements are distinct (`O(k)`) and compute the sum.
   - **Too slow for large `n = 10^5`**.

2. **Optimized Approach: Sliding Window + HashSet (`O(n)`)**:
   - Use a **sliding window of size `k`** to maintain a rolling sum.
   - Use a **HashSet** to track unique elements.
   - If a duplicate is found, shrink the window from the left until uniqueness is restored.

---

## ‚è≥ **Time and Space Complexity**

‚úÖ **Time Complexity:** `O(n)`, since each element is processed at most twice (once added, once removed).  
‚úÖ **Space Complexity:** `O(k)`, for storing elements in the `HashSet`.

## üöÄ **Optimized Rust Implementation**

```rust
use std::collections::HashSet;

impl Solution {
    pub fn maximum_subarray_sum(nums: Vec<i32>, k: i32) -> i64 {
        let k = k as usize;
        let mut max_sum = 0;
        let mut current_sum = 0;
        let mut unique_elements = HashSet::new();
        let mut left = 0;

        for right in 0..nums.len() {
            while unique_elements.contains(&nums[right]) {
                unique_elements.remove(&nums[left]);
                current_sum -= nums[left] as i64;
                left += 1;
            }

            unique_elements.insert(nums[right]);
            current_sum += nums[right] as i64;

            if right - left + 1 == k {
                max_sum = max_sum.max(current_sum);
                unique_elements.remove(&nums[left]);
                current_sum -= nums[left] as i64;
                left += 1;
            }
        }

        max_sum
    }
}
```
