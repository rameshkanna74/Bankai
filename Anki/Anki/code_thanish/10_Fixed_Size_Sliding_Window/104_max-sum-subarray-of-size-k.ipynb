## üß† **Problem Breakdown**

We need to find the **maximum sum of a contiguous subarray** of size `k` from the given array `arr[]`.

### **Key Observations**

1. A **brute-force approach** would:

   - Compute the sum of all possible subarrays of size `k`.
   - **Time Complexity: `O(n * k)`** ‚õî **Too slow for large `n` (up to `10^6`).**

2. **Optimal Approach: Sliding Window**
   - Maintain a **window of size `k`** and compute its sum efficiently.
   - **Sliding Window Technique** allows updating the sum in **O(1)** time per step.
   - **Time Complexity: `O(n)`** ‚úÖ **Optimal!**

---

## üöÄ **Optimal Approach: Sliding Window**

### **Algorithm Design**

1. **Compute the sum of the first `k` elements** ‚Üí `O(k)`.
2. **Slide the window across `arr[]`, updating the sum** by:
   - **Adding the next element (`arr[i]`)**.
   - **Removing the outgoing element (`arr[i-k]`)**.
   - **Update the maximum sum if the new sum is greater**.
3. **Return the maximum sum found**.

---

## ‚è≥ **Time and Space Complexity**

‚úÖ **Time Complexity:** `O(n)`, since we iterate through the array once.  
‚úÖ **Space Complexity:** `O(1)`, as we use only a few extra variables.

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn max_sum_subarray(arr: Vec<i32>, k: usize) -> i32 {
        if arr.len() < k {
            return 0; // Edge case: Invalid input where k > arr.len()
        }

        let mut sum: i32 = arr[..k].iter().sum(); // Initial window sum
        let mut max_sum = sum;

        for i in k..arr.len() {
            sum += arr[i] - arr[i - k]; // Slide the window
            max_sum = max_sum.max(sum);
        }

        max_sum
    }
}
```
