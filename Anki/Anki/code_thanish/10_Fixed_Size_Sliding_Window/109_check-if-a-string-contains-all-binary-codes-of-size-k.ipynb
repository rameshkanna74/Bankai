# ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

---

## 🧠 **Problem Breakdown**

We need to determine if **all possible binary codes of length `k` exist as substrings in `s`**.

### **Key Observations**

1. **Total Unique Binary Codes**:

   - Since `s` consists of binary digits (`0` and `1`), there are at most **`2^k` unique binary substrings** of length `k`.
   - If `s.len() < 2^k + k - 1`, it's **impossible** to have all `2^k` substrings → **Return `false` early**.

2. **Sliding Window + HashSet (`O(n)`)**:
   - We iterate over `s` using a **sliding window of size `k`**.
   - Store encountered substrings in a **HashSet**.
   - If the number of unique substrings reaches `2^k`, return `true`.

---

## ⏳ **Time and Space Complexity**

✅ **Time Complexity:** `O(n)`, since we scan `s` **once** and store at most `2^k` substrings.  
✅ **Space Complexity:** `O(2^k)`, since we store at most `2^k` unique substrings in a HashSet.

---

## 🚀 **Optimized Rust Implementation**

```rust
use std::collections::HashSet;

impl Solution {
    pub fn has_all_codes(s: String, k: i32) -> bool {
        let n = s.len();
        let k = k as usize;
        let required_count = 1 << k; // 2^k possible binary substrings

        if n < k + required_count - 1 {
            return false; // Early exit: not enough length to fit all substrings
        }

        let mut seen = HashSet::with_capacity(required_count);
        let mut hash_value = 0;
        let mask = required_count - 1; // Bitmask to maintain only `k` bits

        // Compute the hash for the first k-length window
        for i in 0..k {
            hash_value = (hash_value << 1) | (s.as_bytes()[i] - b'0') as usize;
        }
        seen.insert(hash_value);

        // Slide over the rest of `s`, maintaining a rolling hash of k bits
        for i in k..n {
            hash_value = ((hash_value << 1) & mask) | (s.as_bytes()[i] - b'0') as usize;
            seen.insert(hash_value);

            if seen.len() == required_count {
                return true;
            }
        }

        false
    }
}
```

✅ **Minimizes space usage (`O(2^k)`, only storing unique binary numbers).**  
✅ **Handles edge cases efficiently (`s.len() < 2^k` exits early).**
