## üß† **Problem Breakdown**

We are given an integer array `nums` and an integer `k`.  
Our goal is to **find the contiguous subarray of length `k` that has the maximum average value** and return this value.

### **Key Observations**

1. A **brute-force approach** would:

   - Compute the sum for all possible subarrays of size `k`.
   - Time Complexity: **O(n \* k)** ‚õî **Too slow for large `n` (up to `10^5`)**.

2. **Optimal Approach: Sliding Window**
   - Maintain a **window of size `k`** and compute its sum efficiently.
   - **Sliding Window Technique** allows updating the sum in **O(1)** time.
   - Time Complexity: **O(n)** ‚úÖ **Optimal!**

---

## üöÄ **Optimal Approach: Sliding Window**

### **Algorithm Design**

1. **Compute the sum of the first `k` elements** ‚Üí `O(k)`.
2. **Slide the window across `nums`**, updating the sum by:
   - **Adding the next element (`nums[i]`)**.
   - **Removing the outgoing element (`nums[i-k]`)**.
   - **Update the maximum sum if the new sum is greater**.
3. **Compute the maximum average** as `max_sum / k`.

---

## ‚è≥ **Time and Space Complexity**

‚úÖ **Time Complexity:** `O(n)`, since we iterate through the array once.  
‚úÖ **Space Complexity:** `O(1)`, as we use only a few extra variables.

---

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_max_average(nums: Vec<i32>, k: i32) -> f64 {
        let k = k as usize;
        let mut sum: i32 = nums[..k].iter().sum(); // Initial window sum
        let mut max_sum = sum;

        for i in k..nums.len() {
            sum += nums[i] - nums[i - k]; // Slide the window
            max_sum = max_sum.max(sum);
        }

        max_sum as f64 / k as f64
    }
}
```
