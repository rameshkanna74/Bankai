## 🧠 **Problem Breakdown**

Given a `txt` string and a `pat` string, we need to count how many times **any anagram of `pat`** appears as a **contiguous substring** of `txt`.

### **Key Observations**

1. **Brute Force Approach (`O(n * m log m)`)**:

   - Iterate over all substrings of length `m` (`m = len(pat)`) in `txt`.
   - Check if each substring is an anagram of `pat` (sorting takes `O(m log m)`).
   - **Too slow for large inputs (`|txt| ≤ 10^5`)**.

2. **Optimized Approach: Sliding Window + Frequency Map (`O(n)`)**:
   - Maintain a **frequency count of characters** in `pat` (`pat_freq`).
   - Use a **sliding window of size `m`** to track character frequencies in `txt`.
   - Compare window frequency with `pat_freq` efficiently.

---

## ⏳ **Time and Space Complexity**

✅ **Time Complexity:** `O(n)`, since each character is processed at most twice (once added, once removed).  
✅ **Space Complexity:** `O(1)`, since we only use frequency arrays of fixed size (`26` for lowercase letters).

## 🚀 **Optimized Rust Implementation**

```rust
use std::collections::HashMap;

impl Solution {
    pub fn count_anagram_occurrences(txt: String, pat: String) -> i32 {
        let txt: Vec<char> = txt.chars().collect();
        let pat: Vec<char> = pat.chars().collect();
        let (n, m) = (txt.len(), pat.len());

        if m > n {
            return 0;
        }

        let mut pat_freq = [0; 26]; // Frequency of characters in `pat`
        let mut window_freq = [0; 26]; // Frequency of characters in sliding window

        // Helper function to convert character to index (a=0, ..., z=25)
        let to_index = |ch: char| -> usize { (ch as u8 - b'a') as usize };

        // Populate frequency array for `pat`
        for &ch in &pat {
            pat_freq[to_index(ch)] += 1;
        }

        let mut count = 0;

        // First window initialization
        for i in 0..m {
            window_freq[to_index(txt[i])] += 1;
        }
        if window_freq == pat_freq {
            count += 1;
        }

        // Sliding window
        for i in m..n {
            window_freq[to_index(txt[i])] += 1;      // Add new character
            window_freq[to_index(txt[i - m])] -= 1;  // Remove old character

            if window_freq == pat_freq {
                count += 1;
            }
        }

        count
    }
}
```

## 📌 **Edge Cases Considered**

1. **Exact match (`txt == pat`)** → Returns `1`.
2. **No anagrams (`txt = "xyz"`, `pat = "abc"`)** → Returns `0`.
3. **All characters the same (`txt = "aaaaaa"`, `pat = "aaa"`)** → Multiple valid windows.
4. **Pattern longer than text (`m > n`)** → Returns `0`.
5. **Pattern contains repeated characters (`txt = "ababab"`, `pat = "ab"`)** → Ensures correct frequency tracking.
6. **Characters from different parts of ASCII (`txt = "abcdefgh"`, `pat = "xyz"`)** → No match case.
7. **Minimum input size (`|pat| = 1`, `|txt| = 1`)** → Tests smallest constraints.
