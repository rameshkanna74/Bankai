impl Solution {
    pub fn generate_matrix(n: i32) -> Vec<Vec<i32>> {
        let mut matrix = vec![vec![0; n as usize]; n as usize];
        let (mut top, mut left) = (0, 0);
        let (mut bottom, mut right) = (n as usize - 1, n as usize - 1);
        let mut num = 1;

        while top <= bottom && left <= right {
            // Move Right
            for col in left..=right {
                matrix[top][col] = num;
                num += 1;
            }
            top += 1;

            // Move Down
            for row in top..=bottom {
                matrix[row][right] = num;
                num += 1;
            }
            right -= 1;

            if top <= bottom {
                // Move Left
                for col in (left..=right).rev() {
                    matrix[bottom][col] = num;
                    num += 1;
                }
                bottom -= 1;
            }

            if left <= right {
                // Move Up
                for row in (top..=bottom).rev() {
                    matrix[row][left] = num;
                    num += 1;
                }
                left += 1;
            }
        }

    }

## **ðŸ“Œ Algorithm Explanation**

### **1ï¸âƒ£ Using Boundary Shrinking**

- Maintain **four boundaries**: `top`, `bottom`, `left`, `right`
- Iterate in **four directions**:
  - **Move Right:** Across the `top` row
  - **Move Down:** Along the `right` column
  - **Move Left:** Across the `bottom` row (if not exceeded)
  - **Move Up:** Along the `left` column (if not exceeded)
- **Update boundaries** after each traversal to **shrink the active area**.

### **2ï¸âƒ£ Steps**

#### **Example:** `n = 3`
