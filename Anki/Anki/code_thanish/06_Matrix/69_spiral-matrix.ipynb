✅ **Best Choice: Boundary Shrinking (`O(1)` Space, `O(mn)` Time)**

- **Uses four boundary markers (`top`, `bottom`, `left`, `right`) instead of extra memory.**

## **📌 Optimized Rust Code (`O(1)` Space)**

```rust
impl Solution {
    pub fn spiral_order(matrix: Vec<Vec<i32>>) -> Vec<i32> {
        let mut result = Vec::with_capacity(matrix.len() * matrix[0].len());
        if matrix.is_empty() { return result; }

        let (mut top, mut left) = (0, 0);
        let (mut bottom, mut right) = (matrix.len() as isize - 1, matrix[0].len() as isize - 1);

        while top <= bottom && left <= right {
            // Move Right
            for col in left..=right {
                result.push(matrix[top as usize][col as usize]);
            }
            top += 1;

            // Move Down
            for row in top..=bottom {
                result.push(matrix[row as usize][right as usize]);
            }
            right -= 1;

            if top <= bottom {
                // Move Left
                for col in (left..=right).rev() {
                    result.push(matrix[bottom as usize][col as usize]);
                }
                bottom -= 1;
            }

            if left <= right {
                // Move Up
                for row in (top..=bottom).rev() {
                    result.push(matrix[row as usize][left as usize]);
                }
                left += 1;
            }
        }
        result
    }
}
```

## **📌 Algorithm Explanation**

### **1️⃣ Using Boundary Shrinking**

- Maintain **four boundaries**: `top`, `bottom`, `left`, `right`
- Iterate in **four directions**:
  - **Move Right:** Across the `top` row
  - **Move Down:** Along the `right` column
  - **Move Left:** Across the `bottom` row (if not exceeded)
  - **Move Up:** Along the `left` column (if not exceeded)
- **Update boundaries** after each traversal to **shrink the active area**.

### **2️⃣ Steps**

#### **Example:** `matrix = [[1,2,3], [4,5,6], [7,8,9]]`

1. **Move Right:** `[1, 2, 3]`
   - `top = 1`
2. **Move Down:** `[6, 9]`
   - `right = 1`
3. **Move Left:** `[8, 7]`
   - `bottom = 1`
4. **Move Up:** `[4]`
   - `left = 1`
5. **Move Right (center):** `[5]`
   - Done! ✅
