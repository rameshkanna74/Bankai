✅ **Best Choice: Iterative DP (`O(n²)`)**

- We **directly construct** the Pascal's Triangle iteratively.
- **No recomputation** of values using binomial coefficients.
- **Memory-efficient** as it builds row by row.

## **📌 Optimized Rust Code (`O(n²)`)**

```rust
impl Solution {
    pub fn generate(num_rows: i32) -> Vec<Vec<i32>> {
        let num_rows = num_rows as usize;
        let mut triangle = vec![vec![1; i + 1] for i in 0..num_rows];

        for i in 2..num_rows {
            for j in 1..i {
                triangle[i][j] = triangle[i - 1][j - 1] + triangle[i - 1][j];
            }
        }

        triangle
    }
}
```

✅ **Time Complexity:** `O(n²)` (Each row requires iterating over `i` elements)

## **📌 Algorithm Explanation**

### **1️⃣ Iterative DP Approach (`O(n²)`)**

- **Step 1:** Initialize `triangle` with `numRows` rows, each filled with `1`s.
- **Step 2:** Iterate from row `2` onward.
- **Step 3:** Update `triangle[i][j]` as **sum of two numbers above** (`triangle[i-1][j-1] + triangle[i-1][j]`).
- **Step 4:** Return the final `triangle`.

### **2️⃣ Handling Edge Cases**

| **Case**                  | **Handled?**                        |
| ------------------------- | ----------------------------------- |
| `numRows = 1`             | ✅ Returns `[[1]]`                  |
| `numRows = 2`             | ✅ Returns `[[1], [1,1]]`           |
| `numRows = 30` (Max Case) | ✅ Efficiently generates in `O(n²)` |
