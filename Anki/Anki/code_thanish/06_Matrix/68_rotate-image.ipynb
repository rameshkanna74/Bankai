```rust
impl Solution {
    pub fn rotate(matrix: &mut Vec<Vec<i32>>) {
        let n = matrix.len();

        // Step 1: Transpose the matrix (swap matrix[i][j] with matrix[j][i])
        for i in 0..n {
            for j in i+1..n {
                matrix.swap(i, j);
            }
        }

        // Step 2: Reverse each row to get the final rotation
        for row in matrix.iter_mut() {
            row.reverse();
        }
    }
}
```

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Using Transpose + Reverse**

- **Transpose:** Convert rows into columns by swapping `matrix[i][j]` with `matrix[j][i]`.
- **Reverse Rows:** Flip each row to complete the 90¬∞ rotation.

### **2Ô∏è‚É£ Steps**

1. **Transpose the Matrix**

   - Swap `matrix[i][j]` with `matrix[j][i]` for all `i < j`.
   - Converts:
     ```
     1  2  3        1  4  7
     4  5  6   ‚Üí    2  5  8
     7  8  9        3  6  9
     ```

2. **Reverse Each Row**
   - Swap elements in each row to complete the rotation:
     ```
     1  4  7        7  4  1
     2  5  8   ‚Üí    8  5  2
     3  6  9        9  6  3
     ```
