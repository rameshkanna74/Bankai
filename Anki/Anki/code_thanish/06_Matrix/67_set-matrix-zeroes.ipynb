impl Solution {
    pub fn set_zeroes(matrix: &mut Vec<Vec<i32>>) {
        let (m, n) = (matrix.len(), matrix[0].len());
        let mut first_col_zero = false;

        // Step 1: Mark zeroes using first row and first column
        for i in 0..m {
            if matrix[i][0] == 0 {
                first_col_zero = true;
            }
            for j in 1..n {
                if matrix[i][j] == 0 {
                    matrix[i][0] = 0; // Mark row
                    matrix[0][j] = 0; // Mark column
                }
            }
        }

        // Step 2: Set zeroes based on marks
        for i in (1..m) {
            for j in (1..n) {
                if matrix[i][0] == 0 || matrix[0][j] == 0 {
                    matrix[i][j] = 0;
                }
            }
        }

        // Step 3: Handle first row
        if matrix[0][0] == 0 {
            for j in 0..n {
                matrix[0][j] = 0;
            }
        }

        // Step 4: Handle first column
        if first_col_zero {
            for i in 0..m {
                matrix[i][0] = 0;
            }
        }
    }
}

## **üìå Algorithm Explanation**

### **1Ô∏è‚É£ Using First Row & Column as Flags**

- Instead of **using extra memory** to store zero locations, we **reuse the first row and column**.
- We **mark** which rows & columns should be zeroed.

### **2Ô∏è‚É£ Steps**

1. **Mark Zero Locations**

   - Iterate through the matrix.
   - If `matrix[i][j] == 0`, mark `matrix[i][0] = 0` and `matrix[0][j] = 0`.
   - Track separately if the **first column itself should be zeroed**.

2. **Update Non-First Row & Column**

   - Use `matrix[i][0]` and `matrix[0][j]` as reference flags.
   - If either is `0`, set `matrix[i][j] = 0`.

3. **Update First Row & Column**
   - If `matrix[0][0] == 0`, zero the first row.
   - If `first_col_zero == true`, zero the first column.
