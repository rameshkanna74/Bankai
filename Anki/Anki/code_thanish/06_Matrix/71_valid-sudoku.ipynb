## **üîπ Approach:**
This problem requires validating a **partially filled 9x9 Sudoku board**. The key constraints are:  
1Ô∏è‚É£ Each row must have unique numbers **(1-9, no duplicates except '.')**.  
2Ô∏è‚É£ Each column must have unique numbers **(1-9, no duplicates except '.')**.  
3Ô∏è‚É£ Each **3x3 sub-grid** must have unique numbers **(1-9, no duplicates except '.')**.  

To efficiently check these conditions, we use **three HashSets**:
- **Set for rows**: To track numbers in each row.
- **Set for columns**: To track numbers in each column.
- **Set for 3x3 sub-boxes**: To track numbers in each 3x3 grid.  

---

## **üîπ Time & Space Complexity:**
‚úÖ **Time Complexity:**  
- **O(1) (Constant Time)** ‚Üí The board is always **9x9**, so we iterate **exactly 81 times**.  

‚úÖ **Space Complexity:**  
- **O(1) (Constant Space)** ‚Üí Three **fixed-size** HashSets of length 9.  

---

## **üîπ Rust Solution**
```rust
use std::collections::HashSet;

impl Solution {
    pub fn is_valid_sudoku(board: Vec<Vec<char>>) -> bool {
        let mut rows: [HashSet<char>; 9] = Default::default();
        let mut cols: [HashSet<char>; 9] = Default::default();
        let mut boxes: [HashSet<char>; 9] = Default::default();

        for i in 0..9 {
            for j in 0..9 {
                let cell = board[i][j];

                if cell == '.' {
                    continue;
                }

                let box_index = (i / 3) * 3 + (j / 3);

                if !rows[i].insert(cell) || !cols[j].insert(cell) || !boxes[box_index].insert(cell) {
                    return false;
                }
            }
        }
        true
    }
}
```

---

## **üîπ Explanation of Code**
1Ô∏è‚É£ **Initialize HashSets for tracking seen numbers.**  
2Ô∏è‚É£ **Iterate through the board** row by row and column by column.  
3Ô∏è‚É£ **Compute the 3x3 sub-box index:**  
   - Each box index is computed as `box_index = (i / 3) * 3 + (j / 3)`.  
4Ô∏è‚É£ **Check for duplicates**:  
   - If a number is already in the row, column, or sub-box, return `false`.  
5Ô∏è‚É£ **If no violations are found, return `true`.**
