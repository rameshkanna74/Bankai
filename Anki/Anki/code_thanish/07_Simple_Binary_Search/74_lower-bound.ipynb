## ðŸ§  **Algorithm Explanation**

### ðŸš€ **Approach: Binary Search**

We use **binary search** to find the **lower bound** efficiently:

1. Initialize `left = 0` and `right = n` (out-of-bounds index).
2. While `left < right`:
   - Compute `mid = left + (right - left) / 2` (avoiding overflow).
   - If `arr[mid] >= x`, update `right = mid` (potential answer, move left).
   - Otherwise, update `left = mid + 1` (move right).
3. Return `left`, which now holds the **smallest index where `arr[idx] >= x`**.

### **âœ… Time Complexity:**

- **O(log n)** â†’ Binary search reduces search space by half at each step.

### **âœ… Space Complexity:**

- **O(1)** â†’ Uses only a few integer variables (`left`, `right`, `mid`).

```rust
fn lower_bound(arr: &[i32], x: i32) -> usize {
    let mut left = 0;
    let mut right = arr.len(); // n (out-of-bounds index)

    while left < right {
        let mid = left + (right - left) / 2;
        if arr[mid] >= x {
            right = mid; // Move left to find the first occurrence
        } else {
            left = mid + 1;
        }
    }

    left // The lower bound index
}
```
