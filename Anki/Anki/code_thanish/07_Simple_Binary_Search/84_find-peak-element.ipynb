## ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

### 🚀 **Problem Breakdown**

- We need to find a **peak element** in an array `nums`, where:

  - A peak element is strictly **greater than its neighbors**.
  - If multiple peaks exist, we can return **any peak index**.
  - The **first and last elements are considered valid peaks** if they are greater than their single neighbor.
  - **Out-of-bounds elements are treated as `-∞` (negative infinity)**.

- The optimal solution must run in **O(log n) time**, implying **Binary Search** is required.

---

## 🧠 **Algorithm Explanation**

### **🔹 Observations**

1. **Naïve Approach (O(n))**:

   - Iterate through the array and find the first peak element.
   - **Downside**: Takes **O(n) time**, which is **not optimal**.

2. **Binary Search Approach (O(log n))**:
   - Start with **left = 0, right = n - 1**.
   - Compute **mid = (left + right) / 2**.
   - Compare `nums[mid]` with `nums[mid + 1]`:
     - If `nums[mid] > nums[mid + 1]`, the **peak lies on the left half** (including `mid`).
     - Otherwise, the **peak lies on the right half**.
   - Repeat until `left == right`, which will be the **peak index**.

---

## ⏳ **Time and Space Complexity Analysis**

- **Time Complexity**: **O(log n)** (Binary search halves the array each step).
- **Space Complexity**: **O(1)** (Uses only integer variables).

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_peak_element(nums: Vec<i32>) -> i32 {
        let (mut left, mut right) = (0, nums.len() - 1);

        while left < right {
            let mid = left + (right - left) / 2;

            if nums[mid] > nums[mid + 1] {
                right = mid; // Peak is in the left half (including mid)
            } else {
                left = mid + 1; // Peak is in the right half
            }
        }

        left as i32 // `left` will be pointing to a peak element
    }
}
```
