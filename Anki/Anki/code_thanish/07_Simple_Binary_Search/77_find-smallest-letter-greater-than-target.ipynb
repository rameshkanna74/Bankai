## ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

The problem requires us to **find the smallest character** in a sorted array that is **greater than a given target**. If such a character **does not exist**, return the **first character in the array**.

Since the input array is **sorted** and we need an **O(log n) solution**, **Binary Search** is the most optimal approach.

---

## 🧠 **Algorithm Explanation**

### 🚀 **Approach: Binary Search**

1. **Initialize Search Space**
   - `left = 0`, `right = letters.len() - 1`
2. **Perform Binary Search**

   - Compute `mid = left + (right - left) / 2` to avoid overflow.
   - If `letters[mid] > target`, move `right` (`right = mid`).
   - Otherwise, move `left` (`left = mid + 1`).

3. **Return the Smallest Greater Character**
   - When the loop exits, `left` points to the **first character greater than `target`**.
   - If no such character exists (`left == letters.len()`), return `letters[0]`.

### **✅ Time Complexity:**

- **O(log n)** → Since we're using Binary Search.

### **✅ Space Complexity:**

- **O(1)** → Uses only a few integer variables (`left`, `right`, `mid`).

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn next_greatest_letter(letters: Vec<char>, target: char) -> char {
        let mut left = 0;
        let mut right = letters.len();

        while left < right {
            let mid = left + (right - left) / 2;
            if letters[mid] > target {
                right = mid;
            } else {
                left = mid + 1;
            }
        }

        // Wrap around to the first character if no greater character is found
        letters[left % letters.len()]
    }
}
```
