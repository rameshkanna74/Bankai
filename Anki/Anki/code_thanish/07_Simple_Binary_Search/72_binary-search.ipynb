### Algorithm Explanation

The problem is a classic **Binary Search** problem where the goal is to find the index of a target value in a sorted array. Binary Search works by repeatedly dividing the search space in half, ensuring that we can find the target value in **O(log n)** time complexity, which is optimal for sorted data.

---

### Time Complexity

- **Time Complexity**: **O(log n)**  
  The binary search algorithm works by halving the search range at each step, so the number of comparisons grows logarithmically with respect to the size of the input array `nums`. This is the optimal time complexity for searching in a sorted array.

- **Space Complexity**: **O(1)**  
  Since we are using a constant amount of extra space (i.e., only variables for the indices), the space complexity is O(1). We are not allocating any additional data structures that depend on the size of the input array.

---

### Edge Cases

1. **Target Not Found**: The target may not be present in the array, and we should return `-1`.
2. **Array Length of 1**: The input array may have a single element. If the element is the target, return its index; otherwise, return `-1`.
3. **Target at First or Last Position**: The target may be at the start or end of the array, and the algorithm should efficiently find it without performing unnecessary steps.
4. **Target at the Middle**: If the target is located exactly in the middle of the search range, the algorithm should return it immediately.
5. **Empty Array**: The array is guaranteed to have at least one element based on the problem constraints, so we don't need to handle the empty array case explicitly.

### Core DSA Concepts

- **Binary Search**: This is the main algorithm used here. It works by comparing the target value with the middle element of the array, then discarding half of the remaining search space based on whether the target is smaller or larger than the middle element. This reduces the search space by half at each step.
- **Arrays**: The input is an array of integers sorted in ascending order.

### Algorithm Design & Reasoning

We are given a sorted array and a target. By applying binary search:

1. We initialize two pointers: `left` at the beginning of the array and `right` at the end.
2. We calculate the middle index and compare the value at this index with the target:
   - If the target is found, we return the index.
   - If the target is less than the middle value, we move the `right` pointer to `mid - 1`.
   - If the target is greater than the middle value, we move the `left` pointer to `mid + 1`.
3. If the `left` pointer exceeds the `right` pointer, it means the target is not present, so we return `-1`.

### Rust Code Implementation

```rust
impl Solution {
    pub fn search(nums: Vec<i32>, target: i32) -> i32 {
        let mut left = 0;
        let mut right = nums.len() as i32 - 1;
        
        while left <= right {
            let mid = left + (right - left) / 2; // Avoid potential overflow
            
            if nums[mid as usize] == target {
                return mid; // Target found at mid
            } else if nums[mid as usize] < target {
                left = mid + 1; // Search in the right half
            } else {
                right = mid - 1; // Search in the left half
            }
        }
        
        -1 // Target not found
    }
}
```

### Explanation of the Code

- We initialize `left` and `right` as indices representing the bounds of the search space.
- The while loop continues until `left` exceeds `right`, meaning there is no more search space.
- In each iteration:
  - We calculate the middle index `mid` using `(left + (right - left) / 2)`, which is a common trick to avoid potential overflow in cases where `left` and `right` are very large.
  - If the element at index `mid` is the target, we return `mid`.
  - If the target is greater than the middle element, we adjust `left` to `mid + 1`.
  - If the target is smaller than the middle element, we adjust `right` to `mid - 1`.
- If the target is not found by the time the loop finishes, we return `-1`.
