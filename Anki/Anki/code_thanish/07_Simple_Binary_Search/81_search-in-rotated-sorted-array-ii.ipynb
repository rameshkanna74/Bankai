## 🧠 **Algorithm Explanation**

### 🔹 **Modified Binary Search (Handling Duplicates)**

Since **binary search** typically works on sorted data, we need a way to handle **rotation** and **duplicates** while still maintaining `O(log n)` efficiency where possible.

1. **Duplicates Handling (Edge Case)**

   - If `nums[left] == nums[mid] == nums[right]`, we **cannot determine the sorted half** reliably.
   - **Solution:** **Increment `left` and decrement `right`** to skip duplicates.

2. **Identify the Sorted Half**

   - If `nums[left] <= nums[mid]`, then the **left half is sorted**.
   - Else, the **right half is sorted**.

3. **Search Within the Sorted Half**

   - If `target` is within the sorted half, continue searching in that half.
   - Otherwise, switch to the unsorted half.

4. **Repeat Until `left > right`**
   - If `nums[mid] == target`, return `true`.
   - If no match is found, return `false`.

### **Time Complexity Analysis**

- **Best case:** `O(log n)` (when no duplicates exist or minimal skips are needed).
- **Worst case:** `O(n)` (when all elements are the same, forcing linear search).
- **Average case:** Slightly worse than `O(log n)` due to possible duplicate skips.

### **Space Complexity Analysis**

- **O(1)** → Uses only a few integer variables (`left`, `right`, `mid`).

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn search(nums: Vec<i32>, target: i32) -> bool {
        let (mut left, mut right) = (0, nums.len() as i32 - 1);

        while left <= right {
            let mid = left + (right - left) / 2;

            // If target is found, return true
            if nums[mid as usize] == target {
                return true;
            }

            // Handle duplicates: If left, mid, and right are the same, shrink the window
            if nums[left as usize] == nums[mid as usize] && nums[mid as usize] == nums[right as usize] {
                left += 1;
                right -= 1;
                continue;
            }

            // Left half is sorted
            if nums[left as usize] <= nums[mid as usize] {
                // Target lies in the sorted left half
                if nums[left as usize] <= target && target < nums[mid as usize] {
                    right = mid - 1;
                } else {
                    left = mid + 1;
                }
            }
            // Right half is sorted
            else {
                // Target lies in the sorted right half
                if nums[mid as usize] < target && target <= nums[right as usize] {
                    left = mid + 1;
                } else {
                    right = mid - 1;
                }
            }
        }

        false
    }
}
```
