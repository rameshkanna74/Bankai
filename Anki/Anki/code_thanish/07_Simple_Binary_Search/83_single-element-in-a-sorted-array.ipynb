### üöÄ **Problem Breakdown**

- The input array is **sorted** and every element appears **exactly twice**, except for **one element** that appears **once**.
- We must **find the single element in O(log n) time** and **O(1) space**.
- **Binary Search** is the optimal approach, as brute force would take **O(n)** time.

---

## üß† **Algorithm Explanation**

### **üîπ Observations**

1. **Even Index & Pairing**:

   - Since every number (except one) appears **twice**, the pairs will always start at **even indices**:  
     Example: `[1, 1, 2, 2, 3, 3, 4]`
   - The **single element disrupts** this pattern.

2. **Binary Search Strategy**:

   - Use **mid-point index**:  
     If `nums[mid] == nums[mid ^ 1]`, the **single element** is in the right half.
   - Otherwise, it's in the left half.

3. **Bitwise Trick (`mid ^ 1`)**:
   - `mid ^ 1` toggles between the **pair indices**:
     - If `mid` is **even**, it becomes `mid + 1` (next pair).
     - If `mid` is **odd**, it becomes `mid - 1` (previous pair).
   - This helps us compare elements efficiently.

- **Time Complexity**: **O(log n)** (Binary search halves the search space each iteration).
- **Space Complexity**: **O(1)** (Uses only integer variables).

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn single_non_duplicate(nums: Vec<i32>) -> i32 {
        let (mut left, mut right) = (0, nums.len() - 1);

        while left < right {
            let mid = left + (right - left) / 2;

            // Check if mid is part of a valid pair
            if nums[mid] == nums[mid ^ 1] {
                left = mid + 1; // The single element is in the right half
            } else {
                right = mid; // The single element is in the left half
            }
        }

        nums[left] // `left` will point to the single element
    }
}
```
