## ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

The problem **Guess Number Higher or Lower (LeetCode 374)** is a **binary search** problem that requires finding the unknown number with minimal guesses using a **logarithmic** approach.

---

### **🧠 Algorithm Explanation**
Since we can only ask if our guess is **higher, lower, or correct**, we can **use Binary Search** to find the picked number efficiently.

#### **🚀 Approach: Binary Search**
1. Initialize **left = 1** and **right = n**.
2. While `left <= right`:
   - Compute `mid = left + (right - left) / 2` (avoiding overflow).
   - Use the given `guess(mid)` API:
     - If `guess(mid) == 0`, return `mid` (correct answer).
     - If `guess(mid) == -1`, update `right = mid - 1` (search left).
     - If `guess(mid) == 1`, update `left = mid + 1` (search right).
3. Return `-1` (this should never be reached since the pick is always valid).

✅ **Time Complexity:**  
- **O(log n)** → Since we halve the search space at every step.

✅ **Space Complexity:**  
- **O(1)** → Uses only a few integer variables (`left`, `right`, `mid`).

    pub unsafe fn guessNumber(n: i32) -> i32 {
        let mut left = 1;
        let mut right = n;

        while left <= right {
            let mid = left + (right - left) / 2;

            match guess(mid) {
                0 => return mid,    // Correct guess
                -1 => right = mid - 1, // Guess is too high, move left
                1 => left = mid + 1,  // Guess is too low, move right
                _ => unreachable!(),  // The API should never return an invalid value
            }
        }

        -1 // Should never reach here
    }
