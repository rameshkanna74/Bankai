## ✅ **Optimized, Production-Quality, and Interview-Ready Rust Solution**

The problem requires us to find the **index of `target`** in a sorted array. If `target` is not found, return the **index where it should be inserted** while maintaining the sorted order.

Since the array is **sorted** and we need an **O(log n) solution**, **Binary Search** is the optimal approach.

---

## 🧠 **Algorithm Explanation**

### 🚀 **Approach: Binary Search**

1. **Initialize Search Space**
   - `left = 0`, `right = nums.len() - 1`
2. **Perform Binary Search**

   - Compute `mid = left + (right - left) / 2` (avoiding overflow).
   - If `nums[mid] == target`, return `mid` (target found).
   - If `nums[mid] < target`, move right (`left = mid + 1`).
   - Otherwise, move left (`right = mid - 1`).

3. **Return `left` as the Insert Position**
   - When the loop ends, `left` holds the **smallest index where `nums[left] >= target`**.
   - If `target` is larger than all elements, `left == nums.len()`, correctly indicating the insert position at the end.

### **✅ Time Complexity:**

- **O(log n)** → Since we're using Binary Search.

### **✅ Space Complexity:**

- **O(1)** → Uses only a few integer variables (`left`, `right`, `mid`).

---

## 🦀 **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn search_insert(nums: Vec<i32>, target: i32) -> i32 {
        let mut left = 0;
        let mut right = nums.len() as i32; // `right` set to nums.len() for proper insert position

        while left < right {
            let mid = left + (right - left) / 2;
            if nums[mid as usize] < target {
                left = mid + 1;
            } else {
                right = mid;
            }
        }

        left
    }
}
```
