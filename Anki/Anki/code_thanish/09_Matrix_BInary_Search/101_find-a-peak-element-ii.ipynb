## üß† **Problem Breakdown**

We are given an **`m x n`** matrix where:

1. **No two adjacent cells are equal**.
2. We need to find any **peak element** `(i, j)`, where:
   - `mat[i][j] > mat[i-1][j]` (above)
   - `mat[i][j] > mat[i+1][j]` (below)
   - `mat[i][j] > mat[i][j-1]` (left)
   - `mat[i][j] > mat[i][j+1]` (right)
3. **Time complexity must be `O(m log(n))` or `O(n log(m))`.**

---

## üöÄ **Optimal Approach: Binary Search on Columns**

### **Key Observations**

- A brute-force approach (`O(m * n)`) is too slow.
- **Binary Search is ideal** since we are asked for `O(m log(n))` or `O(n log(m))`.
- We can efficiently find a peak using **binary search on columns**.

### **Algorithm Design (Binary Search on Columns)**

1. **Start with the middle column** `mid_col = (left + right) / 2`.
2. **Find the row with the max value in this column** (it is a good peak candidate).
3. **Check left and right neighbors**:
   - If the peak candidate is **greater than both left and right neighbors**, return it.
   - If the **left neighbor is greater**, move the search left (`right = mid_col - 1`).
   - If the **right neighbor is greater**, move the search right (`left = mid_col + 1`).
4. **Repeat until a peak is found**.

### **Why is this Efficient?**

- Each step reduces the search space **by half**, leading to `O(m log(n))` time complexity.

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_peak_grid(mat: Vec<Vec<i32>>) -> Vec<i32> {
        let (m, n) = (mat.len(), mat[0].len());
        let mut left = 0;
        let mut right = n - 1;

        while left <= right {
            let mid_col = (left + right) / 2;

            // Find the row index of the maximum value in the middle column
            let max_row = (0..m).max_by_key(|&r| mat[r][mid_col]).unwrap();

            let left_val = if mid_col > 0 { mat[max_row][mid_col - 1] } else { -1 };
            let right_val = if mid_col < n - 1 { mat[max_row][mid_col + 1] } else { -1 };

            if mat[max_row][mid_col] > left_val && mat[max_row][mid_col] > right_val {
                return vec![max_row as i32, mid_col as i32];
            } else if left_val > mat[max_row][mid_col] {
                right = mid_col - 1;
            } else {
                left = mid_col + 1;
            }
        }

        unreachable!() // The problem guarantees a peak exists, so we will always return.
    }
}
```
