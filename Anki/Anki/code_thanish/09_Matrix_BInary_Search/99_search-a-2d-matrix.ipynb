## üß† **Problem Breakdown**

We are given an `m x n` matrix where:

1. **Each row is sorted in non-decreasing order.**
2. **The first element of each row is greater than the last element of the previous row.**

Effectively, **the entire matrix can be treated as a sorted 1D array** when **flattened** row-wise.

### **Key Observations**

- **Flattening the matrix gives a sorted 1D array** of size `m * n`.
- **We can apply Binary Search on this virtual array** in `O(log(m * n))` time.
- **Converting 1D indices back to 2D indices**:
  - Row index: `row = mid / n`
  - Column index: `col = mid % n`

---

## üöÄ **Algorithm Explanation**

### **üîπ Approach: Treat the Matrix as a 1D Sorted Array**

1. **Binary Search on the virtual 1D array**

   - Compute `mid = (left + right) / 2`.
   - Convert `mid` to **row and column indices** using:
     - `row = mid / n`
     - `col = mid % n`
   - Compare `matrix[row][col]` with `target`:
     - If **equal**, return `true`.
     - If **less than `target`**, move right (`left = mid + 1`).
     - If **greater than `target`**, move left (`right = mid - 1`).

2. **Binary Search runs in `O(log(m * n))`**, which is the best possible time complexity for this problem.

```rust
impl Solution {
    pub fn search_matrix(matrix: Vec<Vec<i32>>, target: i32) -> bool {
        let (m, n) = (matrix.len(), matrix[0].len());
        let (mut left, mut right) = (0, m * n - 1);

        while left <= right {
            let mid = left + (right - left) / 2;
            let row = mid / n;
            let col = mid % n;
            let mid_val = matrix[row][col];

            if mid_val == target {
                return true;
            } else if mid_val < target {
                left = mid + 1;
            } else {
                right = mid - 1;
            }
        }
        false
    }
}
```
