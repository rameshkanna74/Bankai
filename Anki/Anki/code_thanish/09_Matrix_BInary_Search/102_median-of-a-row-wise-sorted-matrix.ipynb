## üß† **Problem Breakdown**

We are given a **row-wise sorted matrix** of size `m x n` where `m` and `n` are **always odd**.  
Our goal is to **find the median** of the matrix.

### **Key Observations**

1. Since `m √ó n` is **always odd**, the **median is the middle element** of the sorted matrix.
2. **Na√Øve Approach (Sorting)**:

   - Flatten the matrix into an array ‚Üí `O(m * n)`.
   - Sort the array ‚Üí `O(m * n log(m * n))`.
   - Return the middle element.
   - **Total Complexity:** `O(m * n log(m * n))` ‚õî **Too slow for large matrices.**

3. **Optimal Approach (Binary Search)**
   - Since the matrix is **row-wise sorted**, we can apply **binary search on values** (not indices).
   - The **median** must be the `k = (m * n) / 2`-th smallest element.
   - **Efficiently count how many elements are ‚â§ `x`** using **binary search in each row**.
   - Adjust the search range based on this count.
   - **Total Complexity:** `O(32 * m * log(n)) ‚âà O(m * log(n))` ‚úÖ **Optimal!**

---

## üöÄ **Optimal Approach: Binary Search on the Value Range**

### **Algorithm Design**

1. **Set Binary Search Boundaries**:

   - The smallest element in the matrix is in the first column ‚Üí `low = mat[0][0]`.
   - The largest element in the matrix is in the last column ‚Üí `high = mat[m-1][n-1]`.

2. **Binary Search on Median Value**:
   - Find `mid = (low + high) / 2`.
   - **Count elements ‚â§ `mid`** using `O(log(n))` **binary search in each row**.
   - If count **‚â§ median index**, move `low = mid + 1` (median must be larger).
   - Else, move `high = mid` (median must be smaller).
   - **Repeat until `low == high`**, which will be the median.

---

## ‚è≥ **Time and Space Complexity**

‚úÖ **Time Complexity:** `O(m log(n))`

- **Binary search range is limited to 32 steps (`log(10^9) ‚âà 30`).**
- **Each step involves `m` binary searches (`O(log(n))` per row).**
- **Total = `O(32 * m * log(n)) ‚âà O(m log(n))`** ‚úÖ

‚úÖ **Space Complexity:** `O(1)`

- We use **only constant extra space**.

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn find_median(matrix: Vec<Vec<i32>>) -> i32 {
        let (m, n) = (matrix.len(), matrix[0].len());
        let median_pos = (m * n) / 2; // 0-based index

        let mut low = matrix[0][0];
        let mut high = matrix[m - 1][n - 1];

        while low < high {
            let mid = low + (high - low) / 2;
            let mut count = 0;

            // Count elements <= mid using binary search in each row
            for row in &matrix {
                count += row.partition_point(|&x| x <= mid);
            }

            if count <= median_pos {
                low = mid + 1;
            } else {
                high = mid;
            }
        }

        low
    }
}
```
