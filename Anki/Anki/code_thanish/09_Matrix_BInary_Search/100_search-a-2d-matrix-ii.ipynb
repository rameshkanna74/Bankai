## üß† **Problem Breakdown**

We are given an **`m x n`** matrix with the following properties:

1. **Each row is sorted in ascending order** (left to right).
2. **Each column is sorted in ascending order** (top to bottom).

We need to determine if a given `target` exists in the matrix.

### **Key Observations**

- Since **rows and columns are sorted**, we can exploit this structure for efficient searching.
- **Brute force (`O(m * n)`) is too slow** ‚Üí Need an **optimized approach**.
- **Binary Search (`O(log(m * n))`) is hard to apply directly** because sub-matrices are not necessarily sorted.
- **Optimal approach: Start searching from the top-right or bottom-left** (explained below).

---

## üöÄ **Algorithm Explanation**

### **üîπ Approach: Start from the Top-Right or Bottom-Left**

1. **Start at the top-right corner** (or bottom-left corner).
   - `row = 0`, `col = n - 1` (top-right).
2. **Iterate while within bounds (`row < m` and `col >= 0`):**
   - If `matrix[row][col] == target`, return `true`.
   - If `matrix[row][col] > target`, **move left** (`col -= 1`) ‚Üí Eliminates the current column.
   - If `matrix[row][col] < target`, **move down** (`row += 1`) ‚Üí Eliminates the current row.
3. **If we exit the loop, return `false` (target not found).**

### **Why does this work?**

- **Every step eliminates either a row or a column.**
- **Since we only traverse `O(m + n)` elements**, this is highly efficient.

‚úÖ **Best-case:** `O(1)` (if found early).  
‚úÖ **Worst-case:** `O(m + n)` (full traversal in diagonal).

---

## ‚è≥ **Time and Space Complexity**

‚úÖ **Time Complexity:** `O(m + n)` ‚Üí **Efficient for large matrices**  
‚úÖ **Space Complexity:** `O(1)` ‚Üí **No extra space used**

## ü¶Ä **Optimized Rust Implementation**

```rust
impl Solution {
    pub fn search_matrix(matrix: Vec<Vec<i32>>, target: i32) -> bool {
        let (m, n) = (matrix.len(), matrix[0].len());
        let (mut row, mut col) = (0, n - 1);

        while row < m && col as isize >= 0 {
            let val = matrix[row][col];

            if val == target {
                return true;
            } else if val > target {
                col -= 1; // Move left
            } else {
                row += 1; // Move down
            }
        }
        false
    }
}
```
