```python
from pdf2image import convert_from_path

# Path to your PDF file
pdf_path = "/home/kanna/Django/dev/drf/pdf2image/24120048230043.pdf"
# Output folder to save images
output_folder = "output_images"

# DPI for high-quality images (e.g., 300 or higher)
dpi = 300

# Convert PDF to images
images = convert_from_path(pdf_path, dpi=dpi)

# Save images as files
for i, image in enumerate(images):
    image_path = f"{output_folder}/page_{i+1}.png"  # PNG ensures no quality loss
    image.save(image_path, "PNG")
    print(f"Saved: {image_path}")
```

```python
import os
from pdf2image import convert_from_path

# Path to your PDF file
pdf_path = "/home/kanna/Django/dev/drf/pdf2image/24120048230043.pdf"

# Output folder to save images
output_folder = "/home/kanna/Django/dev/drf/output_images"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# DPI for high-quality images
dpi = 300

# Convert PDF to images
images = convert_from_path(pdf_path, dpi=dpi)

# Save images as files
for i, image in enumerate(images):
    image_path = f"{output_folder}/page_{i+1}.png"  # PNG ensures no quality loss
    image.save(image_path, "PNG")
    print(f"Saved: {image_path}")
```
