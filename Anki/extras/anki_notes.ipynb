        // Initialize pointers for both arrays
        let mut i = 0;

            let val1 = nums1[i][1];
            let id2 = nums2[j][0];
            let val2 = nums2[j][1];

                result.push(vec![id1, val1]);
                i += 1;
            } else if id2 < id1 {
                // Id2 is smaller, add it to result

                j += 1;

                // Ids are equal, sum the values

                i += 1;

```rs
fn count_factors(n: i32) -> i32 {
    if n == 0 {
        return 0; // Special case: infinite divisors
    }
    let n = n.abs(); // Use absolute value to handle negatives
    let mut divisor_count = 0;

    // Loop through all integers from 1 to sqrt(n)
    for i in 1..=((n as f64).sqrt() as i32) {
        if n % i == 0 {
            divisor_count += 1; // i is a divisor
            if i != n / i {
                divisor_count += 1; // Pair divisor
            }
        }
    }
    divisor_count
}
```

```rs
fn is_prime(n: i64) -> bool {
    // Handle edge cases for numbers <= 1
    if n <= 1 {
        return false; // Not prime
    }
    
    // Handle small primes
    if n == 2 || n == 3 {
        return true; // Prime
    }
    
    // Eliminate even numbers and multiples of 3
    if n % 2 == 0 || n % 3 == 0 {
        return false; // Not prime
    }
    
    // Check divisors from 5 to sqrt(n), using 6k Â± 1 optimization
    let limit = (n as f64).sqrt() as i64;
    let mut i = 5;
    while i <= limit {
        if n % i == 0 || n % (i + 2) == 0 {
            return false; // Not prime
        }
        i += 6; // Increment by 6 to check next potential factors
    }
    
    true // Prime
}
```

```rs
fn unique_prime_factors(mut n: i64) -> Vec<i64> {
    let mut prime_factors = Vec::new();
    
    // Handle edge case for numbers <= 1
    if n <= 1 {
        return prime_factors; // Return an empty vector
    }

    // Check divisibility by 2
    if n % 2 == 0 {
        prime_factors.push(2);
        while n % 2 == 0 {
            n /= 2;
        }
    }

    // Check for odd prime factors from 3 to sqrt(n)
    let mut divisor = 3;
    while divisor * divisor <= n {
        if n % divisor == 0 {
            prime_factors.push(divisor);
            while n % divisor == 0 {
                n /= divisor;
            }
        }
        divisor += 2;
    }

    // If n > 1, it is a prime factor
    if n > 1 {
        prime_factors.push(n);
    }

    prime_factors
}

```

```rs
impl Solution {
    pub fn merge_alternately(word1: String, word2: String) -> String {
        let len1 = word1.len();
        let len2 = word2.len();
        let min_len = len1.min(len2);
        
        let mut merged_string = String::with_capacity(len1 + len2);
        
        // Use iterators for better performance
        let mut chars1 = word1.chars();
        let mut chars2 = word2.chars();

        // Alternate adding characters from both words
        for _ in 0..min_len {
            merged_string.push(chars1.next().unwrap());
            merged_string.push(chars2.next().unwrap());
        }

        // Append the remainder of the longer word
        if len1 > min_len {
            merged_string.push_str(&word1[min_len..]);
        }
        if len2 > min_len {
            merged_string.push_str(&word2[min_len..]);
        }

        merged_string
    }
}

```
