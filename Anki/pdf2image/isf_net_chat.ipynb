```python
import subprocess
import os

def find_apkg_files():
    start_directory = "/home/kanna/"
    print("Searching in:", start_directory)

    # Use subprocess to find .apkg files more efficiently
    find_command = ["find", start_directory, "-type", "f", "-name", "*.apkg", "-print0"]
    result = subprocess.run(find_command, capture_output=True)

    # Print the output for debugging
    if result.stdout:
        print("Found files:", result.stdout.decode("utf-8").split("\0"))

        # Use fzf to select a file from the results
        fzf_command = ["fzf", "--read0"]
        # Decode the stdout and pass it to fzf
        selected_file = subprocess.run(
            fzf_command, input=result.stdout.decode("utf-8"), capture_output=True, text=True
        ).stdout.strip()

        # Save the selected file to output.txt
        if selected_file:
            with open("selected_file.txt", "w") as f:
                f.write(selected_file)
            print(f"Selected file saved: {selected_file}")
        else:
            print("No file selected.")
    else:
        print("No .apkg files found.")


if __name__ == "__main__":
    find_apkg_files()
```

To prepare for the Cisco Certified Network Associate (CCNA) exam in 3 months and become fully proficient, you’ll need a structured and efficient approach to cover the necessary topics. The CCNA exam tests your understanding of networking concepts, including the basics of networking, IP addressing, subnetting, routing and switching, security, and troubleshooting. Below is a detailed 3-month roadmap to prepare for the CCNA exam, with a focus on mastering the concepts:

### **Month 1: Build a Strong Foundation**

#### **Week 1: Introduction to Networking and CCNA Overview**

1. **Study Resources:**
   - Start with the **Cisco CCNA 200-301 Official Cert Guide** by Wendell Odom or online courses such as **CBT Nuggets** or **Udemy (Network Chuck or David Bombal)**.
   - You can also use **Cisco Networking Academy** for structured learning.
2. **Key Concepts:**
   - What is networking? Understand **network types** (LAN, WAN, MAN), **network devices** (routers, switches, hubs, etc.), and their roles.
   - Learn about **OSI model** and **TCP/IP model**.
   - Types of IP addresses (IPv4 vs IPv6), basic **IP addressing** concepts, and **subnetting** basics.
3. **Hands-on Practice:**
   - Set up **Cisco Packet Tracer** or use **GNS3** or **Cisco VIRL** for simulation labs.
   - Practice basic network configurations like **ping**, **ipconfig**, and **traceroute**.

#### **Week 2: Networking Devices and Basic Routing**

1. **Key Concepts:**
   - Deep dive into **routers, switches, and hubs**. Understand their functions and differences.
   - **Routing protocols**: Introduction to **RIP**, **OSPF**, and **EIGRP**.
   - **Static routing** and **dynamic routing** protocols.
   - **VLANs** and their role in network segmentation.
2. **Hands-on Practice:**
   - Set up and configure a basic **router** and **switch** in Packet Tracer.
   - Create and configure **VLANs**.
   - Implement basic **static routing** and explore the **routing table**.
3. **CCNA Practice Tests:**
   - Start taking **daily quizzes** on networking basics to reinforce learning.

#### **Week 3: IP Addressing and Subnetting**

1. **Key Concepts:**
   - Master **IP subnetting**, including **subnet masks**, **CIDR** notation, and **VLSM**.
   - Learn **private and public IP address ranges**.
   - Understand **IP address classes** (A, B, C, D, E).
2. **Hands-on Practice:**

   - Practice **subnetting exercises** daily to gain fluency.
   - Use tools like **Subnetting Practice Apps** or websites to improve subnetting speed.

3. **CCNA Practice Tests:**
   - Focus on subnetting practice exams.

#### **Week 4: Network Layer Protocols and Routing**

1. **Key Concepts:**
   - Study **IP routing** in depth, including **dynamic routing** and **distance-vector** protocols (RIP).
   - Understand **OSPF**: its role, process, and configuration.
2. **Hands-on Practice:**
   - Configure **OSPF** and **RIP** on routers in your lab environment.
   - Practice **network design** with multiple routers.

### **Month 2: Routing, Switching, and Network Security**

#### **Week 5: Switching and VLANs**

1. **Key Concepts:**
   - Understand **switching concepts**, including **Layer 2 vs Layer 3 switching**, and **MAC addresses**.
   - Learn about **Spanning Tree Protocol (STP)** and its role in preventing loops.
   - Master **VLAN configuration** and **Inter-VLAN routing**.
2. **Hands-on Practice:**
   - Configure switches to create and assign **VLANs**.
   - Practice **Trunking** with **802.1q**.
   - Configure **Inter-VLAN Routing** using a router or Layer 3 switch.

#### **Week 6: Routing Protocols and Network Design**

1. **Key Concepts:**
   - Dive deeper into **OSPF**, covering **OSPF areas**, **DR/BDR**, and **LSA types**.
   - Learn about **EIGRP** (configuration, advantages, and metrics).
   - Learn **Route Summarization**, **load balancing**, and **redistribution**.
2. **Hands-on Practice:**
   - Configure **OSPF** and **EIGRP** in your lab.
   - Practice **Route Summarization** and **Default Routing**.

#### **Week 7: Network Security Basics**

1. **Key Concepts:**
   - Study **ACLs (Access Control Lists)**: standard and extended ACLs for securing a network.
   - Introduction to **NAT (Network Address Translation)** and **PAT (Port Address Translation)**.
   - Learn about **VPNs** and **Firewall** basics.
2. **Hands-on Practice:**
   - Configure basic **ACLs** to filter traffic.
   - Practice **NAT** and **PAT** configuration on routers.
   - Explore setting up a **VPN**.

#### **Week 8: Troubleshooting and Network Monitoring**

1. **Key Concepts:**
   - Learn troubleshooting techniques for **network connectivity issues**, **routing problems**, and **VLAN issues**.
   - Study **ping**, **traceroute**, and **debugging** commands.
2. **Hands-on Practice:**
   - Simulate network issues in Packet Tracer or GNS3 and practice troubleshooting.
   - Use network monitoring tools like **Wireshark** to capture and analyze network traffic.

### **Month 3: Advanced Topics, Mock Tests, and Exam Preparation**

#### **Week 9: Advanced Routing, Switching, and Security**

1. **Key Concepts:**
   - Learn about **Frame Relay**, **WAN technologies**, and **PPP**.
   - Explore advanced **QoS (Quality of Service)** concepts.
   - Study more on **Wireless LANs (Wi-Fi)** and **Security Features** like **AAA** (Authentication, Authorization, Accounting).
2. **Hands-on Practice:**
   - Set up advanced WAN protocols like **Frame Relay** and **PPP**.
   - Configure **QoS** and **Wireless Security** on your network devices.

#### **Week 10: Review and Deep Dive into Weak Areas**

1. **Key Concepts:**
   - Go back and **review your weak areas**: subnetting, routing protocols, security, or VLAN configurations.
   - Focus on the most important topics that come up frequently in CCNA exams.
2. **Practice:**
   - Revisit any practice tests or quiz results that show areas of improvement.
   - Reinforce concepts through additional practical labs.

#### **Week 11: Full-Length Mock Exams and Practice Tests**

1. **Key Activities:**
   - Take **full-length CCNA practice exams** under timed conditions.
   - Analyze the results and review questions that you got wrong.
   - Revisit key topics like **subnetting**, **OSPF**, and **VLANs**.
2. **Resources:**
   - Use resources like **Boson ExSim**, **ExamCompass**, or **CCNA Study Guides** for mock exams.

#### **Week 12: Final Revision and Exam Preparation**

1. **Key Activities:**

   - Perform a final **review** of all topics, especially the ones you struggle with.
   - Take a final set of **practice exams** to ensure your readiness.
   - Review important commands, configurations, and common troubleshooting steps.

2. **Exam Day Prep:**
   - Relax and get ready for the exam by reviewing basic concepts and getting a good night’s sleep.

### **Additional Tips:**

- **Stay Consistent**: Dedicate at least 2-3 hours per day to studying and hands-on practice.
- **Use Multiple Resources**: Mix textbooks, video tutorials, and labs for a well-rounded approach.
- **Join Study Groups**: Join online forums or study groups (e.g., Reddit, Cisco forums) to ask questions and share knowledge.
- **Simulate Real Environments**: Use Packet Tracer or GNS3 to simulate different network topologies and troubleshoot.

By following this roadmap, you'll be well-prepared to take the CCNA exam after 3 months and gain solid proficiency in networking concepts, which will complement your skills as a Python backend developer. Best of luck with your certification journey!
