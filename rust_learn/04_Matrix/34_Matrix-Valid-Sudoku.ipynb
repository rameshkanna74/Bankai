36. Valid Sudoku

Determine if a 9 x 9 Sudoku board is valid. Only the filled cells need to be validated according to the following rules:

    Each row must contain the digits 1-9 without repetition.
    Each column must contain the digits 1-9 without repetition.
    Each of the nine 3 x 3 sub-boxes of the grid must contain the digits 1-9 without repetition.

Note:

    A Sudoku board (partially filled) could be valid but is not necessarily solvable.
    Only the filled cells need to be validated according to the mentioned rules.

,[".","6",".",".",".",".","2","8","."]

Output: true

To determine if a 9x9 Sudoku board is valid, you need to ensure that each row, each column, and each 3x3 sub-box contain no repeating digits (1-9) and respect the Sudoku rules. You can achieve this efficiently using sets to track occurrences.

### Approach

1. **Use Sets for Validation**:

   - Utilize three sets for each type of validation:
     - One set for each row.
     - One set for each column.
     - One set for each 3x3 sub-box.

2. **Process the Board**:

   - Iterate through each cell of the board.
   - If the cell contains a digit (not '.'), check if the digit is already in the corresponding row, column, or 3x3 sub-box set.
   - If the digit is already present in any of these sets, the board is invalid.
   - Otherwise, add the digit to the respective sets.

3. **Check Sub-boxes**:
   - To identify the correct sub-box, calculate its index based on the current cell's row and column. The index can be derived from `(row / 3) * 3 + (col / 3)`.

### Rust Implementation

Here's how you can implement the validation in Rust:

impl Solution {
    pub fn is_valid_sudoku(board: Vec<Vec<char>>) -> bool {
        // Define sets to track seen digits
        let mut rows = vec![std::collections::HashSet::new(); 9];
        let mut cols = vec![std::collections::HashSet::new(); 9];
        let mut boxes = vec![std::collections::HashSet::new(); 9];

        for row in 0..9 {
            for col in 0..9 {
                let cell = board[row][col];
                if cell != '.' {
                    // Calculate the index of the 3x3 sub-box
                    let box_index = (row / 3) * 3 + (col / 3);

                    // Check if the digit is already in the current row, column, or sub-box
                    if rows[row].contains(&cell) || cols[col].contains(&cell) || boxes[box_index].contains(&cell) {
                        return false;
                    }

                    // Add the digit to the respective sets
                    rows[row].insert(cell);
                    cols[col].insert(cell);
                    boxes[box_index].insert(cell);
                }
            }
        }

        true
    }
}

   - Otherwise, add the digit to the appropriate sets.

3. **Return Result**:
   - If no conflicts are found, return `true`, indicating the board is valid.

pub fn is_valid_sudoku(board: Vec<Vec<char>>) -> bool {
    // Use arrays of bool to track digits (1-9)
    let mut rows = [[false; 9]; 9];
    let mut cols = [[false; 9]; 9];
    let mut boxes = [[false; 9]; 9];

    for row in 0..9 {
        for col in 0..9 {
            let cell = board[row][col];
            if cell != '.' {
                let digit = (cell as usize) - ('1' as usize); // Convert char to index (0-8)
                let box_index = (row / 3) * 3 + (col / 3);

                // Check if the digit is already present in the current row, column, or sub-box
                if rows[row][digit] || cols[col][digit] || boxes[box_index][digit] {
                    return false;
                }

                // Mark the digit as seen
                rows[row][digit] = true;
                cols[col][digit] = true;
                boxes[box_index][digit] = true;
            }
        }
    }

    true
}
