impl Solution {
    pub fn spiral_order(matrix: Vec<Vec<i32>>) -> Vec<i32> {
        let mut result = Vec::new();
        if matrix.is_empty() {
            return result;
        }

        // Define the boundaries for the spiral traversal
        let (mut top, mut bottom) = (0, matrix.len() as i32 - 1);
        let (mut left, mut right) = (0, matrix[0].len() as i32 - 1);

        // Continue traversing until the pointers overlap
        while top <= bottom && left <= right {
            // Traverse from left to right along the top row
            for col in left..=right {
                result.push(matrix[top as usize][col as usize]);
            }
            top += 1; // Move the top boundary down

            // Traverse from top to bottom along the right column
            for row in top..=bottom {
                result.push(matrix[row as usize][right as usize]);
            }
            right -= 1; // Move the right boundary left

            // Traverse from right to left along the bottom row (if still within bounds)
            if top <= bottom {
                for col in (left..=right).rev() {
                    result.push(matrix[bottom as usize][col as usize]);
                }
                bottom -= 1; // Move the bottom boundary up
            }

            // Traverse from bottom to top along the left column (if still within bounds)
            if left <= right {
                for row in (top..=bottom).rev() {
                    result.push(matrix[row as usize][left as usize]);
                }
                left += 1; // Move the left boundary right
            }
        }

        result
    }
}

### Explanation:

1. **Initialization**:
    - We start by defining `top`, `bottom`, `left`, and `right` boundaries to track the edges of the matrix as we spiral inward.
    - `top` and `bottom` represent the row boundaries, while `left` and `right` represent the column boundaries.

2. **Spiral Traversal**:
    - **Left to Right (Top Row)**: Traverse the top boundary from left to right and then move the top boundary down.
    - **Top to Bottom (Right Column)**: Traverse the right boundary from top to bottom and then move the right boundary left.
    - **Right to Left (Bottom Row)**: Traverse the bottom boundary from right to left, only if the top boundary is still above the bottom, then move the bottom boundary up.
    - **Bottom to Top (Left Column)**: Traverse the left boundary from bottom to top, only if the left boundary is still to the left of the right, then move the left boundary right.
    - This process repeats until the boundaries overlap.

3. **Edge Case**:
    - If the matrix is empty, the function returns an empty result.

### Complexity:

- **Time Complexity**: \(O(m \times n)\), where \(m\) and \(n\) are the dimensions of the matrix. Every element is visited exactly once.
- **Space Complexity**: \(O(1)\) additional space is used, excluding the space needed for the output list.

This code efficiently handles the spiral order traversal and is suitable for matrices of various sizes and shapes.

````rust
pub fn spiral_order(matrix: Vec<Vec<i32>>) -> Vec<i32> {
    let mut result = Vec::new();
    if matrix.is_empty() {
        return result;
    }

    // Define the boundaries for the spiral traversal
    let (mut top, mut bottom) = (0, matrix.len() as i32 - 1);
    let (mut left, mut right) = (0, matrix[0].len() as i32 - 1);

    // Continue traversing until the pointers overlap
    while top <= bottom && left <= right {
        // Traverse from left to right along the top row
        for col in left..=right {
            result.push(matrix[top as usize][col as usize]);
        }
        top += 1; // Move the top boundary down

        // Traverse from top to bottom along the right column
        for row in top..=bottom {
            result.push(matrix[row as usize][right as usize]);
        }
        right -= 1; // Move the right boundary left

        // Traverse from right to left along the bottom row (if still within bounds)
        if top <= bottom {
            for col in (left..=right).rev() {
                result.push(matrix[bottom as usize][col as usize]);
            }
            bottom -= 1; // Move the bottom boundary up
        }

        // Traverse from bottom to top along the left column (if still within bounds)
        if left <= right {
            for row in (top..=bottom).rev() {
                result.push(matrix[row as usize][left as usize]);
            }
            left += 1; // Move the left boundary right
        }
    }

    result
}


// Example usage
fn main() {
    // Example case 1
    let matrix1 = vec![
        vec![1, 2, 3],
        vec![4, 5, 6],
        vec![7, 8, 9]
    ];
    let result1 = spiral_order(matrix1);
    println!("Spiral order for example 1: {:?}", result1); // Expected output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

    // Example case 2
    let matrix2 = vec![
        vec![1, 2, 3, 4],
        vec![5, 6, 7, 8],
        vec![9, 10, 11, 12]
    ];
    let result2 = spiral_order(matrix2);
    println!("Spiral order for example 2: {:?}", result2); // Expected output: [1, 2, 3, 4, 8, 12, 11, 10, 9, 5, 6, 7]
}

main()
````
