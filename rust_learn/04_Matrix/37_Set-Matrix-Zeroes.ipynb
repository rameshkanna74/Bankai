73. Set Matrix Zeroes
    Medium
    Topics
    Companies
    Hint

Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

You must do it in place.

Example 1:

Input: matrix = [[1,1,1],[1,0,1],[1,1,1]]
Output: [[1,0,1],[0,0,0],[1,0,1]]

Example 2:

Input: matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
Output: [[0,0,0,0],[0,4,5,0],[0,3,1,0]]

Constraints:

    m == matrix.length
    n == matrix[0].length
    1 <= m, n <= 200
    -231 <= matrix[i][j] <= 231 - 1

Follow up:

    A straightforward solution using O(mn) space is probably a bad idea.
    A simple improvement uses O(m + n) space, but still not the best solution.
    Could you devise a constant space solution?

Here's a Rust solution for the "Set Matrix Zeroes" problem with optimal time and space complexity. The approach uses constant space by utilizing the first row and first column to mark zeroes.

### Rust Solution:

```rust
impl Solution {
    pub fn set_zeroes(matrix: &mut Vec<Vec<i32>>) {
        if matrix.is_empty() || matrix[0].is_empty() {
            return;
        }

        let mut first_row_has_zero = false;
        let mut first_col_has_zero = false;

        let num_rows = matrix.len();
        let num_cols = matrix[0].len();

        // Check if the first row has any zeroes
        for j in 0..num_cols {
            if matrix[0][j] == 0 {
                first_row_has_zero = true;
                break;
            }
        }

        // Check if the first column has any zeroes
        for i in 0..num_rows {
            if matrix[i][0] == 0 {
                first_col_has_zero = true;
                break;
            }
        }

        // Use first row and first column to mark zeroes for other cells
        for i in 1..num_rows {
            for j in 1..num_cols {
                if matrix[i][j] == 0 {
                    matrix[i][0] = 0;  // Mark the start of the row
                    matrix[0][j] = 0;  // Mark the start of the column
                }
            }
        }

        // Set zeroes for marked rows and columns
        for i in 1..num_rows {
            if matrix[i][0] == 0 {
                for j in 1..num_cols {
                    matrix[i][j] = 0;
                }
            }
        }

        for j in 1..num_cols {
            if matrix[0][j] == 0 {
                for i in 1..num_rows {
                    matrix[i][j] = 0;
                }
            }
        }

        // Set zeroes for the first row if needed
        if first_row_has_zero {
            for j in 0..num_cols {
                matrix[0][j] = 0;
            }
        }

        // Set zeroes for the first column if needed
        if first_col_has_zero {
            for i in 0..num_rows {
                matrix[i][0] = 0;
            }
        }
    }
}

fn main() {
    let mut matrix1 = vec![
        vec![1, 1, 1],
        vec![1, 0, 1],
        vec![1, 1, 1],
    ];
    Solution::set_zeroes(&mut matrix1);
    println!("Output for example 1: {:?}", matrix1);

    let mut matrix2 = vec![
        vec![0, 1, 2, 0],
        vec![3, 4, 5, 2],
        vec![1, 3, 1, 5],
    ];
    Solution::set_zeroes(&mut matrix2);
    println!("Output for example 2: {:?}", matrix2);
}
```

### Explanation:

1. **Initial Checks:** Determine if the first row and/or the first column need to be zeroed. This helps us manage cases where the first row or column itself has zeroes.

2. **Marking:** Use the first row and first column to mark which rows and columns should be zeroed.

3. **Zeroing Rows and Columns:** Based on the markings in the first row and column, set the appropriate rows and columns to zero.

4. **Final Adjustments:** Handle zeroing of the first row and column if they were originally supposed to be zeroed.

### Complexity:

- **Time Complexity:** O(m \* n), where `m` is the number of rows and `n` is the number of columns. This is because we traverse the matrix a few times.
- **Space Complexity:** O(1), as we use only a few extra variables and modify the matrix in place.

````rust
pub fn set_zeroes(matrix: &mut Vec<Vec<i32>>) {
    if matrix.is_empty() || matrix[0].is_empty() {
        return;
    }
    
    let mut first_row_has_zero = false;
    let mut first_col_has_zero = false;
    
    let num_rows = matrix.len();
    let num_cols = matrix[0].len();
    
    // Check if the first row has any zeroes
    for j in 0..num_cols {
        if matrix[0][j] == 0 {
            first_row_has_zero = true;
            break;
        }
    }
    
    // Check if the first column has any zeroes
    for i in 0..num_rows {
        if matrix[i][0] == 0 {
            first_col_has_zero = true;
            break;
        }
    }
    
    // Use first row and first column to mark zeroes for other cells
    for i in 1..num_rows {
        for j in 1..num_cols {
            if matrix[i][j] == 0 {
                matrix[i][0] = 0;  // Mark the start of the row
                matrix[0][j] = 0;  // Mark the start of the column
            }
        }
    }
    
    // Set zeroes for marked rows and columns
    for i in 1..num_rows {
        if matrix[i][0] == 0 {
            for j in 1..num_cols {
                matrix[i][j] = 0;
            }
        }
    }
    
    for j in 1..num_cols {
        if matrix[0][j] == 0 {
            for i in 1..num_rows {
                matrix[i][j] = 0;
            }
        }
    }
    
    // Set zeroes for the first row if needed
    if first_row_has_zero {
        for j in 0..num_cols {
            matrix[0][j] = 0;
        }
    }
    
    // Set zeroes for the first column if needed
    if first_col_has_zero {
        for i in 0..num_rows {
            matrix[i][0] = 0;
        }
    }
}

fn main() {
    let mut matrix1 = vec![
        vec![1, 1, 1],
        vec![1, 0, 1],
        vec![1, 1, 1],
    ];
    set_zeroes(&mut matrix1);
    println!("Output for example 1: {:?}", matrix1);

    let mut matrix2 = vec![
        vec![0, 1, 2, 0],
        vec![3, 4, 5, 2],
        vec![1, 3, 1, 5],
    ];
    set_zeroes(&mut matrix2);
    println!("Output for example 2: {:?}", matrix2);
}

main()
````
