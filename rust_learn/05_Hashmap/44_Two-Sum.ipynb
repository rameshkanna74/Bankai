1. Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]

Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

 

Constraints:

    2 <= nums.length <= 104
    -109 <= nums[i] <= 109
    -109 <= target <= 109
    Only one valid answer exists.

 
Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

use std::collections::HashMap;

impl Solution {
    pub fn two_sum(nums: Vec<i32>, target: i32) -> Vec<i32> {
        // Create a HashMap to store the value and its index from the nums array.
        let mut seen_numbers: HashMap<i32, usize> = HashMap::new();

        // Iterate through the nums array.
        for (i, &num) in nums.iter().enumerate() {
            // Calculate the complement that, when added to the current number, equals the target.
            let complement = target - num;

            // Check if the complement exists in the HashMap.
            if let Some(&complement_index) = seen_numbers.get(&complement) {
                // If found, return the indices of the current number and its complement.
                return vec![complement_index as i32, i as i32];
            }

            // If complement is not found, add the current number and its index to the HashMap.
            seen_numbers.insert(num, i);
        }

        // The problem statement guarantees that there is exactly one solution, so this code should never be reached.
        unreachable!()
    }
}

In Rust, `isize` and `usize` are integer types that are used to represent sizes and indices. They have some important differences:

### `isize` vs. `usize`

1. **Signed vs. Unsigned:**

   - `isize` is a signed integer type, meaning it can represent both positive and negative numbers.
   - `usize` is an unsigned integer type, meaning it can only represent non-negative numbers.

2. **Purpose:**

   - `isize` is typically used when you need to represent a number that can be negative, such as offsets or differences in values.
   - `usize` is used for sizes and indexing, where negative numbers do not make sense, such as array indices and the length of collections.

3. **Size:**
   - Both `isize` and `usize` are platform-dependent. They have the same size as the platformâ€™s pointer width.
   - On a 32-bit system, both `isize` and `usize` are 32 bits.
   - On a 64-bit system, both `isize` and `usize` are 64 bits.

### Use Cases

- **`isize`:**

  - Useful for calculations where negative values are possible, such as when working with differences or distances.
  - For example, in a function that calculates the offset from a starting index, `isize` might be used to represent both positive and negative offsets.

- **`usize`:**
  - Commonly used for indexing into arrays and slices, as well as representing sizes or lengths. Since indices cannot be negative, `usize` is the appropriate choice.
  - For example, when iterating over elements in a vector, you typically use `usize` to access the index.
