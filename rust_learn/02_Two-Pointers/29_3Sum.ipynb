15. 3Sum

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation:
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.

Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.

Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

Constraints:

    3 <= nums.length <= 3000
    -105 <= nums[i] <= 105

impl Solution {
    pub fn three_sum(mut nums: Vec<i32>) -> Vec<Vec<i32>> {
        let mut result = Vec::new();
        let n = nums.len();

        // Step 1: Sort the array
        nums.sort_unstable();

        // Step 2: Iterate over the array, using a pointer technique
        for i in 0..n - 2 {
            // Skip duplicate elements to avoid duplicate triplets in the result
            if i > 0 && nums[i] == nums[i - 1] {
                continue;
            }

            let mut left = i + 1;
            let mut right = n - 1;

            // Step 3: Use two pointers to find the remaining two elements
            while left < right {
                let sum = nums[i] + nums[left] + nums[right];

                if sum == 0 {
                    // Found a valid triplet
                    result.push(vec![nums[i], nums[left], nums[right]]);

                    // Move the left pointer to skip duplicates
                    while left < right && nums[left] == nums[left + 1] {
                        left += 1;
                    }
                    // Move the right pointer to skip duplicates
                    while left < right && nums[right] == nums[right - 1] {
                        right -= 1;
                    }

                    // Move both pointers inward
                    left += 1;
                    right -= 1;
                } else if sum < 0 {
                    // If the sum is less than zero, move the left pointer to increase the sum
                    left += 1;
                } else {
                    // If the sum is greater than zero, move the right pointer to decrease the sum
                    right -= 1;
                }
            }
        }

        result
    }

### Explanation:

1. **Sorting**: We first sort the `nums` array. Sorting helps in easily skipping duplicates and using the two-pointer technique efficiently.

2. **Main Loop**: We iterate through the array with an index `i` and treat `nums[i]` as the first element of the triplet.

3. **Two Pointers**: For each `nums[i]`, we set up two pointers: `left` starting from `i + 1` and `right` starting from the end of the array. We check the sum of `nums[i]`, `nums[left]`, and `nums[right]`.

4. **Checking Sum**:

   - If the sum equals 0, we found a valid triplet.
   - If the sum is less than 0, we move the `left` pointer to the right to increase the sum.
   - If the sum is greater than 0, we move the `right` pointer to the left to decrease the sum.

5. **Skipping Duplicates**: After finding a valid triplet, we skip over any duplicates to ensure that we do not add duplicate triplets to the result.

6. **Time Complexity**: \(O(n^2)\) - The array is sorted in \(O(n \log n)\) and the two-pointer technique runs in \(O(n^2)\).

7. **Space Complexity**: \(O(1)\) for the sorting in place, excluding the space required for the output.

This solution is efficient for the problem constraints and avoids unnecessary duplicates in the results.
