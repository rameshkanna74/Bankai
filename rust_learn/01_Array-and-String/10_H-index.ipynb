````python
def calculateHIndex(citations):
    # Sort the citations in non-decreasing order
    citations.sort(reverse=True)

    # Iterate over the sorted citations to find the h-index
    h_index = 0
    for i, citation in enumerate(citations):
        # The h-index is the highest number where citations[i] >= i+1
        if citation >= i + 1:
            h_index = i + 1
        else:
            break

    return h_index
````

````python
# Example usage:
example_1 = [3, 0, 6, 1, 5]
example_2 = [1, 3, 1]

print(f"H-index for example 1: {calculateHIndex(example_1)}")  # Output: 3
print(f"H-index for example 2: {calculateHIndex(example_2)}")  # Output: 1
````

### Explanation:

1. **Sorting the citations**: We first sort the list of citations in descending order. This allows us to easily compare the citation counts with their corresponding indices.
2. **Finding the h-index**: We iterate through the sorted list to find the highest index `i` where the number of citations at that index is greater than or equal to `i + 1`. This is the definition of the h-index.
3. **Stopping condition**: If a paper has fewer citations than its index (plus one), we stop and return the h-index found so far.

### Time and Space Complexity:

- **Time Complexity**: O(n log n) – The most time-consuming operation is sorting the citations, which is O(n log n). The subsequent iteration is O(n).
- **Space Complexity**: O(1) – We are sorting the array in place and using a constant amount of extra space.

This code effectively calculates the h-index of a researcher based on their citation counts, following the optimal time and space complexity given the constraints.
