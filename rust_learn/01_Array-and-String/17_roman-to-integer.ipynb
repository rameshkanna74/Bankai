13. Roman to Integer

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol Value
I 1
V 5
X 10
L 50
C 100
D 500
M 1000

For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

    I can be placed before V (5) and X (10) to make 4 and 9.
    X can be placed before L (50) and C (100) to make 40 and 90.
    C can be placed before D (500) and M (1000) to make 400 and 900.

Given a roman numeral, convert it to an integer.

Example 1:

Input: s = "III"
Output: 3
Explanation: III = 3.

Example 2:

Input: s = "LVIII"
Output: 58
Explanation: L = 50, V= 5, III = 3.

Example 3:

Input: s = "MCMXCIV"
Output: 1994
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.

Constraints:

    1 <= s.length <= 15
    s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M').
    It is guaranteed that s is a valid roman numeral in the range [1, 3999].

To convert a Roman numeral to an integer efficiently, we can leverage a dictionary to map Roman numeral symbols to their integer values. Then, we can iterate through the string from left to right, checking if the current symbol is smaller than the next symbol. If so, subtract the current symbol's value from the total; otherwise, add it. This approach ensures an \(O(n)\) time complexity and \(O(1)\) space complexity.

### Approach
1. **Create a Mapping:** Use a hash map to store the value of each Roman numeral symbol.
2. **Iterate Through the String:** For each character in the string, check if it should be added or subtracted based on its relationship with the next character.
3. **Calculate the Result:** Maintain a running total based on the above checks.

### Complexity
- **Time Complexity:** \(O(n)\), where \(n\) is the length of the string.
- **Space Complexity:** \(O(1)\), as the space used does not grow with the input size.

```rust
use std::collections::HashMap;

impl Solution {
    /// Converts a Roman numeral string to its integer representation.
    ///
    /// # Arguments
    ///
    /// * `s` - A string representing a Roman numeral.
    ///
    /// # Returns
    ///
    /// The integer representation of the Roman numeral.
    pub fn roman_to_int(s: String) -> i32 {
        // Create a mapping from Roman numerals to their integer values
        let roman_to_value: HashMap<char, i32> = [
            ('I', 1),
            ('V', 5),
            ('X', 10),
            ('L', 50),
            ('C', 100),
            ('D', 500),
            ('M', 1000)
        ].iter().cloned().collect();
        
        // Convert the string to a vector of characters for easier indexing
        let chars: Vec<char> = s.chars().collect();
        let mut total_value = 0;
        let mut i = 0;
        
        while i < chars.len() {
            // Get the value of the current Roman numeral
            let current_value = roman_to_value[&chars[i]];
            
            // Check if this is not the last character and if the next character's value is greater
            if i + 1 < chars.len() {
                let next_value = roman_to_value[&chars[i + 1]];
                if current_value < next_value {
                    // If so, subtract the current value
                    total_value -= current_value;
                } else {
                    // Otherwise, add the current value
                    total_value += current_value;
                }
            } else {
                // For the last character, just add its value
                total_value += current_value;
            }
            
            i += 1;
        }
        
        total_value
    }
}

fn main() {
    // Example cases
    let roman1 = "III".to_string();
    let roman2 = "LVIII".to_string();
    let roman3 = "MCMXCIV".to_string();
    
    println!("Roman '{}' converts to {}", roman1, Solution::roman_to_int(roman1)); // Output: 3
    println!("Roman '{}' converts to {}", roman2, Solution::roman_to_int(roman2)); // Output: 58
    println!("Roman '{}' converts to {}", roman3, Solution::roman_to_int(roman3)); // Output: 1994
}
```

### Explanation
1. **Mapping Creation:** A `HashMap` is created to store Roman numeral symbols and their corresponding integer values.
2. **Processing the String:** Convert the input string to a vector of characters for indexed access. Iterate through each character, check if its value should be added or subtracted based on the value of the next character.
3. **Final Result:** Sum the values as determined by the above rules and return the result.

This solution ensures that we handle the conversion efficiently while adhering to the constraints provided.
