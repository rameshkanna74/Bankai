12. Integer to Roman

Seven different symbols represent Roman numerals with the following values:
Symbol Value
I 1
V 5
X 10
L 50
C 100
D 500
M 1000

Roman numerals are formed by appending the conversions of decimal place values from highest to lowest. Converting a decimal place value into a Roman numeral has the following rules:

    If the value does not start with 4 or 9, select the symbol of the maximal value that can be subtracted from the input, append that symbol to the result, subtract its value, and convert the remainder to a Roman numeral.
    If the value starts with 4 or 9 use the subtractive form representing one symbol subtracted from the following symbol, for example, 4 is 1 (I) less than 5 (V): IV and 9 is 1 (I) less than 10 (X): IX. Only the following subtractive forms are used: 4 (IV), 9 (IX), 40 (XL), 90 (XC), 400 (CD) and 900 (CM).
    Only powers of 10 (I, X, C, M) can be appended consecutively at most 3 times to represent multiples of 10. You cannot append 5 (V), 50 (L), or 500 (D) multiple times. If you need to append a symbol 4 times use the subtractive form.

Given an integer, convert it to a Roman numeral.

Example 1:

Input: num = 3749

Output: "MMMDCCXLIX"

Explanation:

3000 = MMM as 1000 (M) + 1000 (M) + 1000 (M)
700 = DCC as 500 (D) + 100 (C) + 100 (C)
40 = XL as 10 (X) less of 50 (L)
9 = IX as 1 (I) less of 10 (X)
Note: 49 is not 1 (I) less of 50 (L) because the conversion is based on decimal places

Example 2:

Input: num = 58

Output: "LVIII"

Explanation:

50 = L
8 = VIII

Example 3:

Input: num = 1994

Output: "MCMXCIV"

Explanation:

1000 = M
900 = CM
90 = XC
4 = IV

Constraints:

    1 <= num <= 3999

To convert an integer to a Roman numeral efficiently, we need to follow a systematic approach. We'll use a list of tuples that represent the Roman numeral symbols and their corresponding values in descending order. By iterating through this list, we can build the Roman numeral string by appending the appropriate symbols while reducing the integer value accordingly.

### Approach
1. **Define Roman Numeral Symbols:** Use a list of tuples to store the Roman numeral symbols and their values in descending order.
2. **Iterate Through the List:** For each tuple, append the Roman numeral symbols to the result string as many times as the current value can be subtracted from the number.
3. **Update the Number:** Subtract the value of the Roman numeral from the number and continue until the number is reduced to zero.

### Complexity
- **Time Complexity:** \(O(1)\), because the number of Roman numeral symbols is fixed, and the algorithm performs a constant amount of work for each symbol.
- **Space Complexity:** \(O(1)\), as the space used does not grow with the input size.

impl Solution {
    /// Converts an integer to a Roman numeral.

    pub fn int_to_roman(num: i32) -> String {
        // Define the Roman numeral symbols and their values in descending order
        let roman_values = [
            (1000, "M"),
            (900, "CM"),
            (500, "D"),
            (400, "CD"),
            (100, "C"),
            (90, "XC"),
            (50, "L"),
            (40, "XL"),
            (10, "X"),
            (9, "IX"),
            (5, "V"),
            (4, "IV"),
            (1, "I"),
        ];

        let mut number = num;
        let mut result = String::new();
        
        // Iterate through the list of Roman numeral values
        for &(value, symbol) in &roman_values {

                result.push_str(symbol); // Append the symbol to the result
                number -= value;        // Subtract the value from the number
            }
        }
        
        result
    }
}
