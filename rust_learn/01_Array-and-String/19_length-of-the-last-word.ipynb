To determine the length of the last word in a string `s` that consists of words and spaces, you can follow these steps:

1. **Trim Trailing Spaces**: Remove any trailing spaces from the string to ensure that the last word isn't accidentally treated as empty due to extra spaces at the end.

2. **Find the Last Word**: Identify the last word in the string by splitting the string based on spaces. The last element of the resulting list will be the last word.

3. **Calculate the Length**: Return the length of this last word.

Here is a Python function that performs these steps efficiently:

```python
def length_of_last_word(s: str) -> int:
    # Strip trailing spaces and split the string by spaces
    words = s.strip().split()
    # Return the length of the last word in the list
    return len(words[-1])

# Example cases
example1 = "Hello World"
example2 = "   fly me   to   the moon  "
example3 = "luffy is still joyboy"

# Run the examples
output1 = length_of_last_word(example1)
output2 = length_of_last_word(example2)
output3 = length_of_last_word(example3)

print(f"Input: s = \"{example1}\"\nOutput: {output1}")  # Expected: 5
print(f"Input: s = \"{example2}\"\nOutput: {output2}")  # Expected: 4
print(f"Input: s = \"{example3}\"\nOutput: {output3}")  # Expected: 6
```

### Explanation:

1. **`strip()`**:
   - The `strip()` method removes any leading and trailing whitespace characters from the string. This helps in ensuring that any extra spaces at the end of the string do not affect the result.

2. **`split()`**:
   - The `split()` method, when used without arguments, splits the string based on any whitespace (including spaces) and returns a list of words. Consecutive spaces are treated as a single delimiter.

3. **`len(words[-1])`**:
   - Accessing `words[-1]` gives the last word in the list, and `len()` computes its length.

### Time and Space Complexity:

- **Time Complexity**: O(n), where n is the length of the input string `s`. This is because both `strip()` and `split()` operations iterate over the string.
- **Space Complexity**: O(n), where n is the length of the input string `s`. The space is used to store the list of words generated by `split()`.

This method efficiently handles the input constraints and ensures that the last word's length is calculated correctly.
