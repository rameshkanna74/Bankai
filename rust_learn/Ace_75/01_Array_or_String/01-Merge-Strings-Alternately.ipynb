        let mut merged_string = String::new();

### Time Complexity

1. **Character Iteration**: The loop iterates through the characters of the shorter string, which is \(O(\min(\text{len1}, \text{len2}))\).
2. **String Slicing**: After the loop, the longer string's remaining characters are appended. This operation is \(O(\max(\text{len1}, \text{len2}))\), but since the length of the remaining string is always at most one of the input string lengths, the complexity remains linear.

   **Total Time Complexity**: \(O(\text{len1} + \text{len2})\).

### Space Complexity

1. **Result String**: The merged string requires space proportional to the sum of the lengths of `word1` and `word2`, so \(O(\text{len1} + \text{len2})\).
2. **Temporary Iterators**: The `chars()` iterators for `word1` and `word2` do not require additional storage for all characters; they iterate lazily, so their space usage is negligible.

   **Total Space Complexity**: `(O(text{len1} + text{len2}))`.
