76. Minimum Window Substring

Given two strings s and t of lengths m and n respectively, return the minimum window
substring
of s such that every character in t (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.

Example 1:

Input: s = "ADOBECODEBANC", t = "ABC"
Output: "BANC"
Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.

Example 2:

Input: s = "a", t = "a"
Output: "a"
Explanation: The entire string s is the minimum window.

Example 3:

Input: s = "a", t = "aa"
Output: ""
Explanation: Both 'a's from t must be included in the window.
Since the largest window of s only has one 'a', return empty string.

Constraints:

    m == s.length
    n == t.length
    1 <= m, n <= 105
    s and t consist of uppercase and lowercase English letters.

 

Follow up: Could you find an algorithm that runs in O(m + n) time?

To solve the "Minimum Window Substring" problem efficiently, we can use the sliding window technique with two pointers. This approach will help us achieve the required time complexity of \( O(m + n) \), where \( m \) and \( n \) are the lengths of the input strings `s` and `t`, respectively.

Here's a detailed Rust implementation of the solution:

### Explanation

1. **Count Characters in `t`**:
   - We need a frequency map of characters in `t` to ensure that all characters in `t` are included in the window substring.

2. **Sliding Window Technique**:
   - Use two pointers (`left` and `right`) to represent the current window of the string `s`.
   - Expand the window by moving the `right` pointer and include characters in the current window.
   - Once all required characters are included in the window, attempt to shrink the window by moving the `left` pointer.

3. **Update Minimum Window**:
   - Track the minimum window size while trying to shrink the window and update the result accordingly.

Here is the Rust code for the solution:

```rust
use std::collections::HashMap;

pub struct Solution;

impl Solution {
    pub fn min_window(s: String, t: String) -> String {
        let s = s.chars().collect::<Vec<_>>();  // Convert to vector of chars for easier indexing
        let t = t.chars().collect::<Vec<_>>();  // Convert to vector of chars for easier indexing
        
        let mut t_count = HashMap::new();
        for &ch in &t {
            *t_count.entry(ch).or_insert(0) += 1;
        }

        let mut window_count = HashMap::new();
        let mut left = 0;
        let mut min_length = usize::MAX;
        let mut min_window_start = 0;
        let mut required = t_count.len();  // Number of distinct characters in t that we need to match

        for right in 0..s.len() {
            let right_char = s[right];
            *window_count.entry(right_char).or_insert(0) += 1;

            if let Some(&count) = t_count.get(&right_char) {
                if window_count[&right_char] == count {
                    required -= 1;
                }
            }

            while required == 0 {
                let window_length = right - left + 1;
                if window_length < min_length {
                    min_length = window_length;
                    min_window_start = left;
                }

                let left_char = s[left];
                if let Some(&count) = t_count.get(&left_char) {
                    if window_count[&left_char] == count {
                        required += 1;
                    }
                }

                *window_count.get_mut(&left_char).unwrap() -= 1;
                left += 1;
            }
        }

        if min_length == usize::MAX {
            String::new()
        } else {
            s[min_window_start..min_window_start + min_length].iter().collect()
        }
    }
}
```

### Key Points

- **Frequency Maps**: `t_count` keeps track of the frequency of characters in `t`, and `window_count` keeps track of the frequency of characters in the current window of `s`.
- **Two-Pointer Technique**: Expands and contracts the window to find the minimum substring that contains all characters in `t`.
- **Result Calculation**: Once we find a valid window, we try to minimize it and update the result accordingly.

This approach ensures an optimal time complexity of \( O(m + n) \) due to the single pass of the string with two pointers and the efficient hashmap operations.
