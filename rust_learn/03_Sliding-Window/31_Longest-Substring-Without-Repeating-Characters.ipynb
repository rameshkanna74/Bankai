To solve the problem of finding the length of the longest substring without repeating characters, an efficient approach is to use the **sliding window technique** with a **hash set**. This approach ensures that we maintain a substring with unique characters while expanding and contracting the window.

### Approach:

1. **Sliding Window**:

   - Use two pointers, `start` and `end`, to represent the current window of characters.
   - Expand the `end` pointer to include more characters in the window.
   - If a duplicate character is found, move the `start` pointer to reduce the window size until the duplicate is removed.

2. **Hash Set**:
   - Use a hash set to keep track of characters in the current window and quickly check for duplicates.

### Time Complexity:

- The algorithm runs in \(O(n)\) time complexity, where \(n\) is the length of the string. Each character is processed at most twice (once by the `end` pointer and once by the `start` pointer).

### Space Complexity:

- The space complexity is \(O(min(n, m))\), where \(m\) is the size of the character set. In practice, this is constant space for a fixed-size character set (like ASCII).

Here's the Rust implementation with clear comments:

impl Solution {
    pub fn length_of_longest_substring(s: String) -> i32 {
        // Convert the input string to a vector of characters
        let chars: Vec<char> = s.chars().collect();

        // Use a hash set to track characters in the current window
        use std::collections::HashSet;
        let mut char_set = HashSet::new();

        let mut max_length = 0; // To store the maximum length of substring found
        let mut start = 0; // Start index of the sliding window

        // Iterate over each character with the end pointer
        for end in 0..chars.len() {
            // If the character is already in the set, remove characters from the start
            while char_set.contains(&chars[end]) {
                char_set.remove(&chars[start]);
                start += 1;
            }

            // Add the current character to the set
            char_set.insert(chars[end]);

            // Update the maximum length of the substring
            max_length = max_length.max(end - start + 1);
        }

        max_length as i32
    }
}
