122. Best Time to Buy and Sell Stock II

You are given an integer array prices where prices[i] is the price of a given stock on the ith day.

On each day, you may decide to buy and/or sell the stock. You can only hold at most one share of the stock at any time. However, you can buy it then immediately sell it on the same day.

Find and return the maximum profit you can achieve.

Example 1:

Input: prices = [7,1,5,3,6,4]
Output: 7
Explanation: Buy on day 2 (price = 1) and sell on day 3 (price = 5), profit = 5-1 = 4.
Then buy on day 4 (price = 3) and sell on day 5 (price = 6), profit = 6-3 = 3.
Total profit is 4 + 3 = 7.

Example 2:

Input: prices = [1,2,3,4,5]
Output: 4
Explanation: Buy on day 1 (price = 1) and sell on day 5 (price = 5), profit = 5-1 = 4.
Total profit is 4.

Example 3:

Input: prices = [7,6,4,3,1]
Output: 0
Explanation: There is no way to make a positive profit, so we never buy the stock to achieve the maximum profit of 0.

Constraints:

    1 <= prices.length <= 3 * 104
    0 <= prices[i] <= 104

Here’s a Rust solution for the "Best Time to Buy and Sell Stock II" problem. The idea is to take advantage of any opportunity to make a profit by buying and selling on consecutive days if the price is rising.

### Explanation
- We iterate through the `prices` array.
- Whenever we find that the price has increased from the previous day, we calculate the profit from this increase and add it to the total profit.

### Complexity
- **Time Complexity:** \(O(n)\), where \(n\) is the length of the `prices` array. We traverse the array once.
- **Space Complexity:** \(O(1)\), as we only use a few extra variables.

Here’s the code:

```rust
impl Solution {
    /// Calculates the maximum profit that can be achieved from buying and selling stocks.
    /// 
    /// # Arguments
    /// 
    /// * `prices` - A vector of integers where `prices[i]` is the price of the stock on day `i`.
    /// 
    /// # Returns
    /// 
    /// The maximum profit that can be achieved.
    pub fn max_profit(prices: Vec<i32>) -> i32 {
        // Initialize total profit to 0
        let mut total_profit = 0;
        
        // Iterate through the prices array from the second day to the last day
        for i in 1..prices.len() {
            // Calculate the profit if we sell today compared to buying yesterday
            if prices[i] > prices[i - 1] {
                // Add the profit to the total profit
                total_profit += prices[i] - prices[i - 1];
            }
        }
        
        // Return the total profit
        total_profit
    }
}

}
```
