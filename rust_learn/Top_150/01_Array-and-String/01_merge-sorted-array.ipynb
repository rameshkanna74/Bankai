You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

Example 1:

Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.

Example 2:

Input: nums1 = [1], m = 1, nums2 = [], n = 0
Output: [1]
Explanation: The arrays we are merging are [1] and [].
The result of the merge is [1].

Example 3:

Input: nums1 = [0], m = 0, nums2 = [1], n = 1
Output: [1]
Explanation: The arrays we are merging are [] and [1].
The result of the merge is [1].
Note that because m = 0, there are no elements in nums1. The 0 is only there to ensure the merge result can fit in nums1.

Constraints:

    nums1.length == m + n
    nums2.length == n
    0 <= m, n <= 200
    1 <= m + n <= 200
    -109 <= nums1[i], nums2[j] <= 109

Follow up: Can you come up with an algorithm that runs in O(m + n) time?

````rust
fn merge(nums1: &mut Vec<i32>, m: usize, nums2: &Vec<i32>, n: usize) {
    // Index pointers for nums1 and nums2
    let mut index1 = m as isize - 1;
    let mut index2 = n as isize - 1;
    // Index pointer for the last position in nums1
    let mut merge_index = (m + n) as isize - 1;

    // Traverse nums1 and nums2 from the end, and merge them into nums1
    while index2 >= 0 {
        if index1 >= 0 && nums1[index1 as usize] > nums2[index2 as usize] {
            nums1[merge_index as usize] = nums1[index1 as usize];
            index1 -= 1;
        } else {
            nums1[merge_index as usize] = nums2[index2 as usize];
            index2 -= 1;
        }
        merge_index -= 1;
    }
}

fn main() {
    // Example 1
    let mut nums1 = vec![1, 2, 3, 0, 0, 0];
    let nums2 = vec![2, 5, 6];
    let m = 3;
    let n = 3;
    merge(&mut nums1, m, &nums2, n);
    println!("Merged array: {:?}", nums1); // Output: [1, 2, 2, 3, 5, 6]

    // Example 2
    let mut nums1 = vec![1];
    let nums2 = vec![];
    let m = 1;
    let n = 0;
    merge(&mut nums1, m, &nums2, n);
    println!("Merged array: {:?}", nums1); // Output: [1]

    // Example 3
    let mut nums1 = vec![0];
    let nums2 = vec![1];
    let m = 0;
    let n = 1;
    merge(&mut nums1, m, &nums2, n);
    println!("Merged array: {:?}", nums1); // Output: [1]
}

main()
````

### Explanation:

- **Time Complexity**: The solution runs in `O(m + n)` time, where `m` and `n` are the lengths of `nums1` and `nums2`, respectively. This is because we traverse each array only once.
- **Space Complexity**: The solution uses `O(1)` extra space because the merging happens in-place within `nums1`.

### Steps:

1. **Pointers Initialization**:

   - `index1`: Tracks the last valid element in `nums1`.
   - `index2`: Tracks the last element in `nums2`.
   - `merge_index`: Tracks the position in `nums1` where the merged element should be placed.

2. **Merging Process**:

   - Compare elements from `nums1` and `nums2` starting from the end.
   - Place the larger element at the current `merge_index` in `nums1`.
   - Move the corresponding index pointer and repeat until all elements from `nums2` are merged into `nums1`.

3. **Final Output**: The merged and sorted array is stored in `nums1`.
