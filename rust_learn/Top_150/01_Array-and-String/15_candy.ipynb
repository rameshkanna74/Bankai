135. Candy

There are n children standing in a line. Each child is assigned a rating value given in the integer array ratings.

You are giving candies to these children subjected to the following requirements:

    Each child must have at least one candy.
    Children with a higher rating get more candies than their neighbors.

Return the minimum number of candies you need to have to distribute the candies to the children.

Example 1:

Input: ratings = [1,0,2]
Output: 5
Explanation: You can allocate to the first, second and third child with 2, 1, 2 candies respectively.

Example 2:

Input: ratings = [1,2,2]
Output: 4
Explanation: You can allocate to the first, second and third child with 1, 2, 1 candies respectively.
The third child gets 1 candy because it satisfies the above two conditions.

Constraints:

    n == ratings.length
    1 <= n <= 2 * 104
    0 <= ratings[i] <= 2 * 104

To solve the problem of distributing the minimum number of candies to children based on their ratings while satisfying the given conditions, we can use a two-pass approach:

1. **Left-to-Right Pass:** Ensure that each child has more candies than their left neighbor if their rating is higher.
2. **Right-to-Left Pass:** Ensure that each child has more candies than their right neighbor if their rating is higher. This pass will adjust the candy count based on the previous left-to-right pass.

### Approach
- **Initialization:** Start by giving each child one candy.
- **Left-to-Right Pass:** Traverse the list from left to right, and if a child has a higher rating than their left neighbor, increase their candy count.
- **Right-to-Left Pass:** Traverse the list from right to left, and if a child has a higher rating than their right neighbor, adjust the candy count to ensure the condition is met.

### Complexity
- **Time Complexity:** \(O(n)\), where \(n\) is the number of children, due to the two linear passes over the array.
- **Space Complexity:** \(O(n)\), due to the space needed to store the candy counts.

impl Solution {
    /// Returns the minimum number of candies needed to distribute according to the ratings.
    /// 
    /// # Arguments
    /// 
    /// * `ratings` - A vector of integers where `ratings[i]` is the rating of the `i`-th child.
    /// 
    /// # Returns
    /// 
    /// The minimum number of candies needed.
    pub fn candy(ratings: Vec<i32>) -> i32 {
        let n = ratings.len();
        if n == 0 {
            return 0;
        }
        
        // Initialize the candies array with 1 candy per child
        let mut candies = vec![1; n];
        
        // Left-to-right pass
        for i in 1..n {
            if ratings[i] > ratings[i - 1] {
                candies[i] = candies[i - 1] + 1;
            }
        }
        
        // Right-to-left pass
        for i in (0..n - 1).rev() {
            if ratings[i] > ratings[i + 1] {
                candies[i] = candies[i].max(candies[i + 1] + 1);
            }
        }
        
        // Calculate the total number of candies
        candies.iter().sum()
    }
}

- **Initialization:** Start by giving each child 1 candy.
- **Left-to-Right Pass:** Adjust the candy count based on the ratings comparison with the left neighbor.
- **Right-to-Left Pass:** Adjust the candy count based on the ratings comparison with the right neighbor. Ensure that the maximum of the current candies and the adjusted count is used to satisfy both conditions.
- **Total Candies:** Sum up the candies array to get the total minimum candies required.

This approach ensures that both conditions are satisfied and computes the minimum candies efficiently with linear time complexity.

````rust

pub fn candy(ratings: Vec<i32>) -> i32 {
    let n = ratings.len();
    if n == 0 {
        return 0;
    }
    
    // Initialize the candies array with 1 candy per child
    let mut candies = vec![1; n];
    
    // Left-to-right pass
    for i in 1..n {
        if ratings[i] > ratings[i - 1] {
            candies[i] = candies[i - 1] + 1;
        }
    }
    
    // Right-to-left pass
    for i in (0..n - 1).rev() {
        if ratings[i] > ratings[i + 1] {
            candies[i] = candies[i].max(candies[i + 1] + 1);
        }
    }
    
    // Calculate the total number of candies
    candies.iter().sum()
}

fn main() {
    // Example cases
    let ratings1 = vec![1, 0, 2];
    let ratings2 = vec![1, 2, 2];
    
    println!("Minimum candies needed for ratings1: {}", candy(ratings1)); // Output: 5
    println!("Minimum candies needed for ratings2: {}", candy(ratings2)); // Output: 4
}

main()
````

class Solution:
    def candy(self, ratings: List[int]) -> int:
        n = len(ratings)
        total_candies = n
        i = 1

        while i < n:
            if ratings[i] == ratings[i - 1]:
                i += 1
                continue

            current_peak = 0
            while i < n and ratings[i] > ratings[i - 1]:
                current_peak += 1
                total_candies += current_peak
                i += 1
            
            if i == n:
                return total_candies

            current_valley = 0
            while i < n and ratings[i] < ratings[i - 1]:
                current_valley += 1
                total_candies += current_valley
                i += 1

            total_candies -= min(current_peak, current_valley)

        return total_candies
