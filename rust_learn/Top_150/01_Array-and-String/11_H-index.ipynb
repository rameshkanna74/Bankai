274. H-Index

Given an array of integers citations where citations[i] is the number of citations a researcher received for their ith paper, return the researcher's h-index.

According to the definition of h-index on Wikipedia: The h-index is defined as the maximum value of h such that the given researcher has published at least h papers that have each been cited at least h times.

Example 1:

Input: citations = [3,0,6,1,5]
Output: 3
Explanation: [3,0,6,1,5] means the researcher has 5 papers in total and each of them had received 3, 0, 6, 1, 5 citations respectively.
Since the researcher has 3 papers with at least 3 citations each and the remaining two with no more than 3 citations each, their h-index is 3.

Example 2:

Input: citations = [1,3,1]
Output: 1

Constraints:

    n == citations.length
    1 <= n <= 5000
    0 <= citations[i] <= 1000

To solve the H-Index problem efficiently, we can use a sorting-based approach. The idea is to sort the citations array and then determine the maximum value of \( h \) such that there are at least \( h \) papers with at least \( h \) citations.

### Explanation
1. **Sort the Citations:** First, sort the array in descending order.
2. **Determine the H-Index:** Iterate through the sorted array and check the condition for each potential h-index. The h-index is the largest number where the number of papers with at least that many citations is at least equal to the h-index value.

### Complexity
- **Time Complexity:** \(O(n \log n)\), where \(n\) is the length of the `citations` array. This is due to the sorting step.
- **Space Complexity:** \(O(1)\), as we only use a few extra variables.

impl Solution {
    /// Calculates the h-index of a researcher based on their citation counts.
    /// 
    /// # Arguments
    /// 
    /// * `citations` - A vector of integers where `citations[i]` is the number of citations for the `i`-th paper.
    /// 
    /// # Returns
    /// 
    /// The h-index of the researcher.
    pub fn h_index(citations: Vec<i32>) -> i32 {
        // Sort citations in descending order
        let mut sorted_citations = citations;
        sorted_citations.sort_by(|a, b| b.cmp(a));
        
        // Initialize h-index
        let mut h_index = 0;
        
        // Iterate through the sorted citations
        for (index, &citation) in sorted_citations.iter().enumerate() {
            // The number of papers with at least `citation` citations
            let num_papers = index as i32 + 1;
            
            // Check if this number of papers is at least `citation`
            if num_papers >= citation {
                h_index = citation;
                break;
            }
        }
        
        h_index
    }
}

class Solution:
    def hIndex(self, citations: List[int]) -> int:
        papers = len(citations)
        citation_buckets = [0] * (papers + 1)

        for citation in citations:
            citation_buckets[min(citation, papers)] += 1

        cumulative_papers = 0
        for h_index in range(papers, -1, -1):
            cumulative_papers += citation_buckets[h_index]
            if cumulative_papers >= h_index:
                return h_index
