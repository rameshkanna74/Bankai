58. Length of Last Word

Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal
substring
consisting of non-space characters only.

Example 1:

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.

Example 2:

Input: s = " fly me to the moon "
Output: 4
Explanation: The last word is "moon" with length 4.

Example 3:

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.

Constraints:

    1 <= s.length <= 104
    s consists of only English letters and spaces ' '.
    There will be at least one word in s.

To solve the problem of finding the length of the last word in a string efficiently, we can approach it as follows:

### Approach
1. **Trim Trailing Spaces:** Start by trimming any trailing spaces from the string. This ensures that the last word is followed by the end of the string.
2. **Find the Last Word:** Once the string is trimmed, iterate backward to find the start of the last word.
3. **Calculate the Length:** Count the characters of the last word by continuing to move backward until a space is encountered or the start of the string is reached.

### Complexity
- **Time Complexity:** \(O(n)\), where \(n\) is the length of the string. This is because, in the worst case, we may need to iterate through the entire string.
- **Space Complexity:** \(O(1)\), since we only need a few additional variables for counting and iteration.

Here's the Rust implementation:

```rust
impl Solution {
    /// Finds the length of the last word in the given string `s`.
    ///
    /// # Arguments
    ///
    /// * `s` - A string containing words separated by spaces.
    ///
    /// # Returns
    ///
    /// The length of the last word in the string.
    pub fn length_of_last_word(s: String) -> i32 {
        let mut length = 0;
        let mut in_word = false;  // Flag to check if we're in a word
        
        // Iterate backward over the string to find the last word
        for ch in s.chars().rev() {
            if ch != ' ' {
                in_word = true;  // We've found a non-space character, so we're in a word
                length += 1;     // Increment the length of the last word
            } else if in_word {
                // If we encounter a space after finding the last word, we're done
                break;
            }
        }
        
        length
    }
}
```

1. **Iterate Backward:** By iterating backward through the string, we can directly locate the last word without needing to split the entire string or process leading spaces.
2. **In-Word Flag:** The `in_word` flag ensures that we start counting only when we enter the last word, and it allows us to stop counting as soon as we exit that word (encounter a space).
3. **Efficiency:** The method is efficient because it avoids unnecessary space usage and minimizes the operations needed to find the last word.

This solution is concise, clear, and well-suited for large input strings as it operates in linear time with constant space.
