Given a pattern and a string s, find if s follows the same pattern.

Here follow means a full match, such that there is a bijection between a letter in pattern and a non-empty word in s. Specifically:

    Each letter in pattern maps to exactly one unique word in s.
    Each unique word in s maps to exactly one letter in pattern.
    No two letters map to the same word, and no two words map to the same letter.

Example 1:

Input: pattern = "abba", s = "dog cat cat dog"

Output: true

Explanation:

The bijection can be established as:

    'a' maps to "dog".
    'b' maps to "cat".

Example 2:

Input: pattern = "abba", s = "dog cat cat fish"

Output: false

Example 3:

Input: pattern = "aaaa", s = "dog cat cat dog"

Output: false

Constraints:

    1 <= pattern.length <= 300
    pattern contains only lower-case English letters.
    1 <= s.length <= 3000
    s contains only lowercase English letters and spaces ' '.
    s does not contain any leading or trailing spaces.
    All the words in s are separated by a single space.

Here’s a Rust implementation of the word pattern problem, along with explanations, variable naming, and test cases to illustrate how the solution works.

```rust
impl Solution {
    pub fn word_pattern(pattern: String, s: String) -> bool {
        // Split the string `s` into words
        let words: Vec<&str> = s.split_whitespace().collect();
        
        // Check if the length of pattern and number of words match
        if pattern.len() != words.len() {
            return false;
        }

        // Hash maps to establish the bijection
        let mut char_to_word = std::collections::HashMap::new();
        let mut word_to_char = std::collections::HashMap::new();

        // Iterate through the characters in pattern and corresponding words
        for (char, word) in pattern.chars().zip(words.iter()) {
            // Check if the character is already mapped to a word
            if let Some(mapped_word) = char_to_word.get(&char) {
                // If the mapped word does not match the current word, return false
                if mapped_word != word {
                    return false;
                }
            } else {
                // If not mapped, insert the new mapping
                char_to_word.insert(char, *word);
            }

            // Check if the word is already mapped to a character
            if let Some(mapped_char) = word_to_char.get(*word) {
                // If the mapped character does not match the current character, return false
                if mapped_char != &char {
                    return false;
                }
            } else {
                // If not mapped, insert the new mapping
                word_to_char.insert(*word, char);
            }
        }

        // If all checks pass, the pattern matches the string
        true
    }
}

fn main() {
    let pattern1 = String::from("abba");
    let s1 = String::from("dog cat cat dog");
    println!("Example 1: {}", Solution::word_pattern(pattern1, s1)); // Output: true

    let pattern2 = String::from("abba");
    let s2 = String::from("dog cat cat fish");
    println!("Example 2: {}", Solution::word_pattern(pattern2, s2)); // Output: false

    let pattern3 = String::from("aaaa");
    let s3 = String::from("dog cat cat dog");
    println!("Example 3: {}", Solution::word_pattern(pattern3, s3)); // Output: false
}
```

### Explanation of the Code:

1. **Splitting the String**: We split the input string `s` into words using `split_whitespace()`.

2. **Length Check**: Before proceeding, we check if the lengths of the `pattern` and the list of `words` are the same. If not, we can immediately return `false`.

3. **Using HashMaps**: We utilize two hash maps:
   - `char_to_word` to map characters in `pattern` to words in `s`.
   - `word_to_char` to map words in `s` back to characters in `pattern`.

4. **Iterating through Characters and Words**: We iterate through the characters of `pattern` and the words from `s` simultaneously using `zip`.

5. **Mapping Checks**:
   - For each character, we check if it already has a mapped word. If it does, we ensure it matches the current word. If it doesn’t match, we return `false`.
   - Similarly, for each word, we check if it already has a mapped character and verify it matches.

6. **Final Decision**: If all checks are satisfied, we return `true`, indicating that the pattern and the string match according to the specified rules.

### Complexity Analysis:
- **Time Complexity**: O(n), where n is the length of the input string `s`, since we perform a single pass to map characters to words.
- **Space Complexity**: O(n), as we store mappings in the hash maps.
