2028. Find Missing Observations

You have observations of n + m 6-sided dice rolls with each face numbered from 1 to 6. n of the observations went missing, and you only have the observations of m rolls. Fortunately, you have also calculated the average value of the n + m rolls.

You are given an integer array rolls of length m where rolls[i] is the value of the ith observation. You are also given the two integers mean and n.

Return an array of length n containing the missing observations such that the average value of the n + m rolls is exactly mean. If there are multiple valid answers, return any of them. If no such array exists, return an empty array.

The average value of a set of k numbers is the sum of the numbers divided by k.

Note that mean is an integer, so the sum of the n + m rolls should be divisible by n + m.

Example 1:

Input: rolls = [3,2,4,3], mean = 4, n = 2
Output: [6,6]
Explanation: The mean of all n + m rolls is (3 + 2 + 4 + 3 + 6 + 6) / 6 = 4.

Example 2:

Input: rolls = [1,5,6], mean = 3, n = 4
Output: [2,3,2,2]
Explanation: The mean of all n + m rolls is (1 + 5 + 6 + 2 + 3 + 2 + 2) / 7 = 3.

Example 3:

Input: rolls = [1,2,3,4], mean = 6, n = 4
Output: []
Explanation: It is impossible for the mean to be 6 no matter what the 4 missing rolls are.

Constraints:

    m == rolls.length
    1 <= n, m <= 105
    1 <= rolls[i], mean <= 6

### Approach:

1. **Sum Calculation**: The total sum of all the rolls (both given and missing) must equal `mean * (n + m)`, where `n` is the number of missing observations, and `m` is the number of given observations.
2. **Remaining Sum**: Calculate the sum of the given rolls and subtract it from the total required sum to find the sum of the missing rolls.
3. **Feasibility Check**: Ensure that the sum of the missing rolls is within the range `[n, 6 * n]`, since each die roll can only be between 1 and 6.
4. **Distribute Missing Rolls**: Distribute the sum evenly among the missing rolls, ensuring each roll is between 1 and 6.

Here’s the Rust code:

```rust
impl Solution {
    pub fn missing_rolls(rolls: Vec<i32>, mean: i32, n: i32) -> Vec<i32> {
        // Calculate the total number of rolls (m is length of rolls array, n is the number of missing rolls)
        let m = rolls.len() as i32;
        let total_sum_needed = mean * (n + m); // Total sum we need to achieve

        // Sum of the given rolls
        let given_sum: i32 = rolls.iter().sum();

        // The sum that the missing rolls must account for
        let missing_sum = total_sum_needed - given_sum;

        // Each roll can be between 1 and 6, so the missing_sum must be within valid bounds
        if missing_sum < n || missing_sum > 6 * n {
            // It's impossible to fill the missing observations with valid dice rolls
            return vec![];
        }

        // Result vector to store the missing rolls
        let mut result = vec![1; n as usize]; // Start by assuming each missing roll is 1
        let mut remaining_sum = missing_sum - n; // Distribute the remaining sum (after the 1s)

        // Distribute the remaining sum by increasing the rolls, ensuring no roll exceeds 6
        for i in 0..n as usize {
            let add = std::cmp::min(remaining_sum, 5); // We can add at most 5 to any roll (since it's already 1)
            result[i] += add;
            remaining_sum -= add;
            if remaining_sum == 0 {
                break;
            }
        }

        result
    }
}
```
