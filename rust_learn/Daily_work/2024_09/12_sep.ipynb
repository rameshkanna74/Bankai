You are given a string allowed consisting of distinct characters and an array of strings words. A string is consistent if all characters in the string appear in the string allowed.

Return the number of consistent strings in the array words.

Example 1:

Input: allowed = "ab", words = ["ad","bd","aaab","baa","badab"]
Output: 2
Explanation: Strings "aaab" and "baa" are consistent since they only contain characters 'a' and 'b'.

Example 2:

Input: allowed = "abc", words = ["a","b","c","ab","ac","bc","abc"]
Output: 7
Explanation: All strings are consistent.

Example 3:

Input: allowed = "cad", words = ["cc","acd","b","ba","bac","bad","ac","d"]
Output: 4
Explanation: Strings "cc", "acd", "ac", and "d" are consistent.

Constraints:

    1 <= words.length <= 104
    1 <= allowed.length <= 26
    1 <= words[i].length <= 10
    The characters in allowed are distinct.
    words[i] and allowed contain only lowercase English letters.

To solve this problem efficiently in Rust, we need to focus on both time and space complexity. Here's how we can approach the solution:

### Key Insights:

1. **Allowed characters**: Since the `allowed` string contains distinct characters, we can represent them as a set. Checking membership in a set has an average time complexity of O(1).
2. **Words**: For each word in `words`, we can check if all characters in the word exist in the allowed set.
3. **Optimal time complexity**: We aim for O(n \* m), where `n` is the number of words, and `m` is the average length of the words. Checking each character in a word against a set is O(1), and this needs to be done for every character of every word.

### Approach:

1. Convert the `allowed` string into a `HashSet` to quickly check if a character is allowed.
2. For each word, check if all of its characters exist in the `HashSet`.
3. Count how many words are consistent.

### Time Complexity:

- Creating the set from `allowed`: O(allowed.length), which is at most O(26), so it's constant time.
- For each word, checking each character in the set takes O(word.length). For `n` words, the total complexity is O(n \* m), where `m` is the average word length.

### Space Complexity:

- We need space to store the `HashSet`, which takes O(allowed.length). Since the length of `allowed` is at most 26, this is also constant O(1) space.
- No additional significant space is required beyond the input itself.

### Rust Code:

```rust
use std::collections::HashSet;

impl Solution {
    pub fn count_consistent_strings(allowed: String, words: Vec<String>) -> i32 {
        // Create a HashSet from the allowed characters for O(1) lookup.
        let allowed_set: HashSet<char> = allowed.chars().collect();

        // Initialize a counter for consistent strings.
        let mut consistent_count = 0;

        // Iterate over each word in words.
        for word in words {
            // Check if all characters in the word are in the allowed set.
            if word.chars().all(|c| allowed_set.contains(&c)) {
                consistent_count += 1;  // Increment if the word is consistent.
            }
        }

        // Return the total count of consistent strings.
        consistent_count
    }
}
```

````rust
use std::collections::HashSet;


pub struct Solution;

impl Solution {
    pub fn count_consistent_strings(allowed: String, words: Vec<String>) -> i32 {
        // Create a HashSet from the allowed characters for O(1) lookup.
        let allowed_set: HashSet<char> = allowed.chars().collect();

        // Initialize a counter for consistent strings.
        let mut consistent_count = 0;

        // Iterate over each word in words.
        for word in words {
            // Check if all characters in the word are in the allowed set.
            if word.chars().all(|c| allowed_set.contains(&c)) {
                consistent_count += 1;  // Increment if the word is consistent.
            }
        }

        // Return the total count of consistent strings.
        consistent_count
    }
}

fn main() {
    // Example 1
    let allowed = String::from("ab");
    let words = vec![
        String::from("ad"),
        String::from("bd"),
        String::from("aaab"),
        String::from("baa"),
        String::from("badab")
    ];
    let result = Solution::count_consistent_strings(allowed, words);
    println!("Example 1 Output: {}", result); // Output: 2

    // Example 2
    let allowed = String::from("abc");
    let words = vec![
        String::from("a"),
        String::from("b"),
        String::from("c"),
        String::from("ab"),
        String::from("ac"),
        String::from("bc"),
        String::from("abc")
    ];
    let result = Solution::count_consistent_strings(allowed, words);
    println!("Example 2 Output: {}", result); // Output: 7

    // Example 3
    let allowed = String::from("cad");
    let words = vec![
        String::from("cc"),
        String::from("acd"),
        String::from("b"),
        String::from("ba"),
        String::from("bac"),
        String::from("bad"),
        String::from("ac"),
        String::from("d")
    ];
    let result = Solution::count_consistent_strings(allowed, words);
    println!("Example 3 Output: {}", result); // Output: 4
}

main()
````
