Input: nums = [1,2,3], head = [1,2,3,4,5]

Example 3:

Input: nums = [5], head = [1,2,3,4]

Output: [1,2,3,4]

Explanation:

No node has value 5.

    The input is generated such that there is at least one node in the linked list that has a value not present in nums.

To solve the problem efficiently, we can make use of a **HashSet** to store the elements from the `nums` array. This allows for constant-time lookups when checking if a value should be removed from the linked list.

The idea is to traverse the linked list and remove any node whose value exists in the `nums` set. The traversal is done in a single pass, ensuring the time complexity is optimal.

### Approach:

1. **Use a HashSet**:

   - Store the values from `nums` in a `HashSet` to allow for quick lookups while traversing the linked list.

2. **Traverse the Linked List**:

   - As we traverse the linked list, skip over any nodes whose values are in the `HashSet`.

3. **Re-link the Nodes**:
   - For nodes not in the `HashSet`, we link them to maintain the final modified list.

impl Solution {
    pub fn modified_list(nums: Vec<i32>, head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        // Step 1: Create a HashSet for quick lookup of values to remove
        let to_remove: HashSet<i32> = nums.into_iter().collect();

        // Step 2: Use a dummy node to simplify edge cases (e.g., removing the head node)
        let mut dummy = ListNode::new(0);
        dummy.next = head;

        // Step 3: Initialize a pointer to the current node starting at the dummy node
        let mut current = &mut dummy;

        // Step 4: Traverse the linked list
        while let Some(ref mut node) = current.next {
            if to_remove.contains(&node.val) {
                // If the current node's value is in the set, skip this node
                current.next = node.next.take();  // Remove the node
            } else {
                // Otherwise, move to the next node
                current = current.next.as_mut().unwrap();
            }
        }

        // Return the modified list starting from the node after dummy
        dummy.next
    }
}

### Explanation:

1. **HashSet for Removal Lookup**:  
   We first convert the `nums` vector into a `HashSet` to allow O(1) lookup times when determining if a node's value should be removed.

2. **Dummy Node**:  
   We use a dummy node (`dummy`) to handle cases where the head node might be removed. This simplifies pointer management and avoids special cases for the head.

3. **List Traversal**:  
   The `current` pointer traverses the linked list. If the current nodeâ€™s value exists in the `to_remove` set, we adjust the pointers to skip that node. Otherwise, we move `current` to the next node.

4. **Edge Case**:  
   If all nodes are removed or none are removed, the logic remains consistent because of the dummy node.

5. **Time Complexity**:

   - Building the `HashSet` takes **O(k)** where `k` is the number of elements in `nums`.
   - Traversing the linked list takes **O(n)** where `n` is the number of nodes in the linked list.
     Thus, the overall time complexity is **O(n + k)**.

6. **Space Complexity**:
   - The space complexity is **O(k)** for storing the `HashSet` and **O(1)** additional space for pointer manipulation (excluding the input).

- **Example 3**: `nums = [5]`, `head = [1, 2, 3, 4]`, Output: `[1, 2, 3, 4]`

This solution is both time and space efficient, suitable for large inputs.


````rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
  pub val: i32,
  pub next: Option<Box<ListNode>>
}

impl ListNode {
  #[inline]
  fn new(val: i32) -> Self {
    ListNode {
      next: None,
      val
    }
  }
}

use std::collections::HashSet;

pub fn modified_list(nums: Vec<i32>, head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
    // Step 1: Create a HashSet for quick lookup of values to remove
    let to_remove: HashSet<i32> = nums.into_iter().collect();
    
    // Step 2: Use a dummy node to simplify edge cases (e.g., removing the head node)
    let mut dummy = ListNode::new(0);
    dummy.next = head;
    
    // Step 3: Initialize a pointer to the current node starting at the dummy node
    let mut current = &mut dummy;

    // Step 4: Traverse the linked list
    while let Some(ref mut node) = current.next {
        if to_remove.contains(&node.val) {
            // If the current node's value is in the set, skip this node
            current.next = node.next.take();  // Remove the node
        } else {
            // Otherwise, move to the next node
            current = current.next.as_mut().unwrap();
        }
    }
    
    // Return the modified list starting from the node after dummy
    dummy.next
}

fn main() {
    // Example 1
    let nums1 = vec![1, 2, 3];
    let head1 = Some(Box::new(ListNode {
        val: 1,
        next: Some(Box::new(ListNode {
            val: 2,
            next: Some(Box::new(ListNode {
                val: 3,
                next: Some(Box::new(ListNode {
                    val: 4,
                    next: Some(Box::new(ListNode {
                        val: 5,
                        next: None,
                    })),
                })),
            })),
        })),
    }));
    
    let result1 = modified_list(nums1, head1);
    print_list(result1); // Expected Output: [4, 5]
    
    // Example 2
    let nums2 = vec![1];
    let head2 = Some(Box::new(ListNode {
        val: 1,
        next: Some(Box::new(ListNode {
            val: 2,
            next: Some(Box::new(ListNode {
                val: 1,
                next: Some(Box::new(ListNode {
                    val: 2,
                    next: Some(Box::new(ListNode {
                        val: 1,
                        next: Some(Box::new(ListNode {
                            val: 2,
                            next: None,
                        })),
                    })),
                })),
            })),
        })),
    }));
    
    let result2 = modified_list(nums2, head2);
    print_list(result2); // Expected Output: [2, 2, 2]
    
    // Example 3
    let nums3 = vec![5];
    let head3 = Some(Box::new(ListNode {
        val: 1,
        next: Some(Box::new(ListNode {
            val: 2,
            next: Some(Box::new(ListNode {
                val: 3,
                next: Some(Box::new(ListNode {
                    val: 4,
                    next: None,
                })),
            })),
        })),
    }));
    
    let result3 = modified_list(nums3, head3);
    print_list(result3); // Expected Output: [1, 2, 3, 4]
}

// Helper function to print the linked list
fn print_list(mut head: Option<Box<ListNode>>) {
    let mut result = vec![];
    while let Some(node) = head {
        result.push(node.val);
        head = node.next;
    }
    println!("{:?}", result);
}

main()
````
