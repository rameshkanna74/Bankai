There is a directed graph of n nodes with each node labeled from 0 to n - 1. The graph is represented by a 0-indexed 2D integer array graph where graph[i] is an integer array of nodes adjacent to node i, meaning there is an edge from node i to each node in graph[i].

A node is a terminal node if there are no outgoing edges. A node is a safe node if every possible path starting from that node leads to a terminal node (or another safe node).

Return an array containing all the safe nodes of the graph. The answer should be sorted in ascending order.

````rust
impl Solution {
    pub fn eventual_safe_nodes(graph: Vec<Vec<i32>>) -> Vec<i32> {
        fn dfs(u: usize, seen: &mut [bool], cache: &mut [i32], graph:  &Vec<Vec<i32>>) -> bool {
            if graph[u].len() == 0 { return true; }
            if cache[u] != 0 {
                return if cache[u] == 1 { true } else { false };
            }

            let mut res = true;

            for &v in graph[u].iter() {
                let v = v as usize;
                if !seen[v] {
                    seen[v] = true;
                    res &= dfs(v, seen, cache, graph);
                    seen[v] = false;
                } else {
                    return false;
                }
            }

            if res { cache[u] = 1; } 
            else { cache[u] = -1; }

            res
        }

        let n = graph.len();
        let mut seen = vec![false; n];
        let mut cache = vec![0; n];

        let mut res = vec![];
        for u in 0..n {
            seen[u] = true;
            if dfs(u, &mut seen, &mut cache, &graph) {
                res.push(u as i32);
            }
            seen[u] = false;
        }

        res
    }
}
````

The problem you solved involves **detecting safe nodes** in a directed graph, which relates to a **graph traversal** problem, specifically using **Depth-First Search (DFS)** with **cycle detection** and **memoization**.

Here are the key patterns and concepts you can learn from this problem:

### 1. **Graph Traversal (DFS)**:
   - The main pattern here is **Depth-First Search (DFS)**. DFS is a common graph traversal technique used to explore all nodes and edges of a graph.
   - You traverse from each node, marking its status to keep track of its state (whether it's safe or not).
   - **DFS** can be implemented recursively (as shown in the solution) or iteratively using an explicit stack.

### 2. **Cycle Detection**:
   - A key feature of this problem is detecting whether a node (or a path) is part of a **cycle**. If a node is revisited in the DFS traversal, it means you've encountered a cycle, and the node cannot be safe (because a cycle would lead to an infinite loop).
   - You detect cycles using the **visiting** state (`1`), where nodes in the current DFS path are temporarily marked as "visiting."
   - If you encounter a node that is already in the "visiting" state, you've detected a cycle, and the current node cannot be marked as safe.

### 3. **Memoization (Storing States)**:
   - Memoization is used to optimize the DFS traversal by storing the results of previous calculations. In this problem, we store the state of each node (whether it is unvisited, visiting, or safe).
   - This avoids redundant checks for the same node, improving the efficiency of the algorithm, especially for larger graphs.

### 4. **Marking Nodes with States**:
   - You use three states to classify nodes:
     - `0`: Unvisited (node hasn't been processed).
     - `1`: Visiting (node is part of the current DFS path, i.e., being processed).
     - `2`: Safe (all paths lead to terminal or safe nodes).
   - This is a common approach for handling graphs where nodes can have different statuses or classifications.

### 5. **Topological Sorting (Implicit)**:
   - Although this problem doesn't directly ask for a topological sort, the concept is closely related. You are essentially marking nodes as "safe" (which could be thought of as a form of topological order), since all paths from a safe node lead to other safe nodes or terminal nodes.
   - In problems like these, topological sorting often comes into play when determining the safe or final order of nodes (such as in the case of resolving dependencies).

### 6. **Backtracking**:
   - While not strictly a backtracking problem, this problem does share some similarities with backtracking because you're making recursive decisions and then "undoing" those decisions (via state changes) as you backtrack through the graph.

### 7. **Graph Theory Concepts**:
   - **Terminal Nodes**: A node with no outgoing edges is a terminal node. These are inherently safe because they do not lead anywhere.
   - **Safe Nodes**: Nodes that can eventually reach terminal nodes without falling into a cycle are considered safe. You must determine which nodes are safe by considering all possible paths.
   - **Strongly Connected Components (SCC)**: While this problem doesn't explicitly use SCCs, detecting cycles in a directed graph and marking nodes as safe or unsafe shares a concept with SCC detection. SCC algorithms (like Tarjan's or Kosaraju's) identify strongly connected subgraphs where each node is reachable from every other node in that subgraph. If any SCC contains a cycle, the nodes in that SCC are unsafe.

### 8. **Pathfinding in Directed Graphs**:
   - This problem is a variant of pathfinding where you are not just looking for the shortest or any path, but instead looking to categorize nodes based on their ability to reach terminal states without cycles. This highlights how pathfinding can be extended to problems where the structure of paths themselves (such as cycles or safety) matters.
