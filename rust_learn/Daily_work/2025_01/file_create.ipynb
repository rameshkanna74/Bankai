```python
import shutil
import os

# Take the original file path as input
original_file_path = input("Enter the path of the original file (e.g., /path/to/2025_jan_01.ipynb): ")

# Check if the provided file path exists
if not os.path.isfile(original_file_path):
    print("The specified file does not exist. Please check the file path and try again.")
else:
    # Get the file name from the file path
    original_file_name = os.path.basename(original_file_path)

    # Loop to create copies for the next 30 days
    for day in range(2, 32):
        # Construct the new file name
        new_file_name = f'2025_jan_{day:02d}.ipynb'

        # Construct the new file path
        new_file_path = os.path.join(os.path.dirname(original_file_path), new_file_name)

        # Copy the original file to the new file path
        shutil.copy(original_file_path, new_file_path)
        print(f'Copied {original_file_name} to {new_file_name}')
```
