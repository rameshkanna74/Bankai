You are given a 0-indexed string array words.

Let's define a boolean function isPrefixAndSuffix that takes two strings, str1 and str2:

isPrefixAndSuffix(str1, str2) returns true if str1 is both a
prefix
and a
suffix
of str2, and false otherwise.
For example, isPrefixAndSuffix("aba", "ababa") is true because "aba" is a prefix of "ababa" and also a suffix, but isPrefixAndSuffix("abc", "abcd") is false.

Return an integer denoting the number of index pairs (i, j) such that i < j, and isPrefixAndSuffix(words[i], words[j]) is true.

Example 1:

Input: words = ["a","aba","ababa","aa"]
Output: 4
Explanation: In this example, the counted index pairs are:
i = 0 and j = 1 because isPrefixAndSuffix("a", "aba") is true.
i = 0 and j = 2 because isPrefixAndSuffix("a", "ababa") is true.
i = 0 and j = 3 because isPrefixAndSuffix("a", "aa") is true.
i = 1 and j = 2 because isPrefixAndSuffix("aba", "ababa") is true.
Therefore, the answer is 4.
Example 2:

Input: words = ["pa","papa","ma","mama"]
Output: 2
Explanation: In this example, the counted index pairs are:
i = 0 and j = 1 because isPrefixAndSuffix("pa", "papa") is true.
i = 2 and j = 3 because isPrefixAndSuffix("ma", "mama") is true.
Therefore, the answer is 2.  
Example 3:

Input: words = ["abab","ab"]
Output: 0
Explanation: In this example, the only valid index pair is i = 0 and j = 1, and isPrefixAndSuffix("abab", "ab") is false.
Therefore, the answer is 0.

Constraints:

1 <= words.length <= 50
1 <= words[i].length <= 10
words[i] consists only of lowercase English letters.

````rust
pub struct Solution;

impl Solution {
    pub fn count_prefix_suffix_pairs(words: Vec<String>) -> i32 {
        // Initialize a counter to keep track of valid prefix-suffix pairs
        let mut count = 0;

        // Iterate through all possible pairs of indices (i, j) where i < j
        for i in 0..words.len() {
            for j in i + 1..words.len() {
                let prefix = &words[i];
                let candidate = &words[j];
                
                // Check if the current prefix is both a prefix and a suffix of the candidate
                if Solution::is_prefix_and_suffix(prefix, candidate) {
                    count += 1;
                }
            }
        }

        // Return the total count of valid pairs
        count
    }

    // Helper function to check if str1 is both a prefix and a suffix of str2
    fn is_prefix_and_suffix(str1: &str, str2: &str) -> bool {
        // Check if str1 is a prefix of str2
        let is_prefix = str2.starts_with(str1);

        // Check if str1 is a suffix of str2
        let is_suffix = str2.ends_with(str1);

        // Return true only if both conditions are met
        is_prefix && is_suffix
    }
}

// Example usage
fn main() {
    // Example case 1
    let words1 = vec!["a".to_string(), "aba".to_string(), "ababa".to_string(), "aa".to_string()];
    println!(
        "Count of prefix-suffix pairs: {}",
        Solution::count_prefix_suffix_pairs(words1)
    ); // Expected output: 4

    // Example case 2
    let words2 = vec!["pa".to_string(), "papa".to_string(), "ma".to_string(), "mama".to_string()];
    println!(
        "Count of prefix-suffix pairs: {}",
        Solution::count_prefix_suffix_pairs(words2)
    ); // Expected output: 2

    // Example case 3
    let words3 = vec!["abab".to_string(), "ab".to_string()];
    println!(
        "Count of prefix-suffix pairs: {}",
        Solution::count_prefix_suffix_pairs(words3)
    ); // Expected output: 0
}

main()
````
