Given an m x n integer matrix heightMap representing the height of each unit cell in a 2D elevation map, return the volume of water it can trap after raining.

Example 1:

Input: heightMap = [[1,4,3,1,3,2],[3,2,1,3,2,4],[2,3,3,2,3,1]]
Output: 4
Explanation: After the rain, water is trapped between the blocks.
We have two small ponds 1 and 3 units trapped.
The total volume of water trapped is 4.

Input: heightMap = [[3,3,3,3,3],[3,2,2,2,3],[3,2,1,2,3],[3,2,2,2,3],[3,3,3,3,3]]

````python
use std::cmp::Reverse;
use std::collections::BinaryHeap;

impl Solution {
    pub fn trap_rain_water(mut height_map: Vec<Vec<i32>>) -> i32 {
        let (m, n) = (height_map.len(), height_map[0].len());
        let mut borders = BinaryHeap::new();
        let mut visited = vec![vec![false; n]; m];

        for i in 0..n {
            borders.push(Reverse((height_map[0][i], 0, i)));
            visited[0][i] = true;
            borders.push(Reverse((height_map[m - 1][i], m - 1, i)));
            visited[m - 1][i] = true;
        }
        for i in 1..m - 1 {
            borders.push(Reverse((height_map[i][0], i, 0)));
            visited[i][0] = true;
            borders.push(Reverse((height_map[i][n - 1], i, n - 1)));
            visited[i][n - 1] = true;
        }

        let mut volume = 0;

        while let Some(Reverse((h, y, x))) = borders.pop() {
            for w in [0, 1, 0, !0, 0].windows(2) {
                let (y, x) = (y.wrapping_add(w[0]), x.wrapping_add(w[1]));
                if y >= m || x >= n || visited[y][x] {
                    continue;
                }
                if height_map[y][x] < h {
                    volume += h - height_map[y][x];
                    height_map[y][x] = h;
                }
                borders.push(Reverse((height_map[y][x], y, x)));
                visited[y][x] = true;
            }
        }

        volume
    }
}
````

````python
use std::cmp::Reverse;
use std::collections::BinaryHeap;

impl Solution {
fn trap_rain_water(height_map: Vec<Vec<i32>>) -> i32 {
    let (h, w) = (height_map.len(), height_map[0].len());
    let mut water = vec![-1; w * h];

    let mut q = BinaryHeap::with_capacity((w + h) * 2);
    let mut push = |y: usize, x: usize| {
        let height = height_map[y][x];
        water[y * w + x] = height;
        q.push((Reverse(height), y, x));
    };
    for y in 0..h {
        push(y, 0);
        push(y, w - 1);
    }
    for x in 0..w {
        push(0, x);
        push(h - 1, x);
    }

    while let Some((cur, y, x)) = q.pop() {
        let neighbors = [
            (y, x + 1),
            (y, x.wrapping_sub(1)),
            (y + 1, x),
            (y.wrapping_sub(1), x),
        ];

        for (ny, nx) in neighbors {
            if nx >= w || ny >= h || water[ny * w + nx] != -1 {
                continue;
            }
            let new = height_map[ny][nx].max(cur.0);
            water[ny * w + nx] = new;
            q.push((Reverse(new), ny, nx));
        }
    }

    height_map
        .into_iter()
        .flatten()
        .zip(water)
        .map(|(mh, wh)| wh - mh)
        .sum()
}

}
````
