```python
impl Solution {
    pub fn find_the_prefix_common_array(a: Vec<i32>, b: Vec<i32>) -> Vec<i32> {
        let n = a.len();
        let mut freq = vec![0; 51];
        let mut ans = Vec::new();
        let mut cnt = 0;

        for i in 0..n {
            freq[a[i] as usize] += 1;
            if freq[a[i] as usize] == 2 {
                cnt += 1;
            }

            freq[b[i] as usize] += 1;
            if freq[b[i] as usize] == 2 {
                cnt += 1;
            }

            ans.push(cnt);
        }

        ans
    }
}
```
