    Let i, i + 1, ..., j be the indices in the subarray. Then, for each pair of indices i <= i1, i2 <= j, 0 <= |nums[i1] - nums[i2]| <= 2.

Return the total number of continuous subarrays.

A subarray is a contiguous non-empty sequence of elements within an array.

Example 1:

Input: nums = [5,4,2,4]
Output: 8
Explanation:
Continuous subarray of size 1: [5], [4], [2], [4].
Continuous subarray of size 2: [5,4], [4,2], [2,4].
Continuous subarray of size 3: [4,2,4].
Thereare no subarrys of size 4.
Total continuous subarrays = 4 + 3 + 1 = 8.
It can be shown that there are no more continuous subarrays.

Example 2:

Input: nums = [1,2,3]

Explanation:
Continuous subarray of size 1: [1], [2], [3].
Continuous subarray of size 2: [1,2], [2,3].
Continuous subarray of size 3: [1,2,3].
Total continuous subarrays = 3 + 2 + 1 = 6.

Constraints:

    1 <= nums.length <= 105
    1 <= nums[i] <= 109

    pub fn continuous_subarrays(nums: Vec<i32>) -> i64 {
        use std::collections::VecDeque;

        // Monotonic deques to track min and max values in the current window
        let mut min_deque = VecDeque::new();
        let mut max_deque = VecDeque::new();

        let mut left = 0;
        let mut result = 0;

        for right in 0..nums.len() {
            // Maintain the max deque: only keep values smaller than the current value
            while !max_deque.is_empty() && nums[*max_deque.back().unwrap()] <= nums[right] {
                max_deque.pop_back();
            }
            max_deque.push_back(right);

            // Maintain the min deque: only keep values larger than the current value
            while !min_deque.is_empty() && nums[*min_deque.back().unwrap()] >= nums[right] {
                min_deque.pop_back();
            }
            min_deque.push_back(right);

            // Adjust the left pointer to maintain the condition: max - min <= 2
            while nums[*max_deque.front().unwrap()] - nums[*min_deque.front().unwrap()] > 2 {
                left += 1;

                // Remove elements from the deques that are out of the current window
                if *max_deque.front().unwrap() < left {
                    max_deque.pop_front();
                }
                if *min_deque.front().unwrap() < left {
                    min_deque.pop_front();
                }
            }

            // The number of valid subarrays ending at 'right' is (right - left + 1)
            result += (right - left + 1) as i64;
        }

        result

### Explanation of Changes:

1. **Monotonic Deques**:

   - We maintain two deques: `min_deque` for the indices of the minimum values and `max_deque` for the indices of the maximum values.
   - Both deques are kept in a manner such that the front of the deque always contains the smallest (for the min deque) or the largest (for the max deque) value within the current window.

2. **Sliding Window**:

   - We iterate through the array with a `right` pointer. For each element, we adjust the `left` pointer to ensure that the subarray formed by the window `[left, right]` satisfies the condition: `max - min <= 2`.
   - If the condition is violated, the `left` pointer is moved to the right to shrink the window until the condition is met again.

3. **Efficiency**:

   - Every time the `left` pointer is incremented, we ensure to remove any indices from the deques that are no longer part of the current window.
   - The overall time complexity remains **O(n)**, where `n` is the length of the input array, because each index is added and removed from the deques at most once.

4. **Result Calculation**:
   - The number of valid subarrays ending at `right` is `(right - left + 1)`, which is added to the result.
