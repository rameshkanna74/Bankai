884. Uncommon Words from Two Sentences
     Easy
     Topics
     Companies

A sentence is a string of single-space separated words where each word consists only of lowercase letters.

A word is uncommon if it appears exactly once in one of the sentences, and does not appear in the other sentence.

Given two sentences s1 and s2, return a list of all the uncommon words. You may return the answer in any order.

Example 1:

Input: s1 = "this apple is sweet", s2 = "this apple is sour"

Output: ["sweet","sour"]

Explanation:

The word "sweet" appears only in s1, while the word "sour" appears only in s2.

Example 2:

Input: s1 = "apple apple", s2 = "banana"

Output: ["banana"]

Constraints:

    1 <= s1.length, s2.length <= 200
    s1 and s2 consist of lowercase English letters and spaces.
    s1 and s2 do not have leading or trailing spaces.
    All the words in s1 and s2 are separated by a single space.

Here's an efficient Rust solution for the problem "Uncommon Words from Two Sentences" with time complexity \( O(n + m) \), where \( n \) is the number of words in `s1` and \( m \) is the number of words in `s2`. The solution uses a frequency count to identify uncommon words.

### Rust Code:

```rust
use std::collections::HashMap;

impl Solution {
    pub fn uncommon_from_sentences(s1: String, s2: String) -> Vec<String> {
        // Step 1: Create a frequency map for all words in both sentences.
        let mut word_count: HashMap<String, i32> = HashMap::new();

        // Step 2: Define a helper function to update the word count.
        fn add_words_to_map(sentence: &str, word_count: &mut HashMap<String, i32>) {
            for word in sentence.split_whitespace() {
                *word_count.entry(word.to_string()).or_insert(0) += 1;
            }
        }

        // Step 3: Add words from both sentences to the map.
        add_words_to_map(&s1, &mut word_count);
        add_words_to_map(&s2, &mut word_count);

        // Step 4: Collect words that appear exactly once.
        word_count
            .into_iter()
            .filter_map(|(word, count)| if count == 1 { Some(word) } else { None })
            .collect()
    }
}

fn main() {
    // Test case 1
    let s1 = "this apple is sweet".to_string();
    let s2 = "this apple is sour".to_string();
    let result = Solution::uncommon_from_sentences(s1, s2);
    println!("Output: {:?}", result); // Expected Output: ["sweet", "sour"]

    // Test case 2
    let s1 = "apple apple".to_string();
    let s2 = "banana".to_string();
    let result = Solution::uncommon_from_sentences(s1, s2);
    println!("Output: {:?}", result); // Expected Output: ["banana"]
}
```

### Explanation:

1. **Frequency Map**:
   We use a `HashMap` to count the frequency of each word across both sentences (`s1` and `s2`). The key in the map is the word, and the value is the count of how many times it appears.

2. **Helper Function**:
   The helper function `add_words_to_map` processes each sentence, splits it into words using `split_whitespace()`, and updates the word count in the map.

3. **Filtering Uncommon Words**:
   After creating the frequency map, we iterate over it and collect words that have a count of exactly 1. These are the uncommon words since they appear only once between both sentences.

4. **Time Complexity**: \( O(n + m) \)
   We process each word in both `s1` and `s2` once, resulting in linear time relative to the total number of words.

5. **Space Complexity**: \( O(n + m) \)
   We use a `HashMap` to store all words from both sentences.

### Example Cases:

- **Example 1**: For `s1 = "this apple is sweet"` and `s2 = "this apple is sour"`, the words "sweet" (from `s1`) and "sour" (from `s2`) are uncommon, so the result is `["sweet", "sour"]`.
- **Example 2**: For `s1 = "apple apple"` and `s2 = "banana"`, the word "banana" is the only uncommon word, so the result is `["banana"]`.

This solution is efficient for the given constraints \( 1 \leq \text{s1.length}, \text{s2.length} \leq 200 \).


````rust
use std::collections::HashMap;

pub struct Solution;

impl Solution {
    pub fn uncommon_from_sentences(s1: String, s2: String) -> Vec<String> {
        // Step 1: Create a frequency map for all words in both sentences.
        let mut word_count: HashMap<String, i32> = HashMap::new();
        
        // Step 2: Define a helper function to update the word count.
        fn add_words_to_map(sentence: &str, word_count: &mut HashMap<String, i32>) {
            for word in sentence.split_whitespace() {
                *word_count.entry(word.to_string()).or_insert(0) += 1;
            }
        }

        // Step 3: Add words from both sentences to the map.
        add_words_to_map(&s1, &mut word_count);
        add_words_to_map(&s2, &mut word_count);

        // Step 4: Collect words that appear exactly once.
        word_count
            .into_iter()
            .filter_map(|(word, count)| if count == 1 { Some(word) } else { None })
            .collect()
    }
}

fn main() {
    // Test case 1
    let s1 = "this apple is sweet".to_string();
    let s2 = "this apple is sour".to_string();
    let result = Solution::uncommon_from_sentences(s1, s2);
    println!("Output: {:?}", result); // Expected Output: ["sweet", "sour"]

    // Test case 2
    let s1 = "apple apple".to_string();
    let s2 = "banana".to_string();
    let result = Solution::uncommon_from_sentences(s1, s2);
    println!("Output: {:?}", result); // Expected Output: ["banana"]
}

main()
````
