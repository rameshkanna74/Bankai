````python
impl Solution {
    pub fn lexical_order(n: i32) -> Vec<i32> {
        let mut result = Vec::new();
        let mut curr = 1;
        for _ in 0..n {
            result.push(curr);
            if curr * 10 <= n {
                curr *= 10;
            } else {
                while curr % 10 == 9 || curr >= n {
                    curr /= 10;
                }
                curr += 1;
            }
        }
        result
    }
}
````

class Solution:
    def lexicalOrder(self, n: int) -> List[int]:
        res = []
        curr = 1
        for _ in range(n):
            res.append(curr)
            if curr * 10 <= n:
                curr *= 10
            else:
                # backtrack
                if curr >= n:
                    curr //= 10
                curr += 1
                while curr % 10 == 0:
                    curr //= 10
        return res
