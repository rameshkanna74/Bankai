You are given a string word and a non-negative integer k.

Return the total number of substrings of word that contain every vowel ('a', 'e', 'i', 'o', and 'u') at least once and exactly k consonants.

Example 1:

Input: word = "aeioqq", k = 1

Output: 0

Explanation:

There is no substring with every vowel.

Example 2:

Input: word = "aeiou", k = 0

Output: 1

Explanation:

The only substring with every vowel and zero consonants is word[0..4], which is "aeiou".

Example 3:

Input: word = "ieaouqqieaouqq", k = 1

Output: 3

Explanation:

The substrings with every vowel and one consonant are:

word[0..5], which is "ieaouq".
word[6..11], which is "qieaou".
word[7..12], which is "ieaouq".

Constraints:

5 <= word.length <= 2 \* 105
word consists only of lowercase English letters.
0 <= k <= word.length - 5

````rust
impl Solution {
    pub fn count_of_substrings(word: String, k: i32) -> i64 {
        let k = k as usize;
        let mut frequencies = vec![vec![0; 128]; 2];
        frequencies[0][b'a' as usize] = 1;
        frequencies[0][b'e' as usize] = 1;
        frequencies[0][b'i' as usize] = 1;
        frequencies[0][b'o' as usize] = 1;
        frequencies[0][b'u' as usize] = 1;

        let (mut response, mut current_k, mut vowels, mut extra_left) = (0i64, 0, 0, 0);
        let word = word.as_bytes();

        let (mut left, mut right) = (0, 0);
        while right < word.len() {
            let right_char = word[right];

            if frequencies[0][right_char as usize] == 1 {
                if frequencies[1][right_char as usize] == 0 {
                    vowels += 1;
                }
                frequencies[1][right_char as usize] += 1;
            } else {
                current_k += 1;
            }

            while current_k > k {
                let left_char = word[left];
                if frequencies[0][left_char as usize] == 1 {
                    if frequencies[1][left_char as usize] == 1 {
                        vowels -= 1;
                    }
                    frequencies[1][left_char as usize] -= 1;
                } else {
                    current_k -= 1;
                }
                left += 1;
                extra_left = 0;
            }

            while vowels == 5
                && current_k == k
                && left < right
                && frequencies[0][word[left] as usize] == 1
                && frequencies[1][word[left] as usize] > 1
            {
                extra_left += 1;
                frequencies[1][word[left] as usize] -= 1;
                left += 1;
            }

            if current_k == k && vowels == 5 {
                response += 1 + extra_left as i64;
            }

            right += 1;
        }

        response
    }
}
````
