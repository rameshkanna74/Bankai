For a binary tree T, we can define a flip operation as follows: choose any node, and swap the left and right child subtrees.

A binary tree X is flip equivalent to a binary tree Y if and only if we can make X equal to Y after some number of flip operations.

Given the roots of two binary trees root1 and root2, return true if the two trees are flip equivalent or false otherwise.

Example 1:
Flipped Trees Diagram

Input: root1 = [1,2,3,4,5,6,null,null,null,7,8], root2 = [1,3,2,null,6,4,5,null,null,null,null,8,7]
Output: true
Explanation: We flipped at nodes with values 1, 3, and 5.

Example 2:

Input: root1 = [], root2 = []
Output: true

Example 3:

Input: root1 = [], root2 = [1]
Output: false

Constraints:

    The number of nodes in each tree is in the range [0, 100].
    Each tree will have unique node values in the range [0, 99].

```python
use std::rc::Rc;
use std::cell::{Ref, RefCell};

type NodePtr = Rc<RefCell<TreeNode>>;
type NodeOpt = Option<NodePtr>;
type NodeRef<'a> = Ref<'a, TreeNode>;

impl Solution {
    pub fn flip_equiv(root1: NodeOpt, root2: NodeOpt) -> bool {
        if get_val(&root1) == get_val(&root2) {

            let mut stack = vec![(root1.clone(), root2.clone())];

            while let Some(pair) = stack.pop() {
                match pair {
                    (Some(node_ptr1), Some(node_ptr2)) => {
                        let node1 = node_ptr1.borrow();
                        let node2 = node_ptr2.borrow();
                        match compare(&node1, &node2) {
                            Unflipped => {
                                stack.push((node1.right.clone(), 
                                            node2.right.clone()));
                                stack.push((node1.left.clone(),
                                            node2.left.clone()))
                            },
                            Flipped => {
                                stack.push((node1.right.clone(), 
                                            node2.left.clone()));
                                stack.push((node1.left.clone(),
                                            node2.right.clone()))

                            },
                            Dissimilar => {
                                return false;
                            }
                        }
                    },
                    (None, None) => {
                        // Ignore.
                    },
                    _ => { 
                        return false;
                    }
                }
            }
            true
        } else {
            false
        }
    }
}

#[inline]
fn get_val(node: &NodeOpt) -> Option<i32> {
    node.as_ref().map(|n| n.borrow().val)
}

enum NodeComparison {
    Unflipped,
    Flipped,
    Dissimilar
}
use NodeComparison::*;

fn compare(node1: &NodeRef, node2: &NodeRef) -> NodeComparison {
    let v1 = get_val(&node1.left);
    let v2 = get_val(&node1.right);
    let v3 = get_val(&node2.left);
    let v4 = get_val(&node2.right);

    if v1 == v3 && v2 == v4 {
        Unflipped
    } else if v1 == v4 && v2 == v3 {
        Flipped
    } else {
        Dissimilar
    }
}
```
