Given two positive integers n and k, the binary string Sn is formed as follows:

    S1 = "0"
    Si = Si - 1 + "1" + reverse(invert(Si - 1)) for i > 1

Where + denotes the concatenation operation, reverse(x) returns the reversed string x, and invert(x) inverts all the bits in x (0 changes to 1 and 1 changes to 0).

For example, the first four strings in the above sequence are:

    S1 = "0"
    S2 = "011"
    S3 = "0111001"
    S4 = "011100110110001"

Return the kth bit in Sn. It is guaranteed that k is valid for the given n.

Example 1:

Input: n = 3, k = 1
Output: "0"
Explanation: S3 is "0111001".
The 1st bit is "0".

Example 2:

Input: n = 4, k = 11
Output: "1"
Explanation: S4 is "011100110110001".
The 11th bit is "1".

Constraints:

    1 <= n <= 20
    1 <= k <= 2n - 1

   - The length of the string \( S_n \) is calculated as \( 2^n - 1 \).

2. **Recursive Function**:
   - The `find_kth_bit_recursive` function determines where \( k \) lies:
     - If \( k \) is exactly at the middle index, return '1'.
     - If \( k \) is in the left half, recursively find the bit in \( S_{n-1} \).

3. **Bit Inversion**:
   - When accessing the right half, the bit is inverted by checking whether it's '0' or '1'.

### Complexity:
- **Time Complexity**: \( O(n) \), due to the depth of recursion.
- **Space Complexity**: \( O(n) \), due to the recursion stack.

This approach is much more efficient than generating the entire string and works well for the constraints provided.

````rust
impl Solution {
	pub fn find_kth_bit(n: i32, k: i32) -> char {
		let mut s = vec![b'0'];
		for _ in 1..n {
			s.push(b'1');
			for i in (0..s.len() - 1).rev() {
				s.push(b'0' + 1 ^ (s[i] - b'0'));
			}
		}
		s[k as usize - 1] as _
	}
}
````
