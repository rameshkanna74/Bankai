You are given a 2D integer array intervals where intervals[i] = [lefti, righti] represents the inclusive interval [lefti, righti].

You have to divide the intervals into one or more groups such that each interval is in exactly one group, and no two intervals that are in the same group intersect each other.

Return the minimum number of groups you need to make.

Two intervals intersect if there is at least one common number between them. For example, the intervals [1, 5] and [5, 8] intersect.

Example 1:

Input: intervals = [[5,10],[6,8],[1,5],[2,3],[1,10]]
Output: 3
Explanation: We can divide the intervals into the following groups:

- Group 1: [1, 5], [6, 8].
- Group 2: [2, 3], [5, 10].
- Group 3: [1, 10].
  It can be proven that it is not possible to divide the intervals into fewer than 3 groups.

Example 2:

Input: intervals = [[1,3],[5,6],[8,10],[11,13]]
Output: 1
Explanation: None of the intervals overlap, so we can put all of them in one group.

Constraints:

    1 <= intervals.length <= 105
    intervals[i].length == 2
    1 <= lefti <= righti <= 106

impl Solution {
    pub fn min_groups(intervals: Vec<Vec<i32>>) -> i32 {
        // Step 1: Collect start and end times into two separate vectors
        let mut starts: Vec<i32> = Vec::with_capacity(intervals.len());
        let mut ends: Vec<i32> = Vec::with_capacity(intervals.len());

        for interval in &intervals {
            starts.push(interval[0]); // Collect start times
            ends.push(interval[1]); // Collect end times
        }

        // Step 2: Sort the starts and ends
        starts.sort_unstable(); // Sort start times
        ends.sort_unstable(); // Sort end times

        let mut total = 0; // To keep track of active intervals
        let mut max_total = 0; // To record the maximum active intervals at any time
        let mut start_idx = 0; // Index for traversing start times
        let mut end_idx = 0; // Index for traversing end times

        // Step 3: Traverse through sorted starts and ends
        while start_idx < intervals.len() {
            if starts[start_idx] <= ends[end_idx] {
                // A new interval is starting before the current one ends
                total += 1; // Increase active interval count
                start_idx += 1; // Move to the next start
                max_total = max_total.max(total); // Update the maximum if needed
            } else {
                // The current interval has ended
                total -= 1; // Decrease active interval count
                end_idx += 1; // Move to the next end
            }
        }

        max_total // Return the maximum number of overlapping intervals
    }
}
