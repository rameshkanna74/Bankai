You are given a string s.

Your task is to remove all digits by doing this operation repeatedly:

Delete the first digit and the closest non-digit character to its left.
Return the resulting string after removing all digits.

 

Example 1:

Input: s = "abc"

Output: "abc"

Explanation:

There is no digit in the string.

Example 2:

Input: s = "cb34"

Output: ""

Explanation:

First, we apply the operation on s[2], and s becomes "c4".

Then we apply the operation on s[1], and s becomes "".

 

Constraints:

1 <= s.length <= 100
s consists only of lowercase English letters and digits.
The input is generated such that it is possible to delete all digits.

````rust
impl Solution {
    fn clear_digits(mut s: String) -> String {
        let is_digit = |byte: u8| (48u8..=57).contains(&byte);

        let bytes = unsafe { s.as_bytes_mut() };

        let n = bytes.len(); // length of input string
        let mut i = 0usize;  // candidate element under consideration
        let mut j = 0usize;  // length of the result string.

        while i < n {
            if is_digit(bytes[i]) { 
                i += 1; 
                j = j.saturating_sub(1);
                continue;
            }

            bytes.swap(i, j);
            i += 1; j += 1;
        }

        s.truncate(j);
        return s;
    }
}
````

````rust
impl Solution {
    /// Removes all digits from the string following the deletion rule.
    ///
    /// # Arguments
    /// * `s` - The input string containing both letters and digits.
    ///
    /// # Returns
    /// * The final string after all removal operations.
    pub fn clear_digits(s: String) -> String {
        let mut stack:Vec<char> = Vec::new(); // Stack to store non-digit characters

        for ch in s.chars() {
            if ch.is_ascii_digit() {
                // If the stack is not empty, remove the closest non-digit character to the left
                stack.pop();
            } else {
                // Push non-digit characters onto the stack
                stack.push(ch);
            }
        }

        // Convert the stack into a string and return the final result
        stack.into_iter().collect()
    }
}

````
