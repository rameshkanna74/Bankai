1415. The k-th Lexicographical String of All Happy Strings of Length n

    Medium

A happy string is a string that:

consists only of letters of the set ['a', 'b', 'c'].
s[i] != s[i + 1] for all values of i from 1 to s.length - 1 (string is 1-indexed).
For example, strings "abc", "ac", "b" and "abcbabcbcb" are all happy strings and strings "aa", "baa" and "ababbc" are not happy strings.

Given two integers n and k, consider a list of all happy strings of length n sorted in lexicographical order.

Return the kth string of this list or return an empty string if there are less than k happy strings of length n.

Example 1:

Input: n = 1, k = 3
Output: "c"
Explanation: The list ["a", "b", "c"] contains all happy strings of length 1. The third string is "c".
Example 2:

Input: n = 1, k = 4
Output: ""
Explanation: There are only 3 happy strings of length 1.
Example 3:

Input: n = 3, k = 9
Output: "cab"
Explanation: There are 12 different happy string of length 3 ["aba", "abc", "aca", "acb", "bab", "bac", "bca", "bcb", "cab", "cac", "cba", "cbc"]. You will find the 9th string = "cab"

Constraints:

1 <= n <= 10
1 <= k <= 100

````rust
impl Solution {
    pub fn get_happy_string(n: i32, mut k: i32) -> String {
        let mut s = String::with_capacity(n as usize);
        Self::bt(n as usize, &mut k, &mut s);
        s
    }
    fn bt(n: usize, k: &mut i32, s: &mut String) {
        if s.len() == n {
            *k -= 1;
            return;
        }
        for c in 'a'..='c' {
            if s.ends_with(c) {
                continue;
            }
            s.push(c);
            Self::bt(n, k, s);
            if *k == 0 {
                break;
            }
            s.pop();
        }
    }
}
````

````rust
pub struct Solution;

impl Solution {
    /// Returns the k-th lexicographical happy string of length `n`
    /// If there are fewer than `k` happy strings, returns an empty string
    pub fn get_happy_string(n: i32, mut k: i32) -> String {
        // Create a mutable string with pre-allocated capacity to optimize memory usage
        let mut happy_string = String::with_capacity(n as usize);
        
        println!("Starting to generate the {}-th happy string of length {}", k, n);
        
        // Begin backtracking to generate happy strings
        Self::generate_happy_string(n as usize, &mut k, &mut happy_string);
        
        // If k is reduced to zero, we found the k-th happy string; otherwise, return an empty string
        if k == 0 {
            println!("Found the {}-th happy string: {}", k, happy_string);
            happy_string
        } else {
            println!("There are fewer than {} happy strings of length {}", k, n);
            String::new()
        }
    }

    /// Backtracking function to generate happy strings of length `n`
    fn generate_happy_string(target_length: usize, remaining_k: &mut i32, current_string: &mut String) {
        // Base case: If the current string reaches the target length
        if current_string.len() == target_length {
            *remaining_k -= 1; // Decrease the counter as we found a valid happy string
            println!(
                "Generated happy string: {}, remaining_k: {}",
                current_string, remaining_k
            );
            return;
        }

        // Try adding 'a', 'b', and 'c' to the current string
        for next_char in 'a'..='c' {
            // Ensure no consecutive characters are the same
            if current_string.ends_with(next_char) {
                println!(
                    "Skipping char '{}' because it matches the last character of {}",
                    next_char, current_string
                );
                continue;
            }

            // Choose: Add the character to the current string
            current_string.push(next_char);
            println!("Trying char '{}' -> {}", next_char, current_string);

            // Explore: Recursively generate the rest of the string
            Self::generate_happy_string(target_length, remaining_k, current_string);

            // If we've found the k-th string, exit early to save time
            if *remaining_k == 0 {
                return;
            }

            // Undo (backtrack): Remove the last character and try the next one
            println!("Backtracking, removing '{}' -> {}", next_char, current_string);
            current_string.pop();
        }
    }
}

fn main() {
    // Example cases to run and test
    let examples = vec![
        (1, 3), // Expected output: "c"
        (1, 4), // Expected output: ""
        (3, 9), // Expected output: "cab"
    ];

    for (n, k) in examples {
        println!("Input: n = {}, k = {}", n, k);
        let result = Solution::get_happy_string(n, k);
        println!("Output: {}\n", result);
    }
}

main()
````
