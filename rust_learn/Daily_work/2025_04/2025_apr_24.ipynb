2799. Count Complete Subarrays in an Array

Medium

Hint

You are given an array nums consisting of positive integers.

We call a subarray of an array complete if the following condition is satisfied:

The number of distinct elements in the subarray is equal to the number of distinct elements in the whole array.
Return the number of complete subarrays.

A subarray is a contiguous non-empty part of an array.

Example 1:

Input: nums = [1,3,1,2,2]
Output: 4
Explanation: The complete subarrays are the following: [1,3,1,2], [1,3,1,2,2], [3,1,2] and [3,1,2,2].
Example 2:

Input: nums = [5,5,5,5]
Output: 10
Explanation: The array consists only of the integer 5, so any subarray is complete. The number of subarrays that we can choose is 10.

Constraints:

1 <= nums.length <= 1000
1 <= nums[i] <= 2000

````rust
use std::collections::{HashMap, HashSet};

impl Solution {
    fn count_complete_subarrays(nums: Vec<i32>) -> i32 {
        let total: HashSet<_> = nums.iter().collect();
        let mut freq = HashMap::new();
        let mut l = 0;
        let mut res = 0;

        for (r, &num) in nums.iter().enumerate() {
            *freq.entry(num).or_insert(0) += 1;
            while freq.len() == total.len() {
                res += nums.len() as i32 - r as i32;
                let left = nums[l];
                *freq.get_mut(&left).unwrap() -= 1;
                if freq[&left] == 0 {
                    freq.remove(&left);
                }
                l += 1;
            }
        }
        res
    }
}
````

````rust
use std::collections::{HashMap, HashSet};

impl Solution {
    pub fn count_complete_subarrays(nums: Vec<i32>) -> i32 {
        let target = nums.iter().collect::<HashSet<_>>().len();
        let n = nums.len();
        let mut count = 0;
        let mut seen = HashMap::new();
        let mut l = 0;
        let mut res = 0;

        for r in 0..n {
            let entry = seen.entry(nums[r]).or_insert(0);
            *entry += 1;
            if *entry == 1 {
                count += 1;
            }

            while count == target {
                let entry = seen.get_mut(&nums[l]).unwrap();
                *entry -= 1;
                res += n - r;

                if *entry == 0 {
                    seen.remove(&nums[l]);
                    count -= 1;
                }
                l += 1;
            }
        }

        res as i32
    }
}
````
