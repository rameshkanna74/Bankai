You are given an integer array nums and two integers minK and maxK.

A fixed-bound subarray of nums is a subarray that satisfies the following conditions:

The minimum value in the subarray is equal to minK.
The maximum value in the subarray is equal to maxK.
Return the number of fixed-bound subarrays.

A subarray is a contiguous part of an array.

Example 1:

Input: nums = [1,3,5,2,7,5], minK = 1, maxK = 5
Output: 2
Explanation: The fixed-bound subarrays are [1,3,5] and [1,3,5,2].
Example 2:

Input: nums = [1,1,1,1], minK = 1, maxK = 1
Output: 10
Explanation: Every subarray of nums is a fixed-bound subarray. There are 10 possible subarrays.

Constraints:

2 <= nums.length <= 105
1 <= nums[i], minK, maxK <= 106

````rust
impl Solution {
    pub fn count_subarrays(nums: Vec<i32>, min_k: i32, max_k: i32) -> i64 {
        let mut res = 0i64;
        let (mut bad, mut left, mut right) = (-1, -1, -1);
        for (i, &v) in nums.iter().enumerate() {
            if v < min_k || v > max_k { bad = i as i32; }
            if v == min_k { left = i as i32; }
            if v == max_k { right = i as i32; }
            res += std::cmp::max(0, std::cmp::min(left, right) - bad) as i64;
        }
        res
    }
}
````

````rust
impl Solution {
    pub fn count_subarrays(nums: Vec<i32>, min_k: i32, max_k: i32) -> i64 {
        let mut cnt = 0;
        let mut has_mink = false;
        let mut has_maxk = false;
        let mut idx_mink = 0;
        let mut idx_maxk = 0;
        let mut offset = 0;

        for (i, &num) in nums.iter().enumerate() {
            if num < min_k || num > max_k {
                has_mink = false;
                has_maxk = false;
                offset = i + 1;
                continue;
            }

            if num == min_k {
                has_mink = true;
                idx_mink = i;
            }
            if num == max_k {
                has_maxk = true;
                idx_maxk = i;
            }

            if has_mink && has_maxk {
                let smaller_idx = if idx_mink < idx_maxk {
                    idx_mink
                } else {
                    idx_maxk
                };
                cnt += (smaller_idx - offset + 1) as i64;
            }
        }

        cnt
    }
}
````

````rust
class Solution:
    def countSubarrays(self, nums, minK, maxK):
        res = 0
        bad_idx = left_idx = right_idx = -1
        for i, num in enumerate(nums):
            if not (minK <= num <= maxK):
                bad_idx = i
            if num == minK:
                left_idx = i
            if num == maxK:
                right_idx = i
            res += max(0, min(left_idx, right_idx) - bad_idx)
        return res
````
