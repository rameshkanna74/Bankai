````python
impl Solution {
    pub fn difference_of_sums(n: i32, m: i32) -> i32 {
        let total_sum = n * (n + 1) / 2;
        let divisible_count = n / m;
        let divisible_sum = m * divisible_count * (divisible_count + 1) / 2;
        total_sum - 2 * divisible_sum
    }
}
````
