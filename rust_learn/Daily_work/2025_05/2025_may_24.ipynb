````python
impl Solution {
  pub fn find_words_containing(words: Vec<String>, x: char) -> Vec<i32> {
    let mut indices: Vec<i32> = Vec::new();
    for (index, word) in words.iter().enumerate() {
      if (word.contains(x)) {
        indices.push(index as i32);
      }
    }
    return indices;
  }
}
````
