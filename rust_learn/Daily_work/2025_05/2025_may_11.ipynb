````python
impl Solution {
    pub fn three_consecutive_odds(arr: Vec<i32>) -> bool {
        let mut window = 0;
        for i in 0..3.min(arr.len()) {
            window += arr[i] % 2;
        }
        if window == 3 {
            return true;
        }
        for i in 3..arr.len() {
            window += arr[i] % 2;
            window -= arr[i - 3] % 2;
            if window == 3 {
                return true;
            }
        }
        false
    }
}
````
