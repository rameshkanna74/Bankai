1 <= dominoes[i][j] <= 9

````python
impl Solution {
    pub fn num_equiv_domino_pairs(dominoes: Vec<Vec<i32>>) -> i32 {
        let mut count = vec![0; 100];
        let mut res = 0;
        for d in dominoes {
            let val = if d[0] < d[1] { d[0]*10 + d[1] } else { d[1]*10 + d[0] };
            res += count[val as usize];
            count[val as usize] += 1;
        }
        res
    }
}
````

        for a, b in dominoes:

            res += count[val]
