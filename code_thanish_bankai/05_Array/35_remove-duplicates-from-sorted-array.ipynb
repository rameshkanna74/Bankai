To solve the problem of removing duplicates from a sorted array in Rust, we need to modify the array in-place such that only unique elements remain at the start of the array, and the rest of the elements can be ignored.

### Plan:

1. **Input**: A mutable reference to a vector `nums` that is already sorted in non-decreasing order.
2. **Goal**: Remove duplicates in-place such that:
   - The first `k` elements of `nums` contain the unique elements.
   - Return `k`, the number of unique elements.
3. **Approach**:
   - Use a two-pointer technique:
     - One pointer `i` will track the position of the unique elements.
     - The other pointer `j` will scan through the array.
     - Whenever a new unique element is found (i.e., `nums[j] != nums[i]`), increment `i` and copy the new element to the position `i`.
   - After the loop, the first `i+1` elements in `nums` will contain the unique elements, and the rest can be ignored.

### Time and Space Complexity:

- **Time Complexity**: The time complexity is **O(n)**, where `n` is the length of the array. We make a single pass through the array.
- **Space Complexity**: The space complexity is **O(1)**, since we are modifying the array in-place and not using any additional space that scales with input size.

### Rust Code Implementation:

```rust
impl Solution {
    pub fn remove_duplicates(nums: &mut Vec<i32>) -> i32 {
        // If the array is empty, return 0
        if nums.is_empty() {
            return 0;
        }

        // Pointer `i` will track the position of the unique elements
        let mut i = 0;

        // Start scanning the array from the second element (index 1)
        for j in 1..nums.len() {
            // If the current element is different from the last unique element
            if nums[j] != nums[i] {
                // Move `i` to the next position and copy the unique element
                i += 1;
                nums[i] = nums[j];
            }
        }

        // The number of unique elements is i + 1
        (i + 1) as i32
    }
}

    // Example 2

    println!("k = {}, nums = {:?}", k2, &nums2[..k2 as usize]); // Expected: k = 5, nums = [0, 1, 2, 3, 4]
```

### Explanation:

1. **`remove_duplicates` function**:
   - **Input**: A mutable reference to a vector `nums`.
   - **Output**: The number of unique elements in `nums` after duplicates have been removed.
   - **Logic**:
     - If the array is empty, return `0` immediately.
     - Initialize `i = 0`, which will point to the last unique element.
     - Use a loop to iterate through the array starting from index `1` (second element). Compare each element `nums[j]` with the element at `nums[i]` (the last unique element).
     - If `nums[j] != nums[i]`, it means a new unique element is found. Increment `i` and copy `nums[j]` to `nums[i]`.
     - After the loop completes, the first `i + 1` elements in `nums` will contain the unique elements, and the function returns `i + 1` (number of unique elements).

### Edge Cases:

1. **Empty array**: If the array is empty, the function correctly returns `0`.
2. **All elements are the same**: If all elements are the same, the result will be `1`, and the array will contain only one unique element.
3. **Array with one element**: If the array has only one element, it is trivially sorted and contains only one unique element.

This solution is both time-efficient and space-efficient, meeting the problem's constraints and ensuring correctness for large inputs (up to size `3 * 10^4`).


````rust
pub struct Solution;

impl Solution {
    pub fn remove_duplicates(nums: &mut Vec<i32>) -> i32 {
        // If the array is empty, return 0
        if nums.is_empty() {
            return 0;
        }
        
        // Pointer `i` will track the position of the unique elements
        let mut i = 0;
        
        // Start scanning the array from the second element (index 1)
        for j in 1..nums.len() {
            // If the current element is different from the last unique element
            if nums[j] != nums[i] {
                // Move `i` to the next position and copy the unique element
                i += 1;
                nums[i] = nums[j];
            }
        }
        
        // The number of unique elements is i + 1
        (i + 1) as i32
    }
}

fn main() {
    // Example 1
    let mut nums1 = vec![1, 1, 2];
    let k1 = Solution::remove_duplicates(&mut nums1);
    println!("k = {}, nums = {:?}", k1, &nums1[..k1 as usize]); // Expected: k = 2, nums = [1, 2]

    // Example 2
    let mut nums2 = vec![0, 0, 1, 1, 1, 2, 2, 3, 3, 4];
    let k2 = Solution::remove_duplicates(&mut nums2);
    println!("k = {}, nums = {:?}", k2, &nums2[..k2 as usize]); // Expected: k = 5, nums = [0, 1, 2, 3, 4]
}

main()
````
