## âœ… Core Insight & Algorithm Design

### Key Observations:

Two strings are **close** if:

1. They contain the **same set of characters** (since you can't introduce new ones).
2. Their **frequency distributions** can be transformed into each other (via Operation 2).

---

### âœ… Algorithm Steps:

1. Count the frequency of each character in `word1` and `word2`.
2. Ensure both have **the same set of characters**.
3. Sort the **frequency values** and check if they match.

---

### âœ… Time & Space Complexity

- **Time:** `O(n + k log k)`
  - `O(n)` to count characters, where `n = max(word1.len(), word2.len())`
  - `O(k log k)` to sort frequency vectors (`k = 26`, constant)
- **Space:** `O(1)` â€” bounded by 26 lowercase letters

```rust
use std::collections::HashMap;

pub struct Solution;

impl Solution {
    pub fn close_strings(word1: String, word2: String) -> bool {
        if word1.len() != word2.len() {
            return false;
        }

        let mut freq1 = [0; 26];
        let mut freq2 = [0; 26];

        for ch in word1.bytes() {
            freq1[(ch - b'a') as usize] += 1;
        }

        for ch in word2.bytes() {
            freq2[(ch - b'a') as usize] += 1;
        }

        // Characters used must match
        for i in 0..26 {
            if (freq1[i] == 0) ^ (freq2[i] == 0) {
                return false;
            }
        }

        // Frequencies (multiset) must match
        let mut sorted1: Vec<_> = freq1.iter().filter(|&&x| x > 0).cloned().collect();
        let mut sorted2: Vec<_> = freq2.iter().filter(|&&x| x > 0).cloned().collect();
        sorted1.sort_unstable();
        sorted2.sort_unstable();

        sorted1 == sorted2
    }
}
```

````rust

class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:
        freq1 = [0] * 26
        freq2 = [0] * 26

        for ch in word1:
            freq1[ord(ch) - ord('a')] += 1

        for ch in word2:
            freq2[ord(ch) - ord('a')] += 1

        for i in range(26):
            if (freq1[i] == 0 and freq2[i] != 0) or (freq1[i] != 0 and freq2[i] == 0):
                return False

        freq1.sort()
        freq2.sort()

        for i in range(26):
            if freq1[i] != freq2[i]:
                return False

        return True
````
