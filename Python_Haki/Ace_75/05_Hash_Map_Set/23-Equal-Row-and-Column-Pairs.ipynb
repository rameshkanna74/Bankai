## ‚úÖ Problem Summary

You're given an `n x n` matrix. You must count the number of `(row, column)` index pairs where the row vector is **identical** to the column vector.

## ‚úÖ Core Idea & Strategy

### Key Observations:
1. Rows are easy to compare: directly access via `grid[i]`.
2. Columns can be **transposed** or accessed via iteration.
3. To efficiently compare row and column vectors:
   - Hash the row as a vector (or tuple) into a frequency map.
   - Then, generate each column vector and check if it exists in the map.

---

### ‚úÖ Algorithm Steps:
1. Create a `HashMap<Vec<i32>, i32>` to store the frequency of each row.
2. Iterate over each row and store it in the map.
3. For each column, construct it as a `Vec<i32>`.
4. Check if that column exists in the map.
5. If it does, add the count (could be multiple rows matching this column).

---

## ‚úÖ Optimal Rust Solution

```rust
use std::collections::HashMap;

pub struct Solution;

impl Solution {
    pub fn equal_pairs(grid: Vec<Vec<i32>>) -> i32 {
        let n = grid.len();
        let mut row_map: HashMap<Vec<i32>, i32> = HashMap::new();

        // Count all rows
        for row in &grid {
            *row_map.entry(row.clone()).or_insert(0) += 1;
        }

        let mut count = 0;

        // Iterate column-wise
        for col in 0..n {
            let mut col_vec = Vec::with_capacity(n);
            for row in 0..n {
                col_vec.push(grid[row][col]);
            }

            if let Some(&freq) = row_map.get(&col_vec) {
                count += freq;
            }
        }

        count
    }
}
```
