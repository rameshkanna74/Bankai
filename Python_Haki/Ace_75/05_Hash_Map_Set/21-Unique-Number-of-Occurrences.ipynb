## ‚úÖ Problem Summary

Given an array of integers, check if the number of **occurrences** of each value is **unique**.

---

## ‚úÖ Algorithm Design

### Steps:
1. Count the frequency of each integer using a `HashMap<i32, usize>`.
2. Store all frequencies in a `HashSet<usize>` to check for uniqueness.
3. If the number of unique frequencies matches the number of keys in the map, return `true`.

### Why this works:
- HashMaps give **O(1)** average-time complexity for insert and lookup.
- HashSets are ideal for checking uniqueness.

---

## ‚úÖ Time & Space Complexity

- **Time Complexity:** `O(n)`  
   Where `n` is the number of elements in the input array. We make two linear passes:
   - One to count frequencies
   - One to check uniqueness

- **Space Complexity:** `O(n)` in worst case  
   - HashMap stores up to `n` keys (all unique values)
   - HashSet stores up to `n` frequencies (if all counts differ)

---

## ‚úÖ Edge Cases Handled
- Empty input ‚Üí not possible due to constraints `arr.len() >= 1`
- All same values
- All unique values
- Duplicates with same and different frequencies

## ‚úÖ Rust Code (Modular, Idiomatic)

```rust
use std::collections::{HashMap, HashSet};

pub struct Solution;

impl Solution {
    pub fn unique_occurrences(arr: Vec<i32>) -> bool {
        let mut freq_map = HashMap::new();
        
        // Count frequencies
        for num in arr {
            *freq_map.entry(num).or_insert(0) += 1;
        }
        
        let mut seen = HashSet::new();
        
        // Check if all frequencies are unique
        for &count in freq_map.values() {
            if !seen.insert(count) {
                return false;
            }
        }
        
        true
    }
}
```

````rust
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        arr.sort()
        v = []

        i = 0
        while i < len(arr):
            cnt = 1

            # Count occurrences of the current element
            while i + 1 < len(arr) and arr[i] == arr[i + 1]:
                cnt += 1
                i += 1

            v.append(cnt)
            i += 1

        v.sort()

        for i in range(1, len(v)):
            if v[i] == v[i - 1]:
                return False

        return True

````
