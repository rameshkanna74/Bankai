## âœ… Algorithm Overview

This is a **combinatorics problem** disguised as a grid traversal.

To go from `(0, 0)` to `(m - 1, n - 1)`, you need to make:

- `(m - 1)` down moves
- `(n - 1)` right moves

That totals to `(m - 1 + n - 1) = (m + n - 2)` moves, and we must choose where to place the right (or down) moves.

### ðŸ”¢ Formula:

\[
\text{Unique Paths} = \binom{m+n-2}{m-1}
\]

This is **optimal** in both **time** and **space**, avoiding unnecessary DP arrays.

## ðŸ§  DSA Concepts Used

- **Combinatorics**: For choosing move sequences.
- **Math Optimizations**: Computing binomial coefficients iteratively to avoid overflow.

## ðŸš€ Time & Space Complexity

- **Time:** `O(min(m, n))` â€” we compute the binomial coefficient iteratively for the smaller of `m-1` or `n-1`.
- **Space:** `O(1)` â€” no extra memory except a few variables.

This is **optimal** for this problem.

## ðŸ’¡ Edge Case Handling

- **`m == 1` or `n == 1`**: Only one way to reach destination â€” handled naturally by the formula.
- **`m == n == 1`**: Also returns 1.

## ðŸ¦€ Rust Implementation

```rust
pub struct Solution;

impl Solution {
    pub fn unique_paths(m: i32, n: i32) -> i32 {
        fn binomial_coefficient(n: i32, k: i32) -> i32 {
            let (n, k) = (n as u64, k.min(n - k) as u64); // use symmetry
            let mut result = 1u64;
            for i in 0..k {
                result = result * (n - i) / (i + 1);
            }
            result as i32
        }

        binomial_coefficient(m + n - 2, m - 1)
    }
}
```

````python
pub struct Solution;

impl Solution {
    pub fn unique_paths(m: i32, n: i32) -> i32 {
        fn binomial_coefficient(n: i32, k: i32) -> i32 {
            let (n, k) = (n as u64, k.min(n - k) as u64); // use symmetry
            let mut result = 1u64;
            for i in 0..k {
                result = result * (n - i) / (i + 1);
            }
            result as i32
        }

        binomial_coefficient(m + n - 2, m - 1)
    }
}
````

````python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        
        aboveRow = [1] * n

        for _ in range(m - 1):
            currentRow = [1] * n
            for i in range(1, n):
                currentRow[i] = currentRow[i-1] + aboveRow[i]
            aboveRow = currentRow
        
        return aboveRow[-1]
````
