## üî¢ Leetcode 72: Edit Distance

---

### üß† Problem Summary

Given two strings `word1` and `word2`, compute the **minimum number of operations** (insert, delete, replace) to transform `word1` into `word2`.

---

## üí° Core Algorithm: Dynamic Programming (2D, Space Optimized)

### üîë Intuition

Use a **DP matrix** where:

- `dp[i][j]` = minimum operations to convert `word1[..i]` to `word2[..j]`

### üîÅ Recurrence Relation

```text
If word1[i-1] == word2[j-1]:
    dp[i][j] = dp[i-1][j-1]         (no cost)
Else:
    dp[i][j] = 1 + min(
        dp[i-1][j],    // Delete
        dp[i][j-1],    // Insert
        dp[i-1][j-1]   // Replace
    )
```

### üß± Base Cases

- `dp[i][0] = i` (delete all characters from word1)
- `dp[0][j] = j` (insert all characters of word2)

---

## ‚úÖ Time and Space Complexity

- **Time:** `O(m * n)` ‚Äî must compute each subproblem
- **Space:** `O(n)` ‚Äî 2 rolling rows (`prev`, `curr`)

This is optimal for inputs up to length 500.

## ü¶Ä Idiomatic Rust Implementation

```rust
pub struct Solution;

impl Solution {
    pub fn min_distance(word1: String, word2: String) -> i32 {
        let (m, n) = (word1.len(), word2.len());
        let (w1, w2): (Vec<char>, Vec<char>) = (word1.chars().collect(), word2.chars().collect());

        // Early exit for empty cases
        if m == 0 { return n as i32; }
        if n == 0 { return m as i32; }

        let mut prev = (0..=n).collect::<Vec<i32>>();
        let mut curr = vec![0; n + 1];

        for i in 1..=m {
            curr[0] = i as i32;
            for j in 1..=n {
                if w1[i - 1] == w2[j - 1] {
                    curr[j] = prev[j - 1];
                } else {
                    curr[j] = 1 + prev[j].min(curr[j - 1]).min(prev[j - 1]);
                }
            }
            std::mem::swap(&mut prev, &mut curr);
        }

        prev[n]
    }
}
```

````python
pub struct Solution;

impl Solution {
    pub fn min_distance(word1: String, word2: String) -> i32 {
        let (m, n) = (word1.len(), word2.len());
        let (w1, w2): (Vec<char>, Vec<char>) = (word1.chars().collect(), word2.chars().collect());

        // Early exit for empty cases
        if m == 0 { return n as i32; }
        if n == 0 { return m as i32; }

        let mut prev = (0..=n).collect::<Vec<i32>>();
        let mut curr = vec![0; n + 1];

        for i in 1..=m {
            curr[0] = i as i32;
            for j in 1..=n {
                if w1[i - 1] == w2[j - 1] {
                    curr[j] = prev[j - 1];
                } else {
                    curr[j] = 1 + prev[j].min(curr[j - 1]).min(prev[j - 1]);
                }
            }
            std::mem::swap(&mut prev, &mut curr);
        }

        prev[n]
    }
}
````

````python
class Solution:
  def minDistance(self, word1: str, word2: str) -> int:
    m = len(word1)
    n = len(word2)
    # dp[i][j] := min # Of operations to convert word1[0..i) to word2[0..j)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
      dp[i][0] = i

    for j in range(1, n + 1):
      dp[0][j] = j

    for i in range(1, m + 1):
      for j in range(1, n + 1):
        if word1[i - 1] == word2[j - 1]:
          dp[i][j] = dp[i - 1][j - 1]
        else:
          dp[i][j] = min(dp[i - 1][j - 1], dp[i - 1][j], dp[i][j - 1]) + 1

    return dp[m][n]
````
