Input: text1 = "abc", text2 = "abc"

## ðŸ§  Problem Understanding

Given two strings, we must find the **length** of the **Longest Common Subsequence (LCS)** â€” not substring. This is a **classic dynamic programming (DP)** problem.

---

## ðŸš€ Algorithm Explanation

### ðŸ”‘ Key Idea

Use **bottom-up DP** with **space optimization**:

- Let `dp[i][j]` be the length of LCS between:
  - `text1[0..i]`
  - `text2[0..j]`
- We only need **2 rows** at any time, so we optimize space to `O(n)`.

---

## âœ… Time & Space Complexity

- **Time Complexity:** `O(m * n)`
  - Must examine all character pairs.
- **Space Complexity:** `O(n)`
  - We optimize DP to use 2 rolling rows.

---

## ðŸ§ª Edge Case Handling

- One or both strings are empty â†’ LCS = 0.
- Strings with no common characters â†’ LCS = 0.
- Identical strings â†’ LCS = length of the string.

---

## ðŸ¦€ Idiomatic, Interview-Ready Rust Implementation

```rust
pub struct Solution;

impl Solution {
    pub fn longest_common_subsequence(text1: String, text2: String) -> i32 {
        let (s1, s2): (Vec<char>, Vec<char>) = (text1.chars().collect(), text2.chars().collect());
        let (m, n) = (s1.len(), s2.len());

        if m == 0 || n == 0 {
            return 0;
        }

        // Two rolling rows
        let mut prev = vec![0; n + 1];
        let mut curr = vec![0; n + 1];

        for i in 1..=m {
            for j in 1..=n {
                if s1[i - 1] == s2[j - 1] {
                    curr[j] = prev[j - 1] + 1;
                } else {
                    curr[j] = curr[j - 1].max(prev[j]);
                }
            }
            std::mem::swap(&mut prev, &mut curr);
        }

        prev[n]
    }
}
```

````python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        dp = [0] * len(text1)
        longest = 0

        for c in text2:
            cur_length = 0
            for i, val in enumerate(dp):
                if cur_length < val:
                    cur_length = val
                elif c == text1[i]:
                    dp[i] = cur_length + 1
                    longest = max(longest, cur_length + 1)
        
        return longest
````
