### Optimized Rust Solution for **Problem 215: Kth Largest Element in an Array**

## **Problem Explanation**

We are tasked with finding the `k`th largest element in an unsorted integer array `nums`. The problem asks us to do this without sorting the entire array.

### **Core DSA Concepts**

- **Quickselect Algorithm**: This is an optimal algorithm that works similar to quicksort but only partially sorts the array to find the `k`th largest (or smallest) element. It operates in average linear time, `O(n)`.
- **Partitioning**: The idea behind Quickselect is to partition the array around a pivot such that:
  - Elements greater than the pivot are on one side, and elements less than the pivot are on the other side.
  - After partitioning, we know that the pivot is in its correct position.
  - Depending on the position of the pivot, we either continue searching in the left or right partition.

---

### **Algorithm Design**

1. **Quickselect**:
   - The core of the solution is based on the Quickselect algorithm, which is a variant of Quicksort.
   - In Quickselect, we choose a pivot and partition the array such that elements greater than the pivot are on one side, and elements smaller than the pivot are on the other side.
   - If the pivot is at the `k`th position (0-indexed), then weâ€™ve found our answer.
   - Otherwise, we continue the search in the appropriate side of the array.
2. **Time Complexity**:

   - The expected time complexity of Quickselect is `O(n)`, which is more efficient than sorting the entire array with a time complexity of `O(n log n)`. In the worst case (rare), Quickselect can take `O(n^2)` time, but this can be mitigated with randomization.

3. **Space Complexity**:
   - The space complexity is `O(1)` (constant space) because we are not using extra space like in sorting algorithms. Quickselect works in-place.

### **Rust Code Implementation**

```rust
use rand::Rng;

impl Solution {
    pub fn find_kth_largest(nums: Vec<i32>, k: i32) -> i32 {
        let mut nums = nums;
        let k = k as usize;  // Convert k to a zero-based index
        Self::quickselect(&mut nums, 0, nums.len() - 1, nums.len() - k)
    }

    fn quickselect(nums: &mut [i32], left: usize, right: usize, k: usize) -> i32 {
        if left == right {
            return nums[left];
        }

        let pivot_index = Self::partition(nums, left, right);

        if k == pivot_index {
            return nums[k];
        } else if k < pivot_index {
            Self::quickselect(nums, left, pivot_index - 1, k)
        } else {
            Self::quickselect(nums, pivot_index + 1, right, k)
        }
    }

    fn partition(nums: &mut [i32], left: usize, right: usize) -> usize {
        let pivot_index = rand::thread_rng().gen_range(left..=right); // Random pivot for better performance
        nums.swap(pivot_index, right); // Move pivot to the end
        let mut i = left;

        for j in left..right {
            if nums[j] >= nums[right] { // We want descending order
                nums.swap(i, j);
                i += 1;
            }
        }
        nums.swap(i, right); // Move pivot to its correct position
        i
    }
}
```

### **Explanation of the Code**

1. **Initial Setup**:

   - The input `nums` is mutable to allow in-place modifications during Quickselect.
   - The index `k` is adjusted to be zero-based (`k - 1`), which is required for Quickselect.

2. **Quickselect Function**:

   - This function is a recursive implementation of the Quickselect algorithm. It chooses a pivot, partitions the array, and recursively searches for the `k`th largest element.
   - The base case occurs when `left == right`, meaning there is only one element left, which is the result.

3. **Partitioning**:
   - We select a random pivot to improve performance in practice, avoiding worst-case performance (which occurs when the pivot is poorly chosen).
   - The array is rearranged such that all elements greater than or equal to the pivot are on the left side, and the pivot itself is placed in the correct sorted position.
4. **Choosing the Next Search Space**:
   - If the pivot is in the correct position (equal to `k`), we return it.
   - If `k` is smaller than the pivot index, we search in the left partition; otherwise, we search in the right partition.

---

### **Time Complexity Analysis**

- **Best/Average Case**: The expected time complexity of Quickselect is `O(n)`. This is because, on average, each partition splits the array in half, and the algorithm only needs to recurse on one partition each time.
- **Worst Case**: The worst-case time complexity is `O(n^2)`, which occurs when the pivot is poorly chosen. However, this can be mitigated with random pivot selection (as done here).

### **Space Complexity**

- **Space Complexity**: The space complexity is `O(1)` because the Quickselect algorithm is done in-place with no extra space usage apart from the recursive call stack.

````python
use rand::Rng;

impl Solution {
    pub fn find_kth_largest(nums: Vec<i32>, k: i32) -> i32 {
        let mut nums = nums;
        let k = k as usize;  // Convert k to a zero-based index
        Self::quickselect(&mut nums, 0, nums.len() - 1, nums.len() - k)
    }

    fn quickselect(nums: &mut [i32], left: usize, right: usize, k: usize) -> i32 {
        if left == right {
            return nums[left];
        }
        
        let pivot_index = Self::partition(nums, left, right);
        
        if k == pivot_index {
            return nums[k];
        } else if k < pivot_index {
            Self::quickselect(nums, left, pivot_index - 1, k)
        } else {
            Self::quickselect(nums, pivot_index + 1, right, k)
        }
    }

    fn partition(nums: &mut [i32], left: usize, right: usize) -> usize {
        let pivot_index = rand::thread_rng().gen_range(left..=right); // Random pivot for better performance
        nums.swap(pivot_index, right); // Move pivot to the end
        let mut i = left;

        for j in left..right {
            if nums[j] >= nums[right] { // We want descending order
                nums.swap(i, j);
                i += 1;
            }
        }
        nums.swap(i, right); // Move pivot to its correct position
        i
    }
}
````

````python
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        k = len(nums) - k

        def quickSelect(left, right):            
            pivot = nums[right]
            low = left
            high = right

            while low <= high:
                while low <= high and nums[low] < pivot:
                    low += 1
                while low <= high and nums[high] > pivot:
                    high -= 1
                if low <= high:
                    nums[low], nums[high] = nums[high], nums[low]
                    low += 1
                    high -= 1

            if k <= high:
                return quickSelect(left, high)
            elif k >= low:
                return quickSelect(low, right)
            else:
                return nums[k]

        return quickSelect(0, len(nums) - 1)
````
