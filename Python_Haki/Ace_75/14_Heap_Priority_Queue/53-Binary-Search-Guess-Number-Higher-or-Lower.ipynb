## Problem: **374. Guess Number Higher or Lower**

### Problem Summary:

You are tasked with guessing a number chosen by someone between 1 and `n`. After each guess, an API function `guess(num)` returns:

- `-1`: Your guess is too high.
- `1`: Your guess is too low.
- `0`: Your guess is correct.

You need to implement a function `guessNumber` that finds the correct number using the provided API.

---

### Optimal Approach:

#### Algorithm:

This is a classic **binary search** problem. You can efficiently guess the number by repeatedly narrowing the search range based on the feedback from the API.

- **Binary Search Logic**:
  - Start with the full range [1, n].
  - At each step, guess the midpoint of the current range.
  - Depending on whether the feedback is -1, 1, or 0, adjust the range:
    - If feedback is `0`, you've found the number.
    - If feedback is `-1`, the picked number is lower than your guess, so adjust the upper bound.
    - If feedback is `1`, the picked number is higher than your guess, so adjust the lower bound.
- This approach reduces the search space by half at each step, achieving a **logarithmic time complexity**.

---

### Code:

```rust
/**
 * Forward declaration of guess API.
 * @param  num   your guess
 * @return 	     -1 if num is higher than the picked number
 *			      1 if num is lower than the picked number
 *               otherwise return 0
 * unsafe fn guess(num: i32) -> i32 {}
 */

impl Solution {
    pub fn guess_number(n: i32) -> i32 {
        let mut left = 1;
        let mut right = n;

        while left <= right {
            let mid = left + (right - left) / 2;
            match unsafe { guess(mid) } {
                0 => return mid,          // Correct guess
                1 => left = mid + 1,      // Guess too low, adjust left
                -1 => right = mid - 1,    // Guess too high, adjust right
                _ => unreachable!(),      // This case should not occur
            }
        }

        // The loop guarantees we will find the number, so this line is unreachable.
        unreachable!()
    }
}
```

---

### Time and Space Complexity:

- **Time Complexity**: `O(log n)`  
  The binary search approach reduces the search space by half at each step. Hence, the time complexity is logarithmic, which is optimal for problems involving searching within a range.

- **Space Complexity**: `O(1)`  
  We are only using a few variables (`left`, `right`, and `mid`), all of which are scalars. No extra data structures are needed, so the space complexity is constant.

---

### Algorithm Explanation:

- **Binary Search**: The core concept here is binary search, which is an efficient method for searching in a sorted range. By halving the search range with each guess, we ensure that we find the correct number in the least amount of time.

#### Steps:

1. Start with `left = 1` and `right = n`.
2. Repeatedly calculate the midpoint `mid = (left + right) / 2`.
3. Call the `guess(mid)` API to get feedback.
4. Adjust the `left` or `right` pointers based on the feedback:
   - If `guess(mid)` returns `0`, we've found the correct number, and we return `mid`.
   - If `guess(mid)` returns `1`, the target is greater than `mid`, so we set `left = mid + 1`.
   - If `guess(mid)` returns `-1`, the target is smaller than `mid`, so we set `right = mid - 1`.
5. The loop continues until we find the correct number.

---

### Edge Cases:

- **Minimum Bound**: If `n = 1`, we only need one guess and the number is always `1`.
- **Large `n`**: The binary search will efficiently handle even the largest possible value of `n` (up to `2^31 - 1`), thanks to its logarithmic time complexity.
- **Pick at Boundaries**: Whether the number is at the lower bound (`1`) or upper bound (`n`), binary search will efficiently converge on it.

````python
# The guess API is already defined for you.
# @param num, your guess
# @return -1 if num is higher than the picked number
#          1 if num is lower than the picked number
#          otherwise return 0
# def guess(num: int) -> int:

class Solution:
    def guessNumber(self, n: int) -> int:
        left = 1
        right = n

        while left <= right:
            mid = left + (right  - left) // 2

            num = guess(mid)
            if num > 0:
                left = mid + 1
            elif num < 0:
                right = mid - 1
            else:
                return mid
````
