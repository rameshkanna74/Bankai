Here's the most optimized, production-quality, and interview-ready Rust solution for Leetcode 2336. Smallest Number in Infinite Set:

✅ Algorithm Summary
We simulate the infinite set [1, 2, 3, ...] using a monotonic counter and manage reinsertions via:

A min-heap to store manually added back values (BinaryHeap<Reverse<i32>>)

A hash set to prevent duplicate entries into the heap

We always serve the smallest value from either the heap (if not empty) or the next number from the counter.

🧠 Time & Space Complexity
Time Complexity

pop_smallest(): O(log n) due to BinaryHeap pop.

add_back(num): O(log n) for insert.

Space Complexity

Worst case O(n) for the heap and set (when all elements are added back).

This is optimal given constraints (num <= 1000, max 1000 ops).

🦀 Idiomatic Rust Code

```rust
use std::collections::{BinaryHeap, HashSet};
use std::cmp::Reverse;

pub struct SmallestInfiniteSet {
    min_heap: BinaryHeap<Reverse<i32>>,
    seen: HashSet<i32>,
    current: i32,
}

impl SmallestInfiniteSet {
    pub fn new() -> Self {
        SmallestInfiniteSet {
            min_heap: BinaryHeap::new(),
            seen: HashSet::new(),
            current: 1,
        }
    }

    pub fn pop_smallest(&mut self) -> i32 {
        if let Some(Reverse(val)) = self.min_heap.pop() {
            self.seen.remove(&val);
            val
        } else {
            let result = self.current;
            self.current += 1;
            result
        }
    }

    pub fn add_back(&mut self, num: i32) {
        if num < self.current && self.seen.insert(num) {
            self.min_heap.push(Reverse(num));
        }
    }
}
```
