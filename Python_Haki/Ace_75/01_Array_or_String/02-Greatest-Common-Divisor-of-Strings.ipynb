## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn gcd_of_strings(str1: String, str2: String) -> String {
        if str1.clone() + &str2 != str2.clone() + &str1 {
            return "".to_string();
        }
        let gcd_len = gcd(str1.len(), str2.len());
        str1[..gcd_len].to_string()
    }

    fn gcd(a: usize, b: usize) -> usize {
        if b == 0 { a } else { Self::gcd(b, a % b) }
    }
}
```

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N + M + log(min(N, M)))**, where `N = str1.len()` and `M = str2.len()`.
- **String Concatenation Check:** `O(N + M)`.
- **GCD Calculation:** `O(log(min(N, M)))`.

### **Space Complexity:**

- **O(1)** (Extra Space)
- The only extra space used is for storing the result, which is required by the problem.

---

## **Key Optimizations**

1. **Concatenation Check for Validity:**

   - If `str1 + str2` is not equal to `str2 + str1`, then no common divisor string exists.

2. **Greatest Common Divisor (GCD) Optimization:**

   - The length of the **largest common divisor string** is the `gcd` of `str1.len()` and `str2.len()`.

3. **Substring Extraction in Constant Time:**
   - The final substring `str1[..gcd_len]` avoids unnecessary looping or allocations.
