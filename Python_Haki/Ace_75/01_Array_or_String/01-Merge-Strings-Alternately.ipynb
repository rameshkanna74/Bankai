## **Optimal Rust Solution**

```rust
impl Solution {
    pub fn merge_alternately(word1: String, word2: String) -> String {
        let mut result = String::with_capacity(word1.len() + word2.len());
        let mut iter1 = word1.chars();
        let mut iter2 = word2.chars();

        // Merge alternately
        while let (Some(c1), Some(c2)) = (iter1.next(), iter2.next()) {
            result.push(c1);
            result.push(c2);
        }

        // Append remaining characters (at most from one of the strings)
        result.extend(iter1);
        result.extend(iter2);

        result
    }
}
```

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N + M)**, where `N` is `word1.len()` and `M` is `word2.len()`.
- We iterate through both strings exactly once, leading to a linear time complexity.

### **Space Complexity:**

- **O(N + M)** (Output Space)
- The extra space used is only for the output string, which is required per the problem statement.
- No additional data structures are used apart from iterators, making this **memory-efficient**.

## **Key Optimizations**

1. **Preallocate `String` Capacity:**

   - We initialize `result` with `word1.len() + word2.len()` to avoid unnecessary reallocations.

2. **Use Iterators Efficiently:**

   - `chars()` creates an iterator that avoids indexing overhead.
   - `while let` loop ensures minimal branching.
   - `result.extend(iter1)` and `result.extend(iter2)` efficiently append remaining characters.

3. **No Unnecessary Cloning:**
   - No `clone()` or `collect()` calls, ensuring zero redundant allocations.

````rust
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        m = len(word1)
        n = len(word2)
        i = 0
        j = 0
        result = []

        while i < m or j < n:
            if i < m:
                result += word1[i]
                i += 1
            if j < n:
                result += word2[j]
                j += 1

        return "".join(result)
````
