## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn product_except_self(nums: Vec<i32>) -> Vec<i32> {
        let n = nums.len();
        let mut result = vec![1; n];

        // Compute prefix products
        let mut prefix = 1;
        for i in 0..n {
            result[i] = prefix;
            prefix *= nums[i];
        }

        // Compute suffix products and multiply with prefix
        let mut suffix = 1;
        for i in (0..n).rev() {
            result[i] *= suffix;
            suffix *= nums[i];
        }

        result
    }
}
```

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the length of `nums`.
  - **First pass (prefix products)** ‚Üí O(N)
  - **Second pass (suffix products)** ‚Üí O(N)
  - **Total Complexity**: **O(N)**
  - **No nested loops**, ensuring linear time execution.

### **Space Complexity:**

- **O(1) extra space (excluding output array)**.
  - **We only use two extra variables (`prefix`, `suffix`)**.
  - The output array (`result`) is **not considered extra space** (per problem constraints).

---

## **Key Optimizations**

1. **Avoids Division Operation**

   - This prevents **division by zero** edge cases.
   - Ensures robustness for all inputs.

2. **Two-Pass Approach (Prefix & Suffix)**

   - First pass computes **prefix products**.
   - Second pass computes **suffix products** while updating `result`.
   - Avoids using extra arrays, achieving **O(1) space complexity**.

3. **In-Place Modification of `result`**
   - Instead of separate prefix/suffix arrays, we **directly update `result`**.
   - This reduces space complexity from **O(N) ‚Üí O(1)**.

````rust
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        output = [1] * len(nums)
        
        left = 1
        for i in range(len(nums)):
            output[i] *= left
            left *= nums[i]
        
        right = 1
        for i in range(len(nums) - 1, -1, -1):
            output[i] *= right
            right *= nums[i]
    
        return output        
````
