# üß© Leetcode 452 ‚Äî Minimum Number of Arrows to Burst Balloons

## üîç Problem Summary

Given a set of balloons on a 2D plane, each represented by an interval `[xstart, xend]` (denoting the start and end of their horizontal diameter), determine the **minimum number of arrows** required to burst all the balloons. The key observation is that one arrow can burst multiple balloons if it intersects their intervals.

---

## ‚úÖ Core Idea: Greedy Algorithm

### üéØ Greedy Strategy

- **Sort the intervals by their end point**.
- Iterate through the intervals:
  - If the current interval starts after the previous arrow's position, shoot a new arrow.
  - Otherwise, continue with the current arrow, as it can burst the current balloon as well.

This guarantees that the number of arrows used is minimized, as we always try to burst as many balloons as possible with each arrow.

### üîë Why Sorting by End Point?

- Sorting by the end point ensures that the arrow is placed in the position where it can burst the most balloons by covering the right-most interval first. This strategy ensures that we minimize the number of arrows.

## ü¶Ä Idiomatic Rust Implementation

```rust
impl Solution {
    pub fn find_min_arrow_shots(mut points: Vec<Vec<i32>>) -> i32 {
        // Sort intervals by the end point of each balloon
        points.sort_unstable_by_key(|balloon| balloon[1]);

        let mut arrows = 1;
        let mut arrow_position = points[0][1];

        for i in 1..points.len() {
            // If the balloon starts after the current arrow position, shoot a new arrow
            if points[i][0] > arrow_position {
                arrows += 1;
                arrow_position = points[i][1];
            }
        }

        arrows
    }
}
```

### Explanation:

1. **Sorting**: We sort the balloons based on their end points. This allows us to greedily choose the best place to shoot each arrow.
2. **Iterating through the sorted intervals**: For each interval, check if it can be burst by the current arrow (i.e., if its start point is greater than the previous arrow's position). If it can't, shoot a new arrow at the end of the current interval.

class Solution:
    def findMinArrowShots(self, points: List[List[int]]) -> int:
        points.sort(key=lambda x: x[0])
        arrows = 1
        end = points[0][1]
        
        for balloon in points[1:]:
            if balloon[0] > end: 
                arrows += 1  
                end = balloon[1] 
            else:
                end = min(end, balloon[1])
