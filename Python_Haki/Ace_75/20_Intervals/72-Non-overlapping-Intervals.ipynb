## üîç Problem Summary

You‚Äôre given a list of intervals. The task is to **remove the minimum number of intervals** so that the remaining intervals are **non-overlapping**.

---

## ‚úÖ Core Idea: Greedy Algorithm

### üéØ Greedy Strategy

- **Sort intervals by their end time**
- Iterate through intervals and **always keep the one with the earliest end time**
- If an interval overlaps with the previous kept interval, **remove it**

This guarantees maximum non-overlapping intervals, hence minimum removals.

### üîÅ Why Sort by End Time?

This ensures that:

- We leave room for future intervals
- Greedily choosing the earliest `end` minimizes overlap risk

| üìà Scales to        | `10^5` intervals efficiently       |

## ü¶Ä Idiomatic Rust Implementation

```rust
impl Solution {
    pub fn erase_overlap_intervals(mut intervals: Vec<Vec<i32>>) -> i32 {
        // Sort by end time
        intervals.sort_unstable_by_key(|v| v[1]);

        let mut count = 0;
        let mut prev_end = i32::MIN;

        for interval in intervals {
            if interval[0] < prev_end {
                // Overlapping, need to remove it
                count += 1;
            } else {
                // Keep it
                prev_end = interval[1];
            }
        }

        count
    }
}
```

````python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        res = 0

        intervals.sort(key=lambda x: x[1])
        prev_end = intervals[0][1]

        for i in range(1, len(intervals)):
            if prev_end > intervals[i][0]:
                res += 1
            else:
                prev_end = intervals[i][1]
        
        return res
````
