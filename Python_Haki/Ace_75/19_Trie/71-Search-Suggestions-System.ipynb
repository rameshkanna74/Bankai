# üîç **Leetcode 1268 ‚Äî Search Suggestions System**

---

## ‚úÖ Overview

Given a list of `products` and a `searchWord`, return at each step **up to three lexicographically smallest products** that start with the current prefix of `searchWord`.

---

## üß† Algorithm Design

### Key Concepts:

- **Lexicographic sort** ensures smallest words come first.
- **Prefix matching** optimized with binary search (`lower_bound` using `Vec::binary_search_by`).
- This avoids constructing an explicit Trie and is more performant for ‚â§1000 items.

### Why this approach is optimal:

- We sort once: **O(n log n)**
- For each prefix, binary search the range, then linearly collect up to 3 matches: **O(m \* log n + 3m)** where `m = len(searchWord)`

---

## ‚úÖ Rust Solution (Idiomatic & Production-Ready)

```rust
impl Solution {
    pub fn suggested_products(mut products: Vec<String>, search_word: String) -> Vec<Vec<String>> {
        products.sort(); // lexicographically sort once

        let mut result = Vec::with_capacity(search_word.len());
        let mut prefix = String::new();

        for ch in search_word.chars() {
            prefix.push(ch);

            let idx = products
                .binary_search_by(|prod| prod.as_str().cmp(&prefix))
                .unwrap_or_else(|i| i);

            let suggestions = products[idx..]
                .iter()
                .take_while(|prod| prod.starts_with(&prefix))
                .take(3)
                .cloned()
                .collect();

            result.push(suggestions);
        }

        result
    }
}
```

````python
class Solution:
    def suggestedProducts(self, products: List[str], searchWord: str) -> List[List[str]]:
        products.sort()
        ans = []
        left, right = 0, len(products) - 1

        for i in range(len(searchWord)):
            c = searchWord[i]
            res = []

            while left <= right and (len(products[left]) <= i or products[left][i] < c):
                left += 1
            while left <= right and (len(products[right]) <= i or products[right][i] > c):
                right -= 1

            for j in range(3):
                if left + j <= right:
                    res.append(products[left + j])
            ans.append(res)

        return ans
````
