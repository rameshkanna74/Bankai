## üå≥ **208. Implement Trie (Prefix Tree)**

### ‚úÖ TL;DR

A **Trie** is a tree-like data structure where each node represents a **prefix** of a word. This allows efficient insert, search, and prefix-check operations on strings.

---

## üîß Implementation in Idiomatic Rust

```rust
use std::collections::HashMap;

#[derive(Default)]
struct TrieNode {
    children: HashMap<char, TrieNode>,
    is_end: bool,
}

pub struct Trie {
    root: TrieNode,
}

impl Trie {
    /// Initializes the Trie.
    pub fn new() -> Self {
        Trie {
            root: TrieNode::default(),
        }
    }

    /// Inserts a word into the trie.
    pub fn insert(&mut self, word: String) {
        let mut node = &mut self.root;
        for ch in word.chars() {
            node = node.children.entry(ch).or_insert_with(TrieNode::default);
        }
        node.is_end = true;
    }

    /// Returns true if the word is in the trie.
    pub fn search(&self, word: String) -> bool {
        match self.find_node(&word) {
            Some(node) => node.is_end,
            None => false,
        }
    }

    /// Returns true if there is any word in the trie that starts with the given prefix.
    pub fn starts_with(&self, prefix: String) -> bool {
        self.find_node(&prefix).is_some()
    }

    /// Helper: finds the node corresponding to the last char of a word/prefix
    fn find_node(&self, s: &str) -> Option<&TrieNode> {
        let mut node = &self.root;
        for ch in s.chars() {
            match node.children.get(&ch) {
                Some(next) => node = next,
                None => return None,
            }
        }
        Some(node)
    }
}
```

````python
use std::collections::HashMap;

#[derive(Default)]
struct TrieNode {
    children: HashMap<char, TrieNode>,
    is_end: bool,
}

pub struct Trie {
    root: TrieNode,
}

impl Trie {
    /// Initializes the Trie.
    pub fn new() -> Self {
        Trie {
            root: TrieNode::default(),
        }
    }

    /// Inserts a word into the trie.
    pub fn insert(&mut self, word: String) {
        let mut node = &mut self.root;
        for ch in word.chars() {
            node = node.children.entry(ch).or_insert_with(TrieNode::default);
        }
        node.is_end = true;
    }

    /// Returns true if the word is in the trie.
    pub fn search(&self, word: String) -> bool {
        match self.find_node(&word) {
            Some(node) => node.is_end,
            None => false,
        }
    }

    /// Returns true if there is any word in the trie that starts with the given prefix.
    pub fn starts_with(&self, prefix: String) -> bool {
        self.find_node(&prefix).is_some()
    }

    /// Helper: finds the node corresponding to the last char of a word/prefix
    fn find_node(&self, s: &str) -> Option<&TrieNode> {
        let mut node = &self.root;
        for ch in s.chars() {
            match node.children.get(&ch) {
                Some(next) => node = next,
                None => return None,
            }
        }
        Some(node)
    }
}
````

````python
class Trie:

    def __init__(self):
        self.root={}
        
    def insert(self, word: str) -> None:

        cur=self.root

        for letter in word:
            if letter not in cur:
                cur[letter]={}
            cur=cur[letter]

        cur['*']=''

    def search(self, word: str) -> bool:

        cur=self.root
        for letter in word:
            if letter not in cur:
                return False
            cur=cur[letter]

        return '*' in cur
        
    def startsWith(self, prefix: str) -> bool:

        cur=self.root
        for letter in prefix:
            if letter not in cur:
                return False
            cur=cur[letter]

        return True
````
