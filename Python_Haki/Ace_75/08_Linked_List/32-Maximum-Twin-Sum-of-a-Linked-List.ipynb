In a linked list of size n, where n is even, the ith node (0-indexed) of the linked list is known as the twin of the (n-1-i)th node, if 0 <= i <= (n / 2) - 1.

    For example, if n = 4, then node 0 is the twin of node 3, and node 1 is the twin of node 2. These are the only nodes with twins for n = 4.

The twin sum is defined as the sum of a node and its twin.

Given the head of a linked list with even length, return the maximum twin sum of the linked list.

Example 1:

Input: head = [5,4,2,1]
Output: 6
Explanation:
Nodes 0 and 1 are the twins of nodes 3 and 2, respectively. All have twin sum = 6.
There are no other nodes with twins in the linked list.
Thus, the maximum twin sum of the linked list is 6.

Example 2:

Input: head = [4,2,2,3]
Output: 7
Explanation:
The nodes with twins present in this linked list are:

- Node 0 is the twin of node 3 having a twin sum of 4 + 3 = 7.
- Node 1 is the twin of node 2 having a twin sum of 2 + 2 = 4.
  Thus, the maximum twin sum of the linked list is max(7, 4) = 7.

Example 3:

Input: head = [1,100000]
Output: 100001
Explanation:
There is only one node with a twin in the linked list having twin sum of 1 + 100000 = 100001.

Constraints:

    The number of nodes in the list is an even integer in the range [2, 105].
    1 <= Node.val <= 105

### ‚úÖ Algorithm Summary

We solve the problem using the **two-pointer technique** and **in-place reversal**:

1. **Find the middle** using slow & fast pointers.
2. **Reverse the second half** of the list.
3. **Iterate from both halves** simultaneously to calculate twin sums.
4. **Track the maximum** twin sum encountered.

---

### ‚è±Ô∏è Time and Space Complexity

- **Time:** `O(n)`  
  - `n/2` steps to find middle
  - `n/2` steps to reverse second half
  - `n/2` steps to compute twin sums
- **Space:** `O(1)` extra space  
  - We reverse the list **in-place** without additional structures.

### üß† Core Concepts

- **Two-pointer technique** to find the midpoint
- **Linked list reversal**
- **Simultaneous traversal** to compute twin sums

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    pub fn pair_sum(head: Option<Box<ListNode>>) -> i32 {
        // Step 1: Find middle using slow and fast pointers
        let mut slow = &head;
        let mut fast = &head;
        while fast.is_some() && fast.as_ref().unwrap().next.is_some() {
            slow = &slow.as_ref().unwrap().next;
            fast = &fast.as_ref().unwrap().next.as_ref().unwrap().next;
        }

        // Step 2: Reverse second half starting from 'slow'
        let reversed = Self::reverse(slow.clone());

        // Step 3: Compute twin sum
        let mut max_sum = 0;
        let mut first = &head;
        let mut second = &reversed;
        while let (Some(f), Some(s)) = (first, second) {
            max_sum = max_sum.max(f.val + s.val);
            first = &f.next;
            second = &s.next;
        }

        max_sum
    }

    // Helper to reverse linked list
    fn reverse(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev = None;
        while let Some(mut current) = head {
            head = current.next.take(); // Advance head, detach current
            current.next = prev;        // Reverse pointer
            prev = Some(current);       // Move prev forward
        }
        prev
    }
}
```

````python
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    fn new(val: i32) -> Self {
        ListNode { next: None, val }
    }
}

pub struct Solution;

impl Solution {
    pub fn pair_sum(head: Option<Box<ListNode>>) -> i32 {
        // Step 1: Find middle using slow and fast pointers
        let mut slow = &head;
        let mut fast = &head;
        while fast.is_some() && fast.as_ref().unwrap().next.is_some() {
            slow = &slow.as_ref().unwrap().next;
            fast = &fast.as_ref().unwrap().next.as_ref().unwrap().next;
        }

        // Step 2: Reverse second half starting from 'slow'
        let reversed = Self::reverse(slow.clone());

        // Step 3: Compute twin sum
        let mut max_sum = 0;
        let mut first = &head;
        let mut second = &reversed;
        while let (Some(f), Some(s)) = (first, second) {
            max_sum = max_sum.max(f.val + s.val);
            first = &f.next;
            second = &s.next;
        }

        max_sum
    }

    // Helper to reverse linked list
    fn reverse(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev = None;
        while let Some(mut current) = head {
            head = current.next.take(); // Advance head, detach current
            current.next = prev;        // Reverse pointer
            prev = Some(current);       // Move prev forward
        }
        prev
    }
}
````
