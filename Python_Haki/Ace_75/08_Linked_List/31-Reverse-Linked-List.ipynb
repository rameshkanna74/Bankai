## üí° Algorithm Overview

Reversing a singly linked list involves **flipping the direction of `next` pointers** for each node. This can be done:

1. **Iteratively** ‚Äì using a loop with `prev`, `curr`, and `next` pointers.
2. **Recursively** ‚Äì by reversing the rest of the list and adjusting `next`.

---

## üß† Time & Space Complexity

| Metric | Iterative | Recursive              |
| ------ | --------- | ---------------------- |
| Time   | O(n)      | O(n)                   |
| Space  | O(1)      | O(n) due to call stack |

- `n` is the number of nodes in the linked list.
- **Iterative approach** is more space-efficient.

## ü¶Ä Optimized Rust Code ‚Äì Iterative

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    /// Reverses a singly linked list using an iterative approach.
    /// Time: O(n), Space: O(1)
    pub fn reverse_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev_node: Option<Box<ListNode>> = None;
        let mut current_node = head;

        while let Some(mut boxed_node) = current_node {
            let next_node = boxed_node.next.take(); // Save next
            boxed_node.next = prev_node;           // Reverse pointer
            prev_node = Some(boxed_node);          // Move prev forward
            current_node = next_node;              // Move curr forward
        }

        prev_node
    }

    /// Reverses a singly linked list using recursion.
    /// Time: O(n), Space: O(n) due to call stack
    pub fn reverse_list_recursive(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        fn helper(
            node: Option<Box<ListNode>>,
            prev: Option<Box<ListNode>>,
        ) -> Option<Box<ListNode>> {
            match node {
                Some(mut boxed_node) => {
                    let next = boxed_node.next.take();
                    boxed_node.next = prev;
                    helper(next, Some(boxed_node))
                }
                None => prev,
            }
        }

        helper(head, None)
    }
}
```

````python
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

pub struct Solution;

impl Solution {
    /// Reverses a singly linked list using an iterative approach.
    /// Time: O(n), Space: O(1)
    pub fn reverse_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        let mut prev_node: Option<Box<ListNode>> = None;
        let mut current_node = head;

        while let Some(mut boxed_node) = current_node {
            let next_node = boxed_node.next.take(); // Save next
            boxed_node.next = prev_node;           // Reverse pointer
            prev_node = Some(boxed_node);          // Move prev forward
            current_node = next_node;              // Move curr forward
        }

        prev_node
    }

    /// Reverses a singly linked list using recursion.
    /// Time: O(n), Space: O(n) due to call stack
    pub fn reverse_list_recursive(head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        fn helper(
            node: Option<Box<ListNode>>,
            prev: Option<Box<ListNode>>,
        ) -> Option<Box<ListNode>> {
            match node {
                Some(mut boxed_node) => {
                    let next = boxed_node.next.take();
                    boxed_node.next = prev;
                    helper(next, Some(boxed_node))
                }
                None => prev,
            }
        }

        helper(head, None)
    }
}
````

````python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head

        new_head = self.reverseList(head.next)
        head.next.next = head
        head.next = None
        return new_head
````
