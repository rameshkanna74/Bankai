## ‚úÖ Problem Recap

You are given a singly linked list. Rearrange it so that **all nodes at odd indices come first**, followed by **all nodes at even indices** (1-based indexing). Maintain relative order **within** the odd and even groups.

---

## üí° Core Idea

### üß† In-Place Partitioning Using Two Pointers

- Use two pointers: `odd` and `even`, each pointing to their respective start.
- `odd` connects all odd-indexed nodes.
- `even` connects all even-indexed nodes.
- At the end, link `odd`'s last node to the start of the even list.

### üîÅ One-Pass Traversal

- We traverse the list **once**, rearranging pointers in place.
- Maintain **O(1)** space and **O(n)** time complexity.

---

## üï∞ Time and Space Complexity

| Metric | Value | Justification                           |
| ------ | ----- | --------------------------------------- |
| Time   | O(n)  | Traverse each node once                 |
| Space  | O(1)  | In-place rearrangement, no extra memory |

## ü¶Ä Rust Solution (Production-Grade)

```rust
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

impl Solution {
    pub fn odd_even_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        // Edge case: 0 or 1 elements
        if head.is_none() || head.as_ref().unwrap().next.is_none() {
            return head;
        }

        let mut odd = head.as_mut().unwrap();
        let mut even_head = odd.next.take();
        let mut even = even_head.as_mut().unwrap();

        while let Some(mut next_odd) = even.next.take() {
            // Link next odd
            odd.next = Some(next_odd);
            odd = odd.next.as_mut().unwrap();

            // Link next even
            even.next = odd.next.take();
            if let Some(ref mut next_even) = even.next {
                even = next_even;
            } else {
                break;
            }
        }

        // Reconnect even list to the end of odd list
        odd.next = Some(even_head);
        head
    }
}
```

````python
// Definition for singly-linked list.
#[derive(PartialEq, Eq, Clone, Debug)]
pub struct ListNode {
    pub val: i32,
    pub next: Option<Box<ListNode>>,
}

impl ListNode {
    #[inline]
    pub fn new(val: i32) -> Self {
        ListNode { val, next: None }
    }
}

impl Solution {
    pub fn odd_even_list(mut head: Option<Box<ListNode>>) -> Option<Box<ListNode>> {
        // Edge case: 0 or 1 elements
        if head.is_none() || head.as_ref().unwrap().next.is_none() {
            return head;
        }

        let mut odd = head.as_mut().unwrap();
        let mut even_head = odd.next.take();
        let mut even = even_head.as_mut().unwrap();

        while let Some(mut next_odd) = even.next.take() {
            // Link next odd
            odd.next = Some(next_odd);
            odd = odd.next.as_mut().unwrap();

            // Link next even
            even.next = odd.next.take();
            if let Some(ref mut next_even) = even.next {
                even = next_even;
            } else {
                break;
            }
        }

        // Reconnect even list to the end of odd list
        odd.next = Some(even_head);
        head
    }
}
````

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return None
        if not head.next:
            return head
        
        odd_head, even_head = head, head.next
        odd_tail, even_tail = odd_head, even_head

        while odd_tail.next and even_tail.next:
            odd_tail.next = even_tail.next if even_tail else None
            odd_tail = odd_tail.next

            even_tail.next = odd_tail.next if odd_tail else None
            even_tail = even_tail.next
        odd_tail.next = even_head

        return head
````
