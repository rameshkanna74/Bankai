### ‚úÖ Algorithm Explanation

**Core Idea**:  
Use a **stack** to simulate asteroid collisions.

- A **positive number** represents an asteroid moving **right**.
- A **negative number** represents an asteroid moving **left**.
- A **collision** occurs only when a left-moving asteroid (`< 0`) meets a right-moving asteroid (`> 0`).

As we iterate:

- Push all right-moving asteroids directly onto the stack.
- When we encounter a left-moving asteroid:
  - **Pop and destroy** smaller right-moving asteroids.
  - If equal, both explode.
  - If the left asteroid is smaller, it gets destroyed.
  - If it survives all collisions, it‚Äôs pushed to the stack.

---

### üöÄ Time and Space Complexity

- **Time Complexity**: `O(n)`  
  Each asteroid is pushed and popped at most once.
- **Space Complexity**: `O(n)`  
  In the worst case (no collisions), all asteroids are stored.

---

### üß† Edge Cases Handled

- All right- or left-moving asteroids (no collision)
- Multiple consecutive collisions
- Equal-sized collisions
- Alternating directions

```rust
impl Solution {
    pub fn asteroid_collision(asteroids: Vec<i32>) -> Vec<i32> {
        let mut stack: Vec<i32> = Vec::with_capacity(asteroids.len());

        for asteroid in asteroids {
            let mut alive = true;

            while let Some(&top) = stack.last() {
                // Only potential collision if top is right and current is left
                if top > 0 && asteroid < 0 {
                    match top.cmp(&-asteroid) {
                        std::cmp::Ordering::Less => {
                            stack.pop(); // top explodes
                            continue;    // check next top
                        }
                        std::cmp::Ordering::Equal => {
                            stack.pop(); // both explode
                            alive = false;
                            break;
                        }
                        std::cmp::Ordering::Greater => {
                            alive = false; // current asteroid explodes
                            break;
                        }
                    }
                } else {
                    break; // no collision
                }
            }

            if alive {
                stack.push(asteroid);
            }
        }

        stack
    }
}
```

````rust
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        
        res = []

        for a in asteroids:

            while res and a < 0 < res[-1]:
                if -a > res[-1]:
                    res.pop()
                    continue
                elif -a == res[-1]:
                    res.pop()
                break
            else:
                res.append(a)

        return res
````
