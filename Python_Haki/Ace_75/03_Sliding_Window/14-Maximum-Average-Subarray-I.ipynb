## **Optimized Rust Solution**

```rust
impl Solution {
    pub fn find_max_average(nums: Vec<i32>, k: i32) -> f64 {
        let k = k as usize; // Convert k to usize for indexing
        let mut sum: i32 = nums[..k].iter().sum(); // Initial sum of first k elements
        let mut max_sum = sum;

        for i in k..nums.len() {
            sum += nums[i] - nums[i - k]; // Sliding window: add new element, remove old
            max_sum = max_sum.max(sum);
        }

        max_sum as f64 / k as f64
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the number of elements in `nums`.
- We iterate through `nums` **once** using a sliding window, making only **constant-time updates** per iteration.

### **Space Complexity:**

- **O(1)** since we only use a few integer variables (`sum`, `max_sum`, `k`).

---

## **Key Optimizations**

1. **Sliding Window for O(N) Complexity**

   - Compute the sum of the first `k` elements.
   - Slide the window by adding one element and removing one element in **O(1) time**.
   - Maintain a running maximum of the sum.

2. **Avoiding Unnecessary Floating-Point Operations**
   - Work with **integers** (`i32`) until the final division.
   - Convert the result to `f64` only at the **end** to avoid floating-point precision errors.

````rust
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        
        # Initialize curr_sum and max_sum to the sum of the initial k elements
        curr_sum = max_sum = sum(nums[:k])

        # Start the loop from the kth element 
        # Iterate until you reach the end
        for i in range(k, len(nums)):

            # Subtract the left element of the window
            # Add the right element of the window
            curr_sum += nums[i] - nums[i - k]
            
            # Update the max
            max_sum = max(max_sum, curr_sum)

        # Since the problem requires average, we return the average
        return max_sum / k
````
