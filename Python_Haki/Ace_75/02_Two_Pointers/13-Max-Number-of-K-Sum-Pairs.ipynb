## **Optimized Rust Solution**

```rust
use std::collections::HashMap;

impl Solution {
    pub fn max_operations(nums: Vec<i32>, k: i32) -> i32 {
        let mut freq = HashMap::new();
        let mut count = 0;

        for &num in &nums {
            let complement = k - num;
            if let Some(&freq_count) = freq.get(&complement) {
                if freq_count > 0 {
                    *freq.get_mut(&complement).unwrap() -= 1;
                    count += 1;
                    continue;
                }
            }
            *freq.entry(num).or_insert(0) += 1;
        }

        count
    }
}
```

---

## **Algorithm Analysis**

### **Time Complexity:**

- **O(N)**, where `N` is the number of elements in `nums`.
- We iterate through `nums` **once**, using `HashMap` operations (`insert`, `get_mut`), which have **average O(1) complexity**.

### **Space Complexity:**

- **O(N)** in the worst case.
  - If there are no valid pairs, the `HashMap` will store **all `N` elements**.
  - If the array contains many valid pairs, space usage is **less than O(N)**.

---

## **Key Optimizations**

1. **Using HashMap for Frequency Count (O(1) Lookup)**

   - Store the **count of numbers** in a `HashMap`.
   - For each number `num`, check if `k - num` already exists.
   - If found, decrement the count and increase the operation counter.

2. **Avoid Sorting (O(N log N) ‚Üí O(N))**

   - **Alternative approach (two-pointer after sorting) is O(N log N) due to sorting**.
   - **HashMap approach is O(N), making it optimal**.

3. **Efficient HashMap Operations**
   - Use `.entry(num).or_insert(0) += 1` to **minimize duplicate insertions**.
   - Use `get_mut(&complement).unwrap()` to **efficiently update values in-place**.

## **Edge Cases Considered**

1. **No Valid Pairs**

   - Example: `nums = [1, 2, 3], k = 10`
   - ‚úÖ Returns `0`.

2. **All Elements Form Pairs**

   - Example: `nums = [1, 4, 2, 3], k = 5`

3. **Duplicates Without Enough Pairs**

````rust
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        nums.sort()

        left = 0 
        right = len(nums) - 1
        operation = 0 

        while left < right:
            if ((nums[left] + nums[right]) == k):
                operation += 1
                left +=1 
                right -=1
            elif((nums[left] + nums[right]) < k):
                left += 1
            else:
                right -= 1
        return operation
````
