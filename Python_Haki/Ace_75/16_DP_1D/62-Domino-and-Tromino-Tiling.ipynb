## ðŸ’¡ Key Insight (Dynamic Programming)

Let:

- `dp[n]` = Number of ways to fully tile a `2 x n` board.

We define the recurrence:

```
dp[n] = dp[n-1] + dp[n-2] + 2 * sum(dp[0..n-3])
```

### Simplified Recurrence:

Instead of computing the sum every time, maintain a running sum called `sum` such that:

```
dp[i] = 2 * sum + dp[i-1] + dp[i-2]
sum += dp[i-3]
```

```rust
impl Solution {
    pub fn num_tilings(n: i32) -> i32 {
        const MOD: i64 = 1_000_000_007;
        let n = n as usize;

        if n == 1 {
            return 1;
        }
        if n == 2 {
            return 2;
        }
        if n == 3 {
            return 5;
        }

        let mut dp = vec![0i64; n + 1];
        dp[0] = 1;
        dp[1] = 1;
        dp[2] = 2;
        dp[3] = 5;

        let mut sum = dp[0] + dp[1]; // sum of dp[0..=1] used in dp[3]

        for i in 4..=n {
            dp[i] = (2 * sum + dp[i - 1] + dp[i - 2]) % MOD;
            sum = (sum + dp[i - 3]) % MOD;
        }

        dp[n] as i32
    }
}
```

- **DP State**: `dp[i]` = number of ways to tile a `2 x i` board.

  - `2 * sum(dp[0..i-3])`: handle tromino transitions (L-shapes can create asymmetry).

````python
impl Solution {
    pub fn num_tilings(n: i32) -> i32 {
        const MOD: i64 = 1_000_000_007;
        let n = n as usize;

        if n == 1 {
            return 1;
        }
        if n == 2 {
            return 2;
        }
        if n == 3 {
            return 5;
        }

        let mut dp = vec![0i64; n + 1];
        dp[0] = 1;
        dp[1] = 1;
        dp[2] = 2;
        dp[3] = 5;

        let mut sum = dp[0] + dp[1]; // sum of dp[0..=1] used in dp[3]

        for i in 4..=n {
            dp[i] = (2 * sum + dp[i - 1] + dp[i - 2]) % MOD;
            sum = (sum + dp[i - 3]) % MOD;
        }

        dp[n] as i32
    }
}
````

````python
class Solution:
    def numTilings(self, n: int) -> int:
        # dynamic programming 
        # note: F = Full, T = TopMissing, B = BottomMissing

        F = {0: 1, 1: 1}
        T = {1: 0}
        B = {1: 0}

        for i in range(2, n + 1): 
            F[i] = F[i - 1] + F[i - 2] + T[i - 1] + B[i - 1]
            T[i] = F[i - 2] + B[i - 1]
            B[i] = F[i - 2] + T[i - 1]

        return F[n] % (10 ** 9 + 7)
````
