## âœ… Optimal Rust Solution

```rust
impl Solution {
    pub fn tribonacci(n: i32) -> i32 {
        match n {
            0 => 0,
            1 | 2 => 1,
            _ => {
                let (mut a, mut b, mut c) = (0, 1, 1);
                for _ in 3..=n {
                    let next = a + b + c;
                    a = b;
                    b = c;
                    c = next;
                }
                c
            }
        }
    }
}
```

````python
class Solution:
    def tribonacci(self, n: int) -> int:
        if n == 0:
            return 0
        elif n == 1 or n == 2:
            return 1
        else:
            return self.tribonacci(n - 1) + self.tribonacci(n - 2) + self.tribonacci(n - 3)
````
