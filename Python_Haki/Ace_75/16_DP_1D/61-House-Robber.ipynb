# üè† Leetcode 198: House Robber

---

## ‚úÖ Solution Summary

This problem is a classic **Dynamic Programming** scenario, where each decision (to rob or not rob a house) depends on prior decisions. We use a **bottom-up tabulation** strategy with **space optimization** to `O(1)`.

---

## üí° Key Idea

Let `dp[i]` represent the **maximum money** that can be robbed from the first `i+1` houses:

- **Rob house i**: `dp[i-2] + nums[i]`
- **Skip house i**: `dp[i-1]`

So,  
`dp[i] = max(dp[i-1], dp[i-2] + nums[i])`

We only need to keep track of the last two values, not the entire `dp` array.

---

## ü¶Ä Final Rust Implementation

```rust
impl Solution {
    pub fn rob(nums: Vec<i32>) -> i32 {
        let n = nums.len();
        if n == 0 {
            return 0;
        }
        if n == 1 {
            return nums[0];
        }

        let (mut prev2, mut prev1) = (nums[0], nums[0].max(nums[1]));

        for i in 2..n {
            let current = prev2 + nums[i];
            let new_prev1 = prev1.max(current);
            prev2 = prev1;
            prev1 = new_prev1;
        }

        prev1
    }
}
```

### ‚è± Time Complexity: `O(n)`

### üß† Space Complexity: `O(1)`

## üß† Algorithm Design

### DSA Concepts:

- **Dynamic Programming**
- **Greedy Optimization** (each step makes the optimal local choice)
- **Sliding Window** for `dp` (tracking only 2 states)

### Trade-offs:

- We don't store the full DP state (cannot reconstruct path) ‚Äî which is fine since only the max value is needed.

````python
impl Solution {
    pub fn rob(nums: Vec<i32>) -> i32 {
        let n = nums.len();
        if n == 0 {
            return 0;
        }
        if n == 1 {
            return nums[0];
        }

        let (mut prev2, mut prev1) = (nums[0], nums[0].max(nums[1]));

        for i in 2..n {
            let current = prev2 + nums[i];
            let new_prev1 = prev1.max(current);
            prev2 = prev1;
            prev1 = new_prev1;
        }

        prev1
    }
}
````

````python
class Solution:
    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        
        if n == 1:
            return nums[0]
        
        dp = [0] * n
        
        dp[0] = nums[0]
        dp[1] = max(nums[0], nums[1])
        
        for i in range(2, n):
            dp[i] = max(dp[i-1], nums[i] + dp[i-2])
        
        return dp[-1] 
````
