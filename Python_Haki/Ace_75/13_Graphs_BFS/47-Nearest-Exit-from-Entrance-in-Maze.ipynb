### Optimized Rust Solution for **Problem 1926: Nearest Exit from Entrance in Maze**

## **Problem Explanation**

We are given a maze represented by a 2D grid where empty cells are denoted by `'.'` and walls are denoted by `'+'`. We start at a given entrance point, and we need to determine the shortest path to the nearest exit. An exit is any empty cell at the border of the maze. The entrance itself cannot be counted as an exit.

The goal is to return the number of steps in the shortest path from the entrance to the nearest exit, or `-1` if no such path exists.

### **Core DSA Concepts**

- **Graph Representation**: The maze can be considered as an unweighted graph where each cell is a node, and edges exist between adjacent empty cells.
- **Breadth-First Search (BFS)**: BFS is ideal for finding the shortest path in an unweighted grid or graph. By exploring the grid layer by layer, BFS guarantees that we visit nodes in increasing distance from the source.

### **Algorithm Design**

We will perform a **Breadth-First Search (BFS)** starting from the entrance. BFS will explore the neighboring cells level by level, ensuring that we find the nearest exit first.

1. **BFS Approach**:
   - Start BFS from the entrance.
   - For each cell, attempt to move to its 4 adjacent cells (up, down, left, right).
   - If the adjacent cell is within the maze and not a wall, continue the BFS.
   - As soon as an exit is encountered (a border cell that is empty and not the entrance), return the distance.
2. **Edge Case Handling**:
   - If the entrance is surrounded by walls or there are no exits, return `-1`.
   - If the entrance itself is the only cell in the maze, return `-1` because it is not an exit.
3. **Termination Condition**:
   - BFS stops when we either find an exit or exhaust all possible paths (in which case, we return `-1`).

### **Rust Code**

```rust
use std::collections::VecDeque;

impl Solution {
    pub fn nearest_exit(maze: Vec<Vec<char>>, entrance: Vec<i32>) -> i32 {
        let rows = maze.len();
        let cols = maze[0].len();

        // Directions: up, down, left, right
        let directions = vec![(0, 1), (1, 0), (0, -1), (-1, 0)];

        // BFS initialization
        let mut queue = VecDeque::new();
        queue.push_back((entrance[0] as usize, entrance[1] as usize, 0)); // (row, col, distance)

        // Mark the entrance as visited
        let mut maze = maze;
        maze[entrance[0] as usize][entrance[1] as usize] = '+';

        while let Some((r, c, dist)) = queue.pop_front() {
            // Check if the current cell is an exit
            if (r == 0 || r == rows - 1 || c == 0 || c == cols - 1) && maze[r][c] == '.' {
                return dist;
            }

            // Explore neighbors
            for &(dr, dc) in &directions {
                let new_r = r as i32 + dr;
                let new_c = c as i32 + dc;

                if new_r >= 0 && new_r < rows as i32 && new_c >= 0 && new_c < cols as i32 {
                    let new_r = new_r as usize;
                    let new_c = new_c as usize;

                    if maze[new_r][new_c] == '.' {
                        maze[new_r][new_c] = '+';
                        queue.push_back((new_r, new_c, dist + 1));
                    }
                }
            }
        }

        // No exit found
        -1
    }
}
```

### **Explanation of the Code**

1. **BFS Initialization**:

   - We use a queue (`VecDeque`) to store cells that need to be processed. Each entry contains the current row, column, and the distance from the entrance.
   - The entrance cell is marked as visited by changing it to `'+'` so that we don't visit it again.

2. **Main BFS Loop**:

   - For each cell in the queue, we check if it's an exit (i.e., it's an empty cell on the border of the maze).
   - If it's an exit, we return the current distance.
   - If it's not an exit, we explore its 4 neighboring cells (up, down, left, right). If a neighboring cell is valid (within bounds and not a wall), it is added to the queue and marked as visited.

3. **Edge Case Handling**:
   - The entrance itself is not considered an exit, so we only return the distance once we encounter a valid exit cell.
   - If no path leads to an exit, we return `-1`.

### **Time Complexity Analysis**

- **Time Complexity**:
  - **BFS** processes each cell at most once. Since the maze has `m * n` cells, the time complexity is `O(m * n)` where `m` is the number of rows and `n` is the number of columns.
- **Space Complexity**:
  - The space complexity is `O(m * n)` because we store each cell in the maze in the queue and the visited state is modified in-place.
