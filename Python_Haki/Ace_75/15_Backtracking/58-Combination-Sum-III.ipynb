### 🔍 Problem Overview

You must find **all unique combinations of `k` numbers** (from 1 to 9) that **add up to `n`**, with **no repeated numbers** in each combination.

---

## ✅ Rust Implementation

```rust
impl Solution {
    pub fn combination_sum3(k: i32, n: i32) -> Vec<Vec<i32>> {
        fn backtrack(
            k: i32,
            n: i32,
            start: i32,
            path: &mut Vec<i32>,
            result: &mut Vec<Vec<i32>>,
        ) {
            if k == 0 && n == 0 {
                result.push(path.clone());
                return;
            }
            if k == 0 || n <= 0 {
                return;
            }

            for i in start..=9 {
                path.push(i);
                backtrack(k - 1, n - i, i + 1, path, result);
                path.pop(); // backtrack
            }
        }

        let mut result = Vec::new();
        let mut path = Vec::with_capacity(k as usize);
        backtrack(k, n, 1, &mut path, &mut result);
        result
    }
}
```

### 🧠 Space Complexity: `O(k)`

## 🧠 Algorithm Explanation

### Core DSA Concept:

- **Backtracking / DFS**  
  Recursively try numbers 1 through 9, ensuring:
  - Only unused numbers are selected.
  - Sum is tracked.
  - Path is valid only if:
    - `k == 0` (correct count),
    - `n == 0` (exact sum).

### Why it’s optimal:

- **Prunes early** if `k == 0` or `n < 0` (no extra work).
- **No duplicates** since we increment start index in each recursive call.

### Trade-Offs:

- DFS is ideal here due to small input size and combination constraints (no need for dynamic programming).

````python
def rec(n, k, i, sum, ds, ans):
    if len(ds) == k and sum == n:
        ans.append(ds[:])  # Copy list to avoid reference issues
        return
    if sum > n or i >= 10:
        return

    ds.append(i)
    rec(n, k, i + 1, sum + i, ds, ans)
    ds.pop()
    rec(n, k, i + 1, sum, ds, ans)

class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        ans = []
        ds = []
        i = 1
        sum = 0
        rec(n, k, i, sum, ds, ans)
        return ans
        
````
