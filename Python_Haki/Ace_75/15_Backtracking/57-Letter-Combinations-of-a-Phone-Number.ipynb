## ðŸ“ž Problem 17. Letter Combinations of a Phone Number

---

### ðŸ” Problem Summary

Given a string of digits `digits` (ranging from `'2'` to `'9'`), return **all possible letter combinations** the digits could represent using the classic phone keypad layout.

---

### âŒ› Time and Space Complexity

- **Time Complexity**: `O(3^n * 4^m)`  
  - Where `n` is the count of digits that map to 3 letters (`2`, `3`, `4`, `5`, `6`, `8`)  
  - And `m` is the count of digits that map to 4 letters (`7`, `9`)
  - Max combinations: `4^4 = 256` (small)
- **Space Complexity**: `O(3^n * 4^m)` for output and recursion stack

---

### ðŸ§  Algorithm: Backtracking (DFS)

- Use a **recursive DFS** (backtracking) approach to build combinations character-by-character.
- Terminate recursion when the current combination's length equals input length.
- **Edge case**: Return empty vector if input is empty.

```rust
impl Solution {
    pub fn letter_combinations(digits: String) -> Vec<String> {
        if digits.is_empty() {
            return vec![];
        }

        // Mapping from digits to characters
        let phone_map = [
            "",     // 0
            "",     // 1
            "abc",  // 2
            "def",  // 3
            "ghi",  // 4
            "jkl",  // 5
            "mno",  // 6
            "pqrs", // 7
            "tuv",  // 8
            "wxyz", // 9
        ];

        let mut result = Vec::new();
        let mut path = String::with_capacity(digits.len());

        fn backtrack(
            index: usize,
            digits: &[u8],
            phone_map: &[&str],
            path: &mut String,
            result: &mut Vec<String>,
        ) {
            if index == digits.len() {
                result.push(path.clone());
                return;
            }

            let digit = (digits[index] - b'0') as usize;
            for ch in phone_map[digit].chars() {
                path.push(ch);
                backtrack(index + 1, digits, phone_map, path, result);
                path.pop(); // backtrack
            }
        }

        backtrack(0, digits.as_bytes(), &phone_map, &mut path, &mut result);
        result
    }
}
```

````python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        
        digit_to_letters = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz',
        }

        def backtrack(idx, comb):
            if idx == len(digits):
                res.append(comb[:])
                return
            
            for letter in digit_to_letters[digits[idx]]:
                backtrack(idx + 1, comb + letter)

        res = []
        backtrack(0, "")

        return res
````
