### Problem Explanation

The goal is to design an algorithm that collects daily stock prices and computes the span of each stock's price. The span of the price on a given day is defined as the number of consecutive days (starting from the current day and going backwards) where the price is less than or equal to the current day's price.

---

## Approach

### Key Insights:
1. **Stack Data Structure**: The problem can be solved efficiently using a stack to keep track of the indices of the stock prices. The stack helps us efficiently find the span of each day's price by maintaining a decreasing order of stock prices from the most recent to the past prices.
2. **Greedy Strategy**: For each new price, we pop from the stack until we find a price that is greater than or equal to the current price, at which point the span is the difference between the current index and the index of the price found in the stack.

### Algorithm:
1. For each price, if the stack is empty or the current price is less than the price at the top of the stack, we push the current index onto the stack and return a span of 1.
2. If the current price is greater than the price at the top of the stack, we pop the stack, counting how many previous prices are less than or equal to the current price. The span is the difference between the current day and the last popped day's index.

### Time and Space Complexity:
- **Time Complexity**: **O(1)** per `next()` call on average, making the total complexity **O(n)** for `n` calls to `next()`. This is because each index is pushed and popped from the stack at most once.
- **Space Complexity**: **O(n)**, where `n` is the number of calls to `next()`, as we store the indices of the prices in the stack.

### Rust Code Implementation

```rust
struct StockSpanner {
    prices: Vec<i32>,  // stores the stock prices
    stack: Vec<(i32, i32)>,  // stores pairs of (price, span)
}

impl StockSpanner {

    // Creates a new instance of StockSpanner
    fn new() -> Self {
        StockSpanner {
            prices: Vec::new(),
            stack: Vec::new(),
        }
    }
    
    // Adds a new price and calculates its span
    fn next(&mut self, price: i32) -> i32 {
        let mut span = 1;  // The default span is 1 (the price itself)
        
        // While the stack is not empty and the price at the top of the stack is less than or equal to the current price
        while let Some(&(prev_price, prev_span)) = self.stack.last() {
            if prev_price <= price {
                // Pop the stack and add the span of the popped price
                self.stack.pop();
                span += prev_span;
            } else {
                break;
            }
        }
        
        // Push the current price and its calculated span onto the stack
        self.stack.push((price, span));
        
        // Return the span for the current price
        span
    }
}

fn main() {
    // Example usage
    let mut stock_spanner = StockSpanner::new();
    println!("{}", stock_spanner.next(100)); // 1
    println!("{}", stock_spanner.next(80));  // 1
    println!("{}", stock_spanner.next(60));  // 1
    println!("{}", stock_spanner.next(70));  // 2
    println!("{}", stock_spanner.next(60));  // 1
    println!("{}", stock_spanner.next(75));  // 4
    println!("{}", stock_spanner.next(85));  // 6
}
```

---

### Key Points:
- **State Maintenance with Stack**: The `stack` holds pairs of `(price, span)`. This allows us to easily compute the span of any new price by popping the stack until we find a price greater than the current one.
- **Efficient Span Calculation**: By leveraging the stack, we avoid a brute force solution where we would have to scan the entire history of prices for each new price.

### 🧠 Algorithm Explanation

### **Core DSA Concepts**:
- **Stack**: A stack is used to store the prices and their corresponding spans. By maintaining the stack with the prices in decreasing order, we can efficiently compute the span by popping the stack when a lower price is encountered.

### **Design**:
- We iterate through each new price. For each price, we calculate the span by popping prices from the stack that are less than or equal to the current price. The span is the sum of the span of the popped elements plus one (for the current day).
- After processing the current price, we push it onto the stack with its calculated span.

---

### 🔄 Edge Case Handling:
1. **Single Price**: The span is always `1` for the first price.
2. **Monotonically Decreasing Prices**: If prices are always decreasing, the span for each price will be `1`.
3. **Monotonically Increasing Prices**: If prices are always increasing, the span will be increasing as well (e.g., 1, 2, 3, ...).
4. **Large Prices**: The algorithm efficiently handles large values of prices up to `10^5` because the stack approach processes each price in constant time.

This solution is optimal, leveraging a stack for efficient span calculation with **O(1)** amortized time complexity per `next()` call, making it highly scalable for large inputs.

````python
struct StockSpanner {
    prices: Vec<i32>,  // stores the stock prices
    stack: Vec<(i32, i32)>,  // stores pairs of (price, span)
}

impl StockSpanner {

    // Creates a new instance of StockSpanner
    fn new() -> Self {
        StockSpanner {
            prices: Vec::new(),
            stack: Vec::new(),
        }
    }
    
    // Adds a new price and calculates its span
    fn next(&mut self, price: i32) -> i32 {
        let mut span = 1;  // The default span is 1 (the price itself)
        
        // While the stack is not empty and the price at the top of the stack is less than or equal to the current price
        while let Some(&(prev_price, prev_span)) = self.stack.last() {
            if prev_price <= price {
                // Pop the stack and add the span of the popped price
                self.stack.pop();
                span += prev_span;
            } else {
                break;
            }
        }
        
        // Push the current price and its calculated span onto the stack
        self.stack.push((price, span));
        
        // Return the span for the current price
        span
    }
}

fn main() {
    // Example usage
    let mut stock_spanner = StockSpanner::new();
    println!("{}", stock_spanner.next(100)); // 1
    println!("{}", stock_spanner.next(80));  // 1
    println!("{}", stock_spanner.next(60));  // 1
    println!("{}", stock_spanner.next(70));  // 2
    println!("{}", stock_spanner.next(60));  // 1
    println!("{}", stock_spanner.next(75));  // 4
    println!("{}", stock_spanner.next(85));  // 6
}
````

````python
class StockSpanner:

    def __init__(self):
        self.stack = []

    def next(self, price: int) -> int:
        # Start with span = 1 (today itself)
        span = 1
        # Collapse all previous days with price <= today's
        while self.stack and self.stack[-1][0] <= price:
            span += self.stack[-1][1]
            self.stack.pop()
        # Push today's price and its span
        self.stack.append((price, span))
        return span    
````
