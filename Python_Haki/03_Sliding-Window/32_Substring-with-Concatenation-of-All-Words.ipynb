You are given a string s and an array of strings words. All the strings of words are of the same length.

A concatenated string is a string that exactly contains all the strings of any permutation of words concatenated.

    For example, if words = ["ab","cd","ef"], then "abcdef", "abefcd", "cdabef", "cdefab", "efabcd", and "efcdab" are all concatenated strings. "acdbef" is not a concatenated string because it is not the concatenation of any permutation of words.

Return an array of the starting indices of all the concatenated substrings in s. You can return the answer in any order.

Example 1:

Input: s = "barfoothefoobarman", words = ["foo","bar"]

Output: [0,9]

Explanation:

The substring starting at 0 is "barfoo". It is the concatenation of ["bar","foo"] which is a permutation of words.
The substring starting at 9 is "foobar". It is the concatenation of ["foo","bar"] which is a permutation of words.

Example 2:

Input: s = "wordgoodgoodgoodbestword", words = ["word","good","best","word"]

Output: []

Explanation:

There is no concatenated substring.

Example 3:

Input: s = "barfoofoobarthefoobarman", words = ["bar","foo","the"]

Output: [6,9,12]

Explanation:

The substring starting at 6 is "foobarthe". It is the concatenation of ["foo","bar","the"].
The substring starting at 9 is "barthefoo". It is the concatenation of ["bar","the","foo"].
The substring starting at 12 is "thefoobar". It is the concatenation of ["the","foo","bar"].

Constraints:

    1 <= s.length <= 104
    1 <= words.length <= 5000
    1 <= words[i].length <= 30
    s and words[i] consist of lowercase English letters.

To solve the problem of finding all starting indices of substrings in `s` that are concatenations of any permutation of the `words` array, we can use a sliding window approach combined with hash maps. This approach ensures we handle large inputs efficiently.

### Approach:

1. **Calculate Sizes**:

   - Determine the length of each word and the total length of the concatenated substring which is `word_length * words_count`.

2. **Build Frequency Map**:

   - Create a frequency map of the words to track their count in the `words` array.

3. **Sliding Window**:
   - Use a sliding window approach to check substrings of length equal to the total concatenated length of the words.
   - Within this window, use another sliding window to check each word's occurrence and ensure the substring matches the frequency map.

### Time Complexity:

- The time complexity is \(O(n \cdot m)\), where \(n\) is the length of the string `s` and \(m\) is the length of each word. This is because we slide over `s` and within each window of size `words_count * word_length`, we check the word frequencies.

### Space Complexity:

- The space complexity is \(O(w)\), where \(w\) is the number of words in the `words` array, for storing the frequency map and the temporary counts.

Here's the Rust implementation:

```rust
use std::collections::HashMap;

impl Solution {
    pub fn find_substring(s: String, words: Vec<String>) -> Vec<i32> {
        let s_chars: Vec<char> = s.chars().collect();
        let word_count = words.len();
        let word_length = words[0].len();
        let total_length = word_count * word_length;

        // Frequency map for the words
        let mut word_freq = HashMap::new();
        for word in &words {
            *word_freq.entry(word.clone()).or_insert(0) += 1;
        }

        let mut results = Vec::new();

        // Iterate over possible starting points
        for i in 0..word_length {
            let mut start = i;
            let mut current_freq = HashMap::new();
            let mut count = 0;

            for end in (i..s_chars.len()).step_by(word_length) {
                if end + word_length <= s_chars.len() {
                    // Extract the word from the substring
                    let word: String = s_chars[end..end + word_length].iter().collect();
                    if word_freq.contains_key(&word) {
                        // Update current frequency map
                        *current_freq.entry(word.clone()).or_insert(0) += 1;
                        count += 1;

                        // If frequency exceeds, move the start index
                        while current_freq[&word] > word_freq[&word] {
                            let start_word: String = s_chars[start..start + word_length].iter().collect();
                            *current_freq.entry(start_word).or_insert(0) -= 1;
                            count -= 1;
                            start += word_length;
                        }

                        // If all words match, record the start index
                        if count == word_count {
                            results.push(start as i32);
                            let start_word: String = s_chars[start..start + word_length].iter().collect();
                            *current_freq.entry(start_word).or_insert(0) -= 1;
                            count -= 1;
                            start += word_length;
                        }
                    } else {
                        // Reset if the word is not in the list
                        current_freq.clear();
                        count = 0;
                        start = end + word_length;
                    }
                }
            }
        }

        results
    }
}
```

### Explanation:

1. **Initialize Data Structures**:

   - `word_freq`: Tracks the frequency of each word.
   - `current_freq`: Tracks the frequency of words in the current window.

2. **Sliding Window**:

   - We check substrings starting at different offsets (`i` from `0` to `word_length - 1`).
   - For each starting point, slide a window and check the substring to see if it matches the concatenation of all words.

3. **Adjust Window**:

   - If a wordâ€™s count exceeds the expected count, adjust the start of the window to maintain valid word counts.

4. **Collect Results**:
   - When a valid window is found, add the starting index to the results.

```python
from collections import defaultdict

class Solution:
    @staticmethod
    def find_substring(s: str, words: list[str]) -> list[int]:
        if not s or not words:
            return []

        word_count = len(words)
        word_length = len(words[0])
        total_length = word_count * word_length

        # Frequency map for the words
        word_freq = defaultdict(int)
        for word in words:
            word_freq[word] += 1

        results = []

        # Iterate over possible starting points
        for i in range(word_length):
            start = i
            current_freq = defaultdict(int)
            count = 0

            for end in range(i, len(s) - word_length + 1, word_length):
                word = s[end:end + word_length]

                if word in word_freq:
                    # Update current frequency map
                    current_freq[word] += 1
                    count += 1

                    # If frequency exceeds, move the start index
                    while current_freq[word] > word_freq[word]:
                        start_word = s[start:start + word_length]
                        current_freq[start_word] -= 1
                        count -= 1
                        start += word_length

                    # If all words match, record the start index
                    if count == word_count:
                        results.append(start)

                        # Move the start index to check for new possible substrings
                        start_word = s[start:start + word_length]
                        current_freq[start_word] -= 1
                        count -= 1
                        start += word_length
                else:
                    # Reset if the word is not in the list
                    current_freq.clear()
                    count = 0
                    start = end + word_length

        return results
```

### Explanation:

   - `word_freq` keeps track of how many times each word should appear in the substring.

2. **Sliding Window**:

   - Iterate over possible starting points for substrings of the length equal to `word_length`.
   - For each starting point, use a sliding window to check substrings of length `total_length`.

3. **Window Adjustment**:

   - Adjust the window by moving the `start` pointer if the frequency of a word exceeds the expected count.
   - Record the starting index of valid substrings where all words are present in the required count.

4. **Edge Cases**:
   - The function handles empty inputs or cases where no valid concatenation is found.
