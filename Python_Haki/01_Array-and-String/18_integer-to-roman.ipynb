````python
def int_to_roman(num: int) -> str:
    # List of tuples mapping integers to Roman numerals
    val_to_roman = [
        (1000, "M"),
        (900, "CM"),
        (500, "D"),
        (400, "CD"),
        (100, "C"),
        (90, "XC"),
        (50, "L"),
        (40, "XL"),
        (10, "X"),
        (9, "IX"),
        (5, "V"),
        (4, "IV"),
        (1, "I"),
    ]

    result = ""

    # Iterate through the value to Roman mapping
    for value, roman in val_to_roman:
        # While num is greater than or equal to the current value
        while num >= value:
            # Subtract the value from num
            num -= value
            # Append the corresponding Roman numeral to the result
            result += roman

    return result
````

To convert an integer to a Roman numeral, we can systematically break down the number by its decimal places and map each part to its corresponding Roman numeral. The Roman numeral system has a unique representation for certain values, especially when dealing with numbers like 4, 9, 40, 90, 400, and 900, which use subtractive combinations.

### Steps to Convert an Integer to a Roman Numeral

1. **Create a Mapping**: Create a list of tuples where each tuple consists of an integer and its corresponding Roman numeral. This list will help us convert the integer to the Roman numeral by iterating through it in descending order.

2. **Iterate Through the Mapping**: Starting from the largest value in the list, repeatedly subtract the value from the integer and append the corresponding Roman numeral to the result string.

3. **Stop When Done**: Continue the process until the integer is reduced to zero.

### Explanation:

- **Iteration and Subtraction**:
  - The code iterates over the mapping list.
  - For each value, as long as the integer (`num`) is greater than or equal to that value, it subtracts the value from `num` and appends the corresponding Roman numeral to `result`.
  
- **Output**:
  - The process continues until `num` is reduced to zero, at which point the `result` contains the complete Roman numeral.

### Time and Space Complexity:
