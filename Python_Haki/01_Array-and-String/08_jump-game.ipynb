````python
def can_jump(nums):
    # The farthest position we can currently reach
    farthest_reachable = 0

    # Iterate over each index in the array
    for i in range(len(nums)):
        # If the current index is beyond the farthest reachable point, return False
        if i > farthest_reachable:
            return False

        # Update the farthest position that can be reached
        farthest_reachable = max(farthest_reachable, i + nums[i])

        # If the farthest reachable position is beyond or at the last index, return True
        if farthest_reachable >= len(nums) - 1:
            return True

    # If we complete the loop without returning, check if we reached the last index
    return farthest_reachable >= len(nums) - 1
````

````python
# Example usage:
print(can_jump([2, 3, 1, 1, 4]))  # Output: True
print(can_jump([3, 2, 1, 0, 4]))  # Output: False
````

### Explanation:
1. **farthest_reachable**: This variable keeps track of the farthest index you can reach as you iterate through the array.
2. **for loop**: We go through each index to see if it's reachable.
3. **Condition check (`i > farthest_reachable`)**: If the current index is beyond what we can reach, it's impossible to proceed further, so we return `False`.
4. **Updating farthest_reachable**: We keep updating the maximum index we can reach from the current position.
5. **Final check**: If at any point we can reach or surpass the last index, we return `True`.

### Time and Space Complexity:
- **Time Complexity**: O(n) – We only need to traverse the array once.
- **Space Complexity**: O(1) – No extra space is used beyond the variables.

This solution efficiently determines if it's possible to reach the last index of the array.

````python
def can_jump(nums):
    # Initialize the maximum reachable index to 0
    max_reachable = 0
    
    # Iterate through each index in the array
    for current_index, jump_length in enumerate(nums):
        # If the current index is beyond the maximum reachable index, we can't proceed
        if current_index > max_reachable:
            return False
        
        # Update the maximum reachable index
        max_reachable = max(max_reachable, current_index + jump_length)
        
        # If we can reach or exceed the last index, return True
        if max_reachable >= len(nums) - 1:
            return True
    
    # Return False if we finish iterating and haven't reached the last index
    return False

# can_jump([2,3,1,1,4])
can_jump([3,2,1,0,4])
````
