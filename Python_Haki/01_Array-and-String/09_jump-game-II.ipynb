2. **current_end**: This variable represents the farthest index that can be reached with the current number of jumps.
3. **farthest_reachable**: This variable tracks the farthest index that can be reached overall while iterating through the array.
4. **for loop**: The loop iterates through the array up to the second-last element since the last element is the target.
5. **Condition check (`i == current_end`)**: When the loop reaches the end of the current jump range, we increment the jump count and update the range to the farthest reachable index.
6. **Breaking the loop**: If the current range reaches or exceeds the last index, the loop is terminated early as the goal has been achieved.

### Time and Space Complexity:

- **Time Complexity**: O(n) – The algorithm processes each element in the array exactly once.
- **Space Complexity**: O(1) – The solution uses a constant amount of extra space.

This code efficiently computes the minimum number of jumps needed to reach the last index, ensuring optimal performance for large arrays.
